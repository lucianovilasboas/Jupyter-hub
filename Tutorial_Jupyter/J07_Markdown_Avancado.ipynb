## 📚 Explicação Detalhada

### 1. Tabelas

Tabelas em Markdown usam `|` para colunas e `-` para separar cabeçalho do conteúdo:

```
| Nome | Idade | Cidade |
|------|-------|--------|
| Ana  | 25    | SP     |
| João | 32    | RJ     |
```

**Alinhamento:**
- `:---` = esquerda
- `:--:` = centro
- `---:` = direita

```
| Produto | Preço | Estoque |
|:--------|:-----:|--------:|
| Café    | R$12  |   50   |
| Chá     | R$8   |  120   |
```

Resultado:

| Produto | Preço | Estoque |
|:--------|:-----:|--------:|
| Café    | R$12  |   50   |
| Chá     | R$8   |  120   |
| Suco    | R$10  |   80   |

### 2. Links

Links são feitos com `[texto](url)`:

```
[Clique aqui para visitar o Jupyter](https://jupyter.org)
```

Resultado: [Clique aqui para visitar o Jupyter](https://jupyter.org)

**Links com título (tooltip):**
```
[Jupyter](https://jupyter.org "Site oficial do Jupyter")
```

**Links de referência:**
```
[Jupyter][1]
[Python][2]

[1]: https://jupyter.org
[2]: https://python.org
```

### 3. Imagens

Imagens seguem a mesma lógica dos links, mas com `!` na frente:

```
![Texto alternativo](url_da_imagem)
```

Exemplo com logo do Jupyter:
```
![Jupyter Logo](https://jupyter.org/assets/homepage/main-logo.svg)
```

**Imagem com link:**
```
[![Jupyter](https://jupyter.org/assets/homepage/main-logo.svg)](https://jupyter.org)
```

**Três formas de inserir imagens no Jupyter:**
1. URL direta (funciona online)
2. Caminho local (imagem na mesma pasta do notebook)
3. Anexo via **Edit > Insert Image** (incorpora no .ipynb, mas aumenta o arquivo)

### 5. HTML dentro do Markdown

O Jupyter aceita HTML puro dentro de células Markdown. Isso permite personalizações que o Markdown puro não alcança:

```html
<span style="color: red; font-size: 20px;">Texto vermelho gigante!</span>

<div style="background-color: #f0f0f0; padding: 15px; border-radius: 10px;">
    <h3 style="color: #2c3e50; margin-top: 0;">📦 Bloco Destacado</h3>
    <p>Este é um bloco estilizado com HTML!</p>
</div>
```

Resultado:

<span style="color: red; font-size: 20px;">Texto vermelho gigante!</span>

<div style="background-color: #f0f0f0; padding: 15px; border-radius: 10px;">
    <h3 style="color: #2c3e50; margin-top: 0;">📦 Bloco Destacado</h3>
    <p>Este é um bloco estilizado com HTML!</p>
</div>

### 6. Citações e Notas

Use `>` para citações:
```
> "O Jupyter Notebook é a melhor ferramenta para aprender Python."
> — *Todo mundo que já usou*
```

Resultado:
> "O Jupyter Notebook é a melhor ferramenta para aprender Python."
> — *Todo mundo que já usou*

**Citações aninhadas:**
```
> Nível 1
>> Nível 2
>>> Nível 3
```

## 💻 Exemplos Práticos

```````python
# Exemplo 1: Gerando tabela com Python
alunos = [
    ["Ana", 8.5, "Aprovado"],
    ["João", 6.0, "Recuperação"],
    ["Maria", 9.2, "Aprovado"],
    ["Pedro", 4.5, "Reprovado"]
]

print("| Nome  | Nota | Situação   |")
print("|-------|------|------------|")
for nome, nota, sit in alunos:
    print(f"| {nome:5} | {nota:4} | {sit:10} |")
print("\n📊 Tabela gerada dinamicamente!")
```````

```````python
# Exemplo 2: Cardápio com Python
cardapio = {
    "X-Burger": 25.90,
    "X-Salada": 22.50,
    "X-Bacon": 29.90,
    "Milk Shake": 18.00,
    "Batata Frita": 12.00
}

print("🍔 CARDÁPIO DO LANCHES 🍔")
print("=\n" + "=" * 30)
print(f"{'Item':20} {'Preço':>8}")
print("-" * 30)
for item, preco in cardapio.items():
    print(f"{item:20} R$ {preco:>6.2f}")
print("-" * 30)
total = sum(cardapio.values())
print(f"{'TOTAL DO CARDÁPIO':20} R$ {total:>6.2f}")
```````

```````python
# Exemplo 3: Links dinâmicos
sites = [
    ("Jupyter", "https://jupyter.org"),
    ("Python", "https://python.org"),
    ("GitHub", "https://github.com"),
    ("Stack Overflow", "https://stackoverflow.com")
]

print("🔗 LINKS ÚTEIS")
print("=" * 30)
for nome, url in sites:
    print(f"📎 {nome}: {url}")
print("=" * 30)
print("💡 Cole esses links no navegador!")
```````

```````python
# Exemplo 4: Mini-blog com HTML
titulo = "🎮 Review: Zelda Tears of the Kingdom"
nota = 10
comentario = "Simplesmente a melhor aventura da década!"
autor = "João, o Gamer"

print(f"""
<div style='background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
            color: white; padding: 20px; border-radius: 15px;'>
    <h2 style='margin-top:0'>{titulo}</h2>
    <div style='font-size: 24px;'>
        {'⭐' * nota}{'☆' * (10 - nota)}
    </div>
    <p style='font-size: 18px;'><em>"{comentario}"</em></p>
    <p style='text-align: right;'>— <strong>{autor}</strong></p>
</div>
""")
```````

## 📌 Resumo

- **Tabelas**: use `|` e `-`, alinhe com `:`
- **Links**: `[texto](url)` ou `[texto][ref]`
- **Imagens**: `![alt](url)` ou `[![alt](url)](link)`
- **Blocos de código**: ``` ```linguagem + código + ``````
- **HTML**: funciona dentro de células Markdown
- **Citações**: `>` no início da linha
- Markdown avançado transforma notebooks em documentos profissionais

## 💡 Curiosidades

- O Markdown foi criado em 2004 por **John Gruber** com contribuição de **Aaron Swartz** (co-fundador do Reddit).
- O símbolo `>` para citações veio dos emails dos anos 90, onde `>` indicava resposta.
- Você pode usar `\` para **escapar** caracteres especiais: `\*isto não é itálico\*`.
- O Jupyter suporta **LaTeX** para fórmulas matemáticas: `$E = mc^2$` vira $E = mc^2$.

## ✅ Boas Práticas

1. **Tabelas com cabeçalho claro** — sempre separe com `| --- | --- |`
2. **Use `![alt]` descritivo** em imagens — acessibilidade importa
3. **Blocos de código com linguagem** — ` ```python ` é melhor que ` ``` `
4. **HTML com moderação** — Markdown já resolve 90% dos casos
5. **Links em contexto** — "[veja mais](url)" é melhor que "clique aqui"

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Célula markdown não renderizar HTML | O HTML é renderizado sim, mas verifique a sintaxe |
| Link quebrado | Teste o URL antes de inserir |
| Imagem não aparece | Verifique se a URL está acessível |
| Tabela desalinhada | Mantenha o mesmo número de colunas em todas as linhas |
| Bloco de código sem fechamento | Sempre feche com `````` as três crases |
| HTML muito complexo | Pode quebrar a visualização do notebook |
