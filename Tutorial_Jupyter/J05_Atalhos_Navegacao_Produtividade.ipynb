No Jupyter, os **atalhos de teclado** são seus power-ups! Eles transformam seu fluxo de trabalho de "lento e clicando" para "rápido e precisso".

A filosofia dos atalhos no Jupyter:
- 🖱️ **Mouse** = iniciante (ok, mas lento)

## 📚 Explicação Detalhada

### 1. Os Dois Modos (Revisão Rápida)

Lembrando: toda célula vive em dois estados:

| Modo | Borda | Tecla | O que faz |
|------|-------|-------|-----------|
| **Edição** 🟢 | Verde | **Enter** | Escrever conteúdo |
| **Comando** 🔵 | Azul | **Esc** | Gerenciar células |

**ATALHO MAIS IMPORTANTE:** `Enter` para editar, `Esc` para comandar.

> 💡 **Guarde isso**: todos os atalhos de gerenciamento de células funcionam no **modo comando** (borda azul)!

### 2. Atalhos Essenciais — Tabela Completa

**Navegação entre células:**

| Atalho | Ação |
|--------|------|
| ⬆️ / ⬇️ | Navegar entre células |
| **K** / **J** | Cima / Baixo (alternativa) |
| **Ctrl + Enter** | Executar célula atual |
| **Shift + Enter** | Executar e ir para próxima |
| **Alt + Enter** | Executar e inserir abaixo |

**Gerenciamento de células:**

| Atalho | Ação |
|--------|------|
| **A** | Inserir célula **acima** |
| **B** | Inserir célula **abaixo** |
| **DD** | Deletar célula (pressione D duas vezes) |
| **Z** | Desfazer deleção de célula |
| **C** | Copiar célula |
| **X** | Recortar célula |
| **V** | Colar célula abaixo |
| **Shift + V** | Colar célula acima |

**Tipo de célula:**

| Atalho | Ação |
|--------|------|
| **M** | Transformar em Markdown |
| **Y** | Transformar em Code |
| **R** | Transformar em Raw |

**Seleção e manipulação:**

| Atalho | Ação |
|--------|------|
| **Shift + ⬆/⬇** | Selecionar múltiplas células |
| **Shift + M** | Mesclar células selecionadas |
| **Ctrl + S** | Salvar notebook |
| **Ctrl + Shift + -** | Dividir célula no cursor |
| **Ctrl + Click** | Múltiplos cursores na mesma célula |

**Outros úteis:**

| Atalho | Ação |
|--------|------|
| **H** | Mostrar todos os atalhos |
| **O** | Mostrar/esconder saída da célula |
| **L** | Mostrar/esconder números de linha |
| **0 + 0** | Reiniciar kernel |
| **I + I** | Interromper kernel |
| **S** | Salvar (atalho alternativo) |
| **F** | Localizar e substituir |

### 3. O Atalho Secreto: Ctrl + Shift + P

A **paleta de comandos** (Ctrl + Shift + P) é como um "Google" dos atalhos do Jupyter.

Ela mostra **todos** os comandos disponíveis — mesmo aqueles sem atalho definido!

Experimente: pressione `Ctrl + Shift + P` neste notebook e comece a digitar o que você quer fazer (ex: "dividir", "mesclar", "executar").

```python
# Experimente agora!
# 1. Pressione Esc para modo comando
# 2. Pressione H para ver a ajuda de atalhos
print("⌨️  Pressione H no modo comando para ver TODOS os atalhos!")
```

### 4. Dicas de Produtividade

**Fluxo de trabalho eficiente:**

1. **Planeje antes de codar**: escreva o markdown primeiro (títulos, seções)
2. **Use Shift + Enter** como padrão (executa e avança)
3. **A + B** para criar células rapidamente
4. **M + Y** para alternar entre markdown e código
5. **DD** para deletar células erradas (e Z para trazer de volta)

## 📌 Resumo

- **Enter** = modo edição / **Esc** = modo comando
- **A** (acima) / **B** (abaixo) = inserir células
- **DD** = deletar / **Z** = desfazer
- **M** = markdown / **Y** = código
- **Ctrl + Enter** = executar / **Shift + Enter** = executar e avançar
- **Ctrl + Shift + P** = paleta de comandos
- **H** = ajuda de atalhos
- Produtividade máxima = evitar o mouse!

## 💡 Curiosidades

- Os atalhos do Jupyter foram inspirados no **Vim** (editor de texto para programadores avançados) — por isso o modo comando vs edição.
- A tecla **H** de ajuda é uma homenagem ao comando `:help` do Vim.
- O **DD** para deletar foi inspirado no Vim, onde `dd` deleta uma linha inteira.
- Você pode personalizar os atalhos! Veja em **Help > Edit Keyboard Shortcuts**.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Tentar atalho no modo errado | Verifique se está no modo comando (borda azul) |
| Pressionar D uma vez só | DD = D pressionado DUAS vezes rápido |
| Esquecer o Enter para editar | Enter entra no modo edição |
| Misturar Ctrl e Esc | Ctrl = modo edição, Esc = modo comando |
| Não conseguir desfazer deleção | Z funciona IMEDIATAMENTE após DD |
| Pressionar tecla aleatória sem querer | No modo comando, teclas viram comandos! |
