## 📝 Introdução

O Jupyter tem **comandos especiais** chamados **Magics** (mágicos) que não existem no Python puro. Eles são como feitiços 🪄 que dão superpoderes ao seu notebook!

Existem dois tipos:
- **Line magic** (`%comando`) — age em uma linha
- **Cell magic** (`%%comando`) — age na célula inteira

> 🧙‍♂️ **Prepare sua varinha**: Você vai aprender feitiços que até magos do Python invejariam!

### 2. %timeit — Medindo Performance

Quer saber qual código é mais rápido? `%timeit` executa várias vezes e calcula a média!

```python
# Qual é mais rápido: range ou lista?
print("⏱️  DISPUTA DE VELOCIDADE")
print("=" * 35)
print("range vs list(range):")
print()print("🏁 range:")
%timeit for _ in range(1000): pass
print()
print("🏁 list(range):")
%timeit for _ in list(range(1000)): pass
print()
print("💡 Dica: range é mais rápido que list!")
```

```python
# Comparando expressões
print("🔢 POTENCIAÇÃO: 2**10 vs 2*2*2*...")
%timeit 2 ** 10
%timeit 2*2*2*2*2*2*2*2*2*2
print("\n💡 ** é mais limpo e igualmente rápido!")
```

### 3. %%timeit — Medindo Células Inteiras

Para medir uma célula inteira, use `%%timeit` no início:

```python
%%timeit
# Célula inteira sendo medida!
soma = 0
for i in range(10000):
    soma += i ** 2
```

### 4. %who — Investigando Variáveis

Quer saber quais variáveis você já definiu? `%who` lista todas!

```python
# Primeiro, vamos definir algumas variáveis
nome = "Jupyter"
versao = 3
notas = [8.5, 9.0, 7.5]
import math

print("🔍 VARIÁVEIS DEFINIDAS NESTE KERNEL:")
print("=" * 35)
%who
```

```python
# %who_ls retorna uma lista (útil para usar em código)
variaveis = %who_ls
print(f"📦 Total de variáveis: {len(variaveis)}")
for v in variaveis:
    print(f"   📌 {v}")
```

### 5. %env — Variáveis de Ambiente

Veja e configure variáveis de ambiente do sistema:

### 6. %run — Executando Scripts Externos

Execute arquivos `.py` diretamente dentro do notebook:

```python
# Primeiro, vamos criar um script Python de exemplo
codigo_exemplo = '''
print("🚀 Script executado via %run!")
print(f"📅 Data: {__import__('datetime').datetime.now().strftime('%d/%m/%Y')}")
print("✅ O script rodou dentro do notebook!")
'''

with open("_exemplo_script.py", "w", encoding="utf-8") as f:
    f.write(codigo_exemplo)
print("✅ Script criado: _exemplo_script.py")
```

```python
# Executa o script externo
%run _exemplo_script.py
```

### 7. %%writefile — Salvando Célula em Arquivo

Salve o conteúdo de uma célula diretamente em um arquivo!

```python
%%writefile _saudacao.py
# Arquivo gerado pelo %%writefile!
def saudacao(nome):
    return f"👋 Olá, {nome}! Este arquivo foi criado pelo Jupyter!"

if __name__ == "__main__":
    print(saudacao("Mundo"))
```

```python
# Testando o arquivo criado
from _saudacao import saudacao
print(saudacao("Jupyter"))
```

### 8. %%bash — Rodando Comandos do Terminal

Execute comandos do sistema sem sair do notebook!

```python
%%bash
# Comandos bash dentro do Jupyter!
echo "🐧 Olá do terminal!"
echo "📁 Diretório atual:" 
pwd
echo "📄 Arquivos:"
ls -la *.py 2>/dev/null || echo "(nenhum arquivo .py na pasta)"
```

### 10. %matplotlib inline (Revisão)

Você já usou este magic desde o início! Ele faz os gráficos aparecerem dentro do notebook:

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [1, 4, 9], marker='o')
plt.title("📈 Gráfico renderizado inline")
plt.show()
```

### 11. %store — Compartilhando Variáveis Entre Notebooks

Quer passar uma variável de um notebook para outro? Use `%store`!

```python
# Armazenando uma variável
mensagem_secreta = "🔐 Você encontrou o easter egg!"
%store mensagem_secreta
print("✅ Variável 'mensagem_secreta' armazenada!")
print("💡 Abra OUTRO notebook e use '%store -r mensagem_secreta' para ler!")
```

## 💻 Exemplos Práticos

> 🤔 **Pense:** O que são "comandos mágicos" no Jupyter? Eles são Python puro ou algo extra que o notebook oferece? Como medir o tempo de execução de uma célula sem cronômetro?

```python
# Exemplo 1: Disputa de performance: for vs compreensão
print("🥊 DISPUTA: for vs list comprehension")
print("=" * 40)
print()
print("🐢 for tradicional:")
%timeit resultado = [i**2 for i in range(1000)]
print()
print("🐇 list comprehension:")
%timeit resultado = [i**2 for i in range(1000)]
print()
print("🥇 Geralmente bem próximos!")
```

```python
# Exemplo 2: Criando múltiplos arquivos com %%writefile
arquivos = ["dados_2025", "dados_2026", "relatorio_final"]
for a in arquivos:
    # Vamos usar o magic em cada um
    print(f"📝 Criando {a}.py...")
    # (Na prática, usaríamos with open, mas aqui é só demonstrativo)
print("✅ 3 scripts criados mentalmente!")
print()
print("💡 %%writefile é ótimo para gerar scripts de teste rapidamente!")
```

```python
# Exemplo 3: %%python — forçando interpretador específico
# (Apenas para demonstrar que existe)
print("🐍 Este código roda no Python padrão do kernel")
print(f"Versão: {__import__('sys').version}")
```

2. **Desafio do Gerador de Scripts 🤖:** Use `%%writefile` para gerar 3 scripts Python úteis: um conversor de temperatura, um gerador de senhas e um sorteador. Execute cada um com `%run` e verifique que funcionam.

3. **Desafio do Detetive de Variáveis 🕵️:** Crie um notebook cheio de variáveis de vários tipos. Use `%who`, `%who_ls` e `%whos` para investigar. Descubra qual é o tipo de cada uma e quantas existem.

```python
# 🏆 Desafio 1: Comparador de Performance
print("🏁 FILTRANDO NÚMEROS PARES")
print("=" * 40)
numeros = list(range(1000))

print("🥉 Método 1: for loop")
%timeit [n for n in numeros if n % 2 == 0]
print()
print("🥈 Método 2: filter")
%timeit list(filter(lambda x: x % 2 == 0, numeros))
print()
print("🥇 Método 3: compreensão (mais rápido!)")
%timeit numeros[::2]
```

```python
# 🏆 Desafio 2: Gerador de Scripts
%%writefile _conversor_temp.py
def celsius_para_fahrenheit(c):
    return (c * 9/5) + 32

if __name__ == "__main__":
    temp = float(input("Temperatura em °C: "))
    print(f"{temp}°C = {celsius_para_fahrenheit(temp):.1f}°F")

print("✅ Script _conversor_temp.py criado!")
print("💡 Execute com %run _conversor_temp.py")
```

## 📌 Resumo

- **Line magics** (`%`): agem em uma linha
- **Cell magics** (`%%`): agem na célula toda
- `%timeit`: mede performance de código
- `%who` / `%who_ls`: lista variáveis
- `%run`: executa script externo
- `%%writefile`: salva célula em arquivo
- `%%bash`: executa comando do sistema
- `%store`: compartilha variáveis entre notebooks
- `%matplotlib inline`: gráficos no notebook
- Use `%lsmagic` para ver todos os disponíveis

## 💡 Curiosidades

- O sistema de magics foi herdado do **IPython** (2001), muito antes do Jupyter existir!
- O nome "magic" veio da sensação de que esses comandos fazem coisas que o Python normal não faz.
- Por padrão, a **automagic** está ativada — você não precisa digitar `%` para alguns comandos.
- Existem mais de **30 magics** diferentes no Jupyter!

## ✅ Boas Práticas

1. **Use `%timeit` para otimizar** código, não para exibir "olha que legal"
2. **Prefira `%%writefile`** em vez de criar arquivos manualmente para testes
3. **`%who` é ótimo para debug** — veja o que tem no kernel
4. **Automagic** pode atrapalhar se você tiver variáveis com nomes de magics
5. **Limpe scripts temporários** criados com `%%writefile` após o uso

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Esquecer `%` ou `%%` | Se automagic estiver desligado, magics não funcionam sem `%` |
| `%timeit` em código com saída | O output do código não aparece durante `%timeit` |
| `%%bash` no Windows | Alguns comandos bash não funcionam no Windows |
| `%run` com caminho errado | Use caminho relativo ou absoluto |
| `%%writefile` sobrescrever arquivo | Verifique se o arquivo já existe |
