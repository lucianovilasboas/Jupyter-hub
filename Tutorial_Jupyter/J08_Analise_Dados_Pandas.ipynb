## 📚 Explicação Detalhada

### 1. Importando Bibliotecas

No Jupyter, é padrão ter uma **célula de setup** no começo com todos os imports:

```python
# Célula de setup — execute primeiro!
import pandas as pd
import numpy as np

print("✅ pandas importado com sucesso!")
print(f"📦 Versão do pandas: {pd.__version__}")
```

### 2. Criando Nossos Próprios Dados

Vamos criar datasets manualmente para não depender de internet:

```python
# 🥊 LENDAS DO STREET FIGHTER
dados_lutadores = {
    "Lutador": ["Ryu", "Ken", "Chun-Li", "Guile", "Zangief", "Blanka", "Dhalsim", "E. Honda"],
    "País": ["Japão", "EUA", "China", "EUA", "Rússia", "Brasil", "Índia", "Japão"],
    "Altura(cm)": [175, 176, 169, 182, 214, 192, 176, 185],
    "Peso(kg)": [85, 83, 58, 99, 181, 98, 52, 137],
    "Estilo": ["Ansatsuken", "Ansatsuken", "Kung Fu", "Judô", "Luta Livre", "Selvagem", "Ioga", "Sumô"]
}

sf = pd.DataFrame(dados_lutadores)
print("✅ DataFrame dos lutadores criado!")
sf
```

```python
# 🎬 CLÁSSICOS DO STUDIO GHIBLI
dados_filmes = {
    "Filme": ["A Viagem de Chihiro", "Meu Amigo Totoro", "Princesa Mononoke",
              "O Castelo Animado", "O Serviço de Entregas da Kiki", "Ponyo",
              "O Vôo das Borboletas", "O Castelo no Céu"],
    "Ano": [2001, 1988, 1997, 2004, 1989, 2008, 1992, 1986],
    "Nota": [8.6, 8.2, 8.4, 8.2, 7.9, 7.7, 7.8, 8.0],
    "Duração(min)": [125, 86, 133, 119, 103, 101, 94, 124],
    "Diretor": ["Hayao Miyazaki"] * 8
}

ghibli = pd.DataFrame(dados_filmes)
print("✅ DataFrame dos filmes Ghibli criado!")
ghibli
```

### 3. Explorando o DataFrame

O pandas tem métodos incríveis para dar uma "olhadinha" nos dados:

```python
# Ver as primeiras linhas
print("👀 PRIMEIROS LUTADORES:")
sf.head(3)
```

```python
# Ver as últimas linhas
print("👀 ÚLTIMOS FILMES:")
ghibli.tail(3)
```

```python
# Informações gerais
print("📊 INFO DO DATASET DE LUTADORES:")
sf.info()
```

```python
# Estatísticas descritivas
print("📈 ESTATÍSTICAS DOS FILMES GHIBLI:")
ghibli.describe()
```

### 4. Selecionando e Filtrando

O pandas permite consultas poderosas:

```python
# Selecionar uma coluna
print("🎮 ESTILOS DE LUTA:")
sf["Estilo"]
```

```python
# Filtrar linhas
print("🇧🇷 LUTADORES BRASILEIROS:")
sf[sf["País"] == "Brasil"]
```

```python
# Filtrar com múltiplas condições
print("🎬 FILMES GHIBLI COM NOTA > 8.0 E ANO > 1995:")
ghibli[(ghibli["Nota"] > 8.0) & (ghibli["Ano"] > 1995)]
```

```python
# Ordenar dados
print("📊 LUTADORES MAIS PESADOS:")
sf.sort_values("Peso(kg)", ascending=False)
```

```python
# Agrupar e contar
print("🌍 PAÍSES DOS LUTADORES:")
sf.groupby("País").size().reset_index(name="Quantidade")
```

```python
# Média por grupo
print("📐 ALTURA MÉDIA POR ESTILO DE LUTA:")
sf.groupby("Estilo")["Altura(cm)"].mean().round(1)
```

## 💻 Exemplos Práticos

> 🤔 **Pense:** O que o Jupyter tem de especial para análise de dados? Por que não usar só Python puro? Que vantagem as células trazem para explorar uma base de dados?

```python
# Exemplo 1: Detetive de Dados 🕵️
print("🔍 RELATÓRIO DO DETETIVE")
print("=" * 35)
print()
print(f"📊 Total de lutadores: {len(sf)}")
print(f"📊 Total de filmes: {len(ghibli)}")
print()
print(f"🏋️ Lutador mais pesado: {sf.loc[sf['Peso(kg)'].idxmax(), 'Lutador']}")
print(f"  -> {sf['Peso(kg)'].max()} kg")
print(f"🏋️ Lutador mais leve: {sf.loc[sf['Peso(kg)'].idxmin(), 'Lutador']}")
print(f"  -> {sf['Peso(kg)'].min()} kg")
print()
print(f"🎬 Filme mais bem avaliado: {ghibli.loc[ghibli['Nota'].idxmax(), 'Filme']}")
print(f"  -> Nota: {ghibli['Nota'].max()}")
print(f"📅 Filme mais antigo: {ghibli.loc[ghibli['Ano'].idxmin(), 'Filme']} ({ghibli['Ano'].min()})")
```

```python
# Exemplo 2: IMC dos Lutadores (nosso primeiro cálculo!)
sf["IMC"] = sf["Peso(kg)"] / ((sf["Altura(cm)"] / 100) ** 2)
sf["IMC"] = sf["IMC"].round(1)

print("📊 IMC DOS LUTADORES:")
print("-" * 45)
for _, row in sf.iterrows():
    imc = row["IMC"]
    if imc < 18.5:
        faixa = "Abaixo do peso"
    elif imc < 25:
        faixa = "Saudável ✅"
    elif imc < 30:
        faixa = "Sobrepeso"
    else:
        faixa = "Obesidade"
    print(f"{row['Lutador']:10} | IMC: {imc:5.1f} | {faixa}")
```

```python
# Exemplo 3: Décadas dos filmes Ghibli
ghibli["Década"] = (ghibli["Ano"] // 10) * 10

print("📅 FILMES POR DÉCADA:")
print("-" * 30)
for decada in sorted(ghibli["Década"].unique()):
    filmes_dec = ghibli[ghibli["Década"] == decada]
    nomes = ", ".join(filmes_dec["Filme"].values)
    print(f"{decada}s: {len(filmes_dec)} filme(s)")
    print(f"   {nomes}")
    print()
```

```python
# Exemplo 4: Correlação divertida
print("🔍 SERÁ QUE ALTURA E PESO SE RELACIONAM?")
print("-" * 40)
correlacao = sf["Altura(cm)"].corr(sf["Peso(kg)"])
print(f"Correlação altura x peso: {correlacao:.2f}")
print()
if correlacao > 0.7:
    print("✅ Correlação forte! Mais altos tendem a ser mais pesados.")
elif correlacao > 0.3:
    print("📊 Correlação moderada.")
else:
    print("❌ Correlação fraca.")
```

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Esquecer de importar pandas | `import pandas as pd` na primeira célula |
| Caminho do CSV errado | Use o caminho completo ou coloque na mesma pasta |
| Esquecer `()` em métodos | `df.head` não funciona, use `df.head()` |
| Filtro sem parênteses | `df[x > 5 & y < 10`] precisa de `df[(x > 5) & (y < 10)]` |
| Modificar dados e não perceber | Sempre verifique com `head()` após alterações |
