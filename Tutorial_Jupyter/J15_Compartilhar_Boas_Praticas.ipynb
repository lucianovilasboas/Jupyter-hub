### 1. Preparando o Notebook para Compartilhar

Antes de compartilhar, siga estes passos:

**Passo 1: Limpe as saídas**
- Menu: **Cell > All Output > Clear**
- Remove resultados antigos que podem confundir

**Passo 2: Reinicie e execute tudo**
- Menu: **Kernel > Restart & Run All**
- Garante que tudo funciona do zero
- Verifica se não há dependências de ordem ocultas

**Passo 3: Revise**
- Verifique se todas as células executaram sem erro
- Confira se os textos estão claros
- Teste se links e imagens funcionam

````python
# Checklist de preparação
print("📋 CHECKLIST PRÉ-COMPARTILHAMENTO")
print("=" * 40)
print()
checklist = [
    "[ ] Cell > All Output > Clear",
    "[ ] Kernel > Restart & Run All",
    "[ ] Todas as células executaram sem erro?",
    "[ ] Textos estão claros e sem erros?",
    "[ ] Links e imagens funcionam?",
    "[ ] Títulos e seções estão organizados?",
    "[ ] Removeu dados sensíveis ou senhas?",
]
for item in checklist:
    print(f"   {item}")
print()
print("✅ Após marcar tudo, está pronto para compartilhar!")
````

### 2. Exportando para Diferentes Formatos

O Jupyter permite exportar seu notebook em vários formatos:

**File > Download As >**

| Formato | Extensão | Ideal para |
|---------|----------|------------|
| **HTML** | `.html` | Publicar na web, compartilhar como página |
| **PDF** | `.pdf` | Relatórios formais, trabalhos acadêmicos |
| **Markdown** | `.md` | Documentação, GitHub README |
| **Python** | `.py` | Script limpo sem células |
| **Slides** | `.slides` | Apresentações |
| **Notebook** | `.ipynb` | Original (sempre mantenha!) |

> 💡 **Dica**: Para PDF, pode ser necessário instalar LaTeX. O HTML é mais simples e funciona sempre!

````python
# Simulação: selecionando formato de exportação
print("📤 ESCOLHA O FORMATO DE EXPORTAÇÃO:")
print("=" * 35)
print("1. 📄 HTML — Web (recomendado)")
print("2. 📕 PDF — Relatório formal")
print("3. 📝 Markdown — Documentação")
print("4. 🐍 Python — Script puro")
print("5. 📽️  Slides — Apresentação")
print("=" * 35)
print("💡 File > Download As > escolha o formato!")
````

### 3. Compartilhando no GitHub

O GitHub é a plataforma mais popular para compartilhar notebooks!

**Opção 1: Repositório completo**
1. Crie um repositório no GitHub
2. Faça upload dos seus arquivos `.ipynb`
3. O GitHub renderiza notebooks automaticamente!

**Opção 2: Gist (para arquivos individuais)**
1. Acesse [gist.github.com](https://gist.github.com)
2. Cole o conteúdo JSON do seu `.ipynb`
3. Nomeie com extensão `.ipynb`
4. Crie como público ou secreto

**Dica para Git:**
- Adicione `.ipynb_checkpoints/` ao `.gitignore`
- Notebooks geram diffs grandes, mas o GitHub lida bem

````python
# Comandos git básicos para notebook
print("🐙 GIT + JUPYTER")
print("=" * 35)
print("$ git init")
print("$ git add *.ipynb")print("$ git commit -m 'Meu primeiro notebook!'")
print("$ git remote add origin <url>")
print("$ git push -u origin main")print("=" * 35)
print("📌 Lembre-se do .gitignore!")
````

### 4. Nbviewer — Visualizador Universal

O [Nbviewer](https://nbviewer.jupyter.org/) é um serviço GRATUITO que renderiza qualquer notebook público:

**Como usar:**
1. Coloque seu notebook em algum lugar público (GitHub, Gist, Dropbox)
2. Acesse nbviewer.jupyter.org
3. Cole a URL do seu notebook
4. Pronto! Link compartilhável gerado!

**Vantagens:**
- ✅ Não precisa de conta
- ✅ Renderiza gráficos e markdown perfeitamente
- ✅ Link permanente enquanto o arquivo existir
- ✅ Suporta GitHub, Gists, URLs diretas

````python
# Simulação: Compartilhando via Nbviewer
print("🌐 COMPARTILHANDO VIA NBVIEWER")
print("=" * 40)
print("1. Faça upload do .ipynb no GitHub")
print("2. Copie a URL do arquivo")
print("   ex: github.com/usuario/repo/notebook.ipynb")
print("3. Acesse nbviewer.jupyter.org")
print("4. Cole a URL e clique 'Go'!")
print("=" * 40)
print("🎉 Link gerado! Compartilhe com quem quiser!")
````

### 5. Senha e Segurança

Se você roda o Jupyter em um servidor acessível pela rede, é importante proteger com senha:

```bash
jupyter notebook password
```

Isso solicita uma senha que será exigida para acessar o Jupyter pelo navegador.

**Outras práticas de segurança:**
- Não compartilhe tokens de acesso
- Remova dados sensíveis antes de compartilhar
- Use variáveis de ambiente para senhas
- Em servidores, use HTTPS

````python
# Dicas de segurança
print("🔒 SEGURANÇA NO JUPYTER")
print("=" * 35)
print("✅ Use 'jupyter notebook password'")
print("✅ Nunca compartilhe o token")
print("✅ Remova dados sensíveis antes de compartilhar")
print("✅ Use .env para variáveis secretas")
print("✅ Em servidores: configure HTTPS")
````
