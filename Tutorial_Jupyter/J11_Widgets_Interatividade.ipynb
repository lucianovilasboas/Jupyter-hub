## 📝 Introdução

Já imaginou ter **botões**, **sliders** e **menus** dentro do seu notebook? É exatamente isso que os **Widgets** fazem!

Eles transformam seu notebook de um documento estático em uma **aplicação interativa** — como um mini-programa dentro do Jupyter!

> 🎮 **Cenário**: Você é um mago da interface e vai criar controles mágicos que reagem aos comandos do usuário!

## 📚 Explicação Detalhada

### 1. Instalação e Import

Se não tiver instalado, execute no terminal:
```bash
pip install ipywidgets
```

Depois, importe no notebook:

````python
# Instalação e setup — execute uma vez
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

print("✅ ipywidgets pronto para uso!")
````

### 2. Slider — O Controle Deslizante

Use `widgets.IntSlider` ou `widgets.FloatSlider`:

````python
slider = widgets.IntSlider(
    value=5,
    min=0,
    max=10,
    step=1,
    description="Número:",
    style={'description_width': 'initial'}
)
display(slider)
print(f"📊 Arraste o slider! Valor inicial: {slider.value}")
````

### 3. Conectando Widgets a Funções

O verdadeiro poder está em ligar widgets a funções que respondem automaticamente:

Use `widgets.interact()` para conectar um widget a uma função:

````python
def saudacao(nome, repetir=1):
    for i in range(repetir):
        print(f"👋 Olá, {nome}!")

widgets.interact(saudacao,
    nome="Mundo",
    repetir=widgets.IntSlider(min=1, max=5, value=1, description="Repetições:")
);
````

### 4. Botões

Botões executam uma ação quando clicados:

````python
def quando_clicar(b):
    print("🎯 Botão clicado! \U0001f389")
    print("💡 Widgets são demais!")

botao = widgets.Button(
    description="🔘 Clique em mim!",
    button_style='success',  # 'primary', 'success', 'info', 'warning', 'danger'
    layout=widgets.Layout(width='200px', height='50px', font_size='16px')
)
botao.on_click(quando_clicar)
display(botao)
````

### 5. Dropdown (Menu Suspenso)

Para selecionar uma opção entre várias:

````python
def escolher_personagem(personagem):
    poderes = {
        "Ryu": "🐉 Hadouken!",
        "Chun-Li": "🦵 Spinning Bird Kick!",
        "Guile": "✈️ Sonic Boom!",
        "Blanka": "⚡ Electric Thunder!"
    }
    print(f"🎮 {personagem}: {poderes[personagem]}")

widgets.interact(
    escolher_personagem,
    personagem=["Ryu", "Chun-Li", "Guile", "Blanka"]
);
````

### 6. Gráfico Interativo

O exemplo mais poderoso — gráficos que se atualizam com sliders!

````python
def plotar_onda(frequencia=1, amplitude=1, cor="blue"):
    x = np.linspace(0, 4 * np.pi, 200)
    y = amplitude * np.sin(frequencia * x)
    
    plt.figure(figsize=(10, 4))
    plt.plot(x, y, color=cor, linewidth=2)
    plt.title(f"🌊 Onda Senoidal (freq={frequencia}, amp={amplitude})", fontsize=14)
    plt.xlabel("Tempo")
    plt.ylabel("Amplitude")
    plt.grid(True, alpha=0.3)
    plt.ylim(-3, 3)
    plt.show()

widgets.interact(
    plotar_onda,
    frequencia=widgets.FloatSlider(min=0.5, max=3.0, step=0.1, value=1.0),
    amplitude=widgets.FloatSlider(min=0.5, max=2.5, step=0.1, value=1.0),
    cor=widgets.Dropdown(options=['blue', 'red', 'green', 'purple', 'orange'], value='blue')
);
````

### 7. Múltiplos Widgets Juntos

Use `widgets.VBox` (vertical) ou `widgets.HBox` (horizontal) para agrupar:

````python
def calculadora(a, b, operacao):
    if operacao == "+":
        resultado = a + b
    elif operacao == "-":
        resultado = a - b
    elif operacao == "×":
        resultado = a * b
    else:
        resultado = a / b if b != 0 else "Erro: divisão por zero"
    print(f"🧮 {a} {operacao} {b} = {resultado}")

widgets.interact(
    calculadora,
    a=widgets.FloatSlider(min=0, max=100, step=1, value=10),
    b=widgets.FloatSlider(min=0, max=100, step=1, value=5),
    operacao=['+', '-', '×', '÷']
);
````

## 💻 Exemplos Práticos

````python
# Exemplo 1: Mini-jogo de adivinhação
import random

numero_secreto = random.randint(1, 10)
tentativas = 0

def adivinhar(palpite):
    global tentativas
    tentativas += 1
    if palpite == numero_secreto:
        print(f"🎉 ACERTOU! O número era {numero_secreto}!")
        print(f"🏆 Tentativas: {tentativas}")
    elif palpite < numero_secreto:
        print("⬆️ Maior! Tente de novo.")
    else:
        print("⬇️ Menor! Tente de novo.")

widgets.interact(adivinhar, palpite=widgets.IntSlider(min=1, max=10, step=1));
````

````python
# Exemplo 2: Sorteador de Times
def sortear_time():
    times = ["Flamengo", "Palmeiras", "Santos", "São Paulo", "Corinthians", "Grêmio"]
    sorteado = random.choice(times)
    print(f"🎲 Time sorteado: {sorteado} 🏆")

botao_sorteio = widgets.Button(description="🎯 Sortear Time", button_style='primary')
botao_sorteio.on_click(lambda b: sortear_time())
display(botao_sorteio)
````

````python
# Exemplo 3: Controlador de pizza 🍕
def pizza(sabor, borda, tamanho):
    preco_base = {"Pequena": 25, "Média": 35, "Grande": 45}
    adicional_borda = 5 if borda == "Sim" else 0
    total = preco_base[tamanho] + adicional_borda
    
    print("🍕 PEDIDO RECEBIDO!")
    print("=" * 30)
    print(f"Sabor: {sabor}")
    print(f"Borda recheada: {borda}")
    print(f"Tamanho: {tamanho}")
    print(f"💰 Total: R$ {total:.2f}")
    print("=" * 30)
    print("✅ Pedido confirmado! 🎉")

widgets.interact(
    pizza,
    sabor=["Mussarela", "Calabresa", "Portuguesa", "Frango", "Marguerita"],
    borda=["Não", "Sim"],
    tamanho=["Pequena", "Média", "Grande"]
);
````

````python
# 🟡 Exercício 6: Calculadora IMC
def imc(peso, altura):
    valor = peso / (altura / 100) ** 2
    print(f"📊 Seu IMC: {valor:.1f}")
    if valor < 18.5:
        print("⚠️ Abaixo do peso")
    elif valor < 25:
        print("✅ Peso saudável!")
    elif valor < 30:
        print("⚠️ Sobrepeso")
    else:
        print("⚠️ Obesidade")

widgets.interact(imc,
    peso=widgets.IntSlider(min=30, max=200, value=70, description="Peso (kg):"),
    altura=widgets.IntSlider(min=100, max=220, value=170, description="Altura (cm):")
);
````

````python
# 🔴 Exercício 13: Gerenciador de tarefas
tarefas = []

def adicionar_tarefa(b):
    tarefas.append(f"📌 Tarefa {len(tarefas) + 1}")
    print("📋 TAREFAS:")
    for t in tarefas:
        print(f"   [ ] {t}")

btn = widgets.Button(description="➕ Adicionar Tarefa", button_style='success')
btn.on_click(adicionar_tarefa)
display(btn)
````

## 📌 Resumo

- **ipywidgets** traz interatividade para o notebook

- Widgets comuns: `IntSlider`, `FloatSlider`, `Button`, `Dropdown`
- Botões usam `on_click()` para reagir a cliques
- Gráficos interativos combinam widgets + matplotlib
- `VBox`/`HBox` agrupam widgets visualmente

## ✅ Boas Práticas

1. **Instale ipywidgets** com `pip install ipywidgets` ou via Anaconda
2. **Importe no começo** junto com os outros imports
3. **Teste widgets separadamente** antes de combiná-los
4. **Use descrições claras** no parâmetro `description`
5. **Evite muitos widgets** na mesma célula — pode ficar lento

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| `ipywidgets` não instalado | `pip install ipywidgets` |
| Widget não aparece | Use `display(widget)` para mostrá-lo |
| Botão sem ação | Configure `on_click()` com uma função |
| Slider sem resposta | Use `interact()` ou `observe()` |
| Widgets sobrepostos | Use `VBox` ou `HBox` para organizar |
| Muitos widgets lentos | Reduza o número ou use `continuous_update=False` |
