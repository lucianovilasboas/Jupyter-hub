## 📋 Apresentação

Bem-vindo ao **Guia Completo de Jupyter Notebook**! Este material foi criado especialmente para te ensinar a usar a ferramenta que será seu ambiente principal de aprendizado durante todo o curso de Python.

O Jupyter Notebook é um ambiente interativo que combina código, texto explicativo, imagens, gráficos e muito mais em um único documento. É a ferramenta mais usada por cientistas de dados, pesquisadores e educadores no mundo todo.

E o melhor: **você já está usando um Jupyter Notebook agora mesmo!** 👏

## 🎯 Objetivos Gerais

- Dominar o Jupyter Notebook do básico ao avançado
- Aprender a combinar código Python com texto formatado
- Criar análises de dados interativas e visuais
- Desenvolver projetos completos dentro do notebook
- Compartilhar seus trabalhos com o mundo

## 🚀 Como Utilizar Este Material

1. **Comece pela J01** se você nunca usou Jupyter antes
2. **Siga a ordem numérica** — cada nível prepara para o próximo
3. **Execute TODOS os exemplos** — não apenas leia, pratique!
4. **Modifique os exemplos** — quebre, erre, conserte, aprenda
5. **Faça os exercícios** — só se aprende programando

> 💡 **Dica de ouro**: Jupyter é como um caderno mágico — você escreve, ele executa. Use isso a seu favor!
