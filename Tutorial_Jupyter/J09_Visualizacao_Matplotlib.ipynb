## 📚 Explicação Detalhada

### 1. Setup — O Comando Mágico `%matplotlib inline`

Para gráficos aparecerem DENTRO do notebook (e não em janela separada), usamos um **comando mágico**:

```python
# Setup — execute esta célula primeiro!
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Dados que vamos usar
meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun"]
vendas = [120, 150, 135, 170, 190, 210]
lucros = [30, 40, 35, 45, 50, 55]

print("✅ Setup completo! Pronto para plotar!")
```

### 2. Gráfico de Linhas — Seu Primeiro Plot

O `plt.plot()` cria um gráfico de linhas. Simples e poderoso!

```python
plt.figure(figsize=(10, 5))  # Tamanho da figura
plt.plot(meses, vendas, marker='o', linewidth=2, color='#2ecc71')
plt.title("📈 Vendas no Semestre", fontsize=16, fontweight='bold')
plt.xlabel("Mês")
plt.ylabel("Vendas (unidades)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### 3. Múltiplas Linhas

Compare diferentes séries no mesmo gráfico:

```python
plt.figure(figsize=(10, 5))
plt.plot(meses, vendas, marker='o', label='Vendas', color='#3498db', linewidth=2)
plt.plot(meses, lucros, marker='s', label='Lucros', color='#e74c3c', linewidth=2)
plt.title("📊 Vendas vs Lucros", fontsize=16, fontweight='bold')
plt.xlabel("Mês")
plt.ylabel("Valor (R$)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### 4. Gráfico de Barras

Perfeito para comparar categorias:

```python
plt.figure(figsize=(10, 5))
cores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']
bars = plt.bar(meses, vendas, color=cores, edgecolor='black', linewidth=1.2)
plt.title("📊 Vendas por Mês", fontsize=16, fontweight='bold')
plt.xlabel("Mês")
plt.ylabel("Vendas")

# Adicionar valores nas barras
for bar, valor in zip(bars, vendas):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
             str(valor), ha='center', fontweight='bold')

plt.show()
```

### 5. Gráfico de Pizza

Para mostrar proporções e porcentagens:

```python
# Dados divertidos: sabores de pizza 🍕
sabores = ['Mussarela', 'Calabresa', 'Portuguesa', 'Frango', 'Marguerita']
porcentagens = [35, 25, 18, 12, 10]
cores_pizza = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']
destaques = [0.05, 0, 0, 0, 0]  # Destacar o primeiro

plt.figure(figsize=(8, 8))
plt.pie(porcentagens, labels=sabores, autopct='%1.1f%%',
        colors=cores_pizza, explode=destaques, startangle=90,
        shadow=True, textprops={'fontsize': 12})
plt.title("🍕 Preferência de Sabores de Pizza", fontsize=16, fontweight='bold')
plt.show()
```

### 6. Histograma

Mostra a distribuição dos dados:

```python
# Simulando notas de uma turma
np.random.seed(42)
notas = np.random.normal(7.0, 1.5, 100)  # 100 notas, média 7, desvio 1.5
notas = np.clip(notas, 0, 10)  # Limitar entre 0 e 10

plt.figure(figsize=(10, 5))
plt.hist(notas, bins=10, color='#3498db', edgecolor='white', alpha=0.8)
plt.title("📊 Distribuição das Notas", fontsize=16, fontweight='bold')
plt.xlabel("Notas")
plt.ylabel("Frequência")
plt.axvline(notas.mean(), color='red', linestyle='--', linewidth=2, label=f'Média: {notas.mean():.2f}')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"📈 Média: {notas.mean():.2f}")
print(f"📊 Desvio padrão: {notas.std():.2f}")
print(f"🏆 Nota máxima: {notas.max():.2f}")
print(f"😢 Nota mínima: {notas.min():.2f}")
```

### 7. Subplots — Vários Gráficos em Um

Organize múltiplos gráficos lado a lado:

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Gráfico 1: Linha
axes[0].plot(meses, vendas, marker='o', color='#2ecc71', linewidth=2)
axes[0].set_title("Vendas")
axes[0].grid(True, alpha=0.3)

# Gráfico 2: Barra
axes[1].bar(meses, vendas, color='#3498db')
axes[1].set_title("Vendas (Barras)")

# Gráfico 3: Pizza
axes[2].pie(vendas, labels=meses, autopct='%1.0f%%', startangle=90)
axes[2].set_title("Proporção")

plt.suptitle("📊 Três Visões dos Mesmos Dados", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Gráfico dos Lutadores de Street Fighter
import pandas as pd

dados_lutadores = {
    "Lutador": ["Ryu", "Ken", "Chun-Li", "Guile", "Zangief", "Blanka"],
    "Altura": [175, 176, 169, 182, 214, 192],
    "Peso": [85, 83, 58, 99, 181, 98]
}
sf = pd.DataFrame(dados_lutadores)

plt.figure(figsize=(12, 5))
bars = plt.bar(sf["Lutador"], sf["Altura"], color=['#e74c3c', '#3498db', '#2ecc71', '#f39c12', '#9b59b6', '#1abc9c'])
plt.title("📏 Altura dos Lutadores", fontsize=16, fontweight='bold')
plt.ylabel("Altura (cm)")
plt.axhline(y=sf["Altura"].mean(), color='red', linestyle='--', label=f'Média: {sf["Altura"].mean():.0f}cm')
plt.legend()

for bar, altura in zip(bars, sf["Altura"]):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2,
             f'{altura}cm', ha='center', fontweight='bold')
plt.show()
```

```python
# Exemplo 2: Evolução das notas dos filmes Ghibli
filmes = ["Castelo no Céu", "Totoro", "Kiki", "Princesa Mononoke", "Chihiro", "Castelo Animado", "Ponyo"]
anos = [1986, 1988, 1989, 1997, 2001, 2004, 2008]
notas = [8.0, 8.2, 7.9, 8.4, 8.6, 8.2, 7.7]

plt.figure(figsize=(12, 5))
plt.plot(anos, notas, marker='o', linewidth=3, color='#e74c3c', markersize=10)
plt.title("⭐ Evolução das Notas dos Filmes Ghibli", fontsize=16, fontweight='bold')
plt.xlabel("Ano")
plt.ylabel("Nota")
plt.ylim(7, 9)
plt.grid(True, alpha=0.3)

for ano, nota, filme in zip(anos, notas, filmes):
    plt.annotate(filme, (ano, nota), textcoords="offset points",
                 xytext=(0, 15), ha='center', fontsize=8, rotation=45)
plt.show()
```

```python
# Exemplo 3: Sorvete mais popular (barras horizontais)
sabores = ['Chocolate 🍫', 'Baunilha 🍦', 'Morango 🍓', 'Creme 🥛', 'Flocos 🎉']
votos = [42, 31, 15, 8, 4]

plt.figure(figsize=(10, 5))
bars = plt.barh(sabores, votos, color=['#8B4513', '#F5DEB3', '#FF6B6B', '#FFFACD', '#FF69B4'])
plt.title("🍦 Enquete: Melhor Sorvete", fontsize=16, fontweight='bold')
plt.xlabel("Votos")

for bar, voto in zip(bars, votos):
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2,
             f'{voto} votos', va='center', fontweight='bold')
plt.show()
```

## 💡 Curiosidades

- Matplotlib foi criado por **John D. Hunter** em 2003, inspirado no MATLAB.
- O nome "Matplotlib" = **Mat**h + **Plot** + **Lib**rary.
- O comando `%matplotlib inline` é um dos **comandos mágicos** mais famosos do Jupyter.
- Você pode salvar gráficos diretamente com `plt.savefig('grafico.png')`.

## ✅ Boas Práticas

1. **Sempre use `%matplotlib inline`** no começo
2. **Defina o tamanho** com `figsize=(L, A)` para gráficos legíveis
3. **Sempre adicione título e labels** — gráfico sem contexto é inútil
4. **Use `plt.tight_layout()`** em subplots para não sobrepor
5. **Cores consistentes** — mantenha um padrão visual no notebook

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Esquecer `%matplotlib inline` | Gráfico aparece em janela separada — sempre use no início |
| `plt.show()` esquecido | Em alguns casos o gráfico não aparece sem ele |
| Subplots sobrepostos | Use `plt.tight_layout()` |
| Cores conflitantes | Use paletas como `plt.cm.Set2` ou `plt.cm.Pastel1` |
| Dados com escala diferente | Um gráfico pode distorcer a comparação |
| Esquecer labels | Gráfico sem título não comunica nada |
