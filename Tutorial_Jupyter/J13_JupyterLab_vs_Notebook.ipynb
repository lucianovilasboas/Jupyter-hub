## 📝 Introdução

Você está usando o **Jupyter Notebook** clássico — a versão original. Mas existe um irmão mais novo e mais poderoso: o **JupyterLab**! 🦾

Enquanto o Notebook clássico é como um **caderno simples** com uma página por vez, o JupyterLab é como uma **mesa de trabalho** com várias ferramentas ao mesmo tempo!

> 💡 **Analogia**: Notebook = um bloco de papel. JupyterLab = uma mesa com bloco, calculadora, livros e um monitor extra!

## 📚 Explicação Detalhada

### 1. O que é o JupyterLab?

JupyterLab é a **próxima geração** da interface do Jupyter. Foi lançado em 2018 como uma evolução do Notebook clássico.

**Características principais:**
- 🖥️ **Interface com abas** — múltiplos notebooks abertos
- 📐 **Painéis lado a lado** — arraste e organize
- 🖊️ **Editor de texto/ código** integrado
- 💻 **Terminal** embutido
- 📁 **File browser** nativo
- 🔌 **Extensões** mais poderosas
- 🎨 **Temas** dark/light

**Mas atenção:** Tudo que você aprendeu até aqui funciona IGUAL no JupyterLab! Os atalhos, células, markdown, kernel — tudo igual.

### 2. Tabela Comparativa

| Característica | Jupyter Notebook | JupyterLab |
|---------------|:----------------:|:----------:|
| Interface | Aba única | Abas + painéis |
| Múltiplos notebooks | Abas no navegador | Abas + lado a lado |
| Terminal | Não | Sim, integrado |
| Editor de código | Não | Sim |
| File browser | Dashboard separado | Integrado na interface |
| Temas | Limitado | Dark/Light nativo |
| Extensões | Nbextensions | Sistema mais flexível |
| Performance | Leve | Mais recursos, mais pesado |
| Ideal para | Iniciantes, tarefas rápidas | Projetos complexos |

### 3. Como Instalar e Abrir o JupyterLab

Instalação:
```bash
pip install jupyterlab
```

Para abrir:
```bash
jupyter lab
```

Isso abre o JupyterLab no seu navegador. A URL é a mesma: `http://localhost:8888` (ou outra porta).

Você pode ter **ambos** instalados e usar qual preferir!

````python
# Verifique se o JupyterLab está instalado
try:
    import jupyterlab
    print(f"✅ JupyterLab versão {jupyterlab.__version__} instalado!")
    print("🚀 Para abrir: jupyter lab")
except ImportError:
    print("❌ JupyterLab não instalado.")
    print("📦 Instale com: pip install jupyterlab")
````

### 4. O Poder dos Painéis

No JupyterLab, você pode:
- 📊 Arrastar abas para criar **painéis lado a lado**
- 📝 Editar código em um painel enquanto vê os resultados em outro
- 📁 Arrastar arquivos do file browser para o notebook
- 💻 Ter um terminal aberto em um painel separado

**Exemplo de uso:**
- Painel esquerdo: notebook com análise
- Painel direito superior: terminal rodando comandos
- Painel direito inferior: editor de código para script auxiliar

````python
# Simulação visual dos painéis do JupyterLab
print("🖥️  JUPYTERLAB — VISÃO DE PAINÉIS")
print("=" * 45)
print("┌──────────────────┬─────────────────────┐")
print("│                  │                     │")
print("│   📓 Notebook    │   💻 Terminal       │")
print("│   (células de    │   (comandos bash)   │")
print("│    código)       │                     │")
print("│                  ├─────────────────────┤")
print("│                  │   📝 Editor         │")
print("│                  │   (script.py)       │")
print("└──────────────────┴─────────────────────┘")
print("")
print("💡 Tudo visível ao mesmo tempo! ")
````

### 5. Terminal Integrado

No JupyterLab, você tem um terminal diretamente na interface:

- **File > New > Terminal**
- Funciona como o terminal normal do sistema
- Pode executar comandos, instalar pacotes, navegar
- Tudo sem sair do navegador!

**Exemplos de uso do terminal:**
- `pip install pandas`
- `git status`
- `python meu_script.py`
- `ls -la`

````python
# Simulação de comandos no terminal
print("💻 SIMULAÇÃO DE TERMINAL")
print("=" * 35)
print("$ pwd")
import os
print(os.getcwd())
print("$ echo 'Olá do terminal JupyterLab!'")
print("Olá do terminal JupyterLab!")
````

### 6. Editor de Código

O JupyterLab inclui um editor de código que suporta:
- Syntax highlighting para várias linguagens
- Múltiplos cursores
- Indentação automática
- Plugins de Linter

Você pode abrir qualquer arquivo `.py`, `.md`, `.json`, `.csv` diretamente no editor!

````python
# Criando um arquivo que você abriria no editor
with open("_exemplo_editor.py", "w") as f:
    f.write('''
def calcular_media(notas):
    """Calcula a média de uma lista de notas"""
    return sum(notas) / len(notas) if notas else 0

turma = [8.5, 7.0, 9.2, 6.5, 8.0]
print(f"Média da turma: {calcular_media(turma):.2f}")
''')
print("✅ Arquivo _exemplo_editor.py criado!")
print("💡 No JupyterLab, abra-o no editor integrado!")
````

### 7. Quando Usar Cada Um

**Use Jupyter Notebook quando:**
- ✅ Você é iniciante e quer simplicidade
- ✅ Precisa de algo leve e rápido
- ✅ Está fazendo tarefas pequenas
- ✅ Seu computador tem pouca memória RAM

**Use JupyterLab quando:**
- ✅ Trabalha em projetos grandes
- ✅ Precisa de múltiplos arquivos abertos
- ✅ Quer terminal + editor + notebook ao mesmo tempo
- ✅ Gosta de temas escuros 🖤
- ✅ Quer extensões mais avançadas

> 💡 **Dica**: Você pode ter os dois instalados! Use Notebook para tarefas rápidas e JupyterLab para projetos completos.

````python
# Decida qual usar!
import random
resposta = random.choice(["Jupyter Notebook", "JupyterLab"])
print("🔮 A SORTE DECIDIU:")
print(f"   HOJE VOCÊ USA: {resposta.upper()}!")
print("💡 Experimente os dois e escolha seu favorito!")
````

````python
# 🟡 Exercício 10: File browser
import os
print("📁 File Browser (simulação):")
for pasta, subpastas, arquivos in os.walk('.'):
    nivel = pasta.count(os.sep)
    if nivel > 2:
        break
    print(f"{'    ' * nivel}📁 {os.path.basename(pasta) or '.'}/")
    for arquivo in arquivos[:5]:
        print(f"{'    ' * (nivel+1)}📄 {arquivo}")
````

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Achar que Lab substitui Notebook | São complementares, não concorrentes |
| Fechar painel sem querer | Use View > Reset Layout para recuperar |
| Instalar Lab e esquecer Notebook | Ambos podem coexistir |
| Esperar que seja igual ao Notebook | Lab tem mais recursos, mas é diferente |
| Terminal fechado sem querer | File > New > Terminal para abrir de novo |
