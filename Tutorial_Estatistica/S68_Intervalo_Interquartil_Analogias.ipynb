## Zona de Conforto

O **IQR** e a "zona de conforto" dos dados:
os 50% centrais, ignorando os extremos.

Pense numa sala de aula:
- 25% tiram notas muito baixas (Q1 para baixo)
- 25% tiram notas muito altas (Q3 para cima)
- **50% centrais**: a maioria da turma — o IQR

O IQR e como olhar para a "classe media" dos dados:
ignora os mais pobres e mais ricos,
focando no que e tipico.

E muito **robusto**: o CEO ganhar R$ 100.000
nao afeta o IQR dos salarios.


## 📊 Traduzindo

| Zona de Conforto | Estatistica |
|-|-|
| Zona de conforto | IQR — 50% centrais |
| Classe media | Dados entre Q1 e Q3 |
| Muito pobre (Q1 para baixo) | 25% inferiores |
| Muito rico (Q3 para cima) | 25% superiores |
| CEO nao afeta | IQR e robusto a outliers |


## 📏 IQR vs Desvio Padrao

**IQR**: robusto a outliers. Use quando ha extremos.
**Desvio padrao**: usa todos os dados. Melhor para dados simetricos.

O IQR tambem define **outliers**:
valores abaixo de Q1 - 1.5xIQR ou acima de Q3 + 1.5xIQR.


```python
# Zona de conforto
dados = sorted([1000, 1200, 1300, 1500, 1600, 1700, 1800, 2000, 2200, 100000])
n = len(dados)
q1 = dados[n//4]
q3 = dados[3*n//4]
iqr = q3 - q1
print(f"Dados ordenados: {dados}")
print(f"Q1: {q1} | Q3: {q3}")
print(f"IQR (zona de conforto): {iqr}")
print(f"Outliers abaixo de: {q1 - 1.5*iqr}")
print(f"Outliers acima de: {q3 + 1.5*iqr}")
```

## 📌 Resumo

| Zona de Conforto | Estatistica |
|-|-|
| Zona de conforto | IQR |
| Classe media | Q1 a Q3 |
| Ignora extremos | Robusto a outliers |
| CEO nao afeta | Resistente |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
