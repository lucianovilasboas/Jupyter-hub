## Album de Fotografias

Voce voltou de uma viagem incrivel. Tirou 500 fotos.

Agora quer mostrar para os amigos.
Voce vai mostrar **todas as 500 fotos**? Claro que nao!

Voce seleciona:
- As **melhores fotos** (as que contam a historia)
- Um **resumo** do que aconteceu em cada dia
- Talvez um **album** organizado por ordem cronologica

**Descrever dados e exatamente isso!**
Voce tem muitos numeros brutos (as 500 fotos) e precisa
selecionar, resumir e organizar para contar a historia que eles carregam.


## 📊 Traduzindo

| Album de Fotografias | Estatistica |
|-|-|
| 500 fotos da viagem | Conjunto de dados brutos |
| Selecionar as melhores fotos | Selecionar variaveis relevantes |
| Organizar por dia | Agrupar dados por categoria |
| Album com legendas | Tabela com rotulos e metadados |
| Mostrar so as melhores | Estatistica descritiva — resumir sem perder o essencial |


## 📏 Nem Todo Dado Precisa Ser Mostrado

Quando voce mostra as fotos, nao mostra as 500.
Mostra as 30 melhores — aquelas que representam bem a viagem.

Com dados e a mesma coisa:
- Nao precisa mostrar todos os 10.000 registros
- Basta mostrar **medidas resumo**: media, mediana, minimo, maximo
- E **graficos** que revelam padroes


## 📏 Organizar para Entender

Fotos fora de ordem cronologica contam uma historia confusa.
O mesmo vale para dados: precisam ser organizados.

**Organizar dados significa:**
- Ordenar valores (do menor para o maior)
- Agrupar por categorias (por estado, por mes)
- Criar tabelas de frequencia
- Visualizar com graficos


```python
# As 500 fotos da viagem = dados brutos
import random
random.seed(42)
# Simulando 500 registros de gastos na viagem
gastos = [random.gauss(100, 30) for _ in range(500)]

print(f"Total de registros (fotos): {len(gastos)}")
print()

# Selecionando as "melhores fotos" = estatistica descritiva
print("--- Album Resumo da Viagem ---")
print(f"Gasto medio por dia: R$ {sum(gastos)/len(gastos):.2f}")
print(f"Menor gasto: R$ {min(gastos):.2f}")
print(f"Maior gasto: R$ {max(gastos):.2f}")
print(f"Total da viagem: R$ {sum(gastos):.2f}")
```

## 📌 Resumo

| Album de Fotografias | Estatistica |
|-|-|
| 500 fotos | Dados brutos |
| Selecionar as melhores | Selecionar variaveis |
| Album organizado | Dados organizados em tabelas |
| Legendas | Rotulos e metadados |
| 30 fotos contam a historia | Estatistica descritiva |

A analogia do album de fotografias mostra que:

> A estatistica esta em todo lugar — so precisamos do olhar certo para enxerga-la.
