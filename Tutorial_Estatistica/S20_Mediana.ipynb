## 📝 Introdução

A **mediana** é o valor que divide um conjunto de dados ordenados em duas metades iguais: 50% dos dados estão abaixo dela e 50% estão acima.

Enquanto a média é influenciada por todos os valores (inclusive extremos), a mediana depende apenas da **ordem** dos dados. Isso a torna uma medida **robusta** — resistente a outliers.

> 🧠 **Para refletir:** Se em uma reunião de 10 pessoas entra o Elon Musk, a "renda média" do grupo dispara. Mas a "renda mediana" praticamente não muda. Qual delas melhor descreve o grupo?

A mediana é especialmente útil para:
- Dados com outliers
- Distribuições assimétricas
- Variáveis ordinais
- Dados de renda, preços de imóveis e outras variáveis socioeconômicas

## 📚 Explicação Detalhada

### 1. Cálculo da Mediana

**Passo 1:** Ordene os dados do menor para o maior.

**Passo 2:** Encontre o valor do meio.

**Caso 1 — n ímpar:** A mediana é o valor na posição $(n + 1)/2$.

Exemplo: `{3, 5, 7, 9, 11}` → n = 5 → posição (5+1)/2 = 3 → mediana = 7

**Caso 2 — n par:** A mediana é a média dos dois valores centrais, nas posições $n/2$ e $n/2 + 1$.

Exemplo: `{3, 5, 7, 9}` → n = 4 → posições 2 e 3 → mediana = (5 + 7)/2 = 6

### 2. Mediana vs Média

| Característica | Média | Mediana |
|---------------|-------|--------|
| Sensibilidade a outliers | Alta | Baixa (robusta) |
| Uso de todos os dados | Sim | Não (só a ordem) |
| Propriedades matemáticas | Muitas | Poucas |
| Distribuição assimétrica | Distorcida | Representativa |
| Eficiência estatística | Alta | Menor |

### 3. Mediana para Dados Agrupados

Para dados agrupados em classes, a mediana é encontrada pela fórmula:

$$\text{Mediana} = L + \frac{\frac{n}{2} - F_{ac}}{f} \times h$$

Onde:
- $L$ = limite inferior da classe mediana
- $n$ = total de observações
- $F_{ac}$ = frequência acumulada anterior à classe mediana
- $f$ = frequência da classe mediana
- $h$ = amplitude da classe

## 💻 Exemplos Práticos

```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
# Exemplo 1: Mediana com n ímpar
dados_impar = np.array([10, 15, 20, 25, 30])
mediana_impar = np.median(dados_impar)

print("n ímpar (n=5):")
print(f"Dados ordenados: {np.sort(dados_impar)}")
print(f"Mediana: {mediana_impar}")
```

```python
# Exemplo 2: Mediana com n par
dados_par = np.array([10, 15, 20, 25, 30, 35])
mediana_par = np.median(dados_par)

print("n par (n=6):")
print(f"Dados ordenados: {np.sort(dados_par)}")
print(f"Mediana: {mediana_par}")
print("(média dos dois valores centrais: 20 + 25 / 2 = 22.5)")
```

```python
# Exemplo 3: Robustez da mediana a outliers
# Comparando média e mediana na presença de outliers
sem_outlier = np.array([100, 102, 98, 105, 101, 99, 103])
com_outlier = np.array([100, 102, 98, 105, 101, 99, 103, 500])

print("SEM outlier:")
print(f"  Mean:   {np.mean(sem_outlier):.2f}")
print(f"  Median: {np.median(sem_outlier):.2f}")
print()
print("COM outlier (500):")
print(f"  Mean:   {np.mean(com_outlier):.2f}")
print(f"  Median: {np.median(com_outlier):.2f}")
print()
print("A média mudou drasticamente, mas a mediana permaneceu quase igual.")
```

```python
# Exemplo 4: Visualização — Média vs Mediana com outliers
# Gerando dados de renda com assimetria
renda = np.random.exponential(scale=5000, size=1000)
media_r = np.mean(renda)
mediana_r = np.median(renda)

plt.figure(figsize=(10, 5))
plt.hist(renda, bins=40, color='lightcoral', edgecolor='black', alpha=0.7)
plt.axvline(media_r, color='red', linestyle='--', linewidth=2, label=f'Mean = R$ {media_r:.0f}')
plt.axvline(mediana_r, color='blue', linestyle='-.', linewidth=2, label=f'Median = R$ {mediana_r:.0f}')
plt.xlabel('Renda (R$)')
plt.ylabel('Frequência')
plt.title('Distribuição de Renda — Média vs Mediana')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
print(f"A mediana (R$ {mediana_r:.0f}) representa melhor a renda típica do que a média (R$ {media_r:.0f}).")
```

```python
# Exemplo 5: Efeito de outliers progressivos na mediana
dados_base = np.random.normal(50, 10, 100)
medias = []
medianas = []
outliers = [0, 10, 20, 30, 50, 80, 100, 150, 200, 300]

for o in outliers:
    dados = np.append(dados_base, o)
    medias.append(np.mean(dados))
    medianas.append(np.median(dados))

plt.figure(figsize=(10, 5))
plt.plot(outliers, medias, 'o-', color='red', label='Mean', linewidth=2)
plt.plot(outliers, medianas, 's-', color='blue', label='Median', linewidth=2)
plt.axhline(y=np.mean(dados_base), color='red', linestyle=':', alpha=0.5)
plt.axhline(y=np.median(dados_base), color='blue', linestyle=':', alpha=0.5)
plt.xlabel('Valor do Outlier')
plt.ylabel('Medida de Tendência Central')
plt.title('Robustez: Mediana vs Média na Presença de Outliers')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Exemplo 6: Mediana em dados agrupados (simulação)
# Simulando distribuição de idade em faixas etárias
idades = np.random.randint(0, 80, size=500)
faixas = np.array([0, 20, 40, 60, 80])
freq, _ = np.histogram(idades, bins=faixas)
freq_ac = np.cumsum(freq)

n = len(idades)
classe_mediana = np.where(freq_ac >= n/2)[0][0]

L = faixas[classe_mediana]
h = faixas[classe_mediana + 1] - L
f = freq[classe_mediana]
F_ac = freq_ac[classe_mediana - 1] if classe_mediana > 0 else 0

mediana_agrupada = L + ((n/2 - F_ac) / f) * h
mediana_real = np.median(idades)

print("Distribuição de Idades (dados agrupados):")
for i in range(len(faixas)-1):
    print(f"  {faixas[i]:.0f}-{faixas[i+1]:.0f}: {freq[i]} pessoas")
print(f"\nMediana calculada (dados agrupados): {mediana_agrupada:.1f} anos")
print(f"Mediana real (dados brutos):         {mediana_real:.1f} anos")
```

## 📝 Exercícios Resolvidos

```python
# Exercício Resolvido 1
# Calcular mediana das notas de uma turma
notas = np.array([5.5, 7.0, 8.5, 6.0, 9.0, 7.5, 4.5, 8.0, 6.5, 7.0])
notas_ordenadas = np.sort(notas)
med_notas = np.median(notas)
media_notas = np.mean(notas)

print(f"Notas ordenadas: {notas_ordenadas}")
print(f"Média:  {media_notas:.2f}")
print(f"Mediana: {med_notas:.2f}")
print(f"n = {len(notas)} (par) → mediana = média dos {len(notas)//2}º e {len(notas)//2 + 1}º valores")
```

```python
# Exercício Resolvido 2
# Comparação Média vs Mediana em diferentes distribuições
distribuicoes = {
    'Normal (simétrica)': np.random.normal(100, 15, 1000),
    'Exponencial (assim. direita)': np.random.exponential(50, 1000),
    'Uniforme': np.random.uniform(0, 100, 1000)
}

print(f"{'Distribuição':<30} {'Média':<10} {'Mediana':<10} {'Diferença':<10}")
print("-"*60)
for nome, dados in distribuicoes.items():
    m = np.mean(dados)
    med = np.median(dados)
    print(f"{nome:<30} {m:<10.2f} {med:<10.2f} {abs(m-med):<10.2f}")
```

```python
# Exercício Resolvido 3
# Mediana em dados ordinais — nível de satisfação
satisfacao = np.array([1, 2, 2, 3, 3, 3, 4, 4, 5])
# 1 = Muito insatisfeito, 2 = Insatisfeito, 3 = Neutro, 4 = Satisfeito, 5 = Muito satisfeito

mediana_sat = np.median(satisfacao)
rotulos = {1: 'Muito insatisfeito', 2: 'Insatisfeito', 3: 'Neutro', 4: 'Satisfeito', 5: 'Muito satisfeito'}

print(f"Dados de satisfação: {satisfacao}")
print(f"Mediana: {mediana_sat:.0f} → {rotulos[int(mediana_sat)]}")
print("A mediana é a medida ideal para dados ordinais.")
```

## 📝 Resumo do Capítulo

- A **mediana** é o valor que separa os 50% menores dos 50% maiores
- Para **n ímpar**: mediana = valor central após ordenar
- Para **n par**: mediana = média dos dois valores centrais
- A mediana é **robusta** a outliers (diferente da média)
- Em distribuições assimétricas, a mediana representa melhor o centro
- Ideal para dados ordinais, renda, preços de imóveis
- Pode ser calculada para dados agrupados com interpolação

## 🧐 Curiosidades

- O termo "mediana" vem do latim **medianus** (que está no meio).
- Em 1883, **Francis Galton** descobriu que a mediana é mais estável que a média quando os dados têm erros de medição.
- O censo dos EUA usa a **renda mediana** (não a média) para relatar a renda das famílias.
- Nos Jogos Olímpicos, a mediana é usada para calcular a pontuação final em esportes como ginástica e saltos ornamentais, descartando notas extremas.

## ✅ Boas Práticas

- Use a **mediana** quando os dados tiverem outliers ou assimetria
- Sempre reporte **média e mediana** juntos para uma visão completa
- Para dados **ordinais**, a mediana é mais apropriada que a média
- A mediana não é adequada para cálculos algébricos posteriores
- Em box plots, a linha central é a mediana — use-a para comparar grupos

## ⚠️ Erros Comuns

| Erro | Consequência | Correção |
|------|-------------|----------|
| Usar mediana como única medida | Perde informação sobre magnitude | Acompanhe com média |
| Calcular mediana sem ordenar | Resultado errado | Sempre ordene primeiro |
| Ignorar que n par dá intervalo | Ambiguidade | Use np.median (média dos centrais) |
| Usar mediana para inferência paramétrica | Menor poder estatístico | Use média para dados normais |
| Confundir mediana com média | Escolha errada da medida | Lembre-se: mediana = valor do meio |
