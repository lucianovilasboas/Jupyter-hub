## Receita de Bolo

Pense em um bolo. Um bolo delicioso.

A **receita** diz: 3 ovos, 2 xicaras de farinha, 1 de acucar, fermento, leite.

Agora imagine:
- Voce usa **ovos estragados** 🥚🤢
- Ou **farinha vencida**
- Ou **acucar pela metade**

O bolo vai ficar uma porcaria, certo?

**Coletar dados e como escolher os ingredientes.**
Se os ingredientes sao ruins, o resultado final tambem sera.
Nao adianta ser um otimo confeiteiro (ou estatistico) se a materia-prima e ruim.


## 📊 Traduzindo

| Receita de Bolo | Estatistica |
|-|-|
| Ovos frescos | Dados precisos e confiaveis |
| Farinha na medida certa | Amostra representativa |
| Acucar pela metade | Amostra incompleta (vies) |
| Ovo estragado | Dado incorreto ou erro de medicao |
| Batedeira | Ferramenta de coleta (questionario, sensor) |
| Provar a massa antes de assar | Validacao dos dados antes da analise |


## 📏 Ingredientes de Qualidade = Dados de Qualidade

Na confeitaria, ingredientes de qualidade sao o segredo.
Na estatistica, a **qualidade dos dados** e o segredo.

Dados de ma qualidade produzem conclusoes erradas.
E o famoso **"garbage in, garbage out"** (lixo entra, lixo sai).

Exemplos de dados de ma qualidade:
- Pesquisa com perguntas tendenciosas (so uma opcao parece certa)
- Sensor descalibrado (medicoes sistematicamente erradas)
- Amostra viesada (entrevistar so pessoas na saida do metro)


## 📏 A Medida Certa = Amostra Representativa

Se a receita pede 2 xicaras de farinha, colocar 1 ou 3 estraga o bolo.

Na coleta de dados, o **tamanho da amostra** importa:
- Amostra muito pequena → nao representa a populacao
- Amostra viesada → so um tipo de pessoa foi ouvida
- Amostra bem feita → representa a diversidade da populacao

Assim como uma colher de sopa prova a panela inteira,
uma boa amostra representa a populacao inteira.


```python
# Ingredientes bons vs ruins — impacto no resultado
# Vamos simular uma pesquisa de satisfacao
import random
random.seed(42)

# Dados de qualidade: amostra representativa da populacao
populacao = [random.gauss(7, 1.5) for _ in range(10000)]
media_real = sum(populacao)/len(populacao)
print(f"Satisfacao REAL (se entrevistassemos todo mundo): {media_real:.2f}")
print()

# Ingrediente ruim: amostra pequena e viesada
amostra_ruim = random.sample([x for x in populacao if x > 8], 10)
print(f"COM DADOS VIESADOS (so clientes satisfeitos):")
print(f"  Media estimada: {sum(amostra_ruim)/len(amostra_ruim):.2f}")
print(f"  Erro: {abs(sum(amostra_ruim)/len(amostra_ruim) - media_real):.2f}")
print()

# Ingrediente bom: amostra grande e aleatoria
amostra_boa = random.sample(populacao, 100)
print(f"COM DADOS DE QUALIDADE (amostra representativa):")
print(f"  Media estimada: {sum(amostra_boa)/len(amostra_boa):.2f}")
print(f"  Erro: {abs(sum(amostra_boa)/len(amostra_boa) - media_real):.2f}")
```

## 📌 Resumo

| Receita de Bolo | Estatistica |
|-|-|
| Ovos frescos | Dados precisos e confiaveis |
| Farinha na medida | Amostra representativa |
| Ovo estragado | Dado incorreto |
| Provar a massa | Validar os dados |
| Bolo ruim = ingredientes ruins | Conclusao ruim = dados ruins |
