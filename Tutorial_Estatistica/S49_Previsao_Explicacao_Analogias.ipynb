## Previsao do Tempo

O meteorologista da TV diz:
*"Amanha a probabilidade de chuva e de 70%."*

Ele nao esta **adivinhando**. Ele esta:
1. **Explicando** o passado: "nos ultimos 10 anos, sempre que
   tivemos nuvens assim, choveu em 7 de cada 10 vezes"
2. **Prevendo** o futuro: "entao amanha provavelmente vai chover"

Essa e a essencia da estatistica:
usar dados do **passado** para **explicar** padroes
e **prever** o que pode acontecer no futuro.

Previsao nunca e certeza — e **probabilidade**.
Assim como o meteorologista nunca diz "vai chover com 100% de certeza".


## 📊 Traduzindo

| Previsao do Tempo | Estatistica |
|-|-|
| Dados de 10 anos de clima | Dados historicos (passado) |
| Identificar padroes nas nuvens | Analise de correlacao |
| "70% de chance de chuva" | Probabilidade — previsao baseada em dados |
| As vezes erra a previsao | Erro de previsao — sempre ha incerteza |
| Mais sensores = melhor previsao | Mais dados = melhor estimativa |


## 📏 Explicar o Passado

O meteorologista olha para tras.
Como estava o ceu nos dias em que choveu?
Qual a temperatura, umidade, pressao?

Isso e **estatistica descritiva**:
- Resumir dados passados
- Encontrar padroes
- Calcular medias, variabilidades


## 📏 Prever o Futuro

Com base nos padroes do passado, o meteorologista projeta.
Se as condicoes de hoje sao parecidas com dias que
tiveram chuva, ele preve chuva.

Isso e **inferencia estatistica**:
- Usar dados de uma amostra (passado)
- Para concluir sobre algo geral (futuro)
- Sempre com uma margem de erro

Quanto mais dados historicos, melhor a previsao.
Mas nunca sera perfeita — o clima e caotico.
Assim como os dados do mundo real.


```python
# O meteorologista em Python
import random
random.seed(42)

# Dados historicos: temperatura e chuva (0 = nao, 1 = sim)
dias = 1000
temperaturas = [random.gauss(25, 5) for _ in range(dias)]
chuvas = [1 if t + random.gauss(0, 3) < 22 else 0 for t in temperaturas]

# Explicando o passado: quando choveu, qual a temp media?
temp_chuva = [temperaturas[i] for i in range(dias) if chuvas[i] == 1]
media_temp_chuva = sum(temp_chuva)/len(temp_chuva)
print("--- Explicando o Passado ---")
print(f"Dias analisados: {dias}")
print(f"Temperatura media em dias de chuva: {media_temp_chuva:.1f} C")
print(f"Frequencia de chuva: {sum(chuvas)/dias*100:.0f}%")

# Prevendo: se hoje esta 20C, qual a chance de chuva?
temp_hoje = 20
dias_parecidos = [i for i in range(dias) if abs(temperaturas[i] - temp_hoje) < 1]
chance = sum(chuvas[i] for i in dias_parecidos) / len(dias_parecidos)
print(f"\n--- Previsao para hoje ({temp_hoje}C) ---")
print(f"Chance de chuva: {chance*100:.0f}% baseado em {len(dias_parecidos)} dias analogos")
```

## 📌 Resumo

| Previsao do Tempo | Estatistica |
|-|-|
| Dados de 10 anos | Dados historicos |
| Padroes nas nuvens | Correlacoes estatisticas |
| 70% de chance | Probabilidade |
| Erro na previsao | Incerteza |
| Mais sensores | Mais dados = melhor estimativa |
