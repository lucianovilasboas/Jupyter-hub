## Medidor de Combustivel

O medidor de combustivel do seu carro nunca mostra o **nivel exato**.
Ele mostra uma **faixa**: entre 1/4 e 1/2 tanque.
Ou, no maximo, uma estimativa: "380 km ate o proximo posto".

**Estimar a media populacional e a mesma coisa.**

Nao sabemos a media exata de toda a populacao.
Mas podemos calcular um **intervalo de confianca**
que contem o valor verdadeiro com alta probabilidade.

O "medidor" e a nossa amostra.
Quanto mais dados (maior a amostra),
mais preciso o medidor (menor o intervalo).


## 📊 Traduzindo

| Medidor de Combustivel | Estatistica |
|-|-|
| Medidor mostra faixa | Intervalo de confianca |
| Nivel exato de gasolina | Media populacional verdadeira |
| Quanto mais sensores | Maior n = IC mais estreito |
| "380 km ate o proximo" | Estimativa com margem de erro |


## 📏 Calculo do IC para Media

IC = x t (s / n)

- **x**: media da amostra (leitura do medidor)
- **t**: valor critico (tamanho da "rede" de confianca)
- **s / n**: erro padrao (precisao do medidor)


```python
# Medidor de combustivel
import math, statistics as stat
from scipy import stats as sp_stats
# Amostra de 20 carros: consumo medio (km/L)
consumo = [12.5, 13.0, 11.8, 12.2, 13.5, 12.8, 12.0, 13.2, 12.6, 11.5,
           12.9, 13.1, 12.3, 12.7, 13.4, 12.1, 13.3, 12.4, 13.6, 12.0]
media = stat.mean(consumo)
erro = stat.stdev(consumo) / math.sqrt(len(consumo))
t = sp_stats.t.ppf(0.975, df=len(consumo)-1)
ic = (media - t*erro, media + t*erro)
print(f"Consumo medio amostral: {media:.2f} km/L")
print(f"IC 95%: [{ic[0]:.2f}, {ic[1]:.2f}] km/L")
```

## 📌 Resumo

| Medidor de Combustivel | Estatistica |
|-|-|
| Medidor | Intervalo de confianca |
| Nivel exato | Media verdadeira |
| Mais sensores | Maior n, IC mais estreito |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
