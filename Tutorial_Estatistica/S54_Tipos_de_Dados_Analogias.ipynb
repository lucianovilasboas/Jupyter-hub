## Festa de Aniversario

Imagine uma festa de aniversario com 30 convidados.
Voce pode classificar cada convidado de varias formas:

- **Tipo sanguineo**: A, B, AB, O (categorias, sem ordem)
- **Faixa etaria**: crianca, adolescente, adulto, idoso (com ordem)
- **Numero de presentes que trouxe**: 0, 1, 2, 3 (numeros inteiros)
- **Peso do bolo que comeu**: 0.150 kg, 0.200 kg (numeros continuos)

Cada tipo de dado exige um tipo de analise diferente!
Nao faz sentido calcular "media do tipo sanguineo",
assim como nao faz sentido dizer "a media dos nomes".


## 📊 Traduzindo

| Festa de Aniversario | Estatistica |
|-|-|
| Tipo sanguineo (A, B, AB, O) | Dado qualitativo nominal |
| Faixa etaria (crianca, adulto...) | Dado qualitativo ordinal |
| Numero de presentes (0, 1, 2...) | Dado quantitativo discreto |
| Peso do bolo (0.150 kg...) | Dado quantitativo continuo |
| Cor da camiseta | Dado qualitativo nominal |


## 📏 Qualitativo vs Quantitativo

**Qualitativo**: descreve qualidades (cor, tipo, categoria)
- Nominal: sem ordem (times, estados, sangue)
- Ordinal: com ordem (pequeno, medio, grande)

**Quantitativo**: descreve quantidades (numeros)
- Discreto: inteiros (quantidade de filhos, de carros)
- Continuo: valores reais (peso, altura, tempo)

O tipo de dado determina que graficos e testes usar!


```python
# Tipos de dados na pratica
convidados = ["Ana", "Bia", "Carlos", "Dani", "Edu"]
sangue = ["A", "O", "B", "AB", "A"]        # nominal
idade = ["crianca", "adulto", "adulto", "idoso", "crianca"]  # ordinal
presentes = [2, 0, 3, 1, 2]               # discreto
peso_bolo = [0.15, 0.20, 0.35, 0.10, 0.25]  # continuo
print("Tipos de dados na festa:")
print(f"  Sangue (nominal): {sangue}")
print(f"  Idade (ordinal): {idade}")
print(f"  Presentes (discreto): {presentes}")
print(f"  Peso do bolo (continuo): {peso_bolo}")
```

## 📌 Resumo

| Festa de Aniversario | Estatistica |
|-|-|
| Tipo sanguineo | Qualitativo nominal |
| Faixa etaria | Qualitativo ordinal |
| Numero de presentes | Quantitativo discreto |
| Peso do bolo | Quantitativo continuo |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
