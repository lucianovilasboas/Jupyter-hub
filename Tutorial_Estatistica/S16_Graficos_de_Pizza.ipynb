## 📝 Introdução

Os **gráficos de pizza** (pie charts) mostram a proporção de cada categoria como uma fatia de um círculo. O círculo completo representa 100%, e cada fatia representa uma porcentagem do total.

Apesar de populares, os gráficos de pizza são **polêmicos** na visualização de dados. Especialistas como Edward Tufte recomendam cautela: o olho humano tem dificuldade em comparar ângulos e áreas. Neste capítulo, vamos aprender quando usá-los e, mais importante, quando **evitá**-los.

## 📚 Explicação Detalhada

### 1. O que é um Gráfico de Pizza?

Cada fatia corresponde a uma categoria, com ângulo proporcional ao seu valor. O círculo completo = 360° = 100%.

### 2. Quando Usar

- Poucas categorias (2 a 5)
- Partes de um todo (proporções que somam 100%)
- Destacar uma categoria dominante

### 3. Quando NÃO Usar

- Muitas categorias (>5-6)
- Diferenças pequenas entre categorias
- Comparar valores entre diferentes gráficos de pizza
- Mostrar mudanças ao longo do tempo

### 4. Donut Charts

Variação do gráfico de pizza com o centro vazado. Visualmente mais limpo e permite adicionar informação no centro (total, média, etc.).

### 5. Cuidados

- Diferenças pequenas são difíceis de perceber
- Cores similares confundem
- Fatias 3D distorcem a percepção
- Ordenar fatias (maior para menor) ajuda a leitura

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
print('Bibliotecas importadas!')
```

```python
# Criando dados de exemplo — market share
categorias = ['Smartphones', 'Notebooks', 'Tablets', 'Acessórios', 'Software']
share = np.array([35, 25, 15, 15, 10])

print('Market Share por Categoria:')
for cat, val in zip(categorias, share):
    print(f'{cat:15s}: {val}%')
```

```python
# Gráfico de pizza básico
plt.figure(figsize=(8, 8))
plt.pie(share, labels=categorias, autopct='%1.1f%%', startangle=90,
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'])
plt.title('Market Share por Categoria')
plt.axis('equal')  # garante que o círculo seja redondo
plt.show()
```

```python
# Gráfico de pizza com fatia destacada (explode)
plt.figure(figsize=(8, 8))
explode = (0.1, 0, 0, 0, 0)  # destaca a primeira fatia
plt.pie(share, labels=categorias, autopct='%1.1f%%', startangle=90,
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'],
        explode=explode, shadow=True)
plt.title('Market Share — Smartphones em Destaque')
plt.axis('equal')
plt.show()
```

```python
# Donut chart (gráfico de rosca)
plt.figure(figsize=(8, 8))
plt.pie(share, labels=categorias, autopct='%1.1f%%', startangle=90,
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'],
        wedgeprops=dict(width=0.4, edgecolor='white'))
plt.title('Donut Chart — Market Share')
plt.axis('equal')
# Adicionando texto no centro
plt.text(0, 0, 'Total
100%', ha='center', va='center', fontsize=16, fontweight='bold')
plt.show()
```

```python
# Donut chart com total no centro
total_vendas = 1_250_000
plt.figure(figsize=(8, 8))
plt.pie(share, labels=categorias, autopct=lambda p: f'R$ {p*total_vendas/100:,.0f}',
        startangle=90, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'],
        wedgeprops=dict(width=0.4, edgecolor='white'))
plt.title('Distribuição de Vendas por Categoria')
plt.axis('equal')
plt.text(0, 0, f'Total
R$ {total_vendas:,}', ha='center', va='center', fontsize=14, fontweight='bold')
plt.show()
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Orçamento familiar
categorias = ['Moradia', 'Alimentação', 'Transporte', 'Saúde', 'Educação', 'Lazer', 'Outros']
orcamento = np.array([2000, 1200, 800, 500, 400, 300, 200])
orc_pct = orcamento / orcamento.sum() * 100

print('=== Orçamento Familiar ===')
for cat, val, pct in zip(categorias, orcamento, orc_pct):
    print(f'{cat:15s}: R$ {val:5d} ({pct:5.1f}%)')

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(orcamento, labels=categorias, autopct='%1.1f%%', startangle=90,
        colors=plt.cm.Set3(np.linspace(0, 1, len(categorias))))
plt.title('Gráfico de Pizza')
plt.axis('equal')

plt.subplot(1, 2, 2)
plt.pie(orcamento, labels=categorias, autopct='%1.1f%%', startangle=90,
        colors=plt.cm.Set3(np.linspace(0, 1, len(categorias))),
        wedgeprops=dict(width=0.4, edgecolor='white'))
plt.title('Donut Chart')
plt.axis('equal')
plt.text(0, 0, f'R$ {orcamento.sum():,}', ha='center', va='center', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Exemplo 2: Comparação de proporções — problema com pizza!
# Quando temos muitas categorias com valores parecidos
np.random.seed(42)
muitas_cat = [f'Item {i}' for i in range(1, 13)]
valores = np.random.randint(5, 15, size=12)

plt.figure(figsize=(14, 5))

plt.subplot(1, 3, 1)
plt.pie(valores, labels=muitas_cat, autopct='%1.0f%%')
plt.title('Pizza — Muitas categorias (confuso!)')
plt.axis('equal')

plt.subplot(1, 3, 2)
plt.barh(muitas_cat, valores, color='steelblue', edgecolor='black')
plt.title('Barras horizontais (melhor!)')
plt.xlabel('Valor')

# Ordenado
ord_idx = np.argsort(valores)
plt.subplot(1, 3, 3)
plt.barh([muitas_cat[i] for i in ord_idx], [valores[i] for i in ord_idx],
         color='coral', edgecolor='black')
plt.title('Barras ordenadas (ideal!)')
plt.xlabel('Valor')

plt.tight_layout()
plt.show()
print('Com 12 categorias, o gráfico de pizza fica confuso.')
print('Barras horizontais ordenadas são muito mais legíveis.')
```

```python
# Exemplo 3: Diferenças pequenas — outro problema com pizza
pequenas_diff = np.array([33, 32, 35])  # quase iguais!
rotulos_peq = ['Produto A', 'Produto B', 'Produto C']

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.pie(pequenas_diff, labels=rotulos_peq, autopct='%1.1f%%')
plt.title('Pizza — Diferenças pequenas')
plt.axis('equal')

plt.subplot(1, 3, 2)
plt.bar(rotulos_peq, pequenas_diff, color=['gold', 'coral', 'skyblue'], edgecolor='black')
plt.ylim(0, 40)
plt.title('Barras — diferenças visíveis!')
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 3, 3)
# Zoom no eixo y revela diferenças
plt.bar(rotulos_peq, pequenas_diff, color=['gold', 'coral', 'skyblue'], edgecolor='black')
plt.ylim(30, 36)
plt.title('Barras com zoom — diferenças claras')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
print('Diferenças de 2-3% são quase imperceptíveis no pizza, mas claras nas barras.')
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Uma pesquisa com 200 pessoas perguntou o meio de transporte preferido: Carro (80), Ônibus (45), Bicicleta (35), Moto (25), A pé (15). Crie um gráfico de pizza e um donut chart.

```python
# Solução do Exercício 1
transportes = ['Carro', 'Ônibus', 'Bicicleta', 'Moto', 'A pé']
contagens = [80, 45, 35, 25, 15]

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.pie(contagens, labels=transportes, autopct='%1.1f%%', startangle=90,
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'])
plt.title('Transporte Preferido — Pizza')
plt.axis('equal')

plt.subplot(1, 2, 2)
plt.pie(contagens, labels=transportes, autopct='%1.1f%%', startangle=90,
        colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'],
        wedgeprops=dict(width=0.4, edgecolor='white'))
plt.title('Transporte Preferido — Donut')
plt.axis('equal')
plt.text(0, 0, f'n={sum(contagens)}', ha='center', va='center', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Exercício 2:** Explique por que um gráfico de pizza NÃO seria adequado para mostrar a evolução das vendas de uma empresa ao longo de 12 meses.

**Solução:** Um gráfico de pizza não é adequado porque:
1. 12 categorias é demais — as fatias ficam muito pequenas
2. A evolução temporal requer um eixo horizontal (tempo) — use gráfico de linhas
3. Diferenças entre meses consecutivos seriam difíceis de perceber
4. A soma das fatias não representa um "todo" significativo (cada mês é independente)

**Alternativa:** Use um gráfico de linhas ou barras para mostrar tendência temporal.

## 📌 Resumo

- **Gráficos de pizza** mostram proporções como fatias de um círculo.
- **Use com moderação**: máximo 5-6 categorias, diferenças significativas.
- **Donut charts**: variação com centro vazado, mais limpo visualmente.
- **Prefira barras** quando houver muitas categorias ou diferenças pequenas.
- **Nunca use** pizza 3D — distorce a percepção.
- **Sempre ordene** as fatias do maior para o menor.

## 💡 Curiosidades

O gráfico de pizza foi inventado por **William Playfair** em 1801, na obra "Statistical Breviary". Originalmente chamado de **"círculo setorial"** (sector circle), ele só se popularizou no século XX. Curiosamente, o gráfico de pizza é um dos poucos tipos de gráfico que **não** recomendamos usar com frequência — a visualização de dados moderna prefere barras ou treemaps para proporções.

**Edward Tufte**, renomado especialista em visualização de dados, disse: "A única coisa pior que um gráfico de pizza são vários deles."

## ✅ Boas Práticas

1. **Limite a 5-6 categorias** no máximo.
2. **Ordene** do maior para o menor (sentido horário).
3. **Use cores contrastantes** para distinguir fatias.
4. **Prefira donut charts** — são visualmente mais limpos.
5. **Evite pizza 3D** — o ângulo distorce a percepção das proporções.
6. **Considere barras** como alternativa sempre que possível.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Muitas categorias (poluição visual) | Limite a 5-6 ou use barras. |
| Diferenças pequenas imperceptíveis | Use barras com zoom ou tabela. |
| Pizza 3D distorcendo proporções | Nunca use 3D — prefira 2D. |
| Comparar dois pizzas separados | Use barras agrupadas ou empilhadas. |
| Cores similares confundindo | Use paletas contrastantes. |
