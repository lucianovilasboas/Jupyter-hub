## Buffet de Casamento

Voce esta organizando um casamento para 200 convidados.
Precisa saber o que servir no jantar.
Nao vai perguntar para **todos os 200** — levaria horas!

Em vez disso, voce escolhe **alguns convidados** para consultar.
A pergunta e: **como escolher esses convidados?**

- **Aleatorio simples**: sorteia 20 nomes de uma urna
  (justo, mas pode nao pegar vegetarianos)
- **Estratificado**: separa por mesa (familia, amigos, trabalho)
  e sorteia 5 de cada grupo (representa todos os grupos)
- **Por conveniencia**: pergunta so para quem esta perto do buffet
  (rapido, mas viesado — sao os mais famintos!)


## 📊 Traduzindo

| Buffet de Casamento | Estatistica |
|-|-|
| Sortear 20 nomes de uma urna | Amostra aleatoria simples |
| Separar por mesa e sortear de cada | Amostra estratificada |
| Perguntar a cada 5o convidado | Amostra sistematica |
| Perguntar a quem esta perto do buffet | Amostra por conveniencia (viesada) |
| Grupos: familia, amigos, trabalho | Estratos |


## 📏 Cada Metodo Tem um Proposito

**Aleatoria simples**: a mais justa, mas exige lista completa.
**Estratificada**: garante representatividade de subgrupos.
**Sistematica**: pratica (ex: a cada 10 pessoas na fila).
**Conveniencia**: facil, mas quase sempre viesada.

Na pratica, combinamos metodos.
O importante e reconhecer o vies de cada um.


```python
# Tipos de amostra em Python
import random
random.seed(42)

convidados = list(range(1, 201))  # 200 convidados
tipos = ["vegetariano"]*30 + ["carne"]*120 + ["vegano"]*20 + ["crianca"]*30
random.shuffle(tipos)

# Amostra aleatoria simples
simples = random.sample(range(200), 20)
print("Aleatoria simples:", sum(1 for i in simples if tipos[i]=="vegano"), "veganos")

# Amostra estratificada (garantir veganismo)
veganos_idx = [i for i in range(200) if tipos[i]=="vegano"]
estratificada = random.sample(veganos_idx, 5)  # 5 veganos garantidos
print("Estratificada: 5 veganos garantidos na amostra")
```

## 📌 Resumo

| Buffet de Casamento | Estatistica |
|-|-|
| Sortear nomes | Aleatoria simples |
| Separar por mesa | Estratificada |
| Perguntar a cada 5o | Sistematica |
| Perguntar perto do buffet | Conveniencia (viesada) |
