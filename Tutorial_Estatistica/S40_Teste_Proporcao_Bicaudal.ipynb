## 📝 Introdução

O **teste bicaudal** (two-tailed test) para proporção é o mais comum na prática. Ele verifica se a **proporção populacional é diferente** de um valor de referência, sem especificar uma direção. A região de rejeição está dividida entre as duas caudas da distribuição.

Situações típicas:
- "A proporção de votos do candidato mudou após o debate?"
- "A taxa de cura do novo tratamento é diferente da taxa padrão?"
- "A proporção de clientes satisfeitos é diferente de 80%?"

Este teste é não-direcional: estamos interessados em detectar qualquer diferença, seja para mais ou para menos.

## 📚 Explicação Detalhada

### 1. Formulação do Teste Bicaudal

- **H₀**: p = p₀ (a proporção é igual ao valor de referência)
- **H₁**: p ≠ p₀ (a proporção é **diferente** do valor de referência)

A região de rejeição está em **ambas as caudas** da distribuição normal padrão.

### 2. Estatística Z e Valor Crítico

Z = (p̂ - p₀) / √(p₀(1-p₀)/n)

Rejeitamos H₀ se:
- **|Z| > Z_{α/2}** (abordagem do valor crítico)
- **p-valor < α** (abordagem do p-valor)

Onde p-valor = 2 × P(Z > |Z_calc|) = 2 × (1 - Φ(|Z_calc|))

### 3. Comparação com Testes Unicaudais

| Característica | Bicaudal | Unilateral Esquerda | Unilateral Direita |
|---------------|----------|--------------------|--------------------|
| H₁ | p ≠ p₀ | p < p₀ | p > p₀ |
| Região de rejeição | Ambas caudas | Cauda esquerda | Cauda direita |
| α por cauda | α/2 | α | α |
| p-valor | 2 × P(Z > |Z|) | P(Z < Z) | P(Z > Z) |
| Potência | Menor para efeitos direcionais | Maior na direção especificada | Maior na direção especificada |

## 💻 Exemplos Práticos

```python
# Importando bibliotecas
import numpy as np
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
import matplotlib.pyplot as plt

print("Bibliotecas carregadas com sucesso!")
```

```python
# Exemplo 1: Teste de proporcao bicaudal (calculo manual)
# Uma empresa afirma que 60% dos clientes preferem seu produto.
# Uma pesquisa com 250 clientes mostra que 140 preferem o produto.
# Teste se a proporcao é diferente de 60%.

n = 250
x = 140
p0 = 0.60
alpha = 0.05

p_hat = x / n
erro = np.sqrt(p0 * (1 - p0) / n)
z = (p_hat - p0) / erro

# p-valor bicaudal
p_valor = 2 * (1 - stats.norm.cdf(abs(z)))

# Valor critico bicaudal
z_crit = stats.norm.ppf(1 - alpha/2)

print(f"n = {n}, x (preferem) = {x}")
print(f"Proporção amostral (p̂) = {p_hat:.3f}")
print(f"Proporção sob H₀ (p₀) = {p0}")
print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
print(f"Estatística Z = {z:.4f}")
print(f"Valor crítico Z = ±{z_crit:.4f}")
print(f"p-valor = {p_valor:.4f}")
print()
if abs(z) > z_crit or p_valor < alpha:
    print("✅ Rejeitamos H₀: a proporção é diferente de 60%.")
else:
    print("❌ Não rejeitamos H₀: não há evidência de diferença.")
```

```python
# Exemplo 2: Usando proportions_ztest (padrao: bicaudal)
z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0)
# alternative='two-sided' e o padrao

print(f"Usando statsmodels (padrão = bicaudal):")
print(f"x = {x}, n = {n}")
print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
print(f"Estatística Z = {z_stat:.4f}")
print(f"p-valor = {p_val:.4f}")
if p_val < alpha:
    print(f"✅ p-valor {p_val:.4f} < α = {alpha} → Rejeitamos H₀")
else:
    print(f"❌ p-valor {p_val:.4f} ≥ α = {alpha} → Não rejeitamos H₀")
```

```python
# Exemplo 3: Visualizacao do teste bicaudal
z_vals = np.linspace(-4, 4, 1000)
densidade = stats.norm.pdf(z_vals, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(z_vals, densidade, 'b-', label='Normal Padrão')

# Regioes de rejeicao (ambas as caudas)
z_crit = stats.norm.ppf(1 - alpha/2)
x_fill_left = np.linspace(-4, -z_crit, 100)
x_fill_right = np.linspace(z_crit, 4, 100)
plt.fill_between(x_fill_left, stats.norm.pdf(x_fill_left), color='red', alpha=0.3)
plt.fill_between(x_fill_right, stats.norm.pdf(x_fill_right), color='red', alpha=0.3, label='Região de Rejeição')

plt.axvline(z, color='green', linestyle='--', lw=2, label=f'Z calculado = {z:.2f}')
plt.axvline(-z_crit, color='red', linestyle=':', lw=2, label=f'Z crítico = ±{z_crit:.2f}')
plt.axvline(z_crit, color='red', linestyle=':', lw=2)
plt.title('Teste de Proporção Bicaudal')
plt.xlabel('Z')
plt.ylabel('Densidade')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Exemplo 4: Funcao completa para teste bicaudal
def teste_proporcao_bicaudal(x, n, p0, alpha=0.05):
    """Teste bicaudal para proporcao."""
    p_hat = x / n
    erro = np.sqrt(p0 * (1 - p0) / n)
    z = (p_hat - p0) / erro
    p_val = 2 * (1 - stats.norm.cdf(abs(z)))
    z_crit = stats.norm.ppf(1 - alpha/2)
    
    print(f"=== Teste Bicaudal (Proporção) ===")
    print(f"Dados: x = {x}, n = {n}, p̂ = {p_hat:.4f}")
    print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
    print(f"Z = {z:.4f}, Z crítico = ±{z_crit:.4f}")
    print(f"p-valor = {p_val:.4f}, α = {alpha}")
    if abs(z) > z_crit:
        print(f"✅ |Z| ({abs(z):.4f}) > Z crítico ({z_crit:.4f}) → Rejeitamos H₀")
    else:
        print(f"❌ |Z| ({abs(z):.4f}) ≤ Z crítico ({z_crit:.4f}) → Não rejeitamos H₀")
    return {"z": z, "p_valor": p_val, "rejeita": abs(z) > z_crit}

teste_proporcao_bicaudal(140, 250, 0.60)
print()
teste_proporcao_bicaudal(45, 100, 0.50, alpha=0.01)
```

```python
# Exemplo 5: Teste bicaudal com diferentes tamanhos amostrais
# Como o tamanho amostral afeta a decisao?

p0 = 0.50
alpha = 0.05

print("Efeito do tamanho amostral (p̂ = 0.45, p₀ = 0.50):")
print()
for n in [100, 200, 500, 1000, 2000]:
    x = int(0.45 * n)
    z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0)
    print(f"  n = {n:4d}, x = {x:4d}, Z = {z_stat:.4f}, p-valor = {p_val:.4f}", end="")
    if p_val < alpha:
        print(" ✅ Rejeita H₀")
    else:
        print(" ❌ Não rejeita H₀")
```

```python
# Exemplo 6: Comparacao bicaudal vs unilateral
print("COMPARAÇÃO: Teste Bicaudal vs Unilateral")
print("=" * 45)
print()

# Mesmo exemplo: 55 sucessos em 100 tentativas, p0 = 0.50
x, n, p0 = 55, 100, 0.50

z_bi, p_bi = proportions_ztest(count=x, nobs=n, value=p0, alternative='two-sided')
z_sm, p_sm = proportions_ztest(count=x, nobs=n, value=p0, alternative='smaller')
z_la, p_la = proportions_ztest(count=x, nobs=n, value=p0, alternative='larger')

print(f"x = {x}, n = {n}, p₀ = {p0}")
print(f"p̂ = {x/n:.3f}")
print()
print(f"Bicaudal:        Z = {z_bi:.4f}, p = {p_bi:.4f}")
print(f"Unilateral esq.: Z = {z_sm:.4f}, p = {p_sm:.4f}")
print(f"Unilateral dir.: Z = {z_la:.4f}, p = {p_la:.4f}")
print()
print(f"Relação: p(bicaudal) = 2 * min(p(unilateral esq.), p(unilateral dir.))")
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Um medicamento existente tem taxa de cura de 70%. Um novo medicamento cura 78 de 100 pacientes. Teste se a taxa de cura é diferente a 5%.

```python
# Solucao do Exercicio 1
n = 100
x = 78
p0 = 0.70
alpha = 0.05

z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0)
print(f"Proporção amostral: {x/n:.3f}")
print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
print(f"Z = {z_stat:.4f}, p-valor = {p_val:.4f}")
if p_val < alpha:
    print(f"✅ Rejeitamos H₀: a taxa de cura é diferente de 70%.")
else:
    print(f"❌ Não rejeitamos H₀.")
```

**Exercício 2:** Uma pesquisa eleitoral mostra que 520 de 1000 eleitores preferem o candidato A. Teste se a preferência é diferente de 50% a 1%.

```python
# Solucao do Exercicio 2
n = 1000
x = 520
p0 = 0.50
alpha = 0.01

p_hat = x / n
z = (p_hat - p0) / np.sqrt(p0 * (1 - p0) / n)
p_val = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"n = {n}, x = {x}")
print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
print(f"Z = {z:.4f}, p-valor = {p_val:.4f}")
if p_val < alpha:
    print(f"✅ Rejeitamos H₀ a {alpha}: a preferência é diferente de 50%.")
else:
    print(f"❌ Não rejeitamos H₀ a {alpha}.")
```

**Exercício 3:** Uma máquina deve produzir exatamente 5% de peças com defeito. Em 400 peças, 28 são defeituosas. Teste a 5%.

```python
# Solucao do Exercicio 3
n = 400
x = 28
p0 = 0.05
alpha = 0.05

z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0)
print(f"Proporção amostral: {x/n:.4f}")
print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
print(f"Z = {z_stat:.4f}, p-valor = {p_val:.4f}")
if p_val < alpha:
    print("✅ Rejeitamos H₀: a taxa de defeitos é diferente de 5%.")
else:
    print("❌ Não rejeitamos H₀.")
```

## 🏆 Desafios

1. Explique por que o teste bicaudal é mais conservador que o unilateral.
2. Pesquise sobre **intervalo de confiança para proporção** e compare com o resultado do teste bicaudal.
3. Crie uma função que, dado um valor de p₀, encontre o tamanho amostral mínimo necessário para detectar uma diferença δ com potência 0.80.

```python
# Desafio 3: Tamanho amostral para potencia 0.80
from scipy.stats import norm

def tamanho_amostral_proporcao(p0, delta, alpha=0.05, potencia=0.80):
    z_alpha = norm.ppf(1 - alpha/2)
    z_beta = norm.ppf(potencia)
    p1 = p0 + delta
    n = ((z_alpha * np.sqrt(p0*(1-p0)) + z_beta * np.sqrt(p1*(1-p1))) / delta) ** 2
    return int(np.ceil(n))

print("Tamanho amostral necessário (α=0.05, potência=0.80):")
for delta in [0.02, 0.05, 0.10, 0.15]:
    n = tamanho_amostral_proporcao(0.50, delta)
    print(f"  Para detectar δ = {delta:+.2f}: n = {n}")
```

## 📌 Resumo

- O **teste bicaudal** para proporção verifica se p é **diferente** de p₀.
- H₀: p = p₀, H₁: p ≠ p₀. Região de rejeição em ambas as caudas.
- Estatística Z = (p̂ - p₀) / √(p₀(1-p₀)/n).
- p-valor = 2 × P(Z > |Z_calc|).
- O `proportions_ztest` do statsmodels usa teste bicaudal por padrão.
- O teste bicaudal é o padrão recomendado quando não há direção específica.

## 💡 Curiosidades

O teste bicaudal é o padrão na maioria das publicações científicas porque é mais **rigoroso** — exige evidências mais fortes para rejeitar H₀. Estima-se que cerca de 80% dos testes de hipótese publicados na literatura médica são bicaudais. Muitos periódicos científicos **exigem** que os testes sejam bicaudais, a menos que o pesquisador forneça uma justificativa muito clara para o uso de testes unicaudais.

## ✅ Boas Práticas

1. Use o teste bicaudal quando não houver uma hipótese direcional clara.
2. O teste bicaudal é o padrão aceito na maioria das áreas científicas.
3. Ao usar statsmodels, lembre-se que o padrão já é bicaudal.
4. Reporte o intervalo de confiança de 95% junto com o p-valor.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Usar teste bicaudal com H₁ direcional | Se H₁ é direcional, use teste unilateral correspondente |
| Esquecer de dividir α por 2 nas caudas | α para cada cauda é α/2 no teste bicaudal |
| Multiplicar o p-valor por 2 quando desnecessário | Só multiplique se estiver convertendo unilateral → bicaudal |
| Interpretar "não rejeitar H₀" como "provar H₀" | Não rejeitar significa evidência insuficiente, não que H₀ é verdadeira |
