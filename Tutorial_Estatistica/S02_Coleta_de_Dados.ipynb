## 📝 Introdução

Antes de qualquer análise estatística, precisamos de dados. A forma como coletamos esses dados influencia diretamente a qualidade das conclusões que podemos tirar. Uma coleta mal planejada pode introduzir vieses, erros e inconsistências que comprometem todo o trabalho posterior.

Neste capítulo, exploraremos as principais fontes de dados, os métodos de coleta mais comuns e como garantir que os dados coletados sejam confiáveis e representativos do fenômeno que queremos estudar.

## 📚 Explicação Detalhada

### 1. Fontes de Dados

Os dados podem vir de duas grandes categorias:

**Dados Primários**: São coletados diretamente pelo pesquisador para um propósito específico. Exemplos: experimentos controlados, questionários, observações em campo.

**Dados Secundários**: Já foram coletados por outra pessoa ou instituição. Exemplos: dados do IBGE, bases públicas de dados, artigos científicos, relatórios financeiros.

### 2. Métodos de Coleta

- **Survey (Questionário)**: Coleta de opiniões, comportamentos e características por meio de perguntas estruturadas.
- **Experimento**: Ambiente controlado onde o pesquisador manipula variáveis para observar efeitos.
- **Observação**: Registro de fenômenos sem interferência do pesquisador.
- **Dados secundários**: Aproveitamento de bases existentes.

### 3. Qualidade dos Dados

Critérios importantes para avaliar a qualidade:
- **Precisão**: Os dados refletem corretamente a realidade?
- **Completude**: Há dados faltantes (missing data)?
- **Consistência**: Os dados seguem o mesmo padrão?
- **Atualidade**: Os dados são recentes o suficiente?
- **Representatividade**: A amostra representa bem a população?

### 4. Viés (Bias)

Viés é um erro sistemático que distorce os resultados. Tipos comuns:
- **Viés de seleção**: A amostra não representa a população.
- **Viés de resposta**: Os participantes não respondem honestamente.
- **Viés de confirmação**: Buscar dados que confirmem nossas crenças.
- **Viés de sobrevivência**: Analisar apenas os "sobreviventes" de um processo.

```python
# Exemplo: criando um dataset simples para representar dados coletados
# Simulamos uma pesquisa sobre hábitos de leitura com 10 pessoas
import random

# Fixamos a semente para reprodutibilidade
random.seed(42)

# Dados coletados: idade, horas de leitura por semana, gênero literário favorito
idades_entrevistados = [22, 35, 28, 45, 19, 52, 31, 27, 40, 33]
horas_leitura = [5, 2, 8, 3, 10, 1, 6, 4, 7, 3]
generos = ['Ficção', 'Não-ficção', 'Ficção', 'Biografia', 'Fantasia', 'História', 'Ficção', 'Romance', 'Ciência', 'Ficção']

# Organizamos em um dicionário (formato similar a um DataFrame)
dados_pesquisa = {
    'idade': idades_entrevistados,
    'horas_semana': horas_leitura,
    'genero': generos
}

print("=== Pesquisa de Hábitos de Leitura ===")
print("Participantes:", len(dados_pesquisa['idade']))
for i in range(len(dados_pesquisa['idade'])):
    print(f"  {i+1}. Idade: {dados_pesquisa['idade'][i]}, "
          f"Leitura: {dados_pesquisa['horas_semana'][i]}h/sem, "
          f"Gênero: {dados_pesquisa['genero'][i]}")
```

### 2. Simulação de Viés de Seleção

Vamos demonstrar como uma amostra mal coletada pode distorcer os resultados. Imagine que queremos saber a altura média dos alunos de uma escola, mas coletamos dados apenas dos alunos do time de basquete.

```python
# Simulando viés de seleção
# População real: alturas de 100 alunos (média ~1.70m)
random.seed(123)
populacao = [round(random.uniform(1.55, 1.90), 2) for _ in range(100)]
media_real = sum(populacao) / len(populacao)
print(f"Média real da população: {media_real:.2f}m")

# Amostra viesada: apenas alunos do time de basquete (pessoas mais altas)
amostra_basquete = [round(random.uniform(1.75, 2.05), 2) for _ in range(10)]
media_amostra = sum(amostra_basquete) / len(amostra_basquete)
print(f"Média da amostra viesada (basquete): {media_amostra:.2f}m")
print(f"Diferença (viés): {abs(media_real - media_amostra):.2f}m")
```

## 💻 Exemplos Práticos

```python
# Exemplo prático: gerando dados de uma pesquisa simulada
# Criamos um dataset mais completo usando estruturas de dados

nomes = ['Ana', 'Bruno', 'Carla', 'Daniel', 'Eduarda', 'Felipe', 'Gabriela', 'Henrique']
idades = [25, 32, 19, 41, 28, 35, 22, 30]
salarios = [3500, 5200, 1800, 7800, 4200, 6100, 2800, 4700]
cursos = ['ADS', 'Engenharia', 'Medicina', 'Direito', 'ADS', 'Engenharia', 'Psicologia', 'Administração']

print("=== Dataset de Alunos ===")
print(f"{'Nome':<12} {'Idade':<8} {'Salário (R$)':<15} {'Curso':<15}")
print('-' * 50)
for i in range(len(nomes)):
    print(f"{nomes[i]:<12} {idades[i]:<8} {salarios[i]:<15} {cursos[i]:<15}")
```

```python
# Calculando estatísticas básicas dos dados coletados
media_salarios = sum(salarios) / len(salarios)
min_salario = min(salarios)
max_salario = max(salarios)

print("=== Resumo dos Salários ===")
print(f"Média: R$ {media_salarios:.2f}")
print(f"Menor salário: R$ {min_salario:.2f}")
print(f"Maior salário: R$ {max_salario:.2f}")
print(f"Amplitude salarial: R$ {max_salario - min_salario:.2f}")

# Contagem por curso
print("\n=== Distribuição por Curso ===")
cursos_unicos = list(set(cursos))
for curso in cursos_unicos:
    contagem = cursos.count(curso)
    print(f"{curso}: {contagem} aluno(s)")
```

## ✅ Boas Práticas

1. Planeje a coleta antes de começar: defina objetivos, população e método.
2. Documente metadados: origem, data da coleta, metodologia, limitações.
3. Use amostragem aleatória sempre que possível para evitar viés.
4. Colete mais dados do que você acha que precisa (dados extras podem ser úteis).
5. Respeite a privacidade e a ética na coleta de dados pessoais.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Coletar dados sem planejamento | Sempre defina os objetivos antes de coletar. |
| Ignorar dados faltantes | Analise e trate valores ausentes adequadamente. |
| Generalizar a partir de amostras não representativas | Use técnicas de amostragem adequadas. |
| Não documentar a coleta | Mantenha um registro detalhado da metodologia. |
