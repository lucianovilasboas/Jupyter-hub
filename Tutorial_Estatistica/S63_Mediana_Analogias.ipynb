## Meio da Fila do Banco

Ha uma fila de 9 pessoas no banco.
Os tempos de espera (em minutos) sao:
[3, 5, 6, 7, 8, 9, 12, 15, 45]

A **mediana** e o valor do meio: **8 minutos**.
Metade esperou menos de 8 min, metade esperou mais.

Repare que o ultimo cliente esperou 45 min (muita coisa!).
Mas a mediana continua sendo 8 — ela **ignora** o extremo.

A mediana e como o **meio da fila**:
nao importa se alguem fura la na ponta ou no inicio,
o meio da fila continua no mesmo lugar.


## 📊 Traduzindo

| Meio da Fila do Banco | Estatistica |
|-|-|
| Fila de pessoas ordenadas | Dados ordenados |
| A pessoa do meio da fila | Mediana (Q2) — 50o percentil |
| Alguem fura no inicio ou fim | Outlier — nao afeta a mediana |
| Metade na frente, metade atras | 50% dos dados abaixo, 50% acima |


## 📏 Mediana: Robusta e Justa

A mediana e **robusta**: valores extremos nao a afetam.
Por isso e usada para:
- Salarios (CEO nao distorce)
- Precos de imoveis (mansion nao distorce)
- Qualquer dado com outliers

Ela responde: "qual e o valor tipico?"
ignorando os casos extremos.


```python
# Meio da fila
import statistics as stat
tempos = [3, 5, 6, 7, 8, 9, 12, 15, 45]
print(f"Fila ordenada: {tempos}")
print(f"Mediana: {stat.median(tempos)} min (meio da fila)")
print(f"Media: {stat.mean(tempos):.1f} min (distorcida pelo 45!)")
```

## 📌 Resumo

| Meio da Fila do Banco | Estatistica |
|-|-|
| Fila ordenada | Dados ordenados |
| Pessoa do meio | Mediana |
| Furar fila na ponta | Outlier nao afeta |
| Metade de cada lado | 50% abaixo / 50% acima |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
