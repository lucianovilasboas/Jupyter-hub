## Telescopio

Um astronomo aponta um **telescopio** para o ceu.
Ele ve uma pequena regiao do universo (a amostra).
Mas quer concluir sobre o universo inteiro (a populacao).

**Inferencia estatistica** e exatamente isso:
usar uma parte (amostra) para concluir sobre o todo (populacao).

O astronomo nunca visitou todos os planetas.
Assim como nos nunca entrevistamos todos os brasileiros.
Mas com as ferramentas certas (telescopio / estatistica),
podemos fazer boas estimativas.


## 📊 Traduzindo

| Telescopio | Estatistica |
|-|-|
| Telescopio apontado para o ceu | Metodo estatistico (inferencia) |
| Regiao observada do ceu | Amostra |
| Universo inteiro | Populacao |
| Concluir sobre outros planetas | Generalizar para a populacao |
| Margem de erro do telescopio | Margem de erro estatistico |


## 📏 O Poder e o Limite da Inferencia

O astronomo nunca tem 100% de certeza.
Mas com boas observacoes e metodos, chega perto.

**Inferencia = amostra + metodo + margem de erro**

A grande pergunta: a amostra e representativa?
Se o telescopio so aponta para um canto do ceu,
a conclusao pode ser viesada.


```python
# Telescopio em Python
import random
random.seed(42)
populacao = [random.gauss(170, 10) for _ in range(100000)]  # alturas
media_real = sum(populacao)/len(populacao)

# O telescopio (amostra)
amostra = random.sample(populacao, 100)
media_est = sum(amostra)/len(amostra)
print(f"Altura media REAL (universo): {media_real:.1f} cm")
print(f"Estimativa pelo telescopio:   {media_est:.1f} cm")
print(f"Erro do telescopio: {abs(media_real - media_est):.1f} cm")
```

## 📌 Resumo

| Telescopio | Estatistica |
|-|-|
| Telescopio | Metodo de inferencia |
| Regiao do ceu | Amostra |
| Universo | Populacao |
| Margem de erro | Erro amostral |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
