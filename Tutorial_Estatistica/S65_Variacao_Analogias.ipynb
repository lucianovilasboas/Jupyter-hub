## Termometro do Clima

Duas cidades brasileiras:

**Cidade A — Estavel:**
Temperaturas: [22, 23, 22, 24, 23, 22, 23] C
Media: 23C. Variacao: muito baixa.
Voce sabe o que vestir todo dia.

**Cidade B — Louca:**
Temperaturas: [10, 35, 18, 40, 12, 38, 15] C
Media: 23C tambem! Mas a variacao e enorme.
Voce precisa de casaco e biquini na mesma semana.

A **media** e a mesma. A **variacao** e completamente diferente.
A variacao mede o "quanto os dados balancam".


## 📊 Traduzindo

| Termometro do Clima | Estatistica |
|-|-|
| Cidade estavel (22-24C) | Baixa variacao — dados homogeneos |
| Cidade louca (10-40C) | Alta variacao — dados heterogeneos |
| Ambas tem media 23C | Media nao conta a historia completa |
| Oscilacao do termometro | Medidas de variacao: amplitude, variancia |


## 📏 Por Que a Variacao Importa?

A media sozinha pode enganar.
Duas coisas podem ter a mesma media,
mas experiencias completamente diferentes.

**Baixa variacao** = previsivel, confiavel
**Alta variacao** = imprevisivel, arriscado

Nos negocios: um investimento com alta variacao
e mais arriscado que um com baixa variacao.


```python
# Termometro do clima
import statistics as stat
estavel = [22, 23, 22, 24, 23, 22, 23]
louca = [10, 35, 18, 40, 12, 38, 15]

for nome, dados in [("Estavel", estavel), ("Louca", louca)]:
    print(f"Cidade {nome}:")
    print(f"  Media: {stat.mean(dados):.1f}C")
    print(f"  Desvio padrao: {stat.stdev(dados):.1f}C")
    print(f"  Amplitude: {max(dados)-min(dados)}C")
    print()
```

## 📌 Resumo

| Termometro do Clima | Estatistica |
|-|-|
| Cidade estavel | Baixa variacao |
| Cidade louca | Alta variacao |
| Ambas media 23C | Media nao basta |
| Oscilacao | Variacao |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
