## 📝 Introdução

A Tabela Z, também chamada de tabela da distribuição normal padrão, é uma ferramenta fundamental em estatística inferencial. Ela mostra a **probabilidade acumulada** P(Z ≤ z) para diferentes valores de z em uma distribuição normal padrão (média = 0, desvio padrão = 1).

Ela é usada para:
- Encontrar o **p-valor** a partir de uma estatística Z calculada
- Encontrar o **valor crítico** para um dado nível de significância α
- Calcular probabilidades associadas a eventos em uma distribuição normal

Neste capítulo, vamos aprender a consultar a tabela e geraremos a tabela completa com Python.

## 📚 Explicação Detalhada

### 1. Estrutura da Tabela Z

A tabela Z tem a seguinte estrutura:
- **Linhas**: representam a parte inteira e a primeira casa decimal do valor Z (ex: 1.5)
- **Colunas**: representam a segunda casa decimal do valor Z (ex: 0.06)
- **Células**: valor de P(Z ≤ z) — a probabilidade acumulada até aquele ponto

Exemplo: para z = 1.56, vamos na linha 1.5 e coluna 0.06 → P(Z ≤ 1.56) = 0.9406

### 2. Valores Positivos vs Negativos

Para valores **positivos** de Z, a tabela dá P(Z ≤ z) diretamente.

Para valores **negativos** de Z, usamos a simetria da normal:
P(Z ≤ -z) = 1 - P(Z ≤ z) = P(Z > z)

### 3. Usando a Tabela para Encontrar o p-valor

Para um teste **bicaudal**:
- Calcule Z
- Encontre P(Z ≤ |Z|)
- p-valor = 2 × (1 - P(Z ≤ |Z|))

Para um teste **unilateral à direita**:
- p-valor = 1 - P(Z ≤ Z_calc)

Para um teste **unilateral à esquerda**:
- p-valor = P(Z ≤ Z_calc)

## 💻 Exemplos Práticos

```python
# Importando bibliotecas
import numpy as np
from scipy import stats
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

print("Bibliotecas carregadas com sucesso!")
```

```python
# Exemplo 1: Consultando valores na tabela Z com scipy
# Encontrando P(Z <= 1.56)
z = 1.56
prob = stats.norm.cdf(z)
print(f"P(Z ≤ {z}) = {prob:.4f}")
print()

# Encontrando P(Z <= -0.78)
z_neg = -0.78
prob_neg = stats.norm.cdf(z_neg)
print(f"P(Z ≤ {z_neg}) = {prob_neg:.4f}")
print()

# Encontrando P(0.5 <= Z <= 1.5)
prob_entre = stats.norm.cdf(1.5) - stats.norm.cdf(0.5)
print(f"P(0.5 ≤ Z ≤ 1.5) = {prob_entre:.4f}")
```

```python
# Exemplo 2: Encontrando o valor critico (Z critico)
# Para alpha = 0.05, bicaudal:
alpha = 0.05
z_critico = stats.norm.ppf(1 - alpha/2)
print(f"α = {alpha}, bicaudal")
print(f"Z crítico = ±{z_critico:.4f}")
print(f"P(Z ≤ {z_critico:.4f}) = {stats.norm.cdf(z_critico):.4f}")
print()

# Para alpha = 0.05, unilateral a direita:
z_crit_uni = stats.norm.ppf(1 - alpha)
print(f"α = {alpha}, unilateral à direita")
print(f"Z crítico = {z_crit_uni:.4f}")
print(f"P(Z ≤ {z_crit_uni:.4f}) = {stats.norm.cdf(z_crit_uni):.4f}")
print()

# Tabela de valores criticos comuns
alphas = [0.10, 0.05, 0.01, 0.001]
print("Valores Críticos para Diferentes α:")
print(" α  | Bicaudal | Unilateral")
print("-" * 30)
for a in alphas:
    z_bi = stats.norm.ppf(1 - a/2)
    z_uni = stats.norm.ppf(1 - a)
    print(f"{a:.3f} | ±{z_bi:.4f} | {z_uni:.4f}")
```

```python
# Exemplo 3: Usando a tabela para encontrar o p-valor
# Suponha que calculamos Z = 2.13 em um teste bicaudal
z_calc = 2.13
p_valor_bic = 2 * (1 - stats.norm.cdf(abs(z_calc)))

print(f"Estatística Z = {z_calc}")
print(f"P(Z ≤ {z_calc}) = {stats.norm.cdf(z_calc):.4f}")
print(f"P(Z > {z_calc}) = {1 - stats.norm.cdf(z_calc):.4f}")
print(f"p-valor (bicaudal) = {p_valor_bic:.4f}")
print(f"p-valor (unilateral dir.) = {1 - stats.norm.cdf(z_calc):.4f}")
```

```python
# Exemplo 4: Funcao de consulta a tabela Z
def consultar_z(z):
    """Retorna P(Z <= z) e tambem o p-valor para teste bicaudal."""
    prob = stats.norm.cdf(z)
    p_bic = 2 * (1 - stats.norm.cdf(abs(z)))
    print(f"Z = {z:.4f}")
    print(f"P(Z ≤ Z) = {prob:.4f}")
    print(f"P(Z > Z) = {1 - prob:.4f}")
    print(f"p-valor (bicaudal) = {p_bic:.4f}")
    return prob

consultar_z(0.84)
print()
consultar_z(-1.28)
print()
consultar_z(2.58)
```

```python
# GERACAO DA TABELA Z COMPLETA
# Gerando valores de z de 0.0 a 3.9 com passo 0.01

# Primeira parte: z = 0.0 a 1.9
print("=" * 100)
print("TABELA Z - Probabilidade Acumulada P(Z ≤ z) para z ≥ 0")
print("=" * 100)
print()

colunas = [f"0.0{i}" for i in range(10)]
print("z    | " + " | ".join(colunas))
print("-" * 100)

for dec in range(40):  # 0.0 a 3.9
    z_principal = dec / 10
    linha = f"{z_principal:.1f} | "
    for cent in range(10):
        z = z_principal + cent / 100
        prob = stats.norm.cdf(z)
        linha += f"{prob:.4f} | "
    print(linha)
print()

# Segunda parte: z = 4.0 a 5.0
print("-" * 100)
print("Valores para z >= 4.0:")
for z in [4.0, 4.1, 4.2, 4.3, 4.4, 4.5, 5.0]:
    prob = stats.norm.cdf(z)
    print(f"Z = {z:.1f} → P(Z ≤ {z:.1f}) = {prob:.6f}")
```

```python
# TABELA Z COMPLETA EM FORMATO PANDAS (mais legivel)
z_values = np.round(np.arange(0, 4.01, 0.01), 2)
z_principals = np.round(np.arange(0, 4.0, 0.1), 1)

tabela = {}
for zp in z_principals:
    linha = {}
    for cent in range(10):
        z = round(zp + cent/100, 2)
        linha[f"0.{cent:01d}"] = round(stats.norm.cdf(z), 4)
    tabela[zp] = linha

df_z = pd.DataFrame(tabela).T
df_z.columns = [f"0.{i:01d}" for i in range(10)]
print("TABELA Z - P(Z ≤ z) para z ≥ 0")
print("(Inteiro + 1ª decimal na linha, 2ª decimal na coluna)")
print("=" * 80)
df_z.index.name = "z"
print(df_z.to_string())
```

```python
# TABELA Z PARA VALORES NEGATIVOS
# Por simetria: P(Z ≤ -z) = 1 - P(Z ≤ z)
z_values_neg = np.round(np.arange(-3.9, 0.01, 0.01), 2)
z_principals_neg = np.round(np.arange(-3.9, 0.1, 0.1), 1)

tabela_neg = {}
for zp in z_principals_neg:
    linha = {}
    for cent in range(10):
        z = round(zp + cent/100, 2)
        linha[f"0.{cent:01d}"] = round(stats.norm.cdf(z), 4)
    tabela_neg[zp] = linha

df_z_neg = pd.DataFrame(tabela_neg).T
df_z_neg.columns = [f"0.{i:01d}" for i in range(10)]
print("TABELA Z (NEGATIVA) - P(Z ≤ z) para z < 0")
print("=" * 80)
df_z_neg.index.name = "z"
print(df_z_neg.to_string())
```

```python
# Exemplo 5: Como consultar a tabela manualmente
print("COMO CONSULTAR A TABELA Z")
print("=" * 60)
print()

print("1. Para encontrar P(Z ≤ 1.56):")
print("   - Localize a linha 1.5")
print("   - Localize a coluna 0.06")
print(f"   - Resultado: {stats.norm.cdf(1.56):.4f}")
print()

print("2. Para encontrar P(Z ≤ -0.78):")
print("   - Localize a linha 0.7 na tabela negativa")
print("   - Localize a coluna 0.08")
print(f"   - Resultado: {stats.norm.cdf(-0.78):.4f}")
print()

print("3. Para encontrar o valor crítico Z para α = 0.05 (bicaudal):")
print(f"   - Procure na tabela o valor mais próximo de {1 - 0.05/2:.4f}")
z_crit = stats.norm.ppf(1 - 0.05/2)
prob_crit = stats.norm.cdf(z_crit)
print(f"   - O valor Z correspondente é aproximadamente ±{z_crit:.4f}")
print(f"   - P(Z ≤ {z_crit:.4f}) = {prob_crit:.4f}")
```

```python
# Exemplo 6: Aplicacao pratica - encontrando p-valor com a tabela
print("APLICAÇÃO PRÁTICA: Encontrando o p-valor usando a Tabela Z")
print("=" * 60)
print()

# Um fabricante testa se a media de producao é diferente de 500
# Estatistica Z calculada: Z = 2.14

z_obs = 2.14
prob_acum = stats.norm.cdf(z_obs)
p_bicaudal = 2 * (1 - prob_acum)

print(f"H₀: μ = 500 | H₁: μ ≠ 500")
print(f"Z calculado = {z_obs}")
print()
print(f"Passo 1: Consultar P(Z ≤ {z_obs}) na tabela Z")
print(f"         P(Z ≤ {z_obs}) = {prob_acum:.4f}")
print()
print(f"Passo 2: Calcular P(Z > {z_obs})")
print(f"         P(Z > {z_obs}) = 1 - {prob_acum:.4f} = {1 - prob_acum:.4f}")
print()
print(f"Passo 3: Para teste bicaudal, multiplicar por 2")
print(f"         p-valor = 2 × {1 - prob_acum:.4f} = {p_bicaudal:.4f}")
print()
print(f"Passo 4: Decisão (α = 0.05)")
if p_bicaudal < 0.05:
    print(f"         {p_bicaudal:.4f} < 0.05 → Rejeitamos H₀")
else:
    print(f"         {p_bicaudal:.4f} ≥ 0.05 → Não rejeitamos H₀")
```

```python
# Exemplo 7: Tabela Z interativa simples
def buscar_na_tabela_z(z):
    """Busca um valor Z na tabela e retorna a probabilidade."""
    prob = stats.norm.cdf(z)
    sinal = "" if z >= 0 else " (negativo, use simetria)"
    print(f"Z = {z:.2f}{sinal}")
    print(f"P(Z ≤ {z:.2f}) = {prob:.4f}")
    print(f"P(Z > {z:.2f}) = {1 - prob:.4f}")
    return prob

print("Teste alguns valores:")
buscar_na_tabela_z(1.96)
print()
buscar_na_tabela_z(2.58)
print()
buscar_na_tabela_z(1.28)
print()
buscar_na_tabela_z(-1.65)
```

## 📌 Resumo

- A **Tabela Z** mostra P(Z ≤ z) para a distribuição normal padrão.
- Linhas representam a parte inteira + 1ª decimal; colunas a 2ª decimal.
- Para valores negativos, use a simetria: P(Z ≤ -z) = 1 - P(Z ≤ z).
- Usamos a tabela para encontrar p-valores e valores críticos.
- Geramos a tabela completa com `stats.norm.cdf()`.
- A inversão (encontrar Z a partir da probabilidade) usa `stats.norm.ppf()`.

## 💡 Curiosidades

Antes dos computadores, os estatísticos usavam tabelas Z impressas em livros. Essas tabelas tinham precisão de 4 casas decimais e eram consultadas manualmente — um processo trabalhoso que exigia cuidado para evitar erros de leitura. Hoje, com Python, podemos gerar a tabela completa ou consultar valores exatos instantaneamente. Ainda assim, entender a lógica da tabela é fundamental para compreender o que o computador está calculando.

## ✅ Boas Práticas

1. Sempre verifique se o valor Z é positivo ou negativo antes de consultar a tabela.
2. Use `stats.norm.cdf()` para valores exatos em vez de aproximações.
3. Lembre-se: a tabela Z acumula probabilidade da esquerda para a direita.
4. Para testes bicaudais, multiplique a probabilidade da cauda por 2.
5. Familiarize-se com os valores críticos mais comuns: 1.64 (α=0.10), 1.96 (α=0.05), 2.58 (α=0.01).

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Confundir P(Z ≤ z) com P(Z > z) | P(Z > z) = 1 - P(Z ≤ z) |
| Esquecer de dobrar o p-valor no teste bicaudal | Multiplique a probabilidade da cauda por 2 |
| Usar a tabela Z para a distribuição t | A tabela t também existe (próximo capítulo), use a correta |
| Ler a linha/coluna errada na tabela | Confirme: z = parteInteira.decimal1 + decimal2 |
