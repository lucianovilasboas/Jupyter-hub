## 📝 Introdução

A distribuição **t de Student** é uma das distribuições mais importantes da inferência estatística. Ela foi desenvolvida por **William Sealy Gosset** (que publicou sob o pseudônimo "Student") em 1908, enquanto trabalhava na cervejaria Guinness na Irlanda.

Gosset precisava de uma distribuição que funcionasse bem com amostras pequenas e quando o desvio padrão populacional era desconhecido — situações comuns no controle de qualidade industrial. O resultado foi a distribuição t, muito similar à normal, mas com **caudas mais pesadas**, o que a torna mais conservadora para amostras pequenas.

## 📚 Explicação Detalhada

### 1. Distribuição t de Student

A distribuição t é simétrica e tem forma de sino, assim como a normal. Porém, ela tem **caudas mais pesadas** (mais probabilidade nas extremidades), o que reflete a maior incerteza quando usamos o desvio padrão amostral s em vez do populacional σ.

### 2. Graus de Liberdade (gl)

O parâmetro que define a forma da distribuição t é o **grau de liberdade (gl ou df, do inglês *degrees of freedom*)**:
- gl = n - 1 (para estimação de média)
- Quanto menor o gl, mais pesadas as caudas
- Quanto maior o gl, mais a t se aproxima da normal padrão

### 3. Quando usar t vs Z?

| Situação | Usar |
|----------|------|
| σ conhecido E n grande (≥ 30) | Z (normal) |
| σ conhecido E n pequeno E população normal | Z (normal) |
| σ desconhecido E n grande | t (ou Z, pois t ≈ Z) |
| σ desconhecido E n pequeno | **t de Student** |

### 4. Convergência para Normal

À medida que n (e portanto gl) aumenta, a distribuição t converge para a normal padrão. Com gl ≥ 30, a diferença é praticamente desprezível.

```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Bibliotecas importadas com sucesso!")
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Comparando a distribuição t com a normal padrão
# Mostramos como a t tem caudas mais pesadas para gl pequenos

x = np.linspace(-5, 5, 500)

plt.figure(figsize=(12, 6))

# Normal padrão
plt.plot(x, stats.norm.pdf(x), 'k-', lw=2.5, label='Normal Padrão N(0,1)')

# t de Student com diferentes gl
graus = [1, 2, 5, 15, 30]
cores = ['red', 'orange', 'green', 'blue', 'purple']

for gl, cor in zip(graus, cores):
    plt.plot(x, stats.t.pdf(x, df=gl), color=cor, lw=1.5, linestyle='--', 
             label=f't (gl = {gl})')

plt.title('Comparação: Distribuição t de Student vs Normal Padrão', fontsize=14)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-5, 5)
plt.show()
```

```python
# Exemplo 2: Caudas mais pesadas — demonstrando na prática
# Calculamos probabilidades nas extremidades para t vs normal

from scipy import stats

print("=== Probabilidades nas Caudas ===")
print(f"{'Distribuição':<20} {'P(|X| > 2)':<20} {'P(|X| > 3)':<20}")
print("-" * 60)

# Normal padrão
p_normal_2 = 2 * (1 - stats.norm.cdf(2))
p_normal_3 = 2 * (1 - stats.norm.cdf(3))
print(f"{'Normal N(0,1)':<20} {p_normal_2:<20.4f} {p_normal_3:<20.4f}")

# t com diferentes gl
for gl in [1, 2, 5, 10, 30]:
    p_t_2 = 2 * (1 - stats.t.cdf(2, df=gl))
    p_t_3 = 2 * (1 - stats.t.cdf(3, df=gl))
    print(f"{'t (gl=' + str(gl) + ')':<20} {p_t_2:<20.4f} {p_t_3:<20.4f}")

print("\n→ A t tem caudas mais pesadas: mais probabilidade nas extremidades")
```

```python
# Exemplo 3: Convergência da t para a normal com n grande
# Mostramos que com gl ≥ 30 a diferença é mínima

x = np.linspace(-4, 4, 500)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, gl in enumerate([2, 10, 50]):
    ax = axes[i]
    
    y_t = stats.t.pdf(x, df=gl)
    y_norm = stats.norm.pdf(x)
    
    ax.plot(x, y_norm, 'k-', lw=2, label='Normal N(0,1)')
    ax.plot(x, y_t, 'r--', lw=2, label=f't (gl={gl})')
    
    # Diferença
    ax.fill_between(x, y_t, y_norm, alpha=0.2, color='gray', 
                     label=f'Diferença máxima: {np.max(np.abs(y_t - y_norm)):.4f}')
    
    ax.set_title(f'Convergência: t(gl={gl}) vs N(0,1)')
    ax.set_xlabel('Valor')
    ax.set_ylabel('Densidade')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

plt.suptitle('Convergência da Distribuição t para a Normal', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Exemplo 4: Valores críticos — t vs Z
# Comparando os valores críticos para 95% de confiança

niveis = [0.90, 0.95, 0.99]

print("=== Valores Críticos: Z vs t (bilateral) ===")
print(f"{'Nível':<10} {'Z crítico':<15} {'t crítico gl=5':<20} {'t crítico gl=10':<20} {'t crítico gl=30':<20}")
print("-" * 85)

for nivel in niveis:
    alpha = 1 - nivel
    z_crit = stats.norm.ppf(1 - alpha/2)
    
    t_crit_5 = stats.t.ppf(1 - alpha/2, df=5)
    t_crit_10 = stats.t.ppf(1 - alpha/2, df=10)
    t_crit_30 = stats.t.ppf(1 - alpha/2, df=30)
    
    print(f"{nivel:<10.0%} {z_crit:<15.4f} {t_crit_5:<20.4f} {t_crit_10:<20.4f} {t_crit_30:<20.4f}")

print("\n→ Quanto menor o gl, maior o valor crítico (mais conservador)")
```

```python
# Exemplo 5: Aplicação prática — calculando com t
# Amostra de 10 pessoas: tempo de reação a um estímulo (ms)

np.random.seed(42)

# Simulando dados (n = 10)
tempos = np.array([245, 256, 234, 268, 241, 252, 239, 249, 258, 244])
n = len(tempos)

media = np.mean(tempos)
desvio_amostral = np.std(tempos, ddof=1)  # ddof=1 para desvio amostral
erro_padrao = desvio_amostral / np.sqrt(n)

# Valor crítico t para 95% de confiança com gl = n - 1 = 9
gl = n - 1
t_critico = stats.t.ppf(0.975, df=gl)

margem_erro = t_critico * erro_padrao
ic_inferior = media - margem_erro
ic_superior = media + margem_erro

print("=== Análise de Tempo de Reação ===")
print(f"Amostra: {tempos}")
print(f"Tamanho (n): {n}")
print(f"Média amostral: {media:.2f} ms")
print(f"Desvio padrão amostral: {desvio_amostral:.2f} ms")
print(f"Graus de liberdade: {gl}")
print(f"Valor crítico t (95%): {t_critico:.4f}")
print(f"Erro padrão: {erro_padrao:.2f}")
print(f"Margem de erro: {margem_erro:.2f}")
print(f"IC 95% para a média: [{ic_inferior:.2f}, {ic_superior:.2f}] ms")

# Comparação com Z
z_critico = stats.norm.ppf(0.975)
ic_z_inferior = media - z_critico * erro_padrao
ic_z_superior = media + z_critico * erro_padrao
print(f"\nSe usássemos Z (errado para n pequeno):")
print(f"IC 95% (Z): [{ic_z_inferior:.2f}, {ic_z_superior:.2f}] ms")
```

```python
# Exemplo 6: Visualização da t vs Z para IC
# Mostramos como a t produz intervalos mais largos (conservadores)

gl_valores = range(1, 61)
t_crit_95 = [stats.t.ppf(0.975, df=gl) for gl in gl_valores]
z_crit_95 = stats.norm.ppf(0.975)

plt.figure(figsize=(10, 5))
plt.plot(gl_valores, t_crit_95, 'b-', lw=2, label='t crítico (95%)')
plt.axhline(y=z_crit_95, color='r', linestyle='--', lw=2, label=f'Z crítico = {z_crit_95:.3f}')
plt.axvline(x=30, color='gray', linestyle=':', alpha=0.7, label='gl = 30 (referência)')
plt.xlabel('Graus de Liberdade (gl)', fontsize=12)
plt.ylabel('Valor Crítico (95% bilateral)', fontsize=12)
plt.title('Convergência do Valor Crítico t para o Valor Crítico Z', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Para gl = 30, t crítico = {stats.t.ppf(0.975, 30):.4f} ≈ Z crítico = {z_crit_95:.4f}")
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Uma amostra de 16 estudantes teve média de horas de estudo = 12h/semana com desvio padrão amostral = 3.5h. Qual o valor crítico t para um IC de 95%?

```python
# Solução Exercício 1
n = 16
gl = n - 1
t_crit = stats.t.ppf(0.975, df=gl)

print(f"n = {n}")
print(f"gl = {gl}")
print(f"t crítico (95%, bilateral, gl={gl}) = {t_crit:.4f}")
print(f"Se fosse Z: {stats.norm.ppf(0.975):.4f}")
print(f"Diferença: {t_crit - stats.norm.ppf(0.975):.4f}")
```

**Exercício 2:** Com os dados do exercício 1, calcule o IC de 95% para a média popuulacional de horas de estudo.

```python
# Solução Exercício 2
media = 12
s = 3.5
n = 16
gl = n - 1

t_crit = stats.t.ppf(0.975, df=gl)
erro_padrao = s / np.sqrt(n)
margem = t_crit * erro_padrao

print(f"Média amostral: {media}")
print(f"Erro padrão: {erro_padrao:.2f}")
print(f"Margem de erro: {margem:.2f}")
print(f"IC 95%: [{media - margem:.2f}, {media + margem:.2f}] horas/semana")
print(f"\nInterpretação: Estamos 95% confiantes de que a verdadeira média
de horas de estudo está entre {media - margem:.2f} e {media + margem:.2f} h/semana.")
```

## 📌 Resumo

- A **distribuição t de Student** é usada quando σ é desconhecido e/ou n é pequeno.
- É parametrizada pelos **graus de liberdade** (gl = n - 1).
- Tem **caudas mais pesadas** que a normal → intervalos de confiança mais conservadores.
- Com gl ≥ 30, a t se aproxima da normal padrão.
- Gosset ("Student") a desenvolveu para controle de qualidade na cervejaria Guinness.

## 💡 Curiosidades

William Sealy Gosset trabalhava na cervejaria Guinness, que tinha uma política contra a publicação de pesquisas por seus funcionários (para evitar que concorrentes descobrissem segredos industriais). Por isso, Gosset publicou sua descoberta sob o pseudônimo "Student" em 1908. A Guinness só reconheceu oficialmente a autoria décadas depois! Hoje, a distribuição t é uma das ferramentas mais usadas em estatística.

## ✅ Boas Práticas

1. Sempre use a distribuição t quando o desvio padrão populacional for desconhecido.
2. Para n ≥ 30, t e Z dão resultados muito similares — mas t ainda é tecnicamente correto.
3. Verifique se a população é aproximadamente normal antes de usar t com amostras muito pequenas (n < 15).
4. Use `ddof=1` no `np.std` para obter o desvio padrão amostral.
5. Reporte sempre os graus de liberdade ao usar a distribuição t.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Usar Z quando σ é desconhecido (especialmente com n pequeno) | Quando σ é desconhecido, use SEMPRE a distribuição t. |
| Confundir gl com tamanho da amostra | gl = n - 1 para uma amostra. |
| Usar t com dados não normais para n muito pequeno | Para n < 15, a população precisa ser aproximadamente normal. |
| Esquecer ddof=1 no desvio padrão | O desvio padrão amostral usa n-1 no denominador. |
