## Arco e Flecha

Um arqueiro atira 10 flechas no alvo.

**Arqueiro A — Preciso:**
Todas as flechas caem perto do centro.
Pequeno desvio padrao.

**Arqueiro B — Impreciso:**
As flechas se espalham por todo o alvo.
Grande desvio padrao.

Somando tudo, a **media** dos dois arqueiros pode ser a mesma
(ambos acertam perto do centro em media).
Mas a **consistencia** e completamente diferente!

E isso que o desvio padrao mede:
o quanto os valores (flechas) se desviam do centro (media).


## 📊 Traduzindo

| Arco e Flecha | Estatistica |
|-|-|
| Flechas agrupadas no centro | Desvio padrao pequeno — dados concentrados |
| Flechas espalhadas no alvo | Desvio padrao grande — dados dispersos |
| Centro do alvo | Media dos dados |
| Arqueiro A vs B | Baixa vs alta variabilidade |


## 📏 Precisao vs Acuracia

**Acuracia**: a media esta perto do valor verdadeiro?
**Precisao**: os dados estao concentrados? (desvio padrao)

O arqueiro A e preciso (flechas juntas) e acurado (perto do centro).
Mas poderia ser: flechas juntas num canto (preciso, nao acurado).

O desvio padrao mede a **precisao** (dispersao).


```python
# Arco e flecha em Python
import statistics as stat
arqueiro_a = [9.5, 9.7, 9.8, 10.0, 10.1, 10.2, 10.3, 10.5]  # preciso
arqueiro_b = [2.0, 5.0, 8.0, 10.0, 10.0, 12.0, 15.0, 18.0]  # impreciso

for nome, dados in [("Arqueiro A (preciso)", arqueiro_a), ("Arqueiro B (impreciso)", arqueiro_b)]:
    print(f"{nome}:")
    print(f"  Media: {stat.mean(dados):.1f}")
    print(f"  Desvio padrao: {stat.stdev(dados):.2f}")
    print()
```

## 📌 Resumo

| Arco e Flecha | Estatistica |
|-|-|
| Flechas agrupadas | Desvio padrao pequeno |
| Flechas espalhadas | Desvio padrao grande |
| Centro do alvo | Media |
| Precisao | Dispersao |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
