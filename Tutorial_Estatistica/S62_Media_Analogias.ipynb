## Altura da Turma vs o Gigante

Imagine uma turma de 10 alunos.
As alturas sao:
[1.65, 1.70, 1.68, 1.72, 1.66, 1.69, 1.71, 1.67, 1.70, 1.68] metros.

A altura media da turma e **1,686 m** — o ponto de equilibrio.
Se todos se nivelassem, essa seria a altura.

Agora imagine que **o gigante** (2,10 m) entra na turma.
A nova media sobe para **1,724 m**!

So porque uma pessoa entrou, a media mudou.
Isso mostra: **a media e sensivel a valores extremos**.

A media e como uma gangorra: cada valor puxa para o seu lado.
Valores muito grandes ou muito pequenos "puxam" a media com forca.


## 📊 Traduzindo

| Altura da Turma vs o Gigante | Estatistica |
|-|-|
| Altura de cada aluno | Cada valor individual (x) |
| Ponto de equilibrio da gangorra | Media (x ou ) |
| O gigante entra na turma | Outlier — valor extremo |
| A gangorra pende | Media e sensivel a outliers |
| Todos se nivelam | Soma dos desvios = zero |


## 📏 A Gangorra dos Dados

Imagine uma gangorra onde cada valor e um peso.
A media e o ponto onde a gangorra se equilibra.

Se voce coloca um peso muito grande de um lado (um outlier),
o ponto de equilibrio se desloca para aquele lado.

Por isso a media nem sempre e a melhor medida:
- Dados simetricos: media funciona bem
- Dados com outliers: mediana e melhor


```python
# A gangorra da media
alturas = [1.65, 1.70, 1.68, 1.72, 1.66, 1.69, 1.71, 1.67, 1.70, 1.68]
media_sem = sum(alturas)/len(alturas)
print(f"Media SEM o gigante: {media_sem:.3f} m")

alturas_com = alturas + [2.10]  # entra o gigante
media_com = sum(alturas_com)/len(alturas_com)
print(f"Media COM o gigante: {media_com:.3f} m")
print(f"Diferenca: {(media_com - media_sem)*100:.1f} cm!")
```

## 📌 Resumo

| Altura da Turma vs o Gigante | Estatistica |
|-|-|
| Altura de cada aluno | Valor individual |
| Ponto de equilibrio | Media |
| Gigante entrando | Outlier puxa a media |
| Gangorra | Soma dos desvios = zero |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
