## Alvo de Tiro

Num **alvo de tiro**, errar para a **esquerda** OU para a **direita**
e igualmente errado. O importante e acertar o centro.

O **teste bicaudal** funciona assim:
nao importa se a proporcao esta acima ou abaixo do esperado —
o que importa e se ela e **diferente**.

Exemplo: uma moeda pode ser viciada para dar mais caras
OU mais coroas. Qualquer desvio importa.

No teste bicaudal, dividimos o nivel de significancia
nos dois lados: 2,5% para cada lado (total 5%).


## 📊 Traduzindo

| Alvo de Tiro | Estatistica |
|-|-|
| Centro do alvo | H0: proporcao = valor esperado |
| Errar para esquerda | Proporcao menor que H0 |
| Errar para direita | Proporcao maior que H0 |
| Os dois lados importam | Teste bicaudal |
| 2,5% de cada lado | α/2 em cada cauda |


```python
# Alvo de tiro
import math
n = 200
caras = 120
p_obs = caras / n
p0 = 0.5
erro = math.sqrt(p0 * (1-p0) / n)
z = (p_obs - p0) / erro
print(f"Caras observadas: {caras}/{n} = {p_obs*100:.0f}%")
print(f"H0: moeda honesta = 50%")
print(f"Z: {z:.2f}")
if abs(z) > 1.96:
    print("CONCLUSAO: Rejeitamos H0 (para QUALQUER lado)")
else:
    print("CONCLUSAO: Nao rejeitamos H0")
```

## 📌 Resumo

| Alvo de Tiro | Estatistica |
|-|-|
| Centro do alvo | H0 |
| Esquerda ou direita | Desvio em qualquer direcao |
| 2,5% + 2,5% = 5% | α dividido nas duas caudas |
