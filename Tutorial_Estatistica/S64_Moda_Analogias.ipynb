## Sabor de Sorvete

Numa sorveteria, 30 clientes pediram:
- Chocolate: 12 pessoas
- Morango: 8 pessoas
- Baunilha: 6 pessoas
- Flocos: 4 pessoas

O sabor **chocolate** e a **moda**: o valor que aparece mais vezes.
Nao precisa calcular — e so olhar qual aparece mais.

A moda e a unica medida de tendencia central
que funciona para **dados qualitativos**!
Qual e o time de futebol mais torcido? A moda responde.
Nao faz sentido calcular a "media dos times", mas a moda sim.


## 📊 Traduzindo

| Sabor de Sorvete | Estatistica |
|-|-|
| Sabores de sorvete | Categorias dos dados |
| Chocolate: 12 pessoas | Frequencia: 12 ocorrencias |
| Chocolate e o mais pedido | Moda: valor de maior frequencia |
| Empate entre dois sabores | Bimodal: duas modas |


## 📏 Quando Usar a Moda

A moda e util quando:
- Dados sao **categoricos** (nominais)
- Voce quer saber o "mais comum"
- Ha muitas repeticoes de valores

Cuidado: se todos os valores sao diferentes,
nao ha moda (ou todos sao moda).


```python
# Sabor de sorvete
import statistics as stat
pedidos = ["chocolate"]*12 + ["morango"]*8 + ["baunilha"]*6 + ["flocos"]*4
print(f"Total de pedidos: {len(pedidos)}")
print(f"Moda: {stat.mode(pedidos)}")
print(f"Frequencia: {pedidos.count(stat.mode(pedidos))} pessoas")
```

## 📌 Resumo

| Sabor de Sorvete | Estatistica |
|-|-|
| Sabores | Categorias |
| Chocolate: 12 | Frequencia |
| Chocolate mais pedido | Moda |
| Empate | Bimodal |
