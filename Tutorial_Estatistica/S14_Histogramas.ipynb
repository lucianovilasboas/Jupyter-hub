## 📝 Introdução

O **histograma** é um dos gráficos mais importantes em estatística. Ele mostra a distribuição de dados numéricos contínuos dividindo o intervalo em **bins** e representando a frequência de cada bin com barras **adjacentes** (sem espaçamento).

O histograma revela: onde os dados se concentram, se há assimetria, se há múltiplos picos (multimodalidade) e se existem valores extremos.

## 📚 Explicação Detalhada

### 1. O que é um Histograma?

- Eixo **x**: intervalos (bins) dos dados
- Eixo **y**: frequência (contagem) ou densidade
- Barras **adjacentes** (sem espaço)

### 2. Escolha de Bins

- **Regra de Sturges**: k = ⌈log₂(n) + 1⌉ (padrão matplotlib)
- **Freedman-Diaconis**: k = (max-min) / (2 × IQR × n⁻¹/³)
- **Scott**: k = (max-min) / (3.5 × σ × n⁻¹/³)

### 3. Histograma vs Densidade

A **densidade** (área total = 1) permite comparar distribuições de tamanhos diferentes. O **KDE** é uma estimativa suave da densidade.

### 4. Interpretação

- **Simétrica**: metades espelhadas
- **Assimetria positiva**: cauda à direita
- **Assimetria negativa**: cauda à esquerda
- **Unimodal**: um pico | **Bimodal**: dois picos | **Multimodal**: vários picos

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
np.random.seed(42)
print('Bibliotecas importadas!')
```

```python
# Criando dados de exemplo — notas de alunos
notas = np.random.normal(loc=65, scale=15, size=500)
notas = np.clip(notas, 0, 100)
print(f'{len(notas)} observações, Média={notas.mean():.2f}, DP={notas.std():.2f}')
```

```python
# Histograma básico com matplotlib
plt.figure(figsize=(10, 5))
plt.hist(notas, bins=10, color='royalblue', edgecolor='black', alpha=0.7)
plt.title('Histograma das Notas — 10 Bins')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Comparando diferentes números de bins
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
configs = [(5, '5 bins (poucos)'), (10, '10 bins (razoável)'), (30, '30 bins (muitos)'), (50, '50 bins (excessivo)')]
for ax, (bins, tit) in zip(axes.flat, configs):
    ax.hist(notas, bins=bins, color='steelblue', edgecolor='white', alpha=0.8)
    ax.set_title(tit)
    ax.set_xlabel('Nota')
    ax.set_ylabel('Frequência')
plt.tight_layout()
plt.show()
```

```python
# Histograma com densidade + KDE (seaborn)
plt.figure(figsize=(10, 5))
sns.histplot(notas, bins=15, kde=True, color='coral', edgecolor='white')
plt.title('Histograma com Densidade e KDE')
plt.xlabel('Nota')
plt.ylabel('Densidade')
plt.axvline(notas.mean(), color='darkred', linestyle='--', linewidth=2, label=f'Média={notas.mean():.1f}')
plt.axvline(np.median(notas), color='darkblue', linestyle=':', linewidth=2, label=f'Mediana={np.median(notas):.1f}')
plt.legend()
plt.show()
```

```python
# Comparando distribuições sobrepostas
np.random.seed(123)
turma_a = np.random.normal(loc=70, scale=10, size=200)
turma_b = np.random.normal(loc=60, scale=12, size=200)
turma_c = np.random.normal(loc=65, scale=8, size=200)

plt.figure(figsize=(10, 5))
plt.hist(turma_a, bins=15, alpha=0.4, label='Turma A (70, 10)', color='blue')
plt.hist(turma_b, bins=15, alpha=0.4, label='Turma B (60, 12)', color='red')
plt.hist(turma_c, bins=15, alpha=0.4, label='Turma C (65, 8)', color='green')
plt.title('Comparação de Notas entre Turmas')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Histogramas separados por grupo com seaborn
df = pd.DataFrame({
    'Nota': np.concatenate([turma_a, turma_b, turma_c]),
    'Turma': ['A'] * 200 + ['B'] * 200 + ['C'] * 200
})
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Nota', hue='Turma', bins=15, alpha=0.5, palette='Set2')
plt.title('Distribuição por Turma')
plt.show()
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Distribuição bimodal
np.random.seed(42)
g1 = np.random.normal(loc=40, scale=8, size=300)
g2 = np.random.normal(loc=75, scale=8, size=300)
bimodal = np.concatenate([g1, g2])

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.hist(bimodal, bins=20, color='purple', edgecolor='white', alpha=0.7)
plt.title('Distribuição Bimodal')
plt.xlabel('Valor')
plt.ylabel('Frequência')

plt.subplot(1, 2, 2)
sns.histplot(bimodal, bins=20, kde=True, color='purple')
plt.title('Bimodal com KDE')
plt.tight_layout()
plt.show()
print('Dois picos indicam duas populações distintas nos dados.')
```

```python
# Exemplo 2: Salários — assimetria à direita
np.random.seed(7)
salarios = np.random.lognormal(mean=8.5, sigma=0.6, size=1000)
salarios = np.clip(salarios, 1000, 50000)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(salarios, bins=30, color='gold', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(salarios), color='red', ls='--', lw=2, label=f'Média={np.mean(salarios):.0f}')
plt.axvline(np.median(salarios), color='green', ls='-', lw=2, label=f'Mediana={np.median(salarios):.0f}')
plt.title('Salários (assimétrica à direita)')
plt.xlabel('Salário (R$)')
plt.ylabel('Frequência')
plt.legend()

plt.subplot(1, 2, 2)
sns.histplot(salarios, bins=50, kde=True, color='gold')
plt.title('Com KDE — cauda longa à direita')
plt.tight_layout()
plt.show()
print(f'Assimetria: {pd.Series(salarios).skew():.4f}')
```

```python
# Exemplo 3: Regra de Freedman-Diaconis na prática
np.random.seed(42)
dados = np.random.normal(loc=50, scale=15, size=200)

q75, q25 = np.percentile(dados, [75, 25])
iqr = q75 - q25
h = 2 * iqr / (len(dados) ** (1/3))
k_fd = int((dados.max() - dados.min()) / h)
k_st = int(np.ceil(np.log2(len(dados)) + 1))
print(f'Sturges: {k_st} bins | Freedman-Diaconis: {k_fd} bins')

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].hist(dados, bins=k_st, color='teal', edgecolor='white')
axes[0].set_title(f'Sturges — {k_st} bins')
axes[1].hist(dados, bins=k_fd, color='teal', edgecolor='white')
axes[1].set_title(f'Freedman-Diaconis — {k_fd} bins')
plt.tight_layout()
plt.show()
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Gere 300 números de uma distribuição exponencial (scale=30). Crie um histograma com 20 bins e KDE. Calcule a assimetria.

```python
# Solução do Exercício 1
np.random.seed(500)
exp = np.random.exponential(scale=30, size=300)
plt.figure(figsize=(10, 4))
sns.histplot(exp, bins=20, kde=True, color='darkorange', edgecolor='white')
plt.title('Distribuição Exponencial — Assimetria Positiva')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.show()
sk = pd.Series(exp).skew()
print(f'Assimetria: {sk:.4f} — cauda longa à direita (típico de exponencial)')
```

**Exercício 2:** Gere duas normais (μ=50, μ=70, σ=10, n=200 cada). Crie um histograma sobreposto.

```python
# Solução do Exercício 2
np.random.seed(600)
d1 = np.random.normal(50, 10, 200)
d2 = np.random.normal(70, 10, 200)
plt.figure(figsize=(10, 4))
plt.hist(d1, bins=15, alpha=0.5, label='Grupo 1 (μ=50)', color='blue')
plt.hist(d2, bins=15, alpha=0.5, label='Grupo 2 (μ=70)', color='red')
plt.title('Comparação de Duas Distribuições')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
print(f'Grupo 1 — Média: {d1.mean():.2f}, DP: {d1.std():.2f}')
print(f'Grupo 2 — Média: {d2.mean():.2f}, DP: {d2.std():.2f}')
```

## 📌 Resumo

- **Histogramas** mostram a distribuição de dados contínuos com barras adjacentes.
- Escolha de **bins** é crítica: poucos escondem detalhes, muitos criam ruído.
- **Sturges, Freedman-Diaconis, Scott** — regras para número de bins.
- **Densidade** permite comparar distribuições de tamanhos diferentes.
- **KDE** suaviza o histograma revelando a forma da distribuição.
- Histogramas revelam **simetria, assimetria, multimodalidade**.

## 💡 Curiosidades

O **KDE** foi introduzido por **Murray Rosenblatt** (1956) e **Emanuel Parzen** (1962) — por isso chamado de **janela de Parzen-Rosenblatt**. É a base de métodos modernos de ML como detecção de anomalias. A escolha do número de bins ainda é pesquisa ativa!

## ✅ Boas Práticas

1. **Experimente diferentes bins** antes de decidir.
2. **Prefira densidade** ao comparar distribuições de tamanhos diferentes.
3. **Adicione KDE** como complemento.
4. **Rotule eixos** claramente (frequência ou densidade).
5. **Mantenha bins consistentes** ao comparar grupos.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Poucos bins e perda de detalhes | Comece com Sturges e ajuste. |
| Achatar dados multimodais | Reduza largura dos bins até picos aparecerem. |
| Confundir histograma com gráfico de barras | Histograma: dados contínuos, barras adjacentes. |
| Outliers distorcendo escala | Considere zoom ou transformação. |
