## Turma da Faculdade

Pense na distribuicao das **notas** de uma turma grande de faculdade.
- A **maioria** tira notas medianas (5 a 7)
- **Poucos** tiram notas muito baixas (0 a 3)
- **Poucos** tiram notas muito altas (9 a 10)

O grafico disso tem forma de **sino**:
alto no meio, baixo nas pontas.

Essa e a **distribuicao normal**!
Ela aparece em todo lugar:
- Altura das pessoas
- QI da populacao
- Pressao arterial
- Erros de medicao
- Peso de produtos

A natureza AMA a curva normal.


## 📊 Traduzindo

| Turma da Faculdade | Estatistica |
|-|-|
| Maioria: notas medianas | Centro da curva — media |
| Poucos: notas baixas | Cauda esquerda |
| Poucos: notas altas | Cauda direita |
| Forma de sino | Curva normal (gaussiana) |


## 📏 Propriedades da Curva Normal

- Simetrica em torno da media
- Media = Mediana = Moda (no mesmo ponto)
- 68% dos dados estao a 1 desvio da media
- 95% a 2 desvios
- 99,7% a 3 desvios

A "regra 68-95-99,7" e a assinatura da normal.


```python
# A turma da faculdade
import random, matplotlib.pyplot as plt
random.seed(42)
notas = [random.gauss(6.5, 1.5) for _ in range(500)]

plt.figure(figsize=(8, 4))
plt.hist(notas, bins=20, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Distribuicao das Notas — Curva Normal")
plt.xlabel("Nota")
plt.ylabel("Numero de alunos")
plt.axvline(x=sum(notas)/len(notas), color="red", linestyle="--", label="Media")
plt.legend()
plt.show()
```

## 📌 Resumo

| Turma da Faculdade | Estatistica |
|-|-|
| Maioria: notas medianas | Centro |
| Poucos: notas baixas/altas | Caudas |
| Forma de sino | Curva normal |
| Regra 68-95-99,7 | Propriedade |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
