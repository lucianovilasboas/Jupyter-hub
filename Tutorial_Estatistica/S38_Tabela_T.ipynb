## 📝 Introdução

A **Tabela T** (tabela da distribuição t de Student) é usada para encontrar **valores críticos** da distribuição t. Diferentemente da Tabela Z (que acumula probabilidade), a Tabela T geralmente apresenta valores críticos para diferentes níveis de significância (α) e graus de liberdade (gl).

Ela é essencial para:
- Testes T para média (one-sample e two-sample)
- Construção de intervalos de confiança quando σ é desconhecido
- Qualquer situação onde usamos a distribuição t

Como a distribuição t depende dos graus de liberdade, a tabela é organizada de forma diferente da tabela Z.

## 📚 Explicação Detalhada

### 1. Estrutura da Tabela T

A Tabela T típica tem:
- **Linhas**: graus de liberdade (gl = n - 1), geralmente de 1 a 30, depois 40, 50, 60, 100, e ∞
- **Colunas**: níveis de significância (α) para testes unicaudais ou bicaudais
- **Células**: valor crítico t tal que P(T > t) = α (para unicaudal) ou P(|T| > t) = α (para bicaudal)

### 2. Como Consultar

1. Determine os **graus de liberdade** (gl = n - 1)
2. Escolha o **nível de significância** α
3. Decida entre **unicaudal** ou **bicaudal**
4. Encontre o valor na interseção da linha gl com a coluna α

Exemplo: gl = 10, α = 0.05, bicaudal → t = 2.228

### 3. Tabela T vs Tabela Z

| Característica | Tabela Z | Tabela T |
|---------------|---------|---------|
| O que fornece | P(Z ≤ z) (acumulada) | Valores críticos t |
| Depende de | Apenas o valor de Z | Graus de liberdade e α |
| Para gl → ∞ | — | Aproxima-se da normal |
| Uso principal | Proporções, σ conhecido | Médias, σ desconhecido |

## 💻 Exemplos Práticos

```python
# Importando bibliotecas
import numpy as np
from scipy import stats
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

print("Bibliotecas carregadas com sucesso!")
```

```python
# GERACAO DA TABELA T COMPLETA (formato classico)
# Graus de liberdade: 1 a 30, depois 40, 50, 60, 80, 100, 120, infinito

gl_list = list(range(1, 31)) + [40, 50, 60, 80, 100, 120, float('inf')]

# Niveis de significancia para teste bicaudal
alphas_bicaudal = {
    '0.20': 0.20,
    '0.10': 0.10,
    '0.05': 0.05,
    '0.02': 0.02,
    '0.01': 0.01,
    '0.001': 0.001
}

print("TABELA T - Valores Críticos da Distribuição t de Student")
print("=" * 95)
print("O valor na tabela é o t crítico tal que P(|T| > t) = α (bicaudal)")
print("Para unicaudal, use α/2: ex: o valor para α=0.05 bicaudal = α=0.025 unicaudal")
print("=" * 95)
print()

colunas = [f"α={k}" for k in alphas_bicaudal.keys()]
print(f"{'gl':<5} | " + " | ".join(f"{c:>10}" for c in colunas))
print("-" * 95)

for gl in gl_list:
    linha = f"{gl:<5} | "
    for nome_a, a in alphas_bicaudal.items():
        if gl == float('inf'):
            t_crit = stats.norm.ppf(1 - a/2)
        else:
            t_crit = stats.t.ppf(1 - a/2, df=gl)
        linha += f"{t_crit:>10.4f} | "
    print(linha)
```

```python
# TABELA T COMPLETA EM FORMATO PANDAS (mais legivel e consultavel)
gl_list = list(range(1, 31)) + [40, 50, 60, 80, 100, 120, float('inf')]

alphas_uni = {
    'Uni 0.10': 0.10,
    'Uni 0.05': 0.05,
    'Uni 0.025': 0.025,
    'Uni 0.01': 0.01,
    'Uni 0.005': 0.005,
    'Uni 0.0005': 0.0005
}

alphas_bi = {
    'Bi 0.20': 0.20,
    'Bi 0.10': 0.10,
    'Bi 0.05': 0.05,
    'Bi 0.02': 0.02,
    'Bi 0.01': 0.01,
    'Bi 0.001': 0.001
}

# Tabela unicaudal
dados_uni = {}
for gl in gl_list:
    linha = {}
    for nome, a in alphas_uni.items():
        if gl == float('inf'):
            linha[nome] = round(stats.norm.ppf(1 - a), 4)
        else:
            linha[nome] = round(stats.t.ppf(1 - a, df=gl), 4)
    dados_uni[gl] = linha

df_t_uni = pd.DataFrame(dados_uni).T
df_t_uni.index.name = "gl"
print("TABELA T - UNICAUDAL (P(T > t) = α)")
print("=" * 85)
print(df_t_uni.to_string())
print()

# Tabela bicaudal
dados_bi = {}
for gl in gl_list:
    linha = {}
    for nome, a in alphas_bi.items():
        if gl == float('inf'):
            linha[nome] = round(stats.norm.ppf(1 - a/2), 4)
        else:
            linha[nome] = round(stats.t.ppf(1 - a/2, df=gl), 4)
    dados_bi[gl] = linha

df_t_bi = pd.DataFrame(dados_bi).T
df_t_bi.index.name = "gl"
print("TABELA T - BICAUDAL (P(|T| > t) = α)")
print("=" * 85)
print(df_t_bi.to_string())
```

```python
# Exemplo 1: Consultando a tabela T
print("EXEMPLO DE CONSULTA À TABELA T")
print("=" * 50)
print()

# Encontrando t critico para gl = 10, alpha = 0.05, bicaudal
gl = 10
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df=gl)
print(f"Para gl = {gl} e α = {alpha} (bicaudal):")
print(f"  t crítico = ±{t_crit:.4f}")
print()

# Para unicaudal a direita
t_crit_uni = stats.t.ppf(1 - alpha, df=gl)
print(f"Para gl = {gl} e α = {alpha} (unicaudal dir.):")
print(f"  t crítico = {t_crit_uni:.4f}")
print()

# Comparando com a normal
z_crit = stats.norm.ppf(1 - alpha/2)
print(f"Para comparação, Z crítico (α = {alpha}) = ±{z_crit:.4f}")
print(f"Diferença: t - Z = {t_crit - z_crit:.4f}")
```

```python
# Exemplo 2: Tabela T detalhada - variando gl de 1 a 30
print("VALORES CRÍTICOS t PARA α = 0.05")
print("=" * 30)
print(" gl  | Bicaudal | Unicaudal")
print("-" * 30)
for gl in range(1, 31):
    t_bi = stats.t.ppf(0.975, df=gl)
    t_uni = stats.t.ppf(0.95, df=gl)
    print(f"{gl:3d} | {t_bi:.4f}   | {t_uni:.4f}")
```

```python
# Exemplo 3: Como a tabela T converge para a Z
print("CONVERGÊNCIA DA DISTRIBUIÇÃO T PARA A NORMAL")
print("=" * 55)
print("À medida que gl aumenta, t crítico → Z crítico")
print()

z_ref = stats.norm.ppf(0.975)
print(f"Z crítico (α=0.05 bi): {z_ref:.4f}")
print()
print(" gl   | t crítico | Diferença para Z")
print("-" * 40)
for gl in [1, 2, 5, 10, 20, 30, 50, 100, 1000]:
    t_crit = stats.t.ppf(0.975, df=gl)
    diff = t_crit - z_ref
    print(f"{gl:4d} | {t_crit:.4f}   | {diff:+.4f}")
```

```python
# Exemplo 4: Funcao de consulta a tabela T
def consultar_t(gl, alpha=0.05, bicaudal=True):
    """Retorna o valor critico t para dados gl e alpha."""
    if bicaudal:
        prob = 1 - alpha/2
        tipo = "bicaudal"
    else:
        prob = 1 - alpha
        tipo = "unicaudal"
    
    if gl == float('inf'):
        t_crit = stats.norm.ppf(prob)
    else:
        t_crit = stats.t.ppf(prob, df=gl)
    
    print(f"gl = {gl}, α = {alpha}, {tipo}")
    print(f"t crítico = {t_crit:+.4f}")
    print(f"P(T > |t|) = {alpha}")
    return t_crit

consultar_t(8, 0.05, bicaudal=True)
print()
consultar_t(15, 0.01, bicaudal=False)
print()
consultar_t(float('inf'), 0.05, bicaudal=True)
```

```python
# Exemplo 5: Encontrando o p-valor a partir da tabela T
# Suponha que calculamos t = 2.5 com gl = 12

t_obs = 2.5
gl = 12

# p-valor bicaudal
p_bi = 2 * (1 - stats.t.cdf(abs(t_obs), df=gl))

print(f"Estatística t = {t_obs}")
print(f"Graus de liberdade = {gl}")
print()
print(f"p-valor (bicaudal) = {p_bi:.4f}")
print(f"p-valor (unicaudal dir.) = {1 - stats.t.cdf(t_obs, df=gl):.4f}")
print(f"p-valor (unicaudal esq.) = {stats.t.cdf(t_obs, df=gl):.4f}")
print()

# Verificando na tabela
print("Verificação na Tabela T:")
t_crit_005 = stats.t.ppf(0.975, df=gl)
t_crit_001 = stats.t.ppf(0.995, df=gl)
print(f"  t crítico (α=0.05, bi) = {t_crit_005:.4f}")
print(f"  t crítico (α=0.01, bi) = {t_crit_001:.4f}")
if abs(t_obs) > t_crit_001:
    print(f"  {abs(t_obs):.2f} > {t_crit_001:.4f} → p < 0.01")
elif abs(t_obs) > t_crit_005:
    print(f"  {abs(t_obs):.2f} > {t_crit_005:.4f} → p < 0.05")
else:
    print(f"  {abs(t_obs):.2f} < {t_crit_005:.4f} → p > 0.05")
```

```python
# Exemplo 6: Visualizando a convergencia da tabela T
import matplotlib.pyplot as plt

gls = list(range(1, 50))
t_valores = [stats.t.ppf(0.975, df=gl) for gl in gls]
z_linha = stats.norm.ppf(0.975)

plt.figure(figsize=(10, 6))
plt.plot(gls, t_valores, 'b-', lw=2, label='t crítico (α=0.05, bi)')
plt.axhline(z_linha, color='r', linestyle='--', lw=2, label=f'Z crítico = {z_linha:.4f}')
plt.title('Convergência do t crítico para o Z crítico')
plt.xlabel('Graus de Liberdade (gl)')
plt.ylabel('Valor Crítico')
plt.legend()
plt.grid(alpha=0.3)
plt.annotate(f'gl=30: t={stats.t.ppf(0.975, df=30):.4f}',
             xy=(30, stats.t.ppf(0.975, df=30)),
             xytext=(25, stats.t.ppf(0.975, df=30) + 0.1),
             arrowprops=dict(arrowstyle='->'))
plt.show()
```

## 📌 Resumo

- A **Tabela T** fornece valores críticos da distribuição t de Student para diferentes gl e α.
- É usada principalmente em testes T para média quando σ é desconhecido.
- A tabela é organizada com gl nas linhas e α nas colunas.
- Existem duas versões: **unicaudal** e **bicaudal**.
- Para gl → ∞, a tabela T converge para a tabela Z (normal).
- Geramos a tabela completa com `stats.t.ppf()`.

## 💡 Curiosidades

Ao contrário da tabela Z (que é sempre a mesma), a tabela T tem infinitas variações — uma para cada grau de liberdade. Por isso as tabelas T impressas geralmente listam apenas os valores críticos para alguns α comuns, e não a função de distribuição completa. A última linha (gl = ∞) da tabela T é idêntica à tabela Z, confirmando a convergência.

## ✅ Boas Práticas

1. Sempre verifique se está usando a coluna correta: unicaudal ou bicaudal.
2. Lembre-se: a linha gl = ∞ na tabela T equivale à tabela Z.
3. Para gl não listados, arredonde para baixo (mais conservador) ou interpole.
4. Use `stats.t.ppf()` para valores exatos em vez de aproximações da tabela.
5. Familiarize-se com os valores típicos para gl comuns (10, 20, 30).

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Confundir unicaudal com bicaudal | Identifique claramente H₁ antes de consultar a tabela |
| Usar gl errado | gl = n - 1 (sempre o número de observações menos 1) |
| Usar a tabela Z para a distribuição t | Use a tabela T quando σ é desconhecido |
| Interpolar linearmente a tabela T | A relação não é linear; prefira o cálculo exato com scipy |
