## Carteira de Identidade vs Selfie

Pense na sua **Carteira de Identidade** (RG).
Lah tem sua **altura oficial**: 1,70 m, por exemplo.
Essa e a medida "verdadeira" — o **parametro**.

Agora pense numa **selfie** que voce tira hoje.
Na foto, voce parece ter 1,68 m. Ou 1,72 m. Depende do angulo,
da luz, do chao, do sapato.
Essa e a **estimativa** — a **estatistica** amostral.

O **parametro (μ)** e fixo: sua altura real nao muda.
A **estatistica (x̄)** varia: cada selfie (amostra) da um resultado diferente.
O objetivo da estatistica e chegar o mais perto possivel do parametro.


## 📊 Traduzindo

| Carteira de Identidade vs Selfie | Estatistica |
|-|-|
| Altura oficial no RG (1,70 m) | Parametro (μ) — valor verdadeiro, fixo |
| Selfie de hoje (1,68 m) | Estatistica (x̄) — estimativa que varia |
| Outra selfie amanha (1,72 m) | Outra amostra — estimativa diferente |
| Angulo da foto | Erro amostral — distancia entre x̄ e μ |
| Tirar muitas selfies | Distribuicao amostral — varias estimativas |


## 📏 O Parametro e Fixo, a Estimativa Varia

Sua altura real nao mudou entre hoje e amanha.
Mas as selfies (estimativas) podem dar resultados diferentes.

Isso e fundamental em estatistica:
- O **parametro (μ)** e um valor fixo (mas desconhecido)
- A **estatistica (x̄)** varia de amostra para amostra
- Por isso usamos intervalos de confianca — uma faixa de selfies


```python
# O RG vs as selfies em Python
import random, statistics as stat
random.seed(42)

# O "RG" (parametro verdadeiro)
altura_real = 1.70  # metros — fixo
print(f"Altura real (parametro μ): {altura_real:.2f}m")

# Varias "selfies" (amostras)
print("\nSelfies (estimativas x̄) de diferentes amostras:")
for i in range(5):
    amostra = [random.gauss(altura_real, 0.03) for _ in range(10)]
    media = stat.mean(amostra)
    erro = abs(media - altura_real)
    print(f"  Selfie {i+1}: {media:.3f}m (erro: {erro:.3f}m)")
```

## 📌 Resumo

| Carteira de Identidade vs Selfie | Estatistica |
|-|-|
| Altura no RG | Parametro μ (fixo e verdadeiro) |
| Selfie | Estatistica x̄ (varia com a amostra) |
| Angulo da foto | Erro amostral |
| Varias selfies | Distribuicao amostral |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
