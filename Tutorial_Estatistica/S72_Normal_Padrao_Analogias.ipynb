## GPS Estatistico

O **Z-score** e como um GPS para os dados.

Quando voce esta dirigindo, o GPS diz:
*"Voce esta a 2 km do destino."*

O Z-score diz:
*"Voce esta a 2 desvios padrao da media."*

Z = 0: voce esta exatamente na media.
Z = +1: voce esta 1 desvio acima da media.
Z = -2: voce esta 2 desvios abaixo da media.

O GPS padroniza distancias: nao importa se voce esta numa
cidade com ruas em km ou milhas — o GPS sempre mostra km.
O Z-score padroniza escalas: nao importa se sao notas,
alturas ou pesos — o Z-score sempre mostra "desvios da media".


## 📊 Traduzindo

| GPS Estatistico | Estatistica |
|-|-|
| GPS: "Voce esta a 2 km" | Z-score: "Voce esta a 2 desvios" |
| GPS padroniza km | Z-score padroniza qualquer escala |
| Destino | Media dos dados |
| 2 km para o destino | 2 desvios padrao da media |


## 📏 Para Que Serve o Z-score

Comparar coisas em escalas diferentes:
- Joao tem 1,85 m (altura) e tira 8 na prova (nota 0-10).
  O que e mais "impressionante"? O Z-score responde!

Identificar outliers:
- Z > 3 ou Z < -3: valor muito extremo

Calcular probabilidades:
- Com Z, podemos usar a tabela Z para saber a chance
  de um valor ocorrer.


```python
# GPS estatistico
import statistics as stat
alturas = [1.65, 1.70, 1.68, 1.72, 1.85, 1.66, 1.69, 1.71, 1.67, 1.70]
media = stat.mean(alturas)
desvio = stat.stdev(alturas)
z_joao = (1.85 - media) / desvio
print(f"Altura do Joao: 1.85 m")
print(f"Media: {media:.3f} | Desvio: {desvio:.3f}")
print(f"Z-score do Joao: {z_joao:.2f}")
print(f"Significa: Joao esta {z_joao:.2f} desvios ACIMA da media")
```

## 📌 Resumo

| GPS Estatistico | Estatistica |
|-|-|
| GPS: km ate o destino | Z-score: desvios ate a media |
| GPS padroniza | Z-score padroniza |
| Comparar cidades | Comparar escalas diferentes |
