## Escada da Medicao

Imagine uma escada com 4 degraus.
Cada degrau e um nivel de medicao.

**1o degrau — Nominal**: nomes, rotulos.
Ex: times de futebol (Flamengo, Corinthians, Sao Paulo).
So da para dizer se e igual ou diferente.

**2o degrau — Ordinal**: tem ordem, mas nao distancia.
Ex: colocacao no campeonato (1o, 2o, 3o).
Sabemos que 1o e melhor que 2o, mas nao quanto melhor.

**3o degrau — Intervalar**: ordem + distancia fixa.
Ex: temperatura em Celsius. 30C e 10C mais quente que 20C.
Mas 0C nao significa "sem temperatura".

**4o degrau — Razaão**: tudo acima + zero absoluto.
Ex: peso. 0 kg significa "sem peso".
Assim, 20 kg e o dobro de 10 kg (razao faz sentido).


## 📊 Traduzindo

| Escada da Medicao | Estatistica |
|-|-|
| Times (Flamengo, Corinthians) | Nominal — so identifica |
| Colocacao (1o, 2o, 3o) | Ordinal — ordem, sem distancia |
| Temperatura (20C, 30C) | Intervalar — distancia fixa, sem zero real |
| Peso (10 kg, 20 kg) | Razao — zero absoluto, razoes fazem sentido |


## 📏 Subir os Degraus

Cada degrau permite mais operacoes matematicas:
- Nominal: so = ou !=
- Ordinal: =, !=, <, >
- Intervalar: soma, subtracao
- Razao: todas as operacoes, incluindo multiplicacao

Quanto mais alto o degrau, mais informacao o dado carrega.


```python
# Niveis de medicao em Python
nominal = ["Flamengo", "Corinthians", "Sao Paulo"]
ordinal = [1, 2, 3]  # colocacao
intervalar = [20, 25, 30]  # Celsius
razao = [10, 15, 20]  # kg
print("Nominal: so identifica ->", nominal)
print(f"Ordinal: ordem -> {ordinal}, mas media nao faz sentido")
print(f"Intervalar: {intervalar}C, media={sum(intervalar)/len(intervalar):.0f}C (faz sentido)")
print(f"Razao: {razao}kg, 20kg e o dobro de 10kg (razao valida)")
```

| Temperatura | Intervalar |
| Peso | Razao |
