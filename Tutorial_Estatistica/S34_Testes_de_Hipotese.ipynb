## 📝 Introdução

Os testes de hipótese são um dos pilares da estatística inferencial. Eles nos permitem tomar decisões baseadas em dados, respondendo perguntas como: "Este novo medicamento é realmente eficaz?", "A média de notas desta turma é diferente da média nacional?" ou "A proporção de votos a favor mudou após a campanha?".

O processo envolve formular duas hipóteses concorrentes, coletar evidências (dados) e decidir qual delas é mais plausível, controlando a probabilidade de erro. Neste capítulo, exploraremos todos os conceitos fundamentais por trás dos testes de hipótese.

## 📚 Explicação Detalhada

### 1. Hipótese Nula (H₀) e Hipótese Alternativa (H₁)

A **hipótese nula (H₀)** representa o status quo, a afirmação que queremos testar. Geralmente afirma que não há diferença ou efeito. A **hipótese alternativa (H₁)** é o oposto, representando o que queremos provar.

- H₀: μ = 100 (a média populacional é 100)
- H₁: μ ≠ 100 (a média populacional é diferente de 100) — teste bicaudal
- H₁: μ > 100 (teste unilateral à direita)
- H₁: μ < 100 (teste unilateral à esquerda)

### 2. Nível de Significância (α) e Erros Tipo I e II

O **nível de significância (α)** é a probabilidade máxima de rejeitarmos H₀ quando ela é verdadeira (erro tipo I). Valores comuns: 0.05 (5%), 0.01 (1%), 0.10 (10%).

| Decisão \ Realidade | H₀ Verdadeira | H₁ Verdadeira |
|---------------------|--------------|--------------|
| Não rejeitar H₀ | ✅ Decisão correta (1-α) | ❌ Erro Tipo II (β) |
| Rejeitar H₀ | ❌ Erro Tipo I (α) | ✅ Decisão correta (1-β = Potência) |

### 3. Estatística de Teste, p-valor e Região de Rejeição

A **estatística de teste** é um valor calculado a partir dos dados amostrais que usamos para tomar a decisão. O **p-valor** é a probabilidade de observarmos um valor tão extremo quanto o obtido, assumindo que H₀ é verdadeira. Se p-valor < α, rejeitamos H₀.

O **valor crítico** é o ponto que define a fronteira da região de rejeição. Se a estatística de teste cair na região de rejeição, rejeitamos H₀.

### 4. Potência do Teste (Power)

A **potência do teste** (1-β) é a probabilidade de rejeitarmos H₀ corretamente quando H₁ é verdadeira. Depende do tamanho amostral, do α e do tamanho do efeito. Testes com maior potência são mais desejáveis.

## 💻 Exemplos Práticos

```python
# Importando as bibliotecas necessárias
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Configurando semente para reprodutibilidade
np.random.seed(42)

print("Bibliotecas carregadas com sucesso!")
```

```python
# Exemplo 1: Teste Z para média (com desvio padrão conhecido)
# Uma fábrica afirma que suas garrafas têm 500ml. Coletamos uma amostra de 30 garrafas.
# Desvio padrão populacional conhecido: 10ml

# Parâmetros
mu_0 = 500.0        # H0: media = 500
sigma = 10.0        # desvio padrao populacional
n = 30              # tamanho da amostra
alpha = 0.05        # nivel de significancia

# Gerando amostra simulada
amostra = np.random.normal(loc=495, scale=sigma, size=n)
media_amostral = np.mean(amostra)

# Calculando estatistica Z
z_calculado = (media_amostral - mu_0) / (sigma / np.sqrt(n))

# Valor critico para alpha = 0.05 (bicaudal)
z_critico = stats.norm.ppf(1 - alpha / 2)

# p-valor
p_valor = 2 * (1 - stats.norm.cdf(abs(z_calculado)))

print(f"Média amostral: {media_amostral:.2f}")
print(f"H₀: μ = {mu_0}")
print(f"H₁: μ ≠ {mu_0}")
print(f"Estatística Z: {z_calculado:.4f}")
print(f"Valor crítico: ±{z_critico:.4f}")
print(f"p-valor: {p_valor:.4f}")
print(f"α = {alpha}")
print()
if abs(z_calculado) > z_critico:
    print("✅ Rejeitamos H₀ - Há evidência de que a média é diferente de 500ml.")
else:
    print("❌ Não rejeitamos H₀ - Não há evidência suficiente para afirmar diferença.")
```

```python
# Visualizando a regiao de rejeicao
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Distribuição Normal Padrão')

# Regiao de rejeicao (caudas)
x_fill_left = np.linspace(-4, -z_critico, 100)
x_fill_right = np.linspace(z_critico, 4, 100)
plt.fill_between(x_fill_left, stats.norm.pdf(x_fill_left), color='red', alpha=0.3, label='Região de Rejeição')
plt.fill_between(x_fill_right, stats.norm.pdf(x_fill_right), color='red', alpha=0.3)

# Linha da estatistica de teste
plt.axvline(z_calculado, color='green', linestyle='--', lw=2, label=f'Z calculado = {z_calculado:.2f}')
plt.axvline(-z_critico, color='red', linestyle=':', lw=2, label=f'Z crítico = ±{z_critico:.2f}')
plt.axvline(z_critico, color='red', linestyle=':', lw=2)

plt.title('Teste de Hipótese Bicaudal - Região de Rejeição')
plt.xlabel('Z')
plt.ylabel('Densidade')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Exemplo 2: Teste T (desvio padrao desconhecido)
# Uma escola afirma que a nota media de seus alunos é 70.
# Coletamos 20 notas. Nao conhecemos o desvio padrao populacional.

notas = np.array([65, 72, 68, 75, 71, 69, 73, 70, 74, 66,
                  67, 71, 70, 68, 72, 69, 73, 75, 70, 68])
mu_0 = 70.0

# Teste T com scipy
t_stat, p_valor_t = stats.ttest_1samp(notas, mu_0)

print(f"Média amostral: {np.mean(notas):.2f}")
print(f"H₀: μ = {mu_0}")
print(f"H₁: μ ≠ {mu_0}")
print(f"Estatística t: {t_stat:.4f}")
print(f"p-valor: {p_valor_t:.4f}")
print(f"α = 0.05")
if p_valor_t < 0.05:
    print("✅ Rejeitamos H₀")
else:
    print("❌ Não rejeitamos H₀")
```

```python
# Exemplo 3: Interpretacao do p-valor passo a passo
# Simulando 10000 amostras sob H0 para visualizar o p-valor

np.random.seed(123)
n_sim = 10000
n_amostra = 50

# Gerando amostras sob H0 (media = 100, dp = 15)
medias_simuladas = np.zeros(n_sim)
for i in range(n_sim):
    amostra_sim = np.random.normal(loc=100, scale=15, size=n_amostra)
    medias_simuladas[i] = np.mean(amostra_sim)

# Nossa amostra observada
amostra_real = np.random.normal(loc=105, scale=15, size=n_amostra)
media_observada = np.mean(amostra_real)

# p-valor: proporcao de medias simuladas mais extremas que a observada
p_valor_sim = np.mean(medias_simuladas >= media_observada)

print(f"Média observada: {media_observada:.2f}")
print(f"Média das simulações sob H₀: {np.mean(medias_simuladas):.2f}")
print(f"p-valor (simulação): {p_valor_sim:.4f}")
print(f"Interpretação: Se H₀ fosse verdadeira, a probabilidade de observar")
print(f"uma média >= {media_observada:.2f} seria de {p_valor_sim:.2%}.")
print()
if p_valor_sim < 0.05:
    print("Como p-valor < 0.05, rejeitamos H₀.")
else:
    print("Como p-valor >= 0.05, não rejeitamos H₀.")
```

```python
# Exemplo 4: Calculo da potencia do teste
from scipy.stats import norm

def calcular_potencia(mu_real, mu_0, sigma, n, alpha=0.05):
    """Calcula a potencia de um teste Z bicaudal."""
    z_crit = norm.ppf(1 - alpha / 2)
    erro_padrao = sigma / np.sqrt(n)
    z_real = (mu_real - mu_0) / erro_padrao
    potencia = (1 - norm.cdf(z_crit - z_real) + norm.cdf(-z_crit - z_real))
    return potencia

# Testando diferentes tamanhos amostrais
tamanhos = [10, 20, 30, 50, 100, 200]
print("Tamanho vs Potência (para μ_real - μ₀ = 5, σ = 10):")
for n in tamanhos:
    pot = calcular_potencia(mu_real=105, mu_0=100, sigma=10, n=n)
    print(f"  n = {n:3d} → Potência = {pot:.3f}")
print()
print("Quanto maior o n, maior a potência do teste!")
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Um fabricante de pilhas afirma que suas pilhas duram em média 100 horas. Uma amostra de 36 pilhas apresentou média de 97 horas. Sabendo que o desvio padrão populacional é 8 horas, teste a hipótese a 5% de significância.

```python
# Solucao do Exercicio 1
mu_0 = 100          # H0: media = 100
media_obs = 97      # media observada
sigma = 8           # desvio padrao
n = 36              # tamanho amostral
alpha = 0.05        # significancia

# Estatistica de teste
z = (media_obs - mu_0) / (sigma / np.sqrt(n))

# Valor critico (bicaudal)
z_crit = norm.ppf(1 - alpha/2)

# p-valor
p = 2 * norm.cdf(z)  # z negativo = cauda esquerda * 2

print(f"H₀: μ = {mu_0} | H₁: μ ≠ {mu_0}")
print(f"Z calculado: {z:.4f}")
print(f"Valor crítico: ±{z_crit:.4f}")
print(f"p-valor: {p:.4f}")
if abs(z) > z_crit:
    print("✅ Rejeitamos H₀: as pilhas duram menos que o anunciado.")
else:
    print("❌ Não rejeitamos H₀: não há evidência suficiente.")
```

**Exercício 2:** Uma turma de 25 alunos obteve média 73 em uma prova. A média nacional é 70 com desvio padrão 10. A turma é significativamente diferente da média nacional? Use α = 0.01.

```python
# Solucao do Exercicio 2
mu_0 = 70
media_obs = 73
sigma = 10
n = 25
alpha = 0.01

z = (media_obs - mu_0) / (sigma / np.sqrt(n))
z_crit = norm.ppf(1 - alpha/2)
p = 2 * (1 - norm.cdf(abs(z)))

print(f"H₀: μ = {mu_0} | H₁: μ ≠ {mu_0}")
print(f"Z calculado: {z:.4f}")
print(f"Valor crítico: ±{z_crit:.4f}")
print(f"p-valor: {p:.4f}")
if abs(z) > z_crit or p < alpha:
    print("✅ Rejeitamos H₀: a turma é significativamente diferente.")
else:
    print("❌ Não rejeitamos H₀.")
```

**Exercício 3:** Explique a diferença entre erro tipo I e erro tipo II, dando um exemplo concreto na área médica.

**Solução:** Erro tipo I (α) é rejeitar H₀ quando ela é verdadeira — um falso positivo. Exemplo: um teste diz que o paciente tem uma doença quando ele não tem.

Erro tipo II (β) é não rejeitar H₀ quando ela é falsa — um falso negativo. Exemplo: o teste diz que o paciente está saudável quando ele tem a doença.

Em geral, reduzir α aumenta β e vice-versa, a menos que aumentemos o tamanho amostral.

## 📌 Resumo

- **Teste de hipótese**: método para tomar decisões com base em dados amostrais.
- **H₀ (nula)**: afirmação padrão; **H₁ (alternativa)**: o que queremos provar.
- **Erro tipo I (α)**: rejeitar H₀ quando verdadeira.
- **Erro tipo II (β)**: não rejeitar H₀ quando falsa.
- **p-valor**: probabilidade dos dados sob H₀; se p < α, rejeitamos H₀.
- **Valor crítico**: ponto de corte da região de rejeição.
- **Potência (1-β)**: probabilidade de detectar um efeito real.
- Python/scipy fornecem funções prontas para testes de hipótese.

## 💡 Curiosidades

O conceito de teste de hipótese foi formalizado por **Ronald Fisher**, **Jerzy Neyman** e **Egon Pearson** na primeira metade do século XX. Fisher introduziu o p-valor como uma medida de evidência contra H₀, enquanto Neyman e Pearson propuseram a abordagem de aceitação/rejeição com α fixo. O famoso nível α = 0.05 foi uma escolha arbitrária de Fisher, que se tornou padrão na literatura científica — embora recentemente muitos pesquisadores defendam padrões mais rigorosos.

## ✅ Boas Práticas

1. Defina H₀ e H₁ **antes** de olhar os dados.
2. Escolha α com base no contexto (mais rigoroso em áreas críticas como medicina).
3. Reporte sempre o p-valor, não apenas "significante" ou "não significante".
4. Não confunda "não rejeitar H₀" com "provar H₀".
5. Considere o tamanho do efeito, não apenas o p-valor.
6. Use correção para múltiplos testes quando aplicável.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Interpretar p-valor como probabilidade de H₀ ser verdadeira | p-valor é P(dados|H₀), não P(H₀|dados) |
| Usar teste bicaudal quando deveria ser unilateral | Escolha com base na pergunta de pesquisa |
| Ignorar a suposição de normalidade | Verifique a distribuição dos dados antes do teste |
| p-hacking (testar várias hipóteses até achar significância) | Pré-registre suas hipóteses e corrija para testes múltiplos |
| Confundir significância estatística com significância prática | Um efeito pode ser estatisticamente significante mas irrelevante na prática |
