## Moeda Viciada

Voce suspeita que uma moeda e **viciada**.
Uma moeda honesta tem 50% de chance de dar cara.

Voce joga a moeda 100 vezes e obtem **65 caras**.

E uma moeda honesta (H0) ou viciada (H1)?
65% e muito distante de 50% para ser so "azar"?

O **teste Z de proporcao** responde:
se H0 fosse verdadeira (p = 0,5),
qual a chance de observar 65 ou mais caras?

Se essa chance (p-valor) for menor que 5%,
rejeitamos H0: a moeda e viciada.


## 📊 Traduzindo

| Moeda Viciada | Estatistica |
|-|-|
| Moeda honesta (50% cara) | H0: proporcao = 0,5 |
| Moeda viciada | H1: proporcao 0,5 |
| Jogar 100 vezes, 65 caras | Amostra: n=100, p=0,65 |
| Chance de 65 caras ser sorte | p-valor do teste |


```python
# Teste da moeda viciada
import math
n = 100
p_obs = 65 / n  # proporcao observada
p0 = 0.5  # H0
erro = math.sqrt(p0 * (1-p0) / n)
z = (p_obs - p0) / erro
print(f"Proporcao observada: {p_obs*100:.0f}%")
print(f"H0: proporcao = {p0*100:.0f}%")
print(f"Estatistica Z: {z:.2f}")
print(f"Interpretacao: {abs(z):.2f} desvios acima do esperado")
```

## 📌 Resumo

| Moeda Viciada | Estatistica |
|-|-|
| Moeda honesta | H0: p = 0,5 |
| Moeda viciada | H1: p 0,5 |
| 65 caras em 100 | Proporcao observada |
| Chance de ser sorte | p-valor |
