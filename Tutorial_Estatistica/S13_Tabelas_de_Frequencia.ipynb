## 📝 Introdução

As **tabelas de frequência** são uma das ferramentas mais fundamentais da estatística descritiva. Elas organizam os dados em categorias ou intervalos, mostrando quantas vezes cada valor (ou grupo de valores) aparece.

Imagine que você tem as notas de 50 alunos. Em vez de listar todas individualmente, você pode agrupá-las em faixas e contar quantos alunos estão em cada faixa. Isso é uma **tabela de frequência**!

## 📚 Explicação Detalhada

### 1. Frequência Absoluta

Contagem de quantas vezes cada valor aparece.

### 2. Frequência Relativa

Proporção (ou porcentagem) que cada valor representa no total.

### 3. Frequência Acumulada

Soma das frequências de todos os valores até aquele ponto.

### 4. Dados Agrupados em Intervalos (Bins)

Quando temos muitos valores distintos ou dados contínuos, agrupamos em intervalos (bins).

### 5. Tabelas de Contingência

Mostram a frequência conjunta de duas variáveis categóricas.

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
print('Bibliotecas importadas!')
```

```python
# Criando dataset de exemplo
notas = np.random.normal(loc=65, scale=15, size=50).round(1)
notas = np.clip(notas, 0, 100)

generos = np.random.choice(['Masculino', 'Feminino'], size=50, p=[0.48, 0.52])
disciplinas = np.random.choice(['Matemática', 'Português', 'História', 'Ciências', 'Artes'], size=50)

df = pd.DataFrame({
    'Aluno': [f'Aluno_{i+1:02d}' for i in range(50)],
    'Nota': notas,
    'Genero': generos,
    'Disciplina': disciplinas,
    'Aprovado': ['Sim' if n >= 60 else 'Não' for n in notas]
})
print('Dataset criado!')
print(df.head(10))
```

```python
# Tabela de frequência absoluta e relativa
print('=== Frequência Absoluta: Disciplina Favorita ===')
freq_abs = df['Disciplina'].value_counts()
print(freq_abs)

print('\n=== Frequência Relativa: Disciplina (%) ===')
freq_rel = df['Disciplina'].value_counts(normalize=True) * 100
print(freq_rel.round(1))
```

```python
# Tabela de frequência completa
freq_tabela = df['Disciplina'].value_counts().reset_index()
freq_tabela.columns = ['Disciplina', 'Freq. Absoluta']
freq_tabela['Freq. Relativa (%)'] = (df['Disciplina'].value_counts(normalize=True) * 100).round(1).values
freq_tabela['Freq. Acumulada'] = freq_tabela['Freq. Absoluta'].cumsum()
freq_tabela['Freq. Rel. Acumulada (%)'] = freq_tabela['Freq. Relativa (%)'].cumsum().round(1)
print(freq_tabela.to_string(index=False))
```

```python
# Dados agrupados em faixas (bins)
bins = [0, 20, 40, 60, 80, 100]
rotulos = ['0-20', '20-40', '40-60', '60-80', '80-100']

df['Faixa_Nota'] = pd.cut(df['Nota'], bins=bins, labels=rotulos, right=False)
freq_faixas = df['Faixa_Nota'].value_counts().sort_index().reset_index()
freq_faixas.columns = ['Faixa', 'Freq. Absoluta']
freq_faixas['Freq. Relativa (%)'] = (freq_faixas['Freq. Absoluta'] / len(df) * 100).round(1)
freq_faixas['Freq. Acumulada'] = freq_faixas['Freq. Absoluta'].cumsum()
print(freq_faixas.to_string(index=False))
```

```python
# Usando numpy.histogram
contagens, limites = np.histogram(df['Nota'], bins=5, range=(0, 100))
print('Limites:', limites)
print('Contagens:', contagens)

plt.figure(figsize=(8, 4))
plt.bar(range(len(contagens)), contagens, width=0.8, color='royalblue', edgecolor='black')
plt.xticks(range(len(contagens)), [f'{limites[i]:.0f}-{limites[i+1]:.0f}' for i in range(len(contagens))])
plt.title('Distribuição de Frequência das Notas')
plt.xlabel('Faixas')
plt.ylabel('Frequência Absoluta')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Tabela de contingência (crosstab)
print('=== Gênero × Aprovação ===')
print(pd.crosstab(df['Genero'], df['Aprovado']))

print('\n=== Com margens ===')
print(pd.crosstab(df['Genero'], df['Aprovado'], margins=True, margins_name='Total'))

print('\n=== Percentuais por linha ===')
print((pd.crosstab(df['Genero'], df['Aprovado'], normalize='index') * 100).round(1))

print('\n=== Disciplina × Aprovação ===')
print(pd.crosstab(df['Disciplina'], df['Aprovado'], margins=True, margins_name='Total'))
```

```python
# Heatmap da tabela de contingência
plt.figure(figsize=(8, 5))
ct = pd.crosstab(df['Disciplina'], df['Aprovado'])
plt.imshow(ct.values, cmap='Blues', aspect='auto')
plt.xticks(range(len(ct.columns)), ct.columns)
plt.yticks(range(len(ct.index)), ct.index)
plt.title('Tabela de Contingência — Disciplina × Aprovação')
for i in range(len(ct.index)):
    for j in range(len(ct.columns)):
        plt.text(j, i, ct.values[i, j], ha='center', va='center', fontsize=12, fontweight='bold')
plt.colorbar(label='Frequência')
plt.show()
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Pesquisa de satisfação
np.random.seed(100)
satisfacao = np.random.choice(
    ['Muito Insatisfeito', 'Insatisfeito', 'Neutro', 'Satisfeito', 'Muito Satisfeito'],
    size=200, p=[0.05, 0.10, 0.20, 0.45, 0.20]
)
ordem = ['Muito Insatisfeito', 'Insatisfeito', 'Neutro', 'Satisfeito', 'Muito Satisfeito']
freq = pd.Series(satisfacao).value_counts().reindex(ordem)
print('=== Pesquisa de Satisfação ===')
for cat in ordem:
    c = freq[cat]
    p = c / len(satisfacao) * 100
    bar = '█' * int(p / 2)
    print(f'{cat:25s}: {c:3d} ({p:5.1f}%) {bar}')
```

```python
# Exemplo 2: Alturas agrupadas
np.random.seed(200)
alturas = np.random.normal(loc=170, scale=10, size=100).round(1)
bins_alt = [140, 150, 160, 170, 180, 190, 200]
rot_alt = ['140-150', '150-160', '160-170', '170-180', '180-190', '190-200']
faixas_alt = pd.cut(alturas, bins=bins_alt, labels=rot_alt, right=False)
tabela_alt = faixas_alt.value_counts().sort_index().reset_index()
tabela_alt.columns = ['Altura (cm)', 'Frequência']
tabela_alt['Porcentagem'] = (tabela_alt['Frequência'] / len(alturas) * 100).round(1)
print('=== Distribuição de Alturas ===')
print(tabela_alt.to_string(index=False))
print(f'Média: {np.mean(alturas):.1f} cm')
print(f'Desvio Padrão: {np.std(alturas):.1f} cm')
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Crie uma tabela de frequência completa para as cores de carros vendidos: ['Branco', 'Preto', 'Prata', 'Branco', 'Vermelho', 'Preto', 'Branco', 'Azul', 'Prata', 'Branco', 'Preto', 'Branco', 'Vermelho', 'Prata', 'Preto'].

```python
# Solução do Exercício 1
cores = ['Branco', 'Preto', 'Prata', 'Branco', 'Vermelho', 'Preto', 'Branco',
         'Azul', 'Prata', 'Branco', 'Preto', 'Branco', 'Vermelho', 'Prata', 'Preto']
cs = pd.Series(cores)
tab = cs.value_counts().reset_index()
tab.columns = ['Cor', 'Freq. Absoluta']
tab = tab.sort_values('Cor')
tab['Freq. Relativa (%)'] = (tab['Freq. Absoluta'] / len(cores) * 100).round(1)
tab['Freq. Acumulada'] = tab['Freq. Absoluta'].cumsum()
tab['Freq. Rel. Acum. (%)'] = tab['Freq. Relativa (%)'].cumsum().round(1)
print(tab.to_string(index=False))
```

**Exercício 2:** Gere 30 idades aleatórias entre 18 e 70 anos. Agrupe em faixas (18-30, 30-40, 40-50, 50-70) e crie uma tabela de frequência.

```python
# Solução do Exercício 2
np.random.seed(300)
idades = np.random.randint(18, 71, 30)
bins_idade = [18, 30, 40, 50, 70]
rot_idade = ['18-30', '30-40', '40-50', '50-70']
faixas = pd.cut(idades, bins=bins_idade, labels=rot_idade, right=False)
tab = faixas.value_counts().sort_index().reset_index()
tab.columns = ['Faixa Etária', 'Frequência']
tab['Porcentagem (%)'] = (tab['Frequência'] / len(idades) * 100).round(1)
print('Idades:', sorted(idades))
print(tab.to_string(index=False))
```

## 📌 Resumo

- **Tabelas de frequência** organizam dados em categorias ou intervalos.
- **Frequência absoluta**: contagem. **Frequência relativa**: proporção.
- **Frequência acumulada**: soma progressiva.
- **Dados agrupados (bins)**: para dados contínuos.
- **Tabelas de contingência (crosstab)**: frequência conjunta de duas variáveis.
- Principais funções: `value_counts()`, `crosstab()`, `cut()`, `np.histogram()`.

## 💡 Curiosidades

**John Graunt** (1620-1674) é considerado o pai da estatística descritiva. Ele publicou análises de registros de nascimentos e mortes em Londres, percebendo padrões demográficos muito antes da computação. O termo **"histograma"** foi cunhado por **Karl Pearson** em 1891, do grego *histos* (tecido).

## ✅ Boas Práticas

1. **Escolha bins adequados** — poucos escondem detalhes, muitos criam ruído.
2. **Inclua frequência relativa** — permite comparar conjuntos de tamanhos diferentes.
3. **Use crosstab para explorar relações** entre variáveis categóricas.
4. **Ordene categorias** de forma lógica.
5. **Documente os bins** escolhidos.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Bins de tamanhos diferentes sem justificativa | Prefira bins iguais a menos que haja motivo teórico. |
| Muitas categorias na tabela | Agrupe categorias semelhantes. |
| Esquecer ordenação ordinal | Use reindex() para manter ordem lógica. |
| Interpretar causalidade em crosstab | Mostra associação, não causalidade. |
