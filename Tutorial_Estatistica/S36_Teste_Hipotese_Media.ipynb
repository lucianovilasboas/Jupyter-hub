## 📝 Introdução

O teste T de Student para uma amostra (one-sample t-test) é usado quando queremos testar se a **média de uma população** é igual a um valor específico, mas não conhecemos o desvio padrão populacional. Esta é a situação mais comum na prática, pois raramente conhecemos o verdadeiro σ da população.

William Sealy Gosset, que publicou sob o pseudônimo "Student", desenvolveu a distribuição t para lidar com amostras pequenas onde a estimativa do desvio padrão populacional precisa ser feita a partir dos próprios dados.

## 📚 Explicação Detalhada

### 1. Quando Usar o Teste T para uma Amostra

Usamos o teste T quando:
- A variável de interesse é quantitativa contínua
- Não conhecemos o desvio padrão populacional (σ)
- A amostra é aleatória e independente
- A população tem distribuição aproximadamente normal (especialmente para n pequeno)

Para n ≥ 30, o teste T é robusto mesmo para distribuições não normais (pelo Teorema Central do Limite).

### 2. Estatística t e Graus de Liberdade

A estatística t é calculada como:

t = (x̄ - μ₀) / (s / √n)

Onde:
- x̄ é a média amostral
- μ₀ é a média sob H₀
- s é o desvio padrão amostral
- n é o tamanho amostral

Os **graus de liberdade (gl)** são n - 1.

### 3. Distribuição t de Student

A distribuição t é similar à normal padrão, mas com **caudas mais pesadas** (maior variabilidade). Quanto menores os graus de liberdade, mais largas as caudas. À medida que gl aumenta (n → ∞), a distribuição t se aproxima da normal padrão.

### 4. Teste T vs Teste Z

| Característica | Teste Z | Teste T |
|---------------|---------|---------|
| σ conhecido? | Sim | Não |
| Distribuição | Normal Padrão | T de Student |
| Tamanho amostral | Qualquer (se σ for conhecido) | Qualquer (ideal para n pequeno) |
| Uso no mundo real | Raro | Muito comum |

## 💻 Exemplos Práticos

```python
# Importando bibliotecas
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

print("Bibliotecas carregadas com sucesso!")
```

```python
# Exemplo 1: Teste T completo - calculo manual e automatico
# Um fabricante afirma que suas latas de refrigerante contem 350ml.
# Coletamos 15 latas e medimos o volume.

volumes = np.array([348, 352, 347, 350, 351, 349, 346, 353, 350, 348,
                    351, 347, 352, 349, 350])
mu_0 = 350.0
alpha = 0.05

# Calculo manual
n = len(volumes)
media_amostral = np.mean(volumes)
desvio_amostral = np.std(volumes, ddof=1)  # ddof=1 para desvio amostral
gl = n - 1
erro_padrao = desvio_amostral / np.sqrt(n)
t_manual = (media_amostral - mu_0) / erro_padrao
p_manual = 2 * (1 - stats.t.cdf(abs(t_manual), df=gl))
t_critico = stats.t.ppf(1 - alpha/2, df=gl)

print("=== Teste T Manual ===")
print(f"Volumes: {volumes}")
print(f"n = {n}, média = {media_amostral:.2f}, desvio = {desvio_amostral:.2f}")
print(f"H₀: μ = {mu_0} | H₁: μ ≠ {mu_0}")
print(f"t calculado = {t_manual:.4f}")
print(f"t crítico = ±{t_critico:.4f} (gl = {gl})")
print(f"p-valor = {p_manual:.4f}")
if abs(t_manual) > t_critico:
    print("✅ Rejeitamos H₀ (pela região crítica)")
else:
    print("❌ Não rejeitamos H₀ (pela região crítica)")
print()

# Usando scipy
t_scipy, p_scipy = stats.ttest_1samp(volumes, mu_0)
print("=== Teste T com scipy.stats.ttest_1samp ===")
print(f"t = {t_scipy:.4f}, p-valor = {p_scipy:.4f}")
if p_scipy < alpha:
    print("✅ Rejeitamos H₀")
else:
    print("❌ Não rejeitamos H₀")
```

```python
# Exemplo 2: Visualizando a distribuicao t vs normal
x = np.linspace(-4, 4, 1000)

plt.figure(figsize=(12, 6))
plt.plot(x, stats.norm.pdf(x, 0, 1), 'k-', lw=2, label='Normal Padrão')
plt.plot(x, stats.t.pdf(x, df=2), 'r--', lw=2, label='t (gl=2)')
plt.plot(x, stats.t.pdf(x, df=5), 'g--', lw=2, label='t (gl=5)')
plt.plot(x, stats.t.pdf(x, df=15), 'b--', lw=2, label='t (gl=15)')
plt.title('Distribuição t de Student vs Normal Padrão')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Exemplo 3: Teste T com n grande (aproximacao normal)
# Neste caso, t e Z serao muito proximos

np.random.seed(42)
n_grande = 100
amostra_grande = np.random.normal(loc=102, scale=15, size=n_grande)

t_stat, p_val = stats.ttest_1samp(amostra_grande, 100)

print(f"Teste T com n = {n_grande}")
print(f"Média amostral: {np.mean(amostra_grande):.2f}")
print(f"t = {t_stat:.4f}")
print(f"p-valor = {p_val:.4f}")
print()
print("Com n grande, a distribuição t ≈ normal padrão.")
print(f"Valor crítico t (α=0.05): ±{stats.t.ppf(0.975, df=n_grande-1):.4f}")
print(f"Valor crítico Z (α=0.05): ±{stats.norm.ppf(0.975):.4f}")
```

```python
# Exemplo 4: Interpretacao do p-valor com simulacao
# Simulando 10000 amostras sob H0 e comparando com nossa estatistica

np.random.seed(123)
n_sim = 10000
n = 15
mu_0 = 350
t_simulados = np.zeros(n_sim)

for i in range(n_sim):
    amostra_sim = np.random.normal(loc=mu_0, scale=2, size=n)
    t_simulados[i], _ = stats.ttest_1samp(amostra_sim, mu_0)

# Nossa estatistica observada
t_obs = t_scipy

plt.figure(figsize=(10, 6))
plt.hist(t_simulados, bins=40, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.axvline(t_obs, color='red', lw=3, label=f't observado = {t_obs:.2f}')
plt.axvline(-t_obs, color='red', lw=3, linestyle='--')
plt.title('Distribuição das estatísticas t sob H₀')
plt.xlabel('t')
plt.ylabel('Densidade')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Proporção de |t| mais extremos que {abs(t_obs):.2f}: {np.mean(abs(t_simulados) >= abs(t_obs)):.4f}")
print(f"p-valor (scipy): {p_scipy:.4f}")
```

```python
# Exemplo 5: Funcao completa de teste T para media
def teste_t_uma_amostra(dados, mu_0, alpha=0.05):
    """Realiza teste T para uma amostra."""
    n = len(dados)
    media = np.mean(dados)
    desvio = np.std(dados, ddof=1)
    gl = n - 1
    erro = desvio / np.sqrt(n)
    t_stat = (media - mu_0) / erro
    p_val = 2 * (1 - stats.t.cdf(abs(t_stat), df=gl))
    t_crit = stats.t.ppf(1 - alpha/2, df=gl)
    
    print(f"Resultado do Teste T para uma Amostra:")
    print(f"  n = {n}, média = {media:.3f}, desvio = {desvio:.3f}")
    print(f"  H₀: μ = {mu_0} | H₁: μ ≠ {mu_0}")
    print(f"  t = {t_stat:.4f} (gl = {gl})")
    print(f"  Valor crítico: ±{t_crit:.4f}")
    print(f"  p-valor = {p_val:.4f}")
    if abs(t_stat) > t_crit or p_val < alpha:
        print("  ✅ Rejeitamos H₀")
    else:
        print("  ❌ Não rejeitamos H₀")
    return {"t": t_stat, "p": p_val, "gl": gl}

teste_t_uma_amostra(volumes, 350)
print()
teste_t_uma_amostra(np.array([23, 25, 22, 24, 23, 26, 22, 25]), 24.5)
```

**Exercício 1:** Uma academia afirma que seus membros perdem em média 5kg nos primeiros 3 meses. Uma amostra de 12 membros mostrou perdas: [4.5, 5.2, 3.8, 6.1, 4.7, 5.5, 4.0, 5.8, 4.3, 5.1, 4.9, 5.6]. Teste a afirmação a 5%.

```python
# Solucao do Exercicio 1
perdas = np.array([4.5, 5.2, 3.8, 6.1, 4.7, 5.5, 4.0, 5.8, 4.3, 5.1, 4.9, 5.6])
t_stat, p_val = stats.ttest_1samp(perdas, 5.0)
print(f"Média: {np.mean(perdas):.2f}")
print(f"t = {t_stat:.4f}, p-valor = {p_val:.4f}")
if p_val < 0.05:
    print("✅ Rejeitamos H₀: a perda média é diferente de 5kg.")
else:
    print("❌ Não rejeitamos H₀: não há evidência de diferença.")
```

**Exercício 2:** Um professor acredita que a nota média de sua turma é superior a 70. As notas de 20 alunos são: [68, 75, 82, 71, 69, 78, 85, 73, 70, 77, 74, 81, 76, 72, 79, 83, 67, 80, 75, 74]. Teste a 5% (unilateral à direita).

```python
# Solucao do Exercicio 2 (unilateral a direita)
notas = np.array([68, 75, 82, 71, 69, 78, 85, 73, 70, 77,
                  74, 81, 76, 72, 79, 83, 67, 80, 75, 74])
t_stat, p_val = stats.ttest_1samp(notas, 70)
# O p-valor do ttest_1samp é bicaudal. Para unilateral à direita:
p_val_dir = p_val / 2 if t_stat > 0 else 1 - p_val / 2

print(f"Média: {np.mean(notas):.2f}")
print(f"H₀: μ ≤ 70 | H₁: μ > 70")
print(f"t = {t_stat:.4f}, p-valor (bicaudal) = {p_val:.4f}")
print(f"p-valor (unilateral dir.) = {p_val_dir:.4f}")
if p_val_dir < 0.05:
    print("✅ Rejeitamos H₀: a média é superior a 70.")
else:
    print("❌ Não rejeitamos H₀.")
```

**Exercício 3:** Uma dieta promete reduzir o colesterol em 20 mg/dL. Em 8 pacientes, a redução foi: [18, 22, 19, 21, 17, 23, 20, 22]. Teste a 1% se a redução média é diferente de 20.

```python
# Solucao do Exercicio 3
reducao = np.array([18, 22, 19, 21, 17, 23, 20, 22])
t_stat, p_val = stats.ttest_1samp(reducao, 20)
print(f"Redução média: {np.mean(reducao):.2f}")
print(f"t = {t_stat:.4f}, p-valor = {p_val:.4f}")
if p_val < 0.01:
    print("✅ Rejeitamos H₀ a 1%.")
else:
    print("❌ Não rejeitamos H₀ a 1%.")
```

1. Compare o teste T com o teste Z para a mesma amostra (use σ estimado vs σ conhecido). Quão diferentes são os resultados?

```python
# Desafio 1: Comparacao T vs Z
# Simule dados onde 'conhecemos' sigma e compare
np.random.seed(42)
n = 15
mu_real = 52
sigma_real = 10
dados = np.random.normal(loc=mu_real, scale=sigma_real, size=n)

# Teste Z (supondo sigma conhecido)
z = (np.mean(dados) - 50) / (sigma_real / np.sqrt(n))
p_z = 2 * (1 - stats.norm.cdf(abs(z)))

# Teste T (sigma estimado)
t, p_t = stats.ttest_1samp(dados, 50)

print(f"Teste Z: z = {z:.4f}, p = {p_z:.4f}")
print(f"Teste T: t = {t:.4f}, p = {p_t:.4f}")
print(f"Diferença no p-valor: {abs(p_z - p_t):.6f}")
```

## 📌 Resumo

- O **teste T para uma amostra** avalia hipóteses sobre a média populacional.
- Usamos t = (x̄ - μ₀) / (s / √n) com n-1 graus de liberdade.
- A distribuição t tem caudas mais pesadas que a normal, compensando a incerteza de estimar σ.
- Usamos **stats.ttest_1samp** do scipy para realizar o teste automaticamente.
- O teste T é preferível ao Z quando σ é desconhecido (quase sempre na prática).
- Para n grande, a distribuição t se aproxima da normal padrão.

## 💡 Curiosidades

William Sealy Gosset (1876-1937) trabalhava na cervejaria Guinness, na Irlanda, onde desenvolveu o teste t para monitorar a qualidade da cerveja com amostras pequenas. A Guinness não permitia que seus cientistas publicassem sob seus nomes reais, então Gosset usou o pseudônimo "Student". Por isso a distribuição se chama **t de Student**, e não "t de Gosset".

## ✅ Boas Práticas

1. Verifique a normalidade dos dados (especialmente para n < 30) com histogramas ou QQ-plots.
2. Use o teste T mesmo para amostras grandes — é mais conservador e correto quando σ é desconhecido.
3. Para testes unicaudais, lembre-se de ajustar o p-valor (dividir por 2 e verificar a direção).
4. Reporte o intervalo de confiança junto com o teste.
5. Considere usar testes não paramétricos (ex: Wilcoxon) se a normalidade for violada.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Usar teste Z quando σ é desconhecido | Sempre use teste T quando σ for estimado |
| Ignorar a suposição de normalidade | Use Q-Q plot ou teste de Shapiro-Wilk |
| Confundir p-valor bicaudal e unicaudal | Divida o p-valor por 2 para teste unicaudal |
| Usar n no denominador do desvio padrão | Use ddof=1 (desvio amostral, não populacional) |
| Esquecer de verificar outliers | Outliers podem distorcer a média e o teste |
