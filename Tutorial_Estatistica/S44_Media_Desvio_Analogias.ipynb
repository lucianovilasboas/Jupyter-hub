## 🚌 O Onibus que Nunca Vem na Hora

Imagine a seguinte cena: todos os dias voce espera o onibus no mesmo ponto.
Nunca sabe exatamente quando ele vai chegar.

Alguns dias:
- O onibus passa **rapido** (2 minutos de espera)
- Outros dias voce espera **20 minutos**
- Na maioria dos dias, algo **entre 5 e 15 minutos**

Essa experiencia cotidiana ja contem *todos* os conceitos fundamentais da estatistica!
Vamos descobrir como.


## 📊 Traduzindo a Analogia

| Situacao do Onibus | Conceito Estatistico | Simbolo |
|-------------------|---------------------|--------|
| **Todos os dias que o onibus ja passou** (desde que a linha existe) | **Populacao** (todos os dados possiveis) | N |
| **Os dias que voce cronometrou** esta semana | **Amostra** (um pedaco dos dados) | n |
| **A media REAL de espera** (se pudessemos medir todos os dias) | **Media populacional** | μ (mi) |
| **Sua estimativa** baseada nos dias que anotou | **Media amostral** | x̄ (x-barra) |
| **O quanto o tempo varia** de um dia para outro | **Desvio padrao** | σ (sigma) |
| **A confianca no seu palpite** (mais dias anotados = mais certeza) | **Erro padrao** | s / √n |

Vamos explorar cada um deles.


## 📏 Conceito 1: A Media — "Quanto tempo eu espero em media?"

Se alguem perguntar *"quanto tempo voce espera pelo onibus?"*,
voc provavelmente responde algo como *"uns 10 minutos"*.

Esse numero e a **media** — o centro dos seus dados.

### Calculando a media na pratica

Suponha que voce anotou o tempo de espera esta semana:

| Dia | Tempo (min) |
|-----|-------------|
| Segunda | 12 |
| Terca | 8 |
| Quarta | 15 |
| Quinta | 10 |
| Sexta | 5 |

Media = (12 + 8 + 15 + 10 + 5) / 5 = **50 / 5 = 10 minutos**

> **Media** = soma dos valores dividido pela quantidade de valores

Esse e o **ponto de equilibrio**: se o onibus sempre chegasse exatamente no mesmo tempo,
esse tempo seria a media. Mas na vida real, nunca e exato — e ai entra o desvio padrao.


```python
# Exemplo 1: Calculando a media dos tempos de espera
tempos = [12, 8, 15, 10, 5]  # minutos de espera
media = sum(tempos) / len(tempos)
print(f'Tempos de espera: {tempos}')
print(f'Media: {media:.1f} minutos')
print(f'Total de dias anotados (tamanho da amostra): {len(tempos)}')
```

## 📏 Conceito 2: O Desvio Padrao — "Por que nunca sei a hora certa?"

A media sozinha nao conta a historia completa.
Voce pode ter uma media de 10 minutos em duas situacoes muito diferentes:

| Cenario | Seg | Ter | Qua | Qui | Sex | Media |
|---------|-----|-----|-----|-----|-----|-------|
| **Linha confiavel** | 9 | 10 | 11 | 10 | 10 | 10 min |
| **Linha caotica** | 2 | 5 | 10 | 15 | 18 | 10 min |

Nos dois casos a media e 10 minutos. Mas a experiencia de quem espera e COMPLETAMENTE diferente!

**Desvio padrao pequeno (σ ≈ 0,7 min)**:
- O onibus passa sempre entre 9 e 11 minutos
- Voce chega no ponto e sabe que nao vai esperar muito
- Previsivel, confiavel

**Desvio padrao grande (σ ≈ 6,5 min)**:
- O onibus pode passar em 2 min ou em 18 min
- Voce nunca sabe se vai chegar atrasado
- Imprevisivel, estressante

> **Desvio padrao** = o quanto os valores individuais se afastam da media

Na pratica: **quanto maior o desvio padrao, menos voce consegue prever o que vai acontecer**.


```python
# Exemplo 2: Comparando duas linhas de onibus
import statistics as stat

confiavel = [9, 10, 11, 10, 10]   # desvio pequeno
caotica = [2, 5, 10, 15, 18]      # desvio grande

print('Linha CONFIAVEL:')
print(f'  Tempos: {confiavel}')
print(f'  Media: {stat.mean(confiavel):.1f} min')
print(f'  Desvio padrao: {stat.stdev(confiavel):.1f} min')
print(f'  Variacao: de {min(confiavel)} a {max(confiavel)} min')
print()
print('Linha CAOTICA:')
print(f'  Tempos: {caotica}')
print(f'  Media: {stat.mean(caotica):.1f} min')
print(f'  Desvio padrao: {stat.stdev(caotica):.1f} min')
print(f'  Variacao: de {min(caotica)} a {max(caotica)} min')
```

```python
# Exemplo 3: Visualizando a dispersao com um grafico simples
import matplotlib.pyplot as plt

confiavel = [9, 10, 11, 10, 10]
caotica = [2, 5, 10, 15, 18]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharey=True)

# Grafico 1: Linha confiavel
ax1.scatter(confiavel, [0]*len(confiavel), color='green', s=100, zorder=5)
ax1.axvline(x=stat.mean(confiavel), color='darkgreen', linestyle='--', linewidth=2, label=f'Media={stat.mean(confiavel):.0f}')
ax1.set_title('Linha Confiavel (desvio pequeno)')
ax1.set_xlabel('Tempo de espera (min)')
ax1.set_yticks([])
ax1.set_xlim(0, 20)
ax1.legend()
ax1.grid(axis='x', alpha=0.3)

# Grafico 2: Linha caotica
ax2.scatter(caotica, [0]*len(caotica), color='red', s=100, zorder=5)
ax2.axvline(x=stat.mean(caotica), color='darkred', linestyle='--', linewidth=2, label=f'Media={stat.mean(caotica):.0f}')
ax2.set_title('Linha Caotica (desvio grande)')
ax2.set_xlabel('Tempo de espera (min)')
ax2.set_yticks([])
ax2.set_xlim(0, 20)
ax2.legend()
ax2.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
print('Ambas tem media 10, mas a dispersao e completamente diferente!')
```

## 📏 Conceito 3: Populacao vs Amostra — "Quantos dias preciso anotar?"

Na vida real, voce **nunca** vai conseguir cronometrar *todos* os dias que o onibus ja passou.
Seria preciso viajar no tempo! 🕰️

**Populacao**: todos os dias, desde que a linha foi criada (milhares de dados)
**Amostra**: os dias que voce anotou esta semana (5 dados)

A pergunta crucial: **sua amostra representa bem a populacao?**

- Se voce anotou **apenas 3 dias**, seu palpite pode ser muito diferente da realidade
- Se voce anotou **30 dias** (quase um mes), seu palpite e bem mais confiavel

Na estatistica, isso se chama **lei dos grandes numeros**:
> Quanto maior a amostra, mais proxima a media amostral (x̄) fica da media verdadeira (μ)

E o **erro padrao** mede exatamente isso: o quanto sua estimativa pode estar errada.

Erro padrao = desvio padrao / √(tamanho da amostra)

Quanto maior n (dias anotados), menor o erro padrao (mais certeza).


```python
# Exemplo 4: Mais dados = melhor estimativa
import random
import statistics as stat
import math

# Simulando a populacao real (milhares de dias de onibus)
random.seed(42)
populacao = [random.gauss(12, 4) for _ in range(10000)]  # media real = 12, desvio = 4
media_real = stat.mean(populacao)
print(f'Media REAL da populacao (todos os dias): {media_real:.2f} min')
print()

# Testando amostras de diferentes tamanhos
for n in [3, 5, 10, 30, 100]:
    amostra = random.sample(populacao, n)
    media_amostral = stat.mean(amostra)
    erro = abs(media_amostral - media_real)
    desvio_amostral = stat.stdev(amostra)
    erro_padrao = desvio_amostral / math.sqrt(n)
    print(f'n={n:3d}  |  Media amostral = {media_amostral:6.2f}  |  Erro = {erro:.2f}  |  Erro padrao = {erro_padrao:.2f}')
```

## 📌 Resumo

| Conceito | Analogia do Onibus | Frase para lembrar |
|----------|-------------------|-------------------|
| **Media (μ/x̄)** | O tempo medio que voce espera | "O centro dos meus dados" |
| **Desvio padrao (σ)** | O quanto o tempo varia de um dia pro outro | "A confiabilidade da linha" |
| **Populacao (N)** | Todos os dias desde que a linha existe | "O conjunto completo" |
| **Amostra (n)** | Os dias que voce anotou | "Um pedaco representativo" |
| **Erro padrao** | Sua certeza sobre a estimativa | "Mais dados = mais certeza" |

A analogia do onibus nos ajuda a lembrar:

> A media sozinha nao conta a historia — e preciso saber o desvio padrao
> para entender se a linha e confiavel ou caotica.
> E quanto mais dias voce anotar, mais perto da verdade voce chega.
