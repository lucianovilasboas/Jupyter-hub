## 📝 Introdução

Saber apenas a **média** dos dados não é suficiente. Dois conjuntos podem ter a mesma média, mas serem completamente diferentes!

**Exemplo:**
- Turma A: `{5, 5, 5, 5, 5}` — média=5, variação=0
- Turma B: `{0, 0, 5, 10, 10}` — média=5, variação=alta

Ambas têm média 5, mas a realidade é totalmente diferente! A **variação** mede o **espalhamento** dos dados.

> 🧠 **Para refletir:** Um investimento com retorno médio de 10% ao ano é bom. Mas se a variação for de ±30%, você pode perder dinheiro. A variação é crucial para avaliar riscos!

## 📚 Explicação Detalhada

### 1. O que é Variação?

Variação (ou dispersão) é o grau de espalhamento dos dados em torno de uma medida central.

**Por que a variação é importante?**
- Dois conjuntos podem ter a mesma média mas naturezas diferentes
- A variação afeta a **confiabilidade** da média
- Fundamental para **inferência estatística**
- Mede **risco** em finanças e **precisão** em processos

### 2. Principais Medidas de Dispersão

| Medida | Símbolo | Descrição | Unidade |
|--------|---------|-----------|---------|
| **Range** | R | Max - Min | Mesma dos dados |
| **IQR** | IQR | Q3 - Q1 | Mesma dos dados |
| **Variância** | σ², s² | Média dos desvios² | Quadrado da unidade |
| **Desvio Padrão** | σ, s | Raiz da variância | Mesma dos dados |
| **Coef. de Variação** | CV | s/x̄ | Adimensional (%) |

### 3. Baixa vs Alta Variabilidade

**Baixa variabilidade:** dados homogêneos (ex: altura de jogadores de basquete)
**Alta variabilidade:** dados heterogêneos (ex: renda de uma população)

## 💻 Exemplos Práticos

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Exemplo 1: Mesma media, variacoes diferentes
turma_a = np.full(5, 5)
turma_b = np.array([0, 0, 5, 10, 10])
print("Turma A — Todos com nota 5:")
print(f"  Dados: {turma_a}")
print(f"  Media: {np.mean(turma_a):.1f}, DP: {np.std(turma_a):.1f}")
print()
print("Turma B — Notas extremas:")
print(f"  Dados: {turma_b}")
print(f"  Media: {np.mean(turma_b):.1f}, DP: {np.std(turma_b):.1f}")
print()
print("Mesma media, realidades COMPLETAMENTE diferentes!")
```

```python
# Exemplo 2: Visualizacao da variacao
baixa_var = np.random.normal(100, 5, 200)
alta_var = np.random.normal(100, 25, 200)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes[0,0].hist(baixa_var, bins=20, color="green", alpha=0.7, edgecolor="black")
axes[0,0].set_title("Baixa Variabilidade (std=5)")
axes[0,0].set_xlabel("Valor"); axes[0,0].set_ylabel("Frequencia")
axes[0,1].hist(alta_var, bins=20, color="red", alpha=0.7, edgecolor="black")
axes[0,1].set_title("Alta Variabilidade (std=25)")
axes[0,1].set_xlabel("Valor")
axes[1,0].boxplot(baixa_var, vert=True, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
axes[1,0].set_title("Box Plot — Baixa"); axes[1,0].set_ylabel("Valor")
axes[1,1].boxplot(alta_var, vert=True, patch_artist=True, boxprops=dict(facecolor="lightcoral"))
axes[1,1].set_title("Box Plot — Alta")
plt.tight_layout(); plt.show()
```

```python
# Exemplo 3: Todas as medidas de variacao
grupo1 = np.random.randint(18, 25, size=50)
grupo2 = np.random.randint(18, 65, size=50)
def medidas_variacao(dados, nome):
    print(f"\n{'='*40}\n{nome}\n{'='*40}")
    print(f"Range: {np.ptp(dados)}")
    print(f"IQR:   {np.percentile(dados,75)-np.percentile(dados,25):.1f}")
    print(f"Var:   {np.var(dados):.1f}")
    print(f"DP:    {np.std(dados):.1f}")
    print(f"CV:    {np.std(dados)/np.mean(dados)*100:.1f}%")
medidas_variacao(grupo1, "Grupo 1 — Idade: 18-25")
medidas_variacao(grupo2, "Grupo 2 — Idade: 18-65")
```

```python
# Exemplo 4: Variacao na qualidade de produtos
forn_a = np.random.normal(10, 0.1, 100)
forn_b = np.random.normal(10, 0.5, 100)
tol_inf, tol_sup = 9.8, 10.2
def_a = np.sum((forn_a < tol_inf) | (forn_a > tol_sup))
def_b = np.sum((forn_b < tol_inf) | (forn_b > tol_sup))
print("Fornecedor A — Baixa variacao (std=0.1mm):")
print(f"  Defeitos: {def_a}%")
print("Fornecedor B — Alta variacao (std=0.5mm):")
print(f"  Defeitos: {def_b}%")
print("A variacao determina a qualidade!")
```

```python
# Exemplo 5: Coeficiente de Variacao (CV)
alturas = np.random.normal(170, 10, 100)
pesos = np.random.normal(70, 8, 100)
cv_alt = np.std(alturas) / np.mean(alturas) * 100
cv_peso = np.std(pesos) / np.mean(pesos) * 100
print(f"Altura — Media: {np.mean(alturas):.1f}cm, DP: {np.std(alturas):.1f}cm, CV: {cv_alt:.1f}%")
print(f"Peso — Media: {np.mean(pesos):.1f}kg, DP: {np.std(pesos):.1f}kg, CV: {cv_peso:.1f}%")
print("CV permite comparar variabilidade de grandezas em unidades diferentes.")
```

## 📝 Resumo do Capítulo

- **Variação** mede o espalhamento dos dados
- Principais medidas: range, IQR, variância, desvio padrão, CV
- **Variância**: média dos quadrados dos desvios (unidade²)
- **Desvio Padrão**: raiz da variância (mesma unidade)
- **CV**: permite comparar variação entre escalas diferentes
- A variação é essencial para avaliar risco e qualidade

## 🧐 Curiosidades

- O termo **variância** foi introduzido por Ronald Fisher em 1918.
- Em finanças, o DP é chamado de **volatilidade**.
- O **Six Sigma** busca reduzir a variação para < 3.4 defeitos por milhão.

## ✅ Boas Práticas

- Sempre reporte **centro + variação** juntos
- Use **DP** para interpretação direta
- Use **CV** para comparar escalas diferentes
- O **IQR** é mais robusto que o range

## ⚠️ Erros Comuns

| Erro | Consequência | Correção |
|------|-------------|----------|
| Interpretar variância como DP | Unidade confusa | Use sqrt |
| Ignorar a variação | Análise incompleta | Calcule dispersão |
| Comparar DP de escalas diferentes | Inválido | Use CV |
| Confundir σ (pop) com s (amostra) | Viés | Use ddof=1 |
