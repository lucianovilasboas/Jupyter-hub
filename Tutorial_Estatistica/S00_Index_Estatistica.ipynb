## 📋 Apresentação

Bem-vindo ao **Curso Completo de Estatística**! Este material foi cuidadosamente elaborado para servir como um livro interativo de Estatística, adaptado para estudantes brasileiros, com exemplos práticos em Python.

A Estatística é a ciência que nos fornece métodos para coletar, organizar, analisar e interpretar dados. Ela está presente em praticamente todas as áreas do conhecimento: ciência de dados, medicina, economia, engenharia, psicologia, biologia e muito mais.

## 🎯 Objetivos

- Compreender os conceitos fundamentais da Estatística
- Dominar técnicas de estatística descritiva e inferencial
- Aprender a visualizar dados de forma eficaz
- Realizar testes de hipótese e estimações
- Aplicar estatística com Python (NumPy, SciPy, Matplotlib)

## 📋 Pré-requisitos

- **Computador** com Python 3 instalado (recomendado: 3.10 ou superior)
- **Conhecimento básico de Python** (variáveis, listas, funções)
- **Vontade de aprender** 💪

> **Nota:** Se você não conhece Python, recomendamos fazer o Curso de Python deste hub antes.
