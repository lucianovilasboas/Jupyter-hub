## 📝 Introdução

A **média aritmética** (ou simplesmente **mean**) é a medida de tendência central mais conhecida e utilizada. Quando alguém diz "a média" no dia a dia, está quase sempre se referindo à média aritmética.

Formalmente, a média aritmética de um conjunto de valores é a **soma de todos os valores dividida pela quantidade de valores**.

$$\bar{x} = \frac{1}{n}\sum_{i=1}^{n} x_i$$

Onde:
- $\bar{x}$ (lê-se "x barra") é a média da **amostra**
- $\mu$ (letra grega "mi") é a média da **população**
- $n$ é o número de elementos
- $\sum$ (sigma) significa "soma de"

Apesar de simples, a média tem propriedades importantes e limitações que precisamos conhecer para usá-la corretamente.

## 📚 Explicação Detalhada

### 1. Cálculo da Média Aritmética

O cálculo é simples:

$$\text{Mean} = \frac{\text{Soma dos valores}}{\text{Quantidade de valores}}$$

**Exemplo:** Para as notas `{7, 8, 9, 6, 10}`:
$$\bar{x} = \frac{7 + 8 + 9 + 6 + 10}{5} = \frac{40}{5} = 8$$

### 2. Notação: x̄ vs μ

| Símbolo | Nome | Usado para |
|---------|------|-----------|
| $\bar{x}$ | x barra | Média de uma **amostra** |
| $\mu$ | mi | Média de uma **população** |

Em estatística, usamos letras romanas ($\bar{x}$, $s$) para **amostras** e letras gregas ($\mu$, $\sigma$) para **populações**.

### 3. Propriedades da Média

1. **A soma dos desvios é zero:** $\sum(x_i - \bar{x}) = 0$
2. **A média é o ponto de equilíbrio** dos dados
3. **É sensível a outliers:** um valor extremo pode distorcer completamente a média
4. **É única:** para um conjunto de dados, existe apenas uma média
5. **Unidade de medida:** mesma unidade dos dados originais

### 4. Média Ponderada

Quando os valores têm **pesos** diferentes, usamos a média ponderada:

$$\bar{x}_p = \frac{\sum_{i=1}^{n} w_i x_i}{\sum_{i=1}^{n} w_i}$$

Onde $w_i$ é o peso do valor $x_i$.

**Exemplo:** Se uma prova vale peso 3 e um trabalho vale peso 2:
Nota final = (7 × 3 + 9 × 2) / (3 + 2) = (21 + 18) / 5 = 7.8

### 5. Média Geométrica

Usada para taxas de crescimento, porcentagens e dados multiplicativos:

$$\bar{x}_g = \left(\prod_{i=1}^{n} x_i\right)^{1/n}$$

É sempre **menor ou igual** à média aritmética (desigualdade das médias).

## 💻 Exemplos Práticos

```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
# Exemplo 1: Cálculo manual da média
notas = np.array([7.0, 8.5, 9.0, 6.5, 10.0, 7.5, 8.0])

# Cálculo manual
soma = np.sum(notas)
quantidade = len(notas)
media_manual = soma / quantidade

# Cálculo com numpy
media_numpy = np.mean(notas)

print(f"Notas: {notas}")
print(f"Soma: {soma}")
print(f"Quantidade: {quantidade}")
print(f"Média (manual): {media_manual:.2f}")
print(f"Média (numpy):  {media_numpy:.2f}")
```

```python
# Exemplo 2: Sensibilidade a outliers
sem_outlier = np.array([100, 102, 98, 105, 101, 99, 103])
com_outlier = np.array([100, 102, 98, 105, 101, 99, 103, 500])

media_sem = np.mean(sem_outlier)
media_com = np.mean(com_outlier)

print("Dados sem outlier:", sem_outlier)
print(f"Média: {media_sem:.2f}")
print()
print("Dados com outlier (500):", com_outlier)
print(f"Média: {media_com:.2f}")
print()
print(f"Diferença: {media_com - media_sem:.2f}")
print("Um único outlier de 500 aumentou a média em mais de 50 unidades!")
```

```python
# Exemplo 3: Demonstração visual do efeito de outliers
# Gerando dados progressivamente com outliers mais extremos
dados_base = np.random.normal(50, 10, 100)
medias = []
outliers = [0, 10, 20, 30, 50, 80, 100, 150, 200, 300]

for o in outliers:
    dados_com_outlier = np.append(dados_base, o)
    medias.append(np.mean(dados_com_outlier))

plt.figure(figsize=(10, 5))
plt.plot(outliers, medias, 'o-', color='red', linewidth=2)
plt.axhline(y=np.mean(dados_base), color='green', linestyle='--', label=f'Média original = {np.mean(dados_base):.1f}')
plt.xlabel('Valor do Outlier')
plt.ylabel('Média do Conjunto')
plt.title('Efeito de Outliers na Média Aritmética')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Quanto maior o outlier, mais a média se distancia do valor original.")
```

```python
# Exemplo 4: Média Ponderada
notas_prova = np.array([7.0, 8.5, 9.0])  # Pesos 3
notas_trabalho = np.array([10.0, 9.5])  # Pesos 2
notas_exercicio = np.array([8.0, 7.5, 9.0, 8.5])  # Pesos 1

# Usando np.average para média ponderada
todas_notas = np.concatenate([notas_prova, notas_trabalho, notas_exercicio])
pesos = np.array([3, 3, 3, 2, 2, 1, 1, 1, 1])

media_ponderada = np.average(todas_notas, weights=pesos)
media_simples = np.mean(todas_notas)

print("Notas das provas (peso 3):", notas_prova)
print("Notas dos trabalhos (peso 2):", notas_trabalho)
print("Notas dos exercícios (peso 1):", notas_exercicio)
print()
print(f"Média simples:   {media_simples:.2f}")
print(f"Média ponderada: {media_ponderada:.2f}")
print("A média ponderada dá mais peso às provas.")
```

```python
# Exemplo 5: Média Geométrica
# Calculando crescimento médio anual de um investimento
# Taxas de retorno anual: 10%, 8%, 12%, -5%, 15%
taxas = np.array([1.10, 1.08, 1.12, 0.95, 1.15])

media_geom = np.prod(taxas) ** (1 / len(taxas))
media_arit = np.mean(taxas)

print("Fatores de crescimento anual:", taxas)
print(f"Média aritmética: {media_arit:.4f} ({((media_arit - 1) * 100):.2f}% ao ano)")
print(f"Média geométrica: {media_geom:.4f} ({((media_geom - 1) * 100):.2f}% ao ano)")
print()
print("A média geométrica é mais precisa para taxas de crescimento.")
print(f"Crescimento total: {(np.prod(taxas) - 1) * 100:.2f}% em 5 anos")
```

## 📝 Exercícios Resolvidos

```python
# Exercício Resolvido 1
# Calcular a média de temperatura diária em uma semana
temperaturas = np.array([28, 30, 27, 31, 29, 33, 26])
media_temp = np.mean(temperaturas)

print(f"Temperaturas da semana (°C): {temperaturas}")
print(f"Temperatura média: {media_temp:.1f}°C")
print(f"Dias acima da média: {np.sum(temperaturas > media_temp)}")
print(f"Dias abaixo da média: {np.sum(temperaturas < media_temp)}")
```

```python
# Exercício Resolvido 2
# Média ponderada — Cálculo do IGP-M
# O IGP-M é composto por: IPA (60%), IPC (30%), INCC (10%)
ipa = 0.85  # 0.85%
ipc = 0.42  # 0.42%
incc = 0.63  # 0.63%

igp_m = np.average([ipa, ipc, incc], weights=[60, 30, 10])

print(f"IPA (60%): {ipa:.2f}%")
print(f"IPC (30%): {ipc:.2f}%")
print(f"INCC (10%): {incc:.2f}%")
print(f"IGP-M: {igp_m:.2f}%")
```

```python
# Exercício Resolvido 3
# Propriedade: soma dos desvios é zero
dados = np.array([5, 8, 12, 7, 10])
media = np.mean(dados)
desvios = dados - media

print(f"Dados: {dados}")
print(f"Média: {media:.2f}")
print(f"Desvios (x_i - x̄): {desvios}")
print(f"Soma dos desvios: {np.sum(desvios):.4f}")
print("(A soma dos desvios é sempre zero!)")
```

## 📝 Resumo do Capítulo

- A **média aritmética** é a soma dos valores dividida pela quantidade
- Notação: $\bar{x}$ (amostra) e $\mu$ (população)
- A média é **sensível a outliers** — um valor extremo pode distorcê-la
- A **média ponderada** atribui pesos diferentes aos valores
- A **média geométrica** é adequada para taxas e crescimentos percentuais
- A soma dos desvios em relação à média é sempre zero
- A média aritmética ≥ média geométrica (desigualdade das médias)

## 🧐 Curiosidades

- O símbolo $\bar{x}$ foi introduzido por **Ronald Fisher**, um dos pais da estatística moderna.
- Na China antiga, o imperador **Yao** (século 23 a.C.) já usava médias para calcular a produção agrícola.
- A média é tão comum que virou padrão em boletins escolares de praticamente todos os países.
- Em 1954, o statistician **Darrell Huff** escreveu "How to Lie with Statistics", mostrando como médias mal usadas podem enganar.

## ✅ Boas Práticas

- Sempre verifique a **distribuição** dos dados antes de usar a média
- Para dados financeiros, prefira **mediana** à média
- Use **média ponderada** quando os valores têm importâncias diferentes
- Para **taxas de crescimento**, use média geométrica
- Arredonde a média com **1-2 casas decimais** a mais que os dados originais
- **Nunca** use a média sozinha — sempre acompanhe de uma medida de dispersão

## ⚠️ Erros Comuns

| Erro | Consequência | Correção |
|------|-------------|----------|
| Usar média aritmética para taxas | Superestima o crescimento real | Use média geométrica |
| Ignorar pesos diferentes | Cálculo incorreto | Use média ponderada |
| Não verificar outliers | Média distorcida | Olhe o box plot primeiro |
| Confundir $\bar{x}$ com $\mu$ | Inferência incorreta | Distinga amostra de população |
| Calcular média de dados categóricos | Resultado sem sentido | Use moda para categorias |
