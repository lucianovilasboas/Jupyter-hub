## Andares do Predio

Imagine um predio de 100 andares.

O **percentil 90** significa:
"voce esta no 90o andar — 90% dos andares estao abaixo de voce."

Os **quartis** dividem o predio em 4 partes iguais:
- Q1 (25o andar): 25% dos dados estao abaixo
- Q2 (50o andar): a mediana — metade abaixo, metade acima
- Q3 (75o andar): 75% dos dados estao abaixo

Num contexto de renda:
- Se voce esta no percentil 90 da renda,
  voce ganha mais que 90% da populacao!
- Se esta no Q1, 25% ganham menos que voce.


## 📊 Traduzindo

| Andares do Predio | Estatistica |
|-|-|
| Predio de 100 andares | 100 percentis |
| Andar 25 (Q1) | 25o percentil — 25% abaixo |
| Andar 50 (Q2) | 50o percentil — mediana |
| Andar 75 (Q3) | 75o percentil — 75% abaixo |


## 📏 Para Que Servem

Quartis e percentis ajudam a **posicionar** um valor:
- "Seu filho esta no percentil 80 de altura"
  (mais alto que 80% das criancas da idade)
- "Voce esta no Q1 de renda da sua regiao"

Eles dividem os dados em partes iguais,
dando uma visao da distribuicao completa.


```python
# Andares do predio
import statistics as stat
# Simulando rendas
rendas = sorted([1200, 1500, 1800, 2000, 2500, 3000, 3500, 5000, 8000, 12000])
n = len(rendas)

print(f"Rendas ordenadas: {rendas}")
print(f"Q1 (25o andar): {rendas[n//4]}")
print(f"Q2 (50o andar / mediana): {rendas[n//2]}")
print(f"Q3 (75o andar): {rendas[3*n//4]}")
```

## 📌 Resumo

| Andares do Predio | Estatistica |
|-|-|
| Predio de 100 andares | 100 percentis |
| Andar 25 | Q1 |
| Andar 50 | Q2 / Mediana |
| Andar 75 | Q3 |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
