## Julgamento do Inocente

O tribunal esta em sessao.

**H0 (inocente)**: O reu e inocente ate prova em contrario.
**H1 (culpado)**: As evidencias mostram que o reu e culpado.

O juri analisa as evidencias (dados).

- Evidencias FRACAS: o juri mantem H0 (inocente).
- Evidencias FORTES: o juri rejeita H0 (condena).

Mas o juri pode errar:
- **Condenar inocente** = erro tipo I (falso positivo)
- **Absolver culpado** = erro tipo II (falso negativo)

A **significancia (α)** e o "risco que aceitamos
de condenar um inocente".


## 📊 Traduzindo

| Julgamento do Inocente | Estatistica |
|-|-|
| Reu e inocente | H0 — hipotese nula |
| Reu e culpado | H1 — hipotese alternativa |
| Evidencias da acusacao | Dados da amostra |
| Juri condena | Rejeitar H0 |
| Juri absolve | Nao rejeitar H0 |
| Condenar inocente | Erro tipo I |
| Absolver culpado | Erro tipo II |


## 📏 O que e o p-valor?

O **p-valor** responde:
*"Se o reu fosse inocente (H0 verdadeira),
qual a chance de vermos evidencias tao fortes quanto essas?"*

Se essa chance e muito pequena (p < 0,05),
o juri se convence: "nao faz sentido que seja inocente"
e condena (rejeita H0).


```python
# O julgamento em Python
import random
random.seed(42)
# H0: a moeda e honesta (probabilidade = 0.5)
# Vamos jogar 50 vezes e ver
resultados = [random.choice(["cara", "coroa"]) for _ in range(50)]
caras = resultados.count("cara")
print(f"50 jogadas: {caras} caras")
print(f"Proporcao: {caras/50*100:.0f}%")
print()
print("Sera que a moeda e honesta (H0) ou viciada (H1)?")
print("Se cair muito longe de 50%, rejeitamos H0.")
```

## 📌 Resumo

| Julgamento do Inocente | Estatistica |
|-|-|
| Inocente | H0 |
| Culpado | H1 |
| Condenar | Rejeitar H0 |
| Absolver | Nao rejeitar H0 |
| Condenar inocente | Erro tipo I |
| Absolver culpado | Erro tipo II |
