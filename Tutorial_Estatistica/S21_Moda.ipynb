## 📝 Introdução

A **moda** (mode) é a medida de tendência central mais simples: é o **valor que aparece com maior frequência** em um conjunto de dados.

Diferente da média e da mediana, a moda:
- Pode ser usada com **dados qualitativos** (categorias, nomes, cores)
- Pode ter **mais de um valor** (distribuições bimodais ou multimodais)
- Nem sempre existe (distribuição uniforme)

> 🧠 **Para refletir:** Em uma loja de roupas, o tamanho mais vendido (P, M, G) é a moda. Essa informação é mais útil que a média dos tamanhos!

## 📚 Explicação Detalhada

### 1. O Conceito de Moda

A moda é simplesmente o valor que ocorre com maior frequência.

**Características:**
- **Dados discretos:** moda = valor com maior contagem
- **Dados contínuos:** precisa agrupar em classes
- **Dados qualitativos:** moda = categoria mais frequente

### 2. Tipos de Distribuições

| Tipo | Descrição | Exemplo |
|------|-----------|---------|
| **Unimodal** | Uma moda | Altura de adultos |
| **Bimodal** | Duas modas | Duas turmas diferentes |
| **Multimodal** | Três ou mais modas | Preferências de consumo |
| **Uniforme** | Sem moda clara | Dado honesto |

### 3. Moda vs Média vs Mediana

- **Moda**: única para dados qualitativos
- **Moda**: pode não existir ou existir múltiplas
- **Moda**: não é afetada por outliers
- **Moda**: menos usada em análises formais

## 💻 Exemplos Práticos

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)
```

```python
# Exemplo 1: Moda em dados discretos
dados = np.array([1, 2, 2, 3, 3, 3, 4, 4, 5])
moda_scipy = stats.mode(dados, keepdims=True)
print(f"Dados: {dados}")
print(f"Moda (scipy): {moda_scipy[0][0]} (frequencia: {moda_scipy[1][0]})")
valores, contagens = np.unique(dados, return_counts=True)
print("
Tabela de frequencias:")
for v, c in zip(valores, contagens):
    barra = chr(9608) * c
    print(f"  {v}: {barra} ({c}x)")
```

```python
# Exemplo 2: Moda com pandas
series = pd.Series([1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 4, 5])
moda_pd = series.mode()
freq = series.value_counts()
print("Frequencias:")
print(freq)
print(f"\nModa(s) com pandas: {moda_pd.tolist()}")
print("Pandas retorna TODAS as modas.")
```

```python
# Exemplo 3: Unimodal vs bimodal vs multimodal
unimodal = np.random.normal(100, 15, 1000)
bimodal = np.concatenate([np.random.normal(60, 10, 500), np.random.normal(140, 10, 500)])
multimodal = np.concatenate([np.random.normal(30, 5, 333), np.random.normal(60, 5, 333), np.random.normal(90, 5, 334)])
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].hist(unimodal, bins=30, color="skyblue", edgecolor="black")
axes[0].set_title("Unimodal")
axes[0].set_xlabel("Valor"); axes[0].set_ylabel("Frequencia")
axes[1].hist(bimodal, bins=30, color="salmon", edgecolor="black")
axes[1].set_title("Bimodal"); axes[1].set_xlabel("Valor")
axes[2].hist(multimodal, bins=30, color="lightgreen", edgecolor="black")
axes[2].set_title("Multimodal"); axes[2].set_xlabel("Valor")
plt.tight_layout(); plt.show()
```

```python
# Exemplo 4: Moda em dados qualitativos
cores = np.array(["azul", "vermelho", "azul", "verde", "azul",
                 "vermelho", "azul", "verde", "vermelho", "azul"])
series_cores = pd.Series(cores)
moda_cores = series_cores.mode()
print(f"Cores: {cores.tolist()}")
print(f"Frequencias:\n{series_cores.value_counts()}")
print(f"\nModa: {moda_cores.values[0]}")
print("A moda e a unica medida para dados qualitativos!")
```

```python
# Exemplo 5: Moda em dados continuos (agrupados)
alturas = np.random.normal(170, 10, 1000)
bins = np.arange(140, 201, 5)
freq, edges = np.histogram(alturas, bins=bins)
classe_modal = np.argmax(freq)
moda_cont = (edges[classe_modal] + edges[classe_modal + 1]) / 2
print("Distribuicao de alturas:")
for i in range(len(freq)):
    barra = chr(9608) * (freq[i] // 5)
    print(f"  {edges[i]:.0f}-{edges[i+1]:.0f}: {freq[i]:3d} {barra}")
print(f"\nClasse modal: {edges[classe_modal]:.0f}-{edges[classe_modal+1]:.0f}")
print(f"Moda aprox.: {moda_cont:.1f} cm")
```

```python
# Exemplo 6: Distribuicao uniforme (sem moda)
uniforme = np.random.randint(1, 11, 100)
freq_unif = pd.Series(uniforme).value_counts().sort_index()
print("Distribuicao uniforme (valores de 1 a 10):")
for v, c in freq_unif.items():
    print(f"  {v}: {c:2d} {chr(9608) * c}")
print("\nSem moda clara — todos tem frequencia similar.")
```

## 📝 Exercícios Resolvidos

```python
# Exercicio 1: Encontrar a moda das idades
idades = np.array([22, 25, 22, 30, 35, 22, 28, 30, 25, 30, 30, 22])
series_id = pd.Series(idades)
modas = series_id.mode()
print(f"Idades: {idades}")
print(f"Frequencias:\n{series_id.value_counts().sort_index()}")
print(f"\nModa(s): {modas.tolist()}")
```

```python
# Exercicio 2: Classificando modalidade
def classificar_modalidade(dist, bins=20):
    freq, _ = np.histogram(dist, bins=bins)
    picos = np.sum((freq[1:-1] > freq[:-2]) & (freq[1:-1] > freq[2:]))
    if picos <= 1: return "unimodal"
    elif picos == 2: return "bimodal"
    else: return f"multimodal ({picos} picos)"
print(f"Normal: {classificar_modalidade(np.random.normal(0,1,1000))}")
print(f"Bimodal: {classificar_modalidade(bimodal)}")
print(f"Multimodal: {classificar_modalidade(multimodal)}")
print(f"Uniforme: {classificar_modalidade(np.random.uniform(0,1,1000))}")
```

## 📝 Resumo do Capítulo

- **Moda (mode)** é o valor que mais aparece
- **Unimodal**: uma moda | **Bimodal**: duas | **Multimodal**: três ou mais
- Única medida para **dados qualitativos**
- Em dados contínuos, agrupe em classes
- Pode não existir moda
- scipy=primeira moda, pandas=todas as modas

## 🧐 Curiosidades

- O termo **moda** veio do francês **mode**, introduzido por Karl Pearson em 1895.
- Em marketing, a moda é mais útil que a média.
- Em linguística, a moda encontra a palavra mais comum.
- Distribuição uniforme não tem moda.

## ✅ Boas Práticas

- Use moda para **dados categóricos**
- Agrupe dados contínuos em bins
- Reporte **todas** as modas
- Combine com média e mediana

## ⚠️ Erros Comuns

| Erro | Consequência | Correção |
|------|-------------|----------|
| Calcular moda em contínuos sem agrupar | Cada valor único | Agrupe em faixas |
| Ignorar modas secundárias | Perde informação | Reporte todas |
| Achar que moda sempre existe | Frustração | Verifique repetições |
| Confundir moda com mediana | Conceitos trocados | Moda = frequência |
