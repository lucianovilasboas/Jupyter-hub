## 📝 Introdução

O **teste unilateral à esquerda** (left-tailed test) para proporção é usado quando queremos verificar se a **proporção populacional é menor** que um determinado valor. A região de rejeição está na cauda esquerda da distribuição normal.

Situações típicas:
- "A taxa de defeitos é menor que 5%?"
- "A proporção de fumantes diminuiu após a campanha?"
- "Menos de 30% dos alunos reprovam no curso?"

Este teste é direcional: estamos interessados especificamente em detectar se a proporção é **menor** que p₀.

## 📚 Explicação Detalhada

### 1. Formulação do Teste Unilateral à Esquerda

- **H₀**: p ≥ p₀ (a proporção é maior ou igual ao valor de referência)
- **H₁**: p < p₀ (a proporção é menor que o valor de referência)

A região de rejeição está na **cauda esquerda** da distribuição normal padrão.

### 2. Estatística Z e Valor Crítico

Z = (p̂ - p₀) / √(p₀(1-p₀)/n)

Rejeitamos H₀ se:
- **Z calculado < -Zα** (abordagem do valor crítico)
- **p-valor < α** (abordagem do p-valor)

Onde p-valor = P(Z ≤ Z_calc) = Φ(Z_calc) (probabilidade acumulada à esquerda).

### 3. Condições

Mesmas condições do teste Z para proporção:
- n·p₀ ≥ 10 e n·(1-p₀) ≥ 10
- Amostra aleatória independente
- Categoria binária (sucesso/falha)

## 💻 Exemplos Práticos

```python
# Importando bibliotecas
import numpy as np
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
import matplotlib.pyplot as plt

print("Bibliotecas carregadas com sucesso!")
```

```python
# Exemplo 1: Teste de proporcao unilateral a esquerda (calculo manual)
# Uma empresa afirma que a taxa de satisfacao é de pelo menos 85%.
# Uma pesquisa com 300 clientes mostra que 240 estao satisfeitos.
# Teste se a taxa de satisfacao é menor que 85%.

n = 300
x = 240
p0 = 0.85
alpha = 0.05

# Proporcao amostral
p_hat = x / n

# Erro padrao sob H0
erro = np.sqrt(p0 * (1 - p0) / n)

# Estatistica Z
z = (p_hat - p0) / erro

# p-valor (unilateral a esquerda)
p_valor = stats.norm.cdf(z)

# Valor critico para teste unilateral esquerda
z_crit = stats.norm.ppf(alpha)

print(f"n = {n}, x (satisfeitos) = {x}")
print(f"Proporção amostral (p̂) = {p_hat:.3f}")
print(f"Proporção sob H₀ (p₀) = {p0}")
print(f"H₀: p ≥ {p0} | H₁: p < {p0}")
print(f"Estatística Z = {z:.4f}")
print(f"Valor crítico Z = {z_crit:.4f}")
print(f"p-valor = {p_valor:.4f}")
print()
if z < z_crit or p_valor < alpha:
    print("✅ Rejeitamos H₀: a satisfação é menor que 85%.")
else:
    print("❌ Não rejeitamos H₀: não há evidência de queda na satisfação.")
```

```python
# Exemplo 2: Usando proportions_ztest do statsmodels
# alternative='smaller' para teste unilateral a esquerda

z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0, alternative='smaller')

print(f"Usando statsmodels:")
print(f"x = {x}, n = {n}")
print(f"H₀: p ≥ {p0} | H₁: p < {p0}")
print(f"Estatística Z = {z_stat:.4f}")
print(f"p-valor = {p_val:.4f}")
if p_val < alpha:
    print(f"✅ p-valor {p_val:.4f} < α = {alpha} → Rejeitamos H₀")
else:
    print(f"❌ p-valor {p_val:.4f} ≥ α = {alpha} → Não rejeitamos H₀")
```

```python
# Exemplo 3: Visualizacao da regiao de rejeicao a esquerda
z_vals = np.linspace(-4, 4, 1000)
densidade = stats.norm.pdf(z_vals, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(z_vals, densidade, 'b-', label='Normal Padrão')

# Regiao de rejeicao (cauda esquerda)
z_crit = stats.norm.ppf(alpha)
x_fill = np.linspace(-4, z_crit, 100)
plt.fill_between(x_fill, stats.norm.pdf(x_fill), color='red', alpha=0.3, label='Região de Rejeição')

plt.axvline(z, color='green', linestyle='--', lw=2, label=f'Z calculado = {z:.2f}')
plt.axvline(z_crit, color='red', linestyle=':', lw=2, label=f'Z crítico = {z_crit:.2f}')
plt.title('Teste de Proporção Unilateral à Esquerda')
plt.xlabel('Z')
plt.ylabel('Densidade')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Exemplo 4: Diferentes niveis de significancia
n = 500
x = 230
p0 = 0.50
p_hat = x / n

z_calc = (p_hat - p0) / np.sqrt(p0 * (1 - p0) / n)
p_left = stats.norm.cdf(z_calc)

print(f"Proporção amostral: {p_hat:.3f}")
print(f"H₀: p ≥ {p0} | H₁: p < {p0}")
print(f"Z = {z_calc:.4f}")
print(f"p-valor = {p_left:.4f}")
print()
print("Decisão para diferentes α:")
for a in [0.10, 0.05, 0.01]:
    z_crit_a = stats.norm.ppf(a)
    decisao = "Rejeitar H₀" if z_calc < z_crit_a else "Não rejeitar H₀"
    print(f"  α = {a:.2f} → Z crítico = {z_crit_a:.4f} → {decisao}")
```

```python
# Exemplo 5: Funcao completa para teste unilateral esquerda
def teste_proporcao_esquerda(x, n, p0, alpha=0.05):
    """Teste unilateral a esquerda para proporcao."""
    p_hat = x / n
    erro = np.sqrt(p0 * (1 - p0) / n)
    z = (p_hat - p0) / erro
    p_val = stats.norm.cdf(z)
    z_crit = stats.norm.ppf(alpha)
    
    print(f"=== Teste Unilateral à Esquerda (Proporção) ===")
    print(f"Dados: x = {x}, n = {n}, p̂ = {p_hat:.4f}")
    print(f"H₀: p ≥ {p0} | H₁: p < {p0}")
    print(f"Z = {z:.4f}, Z crítico = {z_crit:.4f}")
    print(f"p-valor = {p_val:.4f}, α = {alpha}")
    if z < z_crit:
        print(f"✅ Z ({z:.4f}) < Z crítico ({z_crit:.4f}) → Rejeitamos H₀")
    else:
        print(f"❌ Z ({z:.4f}) ≥ Z crítico ({z_crit:.4f}) → Não rejeitamos H₀")
    return {"z": z, "p_valor": p_val, "rejeita": z < z_crit}

teste_proporcao_esquerda(240, 300, 0.85)
print()
teste_proporcao_esquerda(20, 500, 0.05, alpha=0.01)
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Um hospital afirma que a taxa de infecção é no máximo 3%. Em 800 cirurgias, houve 18 infecções. Teste a 5% se a taxa é menor que 3%.

```python
# Solucao do Exercicio 1
n = 800
x = 18
p0 = 0.03
alpha = 0.05

p_hat = x / n
erro = np.sqrt(p0 * (1 - p0) / n)
z = (p_hat - p0) / erro
p_val = stats.norm.cdf(z)

print(f"Proporção amostral: {p_hat:.4f}")
print(f"H₀: p ≥ {p0} | H₁: p < {p0}")
print(f"Z = {z:.4f}, p-valor = {p_val:.4f}")
if p_val < alpha:
    print(f"✅ Rejeitamos H₀: a taxa de infecção é menor que 3%.")
else:
    print(f"❌ Não rejeitamos H₀.")
```

**Exercício 2:** Um político afirma que pelo menos 40% dos jovens votam. Uma pesquisa com 400 jovens mostra que 140 votam. Teste a 1%.

```python
# Solucao do Exercicio 2
n = 400
x = 140
p0 = 0.40
alpha = 0.01

z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0, alternative='smaller')

print(f"Proporção amostral: {x/n:.3f}")
print(f"H₀: p ≥ {p0} | H₁: p < {p0}")
print(f"Z = {z_stat:.4f}, p-valor = {p_val:.4f}")
if p_val < alpha:
    print(f"✅ Rejeitamos H₀ a {alpha}: menos de 40% votam.")
else:
    print(f"❌ Não rejeitamos H₀ a {alpha}.")
```

**Exercício 3:** Uma loja online afirma que a taxa de devolução é no máximo 8%. Em 600 vendas, 36 foram devolvidas. Teste a 5%.

```python
# Solucao do Exercicio 3
n = 600
x = 36
p0 = 0.08
alpha = 0.05

z_manual = (x/n - p0) / np.sqrt(p0 * (1 - p0) / n)
p_manual = stats.norm.cdf(z_manual)

print(f"Proporção amostral: {x/n:.4f}")
print(f"H₀: p ≥ {p0} | H₁: p < {p0}")
print(f"Z = {z_manual:.4f}, p-valor = {p_manual:.4f}")
if p_manual < alpha:
    print("✅ Rejeitamos H₀: taxa de devolução < 8%.")
else:
    print("❌ Não rejeitamos H₀.")
```

## 📌 Resumo

- O **teste unilateral à esquerda** para proporção verifica se p < p₀.
- H₀: p ≥ p₀, H₁: p < p₀. Região de rejeição na cauda esquerda.
- Estatística Z = (p̂ - p₀) / √(p₀(1-p₀)/n).
- p-valor = P(Z ≤ Z_calc) = Φ(Z_calc).
- Usamos `proportions_ztest(..., alternative='smaller')` do statsmodels.
- Rejeitamos H₀ se Z < -Zα ou p-valor < α.

## 💡 Curiosidades

O teste unilateral é mais **poderoso** (maior capacidade de detectar um efeito) que o bicaudal para detectar mudanças na direção especificada. Porém, só deve ser usado quando temos uma hipótese direcional forte antes de coletar os dados. Usar o teste unilateral depois de ver os dados (porque o bicaudal não deu significância) é uma forma de **p-hacking** e não é ético.

## ✅ Boas Práticas

1. Defina a direção do teste (esquerda/direita) **antes** de coletar os dados.
2. Use o teste unilateral apenas quando há uma justificativa teórica clara.
3. Em caso de dúvida, prefira o teste bicaudal (mais conservador).
4. Reporte sempre o p-valor e a direção do teste utilizado.
5. Verifique as condições de aplicação (n·p₀ ≥ 10) antes de usar o teste Z.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Usar teste unilateral sem justificativa | Sempre justifique ou use bicaudal |
| Confundir o sinal do Z crítico | Unilateral esquerda: Z crítico negativo |
| Trocar H₀ e H₁ | H₀ sempre contém o = (≥, ≤, ou =) |
| Usar alternative='smaller' quando Z é positivo | O teste ainda funciona, mas o p-valor será > 0.5 |
