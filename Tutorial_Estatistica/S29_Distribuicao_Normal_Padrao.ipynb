## 📝 Introdução

Toda distribuição normal pode ser transformada em uma **distribuição normal padrão** — uma versão padronizada com μ = 0 e σ = 1. Isso é feito através do **Z-score**, que nos diz a quantos desvios padrão um valor está da média.

Esta padronização permite que usemos uma única tabela (a tabela Z) para calcular probabilidades para qualquer distribuição normal, sem precisar de tabelas diferentes para cada combinação de μ e σ.

## 📚 Explicação Detalhada

### 1. Z-Score (Escore Padronizado)

O Z-score transforma qualquer valor X de uma distribuição normal em um valor Z da normal padrão:

$$ Z = \frac{X - \mu}{\sigma} $$

Onde:
- **Z**: número de desvios padrão entre X e a média
- **X**: valor original
- **μ**: média da distribuição original
- **σ**: desvio padrão da distribuição original

### 2. Distribuição Normal Padrão N(0,1)

A distribuição normal padrão é simplesmente uma normal com:
- **μ = 0**
- **σ = 1**

Ela serve como distribuição de referência para todas as outras normais.

### 3. Probabilidades Acumuladas e CDF

A **função de distribuição acumulada (CDF)** retorna a probabilidade de Z ser menor ou igual a um valor: P(Z ≤ z).

### 4. Percentis e PPF

A **função ponto percentual (PPF)** é a inversa da CDF: dado uma probabilidade p, retorna o valor z tal que P(Z ≤ z) = p.

```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

print("Bibliotecas importadas com sucesso!")
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Calculando Z-scores manualmente
# Considere notas de um vestibular: μ = 500, σ = 100

mu = 500
sigma = 100

# Notas de 3 alunos
notas = [650, 400, 520]

print("=== Cálculo de Z-scores ===")
for nota in notas:
    z = (nota - mu) / sigma
    print(f"Nota: {nota} → Z = ({nota} - {mu}) / {sigma} = {z:.2f}")

# Interpretação
print("\nInterpretação:")
print("- Z positivo: valor acima da média")
print("- Z negativo: valor abaixo da média")
print("- Z = 0: valor exatamente na média")
```

```python
# Exemplo 2: Visualizando a transformação para normal padrão
# Mostramos como uma distribuição N(500,100) se torna N(0,1) após padronização

np.random.seed(42)

# Dados originais
mu_orig, sigma_orig = 500, 100
dados_originais = np.random.normal(mu_orig, sigma_orig, 10000)

# Padronização (Z-scores)
dados_padronizados = (dados_originais - mu_orig) / sigma_orig

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Original
axes[0].hist(dados_originais, bins=40, density=True, color='skyblue', edgecolor='black')
axes[0].set_title('Distribuição Original N(500, 100)', fontsize=13)
axes[0].set_xlabel('Nota')
axes[0].set_ylabel('Densidade')

# Padronizada
axes[1].hist(dados_padronizados, bins=40, density=True, color='salmon', edgecolor='black')

x = np.linspace(-4, 4, 200)
axes[1].plot(x, stats.norm.pdf(x, 0, 1), 'b-', lw=2, label='N(0,1)')

axes[1].set_title('Distribuição Padronizada N(0, 1)', fontsize=13)
axes[1].set_xlabel('Z-score')
axes[1].set_ylabel('Densidade')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Exemplo 3: Calculando probabilidades com CDF
# Usamos stats.norm.cdf para consultar a normal padrão

from scipy import stats

# P(Z < 1.5)
prob_menos_1_5 = stats.norm.cdf(1.5)

# P(Z > -0.5)
prob_mais_menos_0_5 = 1 - stats.norm.cdf(-0.5)

# P(-1 < Z < 2)
prob_entre_1_2 = stats.norm.cdf(2) - stats.norm.cdf(-1)

print("=== Probabilidades na Normal Padrão ===")
print(f"P(Z < 1.5) = {prob_menos_1_5:.4f} ({prob_menos_1_5*100:.2f}%)")
print(f"P(Z > -0.5) = {prob_mais_menos_0_5:.4f} ({prob_mais_menos_0_5*100:.2f}%)")
print(f"P(-1 < Z < 2) = {prob_entre_1_2:.4f} ({prob_entre_1_2*100:.2f}%)")
```

```python
# Exemplo 4: Calculando percentis com PPF (inversa da CDF)
# Encontrar o valor de Z para uma dada probabilidade acumulada

# Percentil 95 (z tal que P(Z < z) = 0.95)
z_95 = stats.norm.ppf(0.95)

# Percentil 97.5 (bilateral: z tal que P(Z < z) = 0.975)
z_975 = stats.norm.ppf(0.975)

# Percentil 99
z_99 = stats.norm.ppf(0.99)

print("=== Percentis da Normal Padrão ===")
print(f"Percentil 95 (z₀.₀₅): z = {z_95:.4f}")
print(f"Percentil 97.5 (z₀.₀₂₅): z = {z_975:.4f}")
print(f"Percentil 99 (z₀.₀₁): z = {z_99:.4f}")
print(f"\nValores comuns:")
print(f"z₀.₀₅ (unilateral 95%) = {z_95:.4f}")
print(f"z₀.₀₂₅ (bilateral 95%) = {z_975:.4f}")
```

```python
# Exemplo 5: Consulta virtual à Tabela Z
# Construímos uma tabela Z programática para consulta

def consultar_tabela_z(z):
    """Retorna a probabilidade P(Z < z) para um dado z."""
    prob = stats.norm.cdf(z)
    return prob

# Exibimos uma tabela Z parcial
print("=== Tabela Z Virtual (parcial) ===")
print("Tabela: P(Z < z) para valores de z de 0.0 a 3.9")
print()

valores_z = [round(i * 0.1, 1) for i in range(0, 40)]
tabela = []

for z in valores_z:
    prob = consultar_tabela_z(z)
    tabela.append({'z': z, 'P(Z < z)': round(prob, 4)})

df_tabela = pd.DataFrame(tabela)
print(df_tabela.to_string(index=False))

print("\nExemplo de consulta rápida:")
z_exemplo = 1.96
p_exemplo = consultar_tabela_z(z_exemplo)
print(f"P(Z < {z_exemplo}) = {p_exemplo:.4f}")
```

```python
# Exemplo 6: Aplicação completa — notas do vestibular
# Distribuição original: N(500, 100)

mu, sigma = 500, 100

# Problema: qual a nota mínima para estar entre os 10% melhores?
percentil_90 = stats.norm.ppf(0.90, mu, sigma)

# Problema: qual a nota máxima para estar entre os 5% piores?
percentil_5 = stats.norm.ppf(0.05, mu, sigma)

# Problema: qual a probabilidade de tirar entre 450 e 550?
prob_450_550 = stats.norm.cdf(550, mu, sigma) - stats.norm.cdf(450, mu, sigma)

print("=== Aplicação: Vestibular ===")
print(f"Nota mínima para top 10%: {percentil_90:.1f}")
print(f"Nota máxima para bottom 5%: {percentil_5:.1f}")
print(f"P(450 < X < 550): {prob_450_550:.4f} ({prob_450_550*100:.2f}%)")
```

```python
# Exemplo 7: Visualização da CDF na normal padrão

z_vals = np.linspace(-4, 4, 200)
cdf_vals = stats.norm.cdf(z_vals)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# PDF
axes[0].plot(z_vals, stats.norm.pdf(z_vals), 'b-', lw=2)
z_dest = 1.5
mask = z_vals <= z_dest
axes[0].fill_between(z_vals, stats.norm.pdf(z_vals), where=mask, 
                      color='skyblue', alpha=0.5, 
                      label=f'P(Z < {z_dest}) = {stats.norm.cdf(z_dest):.3f}')
axes[0].axvline(z_dest, color='red', linestyle='--', lw=1.5)
axes[0].set_title('PDF — Probabilidade como Área', fontsize=13)
axes[0].set_xlabel('Z')
axes[0].set_ylabel('Densidade')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# CDF
axes[1].plot(z_vals, cdf_vals, 'g-', lw=2)
axes[1].axvline(z_dest, color='red', linestyle='--', lw=1.5)
axes[1].axhline(stats.norm.cdf(z_dest), color='red', linestyle=':', lw=1.5)
axes[1].set_title('CDF — Probabilidade Acumulada', fontsize=13)
axes[1].set_xlabel('Z')
axes[1].set_ylabel('P(Z < z)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Converta para Z-score: uma pessoa com QI = 130 sabendo que QI ~ N(100, 15). Qual a proporção de pessoas com QI maior que 130?

```python
# Solução Exercício 1
mu = 100
sigma = 15
x = 130

z = (x - mu) / sigma
print(f"Z-score: Z = ({x} - {mu}) / {sigma} = {z:.2f}")

proporcao = 1 - stats.norm.cdf(z)
print(f"P(Z > {z:.2f}) = {proporcao:.4f} = {proporcao*100:.2f}%")
```

**Exercício 2:** O peso de recém-nascidos segue N(3.2 kg, 0.5 kg). Qual o peso que separa os 2% mais leves?

```python
# Solução Exercício 2
mu = 3.2
sigma = 0.5

# Encontrar o percentil 2
peso_limite = stats.norm.ppf(0.02, mu, sigma)

print(f"Peso que separa os 2% mais leves: {peso_limite:.3f} kg")
print(f"Ou seja, 2% dos bebês nascem com menos de {peso_limite:.3f} kg")
```

## 📌 Resumo

- O **Z-score** padroniza qualquer valor: $Z = (X - \mu) / \sigma$
- A **distribuição normal padrão** tem μ = 0 e σ = 1, denotada N(0,1).
- **CDF** (cumulative distribution function): $P(Z \leq z)$
- **PPF** (percent point function): inversa da CDF — dado p, retorna z
- Z-scores permitem comparar valores de diferentes distribuições normais.

## 💡 Curiosidades

A tabela Z tradicional, usada antes dos computadores, foi calculada manualmente por Karl Pearson no início do século XX. Levou anos para ser completada! A tabela lista apenas valores de Z positivos (de 0 a 3.99) usando a simetria da distribuição para valores negativos. ## ✅ Boas Práticas

1. Sempre use Z-scores para comparar valores de distribuições diferentes.
2. Lembre-se que Z negativo significa valor abaixo da média.
3. Use `stats.norm.ppf()` para encontrar valores críticos para intervalos de confiança.
4. A simetria da normal: P(Z < -z) = P(Z > z) = 1 - P(Z < z).
5. Documente a unidade dos dados ao reportar Z-scores.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Interpretar Z-score como percentil | Z = 2 não significa 2% — significa 2 desvios acima da média. |
| Esquecer de verificar se os dados são normais | Z-scores só são interpretáveis se a distribuição for aproximadamente normal. |
| Usar CDF quando precisa de PPF (e vice-versa) | CDF: probabilidade → valor. PPF: valor → probabilidade. |
| Confundir Z com percentil | Z = 1.96 corresponde ao percentil 97.5, não 95. |
