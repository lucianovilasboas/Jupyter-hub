## Corda de Pular

Imagine uma **corda de pular** esticada.
Uma ponta e o menor valor, a outra e o maior.
O comprimento da corda e o **intervalo** (range).

Exemplo: salarios de uma empresa
- Menor salario: R$ 1.500
- Maior salario: R$ 25.000
- **Intervalo = R$ 23.500**

O intervalo e facil de calcular e entender.
Mas ele so olha para os extremos — ignora tudo que esta no meio.
E muito sensivel a outliers: se o CEO ganhasse R$ 100.000,
o intervalo pularia para R$ 98.500!


## 📊 Traduzindo

| Corda de Pular | Estatistica |
|-|-|
| Corda esticada | Intervalo (max - min) |
| Ponta esquerda | Valor minimo |
| Ponta direita | Valor maximo |
| Comprimento da corda | Amplitude total |


## 📏 Limitacoes do Intervalo

Intervalo e simples, mas limitado:
- So usa 2 valores (min e max)
- Ignora a distribuicao interna
- Muito sensivel a outliers

Para uma visao mais completa, usamos
desvio padrao ou intervalo interquartil (IQR).


```python
# Corda de pular
dados = [1500, 1800, 2000, 2200, 2500, 3000, 25000]
range_val = max(dados) - min(dados)
print(f"Dados: {dados}")
print(f"Minimo: R$ {min(dados)}")
print(f"Maximo: R$ {max(dados)}")
print(f"Intervalo (range): R$ {range_val}")
```

## 📌 Resumo

| Corda de Pular | Estatistica |
|-|-|
| Corda esticada | Intervalo |
| Ponta esquerda | Minimo |
| Ponta direita | Maximo |
| Comprimento | Amplitude total |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
