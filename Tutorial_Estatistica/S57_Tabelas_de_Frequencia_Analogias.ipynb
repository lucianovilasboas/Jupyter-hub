## Inventario do Mercado

O gerente de um mercado precisa saber o que tem no estoque.
Ele nao vai listar um por um — faria um livro!
Em vez disso, ele faz um **inventario**:

| Produto | Quantidade |
|-------------------|------|
| Arroz | 200 pacotes |
| Feijao | 150 pacotes |
| Macarrao | 300 pacotes |
| Oleo | 100 garrafas |

Isso e uma **tabela de frequencia**!
Ela mostra quantas vezes cada valor aparece.

Em vez de listar todos os 750 itens individualmente,
o gerente ve o resumo em 4 linhas.


## 📊 Traduzindo

| Inventario do Mercado | Estatistica |
|-|-|
| Listar todos os 750 itens um por um | Dados brutos — todos os registros |
| Agrupar por tipo de produto | Criar categorias / classes |
| Contar quantos de cada | Frequencia absoluta |
| Calcular a porcentagem de cada | Frequencia relativa |
| "30% dos produtos sao macarrao" | Interpretacao da tabela |


## 📏 Por que Agrupar?

Listar todos os 750 itens e impossivel de analisar.
Agrupar em categorias revela padroes.

**Frequencia absoluta**: quantos de cada categoria.
**Frequencia relativa**: a proporcao (%)
**Frequencia acumulada**: quantos ate aquela categoria.

Com uma tabela de frequencia, uma olhada ja mostra o essencial.


```python
# Inventario do mercado em Python
produtos = ["arroz"]*200 + ["feijao"]*150 + ["macarrao"]*300 + ["oleo"]*100
total = len(produtos)

print(f"Total de itens: {total}")
print()
print("--- TABELA DE FREQUENCIA ---")
print("Produto          | Freq. Abs. | Freq. Rel.")
print("-"*40)
for p in ["arroz", "feijao", "macarrao", "oleo"]:
    freq_abs = produtos.count(p)
    freq_rel = freq_abs / total * 100
    print(f"{p:15s} | {freq_abs:9d} | {freq_rel:8.1f}%")
```

## 📌 Resumo

| Inventario do Mercado | Estatistica |
|-|-|
| Listar item por item | Dados brutos |
| Agrupar por tipo | Categorias / classes |
| Contar quantos de cada | Frequencia absoluta |
| Calcular % de cada | Frequencia relativa |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
