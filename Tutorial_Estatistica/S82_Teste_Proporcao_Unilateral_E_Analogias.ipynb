## Controle de Qualidade

Uma fabrica de parafusos exige que no maximo **2%** sejam defeituosos.
Se passar disso, a producao precisa ser ajustada.

Voce testa 500 parafusos e encontra **15 defeituosos (3%)**.

A preocupacao e so para **cima**:
se a taxa for maior que 2%, o lote e reprovado.
Se for menor ou igual, esta tudo bem.

Esse e um **teste unilateral a DIREITA** (na verdade).

Mas se o problema fosse: "o rendimento minimo aceitavel e 98%",
ai seria **unilateral a ESQUERDA** — so nos importa
se o valor esta **abaixo** do esperado.


## 📊 Traduzindo

| Controle de Qualidade | Estatistica |
|-|-|
| Maximo 2% defeituosos | H0: p <= 0,02 |
| 15 defeituosos em 500 (3%) | Proporcao observada |
| So importa se passou do limite | Teste unilateral |
| Reprovar o lote | Rejeitar H0 |


```python
# Controle de qualidade
import math
n = 500
defeitos = 15
p_obs = defeitos / n
p0 = 0.02  # limite maximo
erro = math.sqrt(p0 * (1-p0) / n)
z = (p_obs - p0) / erro
print(f"Defeituosos: {defeitos}/{n} = {p_obs*100:.1f}%")
print(f"Limite maximo (H0): {p0*100:.0f}%")
print(f"Z: {z:.2f}")
if z > 1.64:
    print("Acima do limite! Reprovar lote.")
else:
    print("Dentro do limite. Aprovado.")
```

## 📌 Resumo

| Controle de Qualidade | Estatistica |
|-|-|
| Limite maximo 2% | H0 |
| 3% observado | Proporcao amostral |
| So importa se >2% | Teste unilateral |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
