## 📝 Introdução

Na estatística descritiva, descrevemos os dados que temos. Mas muitas vezes queremos ir além: queremos **tirar conclusões** sobre uma população maior a partir de uma amostra. Isso é o campo da **Estatística Inferencial**.

Ao fazer isso, precisamos lidar com a **incerteza**. Uma amostra nunca nos dará uma certeza absoluta sobre a população — sempre existe uma margem de erro. Neste capítulo, veremos como quantificar essa incerteza e fazer conclusões fundamentadas.

## 📚 Explicação Detalhada

### 1. População vs Amostra

- **População**: Todo o grupo que queremos estudar (ex: todos os brasileiros).
- **Amostra**: Um subconjunto da população (ex: 1000 brasileiros entrevistados).

Idealmente, a amostra deve ser **representativa** da população.

### 2. Inferência Estatística

É o processo de usar dados de uma amostra para estimar características da população. Exemplos:
- Estimar a média de altura dos brasileiros a partir de uma amostra.
- Estimar a intenção de voto a partir de uma pesquisa eleitoral.

### 3. Incerteza e Margem de Erro

Toda estimativa tem incerteza. A **margem de erro** quantifica essa incerteza:
- Uma pesquisa eleitoral pode dizer: "Candidato A tem 45% das intenções de voto, com margem de erro de 3 pontos percentuais".
- Isso significa que o valor real provavelmente está entre 42% e 48%.

### 4. Teorema do Limite Central

Um dos conceitos mais importantes da estatística: a distribuição das médias amostrais tende a ser normal (em forma de sino) conforme o tamanho da amostra aumenta, independentemente da distribuição original dos dados.

```python
# Simulação simples: estimando a média de uma população a partir de amostras
import random
import matplotlib.pyplot as plt

random.seed(42)

# População: 10000 números com distribuição uniforme entre 0 e 100
populacao = [random.uniform(0, 100) for _ in range(10000)]
media_real = sum(populacao) / len(populacao)
print(f"Média real da população: {media_real:.2f}")

# Colhemos uma amostra de 50 elementos
amostra = random.sample(populacao, 50)
media_amostral = sum(amostra) / len(amostra)
print(f"Média da amostra (n=50): {media_amostral:.2f}")
print(f"Erro (diferença): {abs(media_real - media_amostral):.2f}")
```

```python
# Demonstração do Teorema do Limite Central
# Coletamos várias amostras e calculamos a média de cada uma

random.seed(123)
medias_amostrais = []
n_amostras = 1000
tamanho_amostra = 30

for _ in range(n_amostras):
    amostra = random.sample(populacao, tamanho_amostra)
    medias_amostrais.append(sum(amostra) / len(amostra))

# A distribuição das médias amostrais se aproxima de uma normal
media_das_medias = sum(medias_amostrais) / len(medias_amostrais)
print(f"Média das {n_amostras} médias amostrais: {media_das_medias:.2f}")
print(f"Média real da população: {media_real:.2f}")

# Visualizando
plt.hist(medias_amostrais, bins=30, color='lightgreen', edgecolor='black')
plt.axvline(media_real, color='red', linestyle='--', label='Média real')
plt.axvline(media_das_medias, color='blue', linestyle='--', label='Média das médias')
plt.title('Distribuição das Médias Amostrais (Teorema do Limite Central)')
plt.xlabel('Média amostral')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

## 💻 Exemplos Práticos

```python
# Exemplo prático: simulando uma pesquisa de intenção de voto
import random

random.seed(456)

# População: 100000 eleitores com preferência real de 55% para o candidato A
populacao_voto = ['A'] * 55000 + ['B'] * 45000
random.shuffle(populacao_voto)

def pesquisa_intencao(tamanho_amostra=1000):
    """Simula uma pesquisa de intenção de voto."""
    amostra = random.sample(populacao_voto, tamanho_amostra)
    votos_a = amostra.count('A')
    return votos_a / tamanho_amostra * 100

print("=== Simulação de Pesquisa Eleitoral ===")
print(f"Intenção real do candidato A: 55%")
print()

for n in [100, 500, 1000, 2000]:
    resultado = pesquisa_intencao(n)
    erro = abs(resultado - 55)
    print(f"Amostra de {n:5d} eleitores: {resultado:.1f}% (erro: {erro:.1f}pp)")
```

```python
# Simulando margem de erro com múltiplas pesquisas
random.seed(789)

resultados = []
for _ in range(100):
    amostra = random.sample(populacao_voto, 1000)
    p = amostra.count('A') / 1000 * 100
    resultados.append(p)

# Calculamos intervalo que contém 95% dos resultados
resultados_ord = sorted(resultados)
inf = resultados_ord[2]   # 2.5%
sup = resultados_ord[96]  # 97.5%

print("=== Margem de Erro (100 pesquisas simuladas) ===")
print(f"95% dos resultados estão entre {inf:.1f}% e {sup:.1f}%")
print(f"Valor real: 55%")
print(f"Margem de erro aproximada: {sup - 55:.1f}pp")

dentro = sum(1 for r in resultados if 52 <= r <= 58)
print(f"Pesquisas dentro da margem de 3pp: {dentro}/100")
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Uma amostra de 200 clientes mostra que 60% estão satisfeitos. Qual é a margem de erro aproximada (para 95% de confiança, use a fórmula 1/√n)?

```python
# Solução do Exercício 1
import math
n = 200
margem_erro = 1 / math.sqrt(n)
print(f"Para n = {n}:")
print(f"Margem de erro ≈ {margem_erro:.1%}")
print(f"IC aproximado (95%): 60% ± {margem_erro:.1%}")
print(f"Ou seja, entre {60 - margem_erro*100:.1f}% e {60 + margem_erro*100:.1f}%")
```

**Exercício 2:** Simule 500 amostras de tamanho 100 de uma população com média 50 e desvio padrão 10. Calcule a média das médias amostrais e compare com a média real.

```python
# Solução do Exercício 2
random.seed(111)
pop = [random.gauss(50, 10) for _ in range(100000)]
media_pop = sum(pop) / len(pop)

medias_500 = []
for _ in range(500):
    amostra = random.sample(pop, 100)
    medias_500.append(sum(amostra) / len(amostra))

media_das_500 = sum(medias_500) / len(medias_500)
print(f"Média da população: {media_pop:.2f}")
print(f"Média das 500 médias amostrais: {media_das_500:.2f}")
print(f"Diferença: {abs(media_pop - media_das_500):.4f}")
```

## 📌 Resumo

- **Inferência**: usar amostras para concluir sobre populações.
- **Margem de erro**: quantifica a incerteza das estimativas.
- **Teorema do Limite Central**: médias amostrais tendem à normalidade.
- Amostras maiores → estimativas mais precisas.
- Simulações em Python ajudam a visualizar conceitos de inferência.

## 💡 Curiosidades

O **Teorema do Limite Central** é considerado um dos pilares da estatística moderna. Foi demonstrado pela primeira vez por Abraham de Moivre em 1733, para o caso especial de variáveis binárias (moedas). Depois, Pierre-Simon Laplace generalizou o teorema em 1810. Hoje, ele justifica porque tantos fenômenos naturais seguem distribuições aproximadamente normais: sempre que um resultado é a soma de muitos fatores independentes, a distribuição tende à normalidade.

## ✅ Boas Práticas

1. Sempre reporte a margem de erro junto com suas estimativas.
2. Especifique o nível de confiança (geralmente 95%).
3. Amostras aleatórias são fundamentais para inferência válida.
4. Não generalize além do que sua amostra representa.
5. Use simulações para verificar a robustez das suas conclusões.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Ignorar a margem de erro | Sempre reporte incerteza junto com a estimativa. |
| Generalizar para além da população amostrada | A amostra representa apenas a população da qual foi retirada. |
| Confundir significância estatística com importância prática | Um resultado pode ser significativo mas irrelevante na prática. |
| Acreditar que uma amostra grande elimina viés | Viés é um erro sistemático, não se reduz com tamanho da amostra. |
