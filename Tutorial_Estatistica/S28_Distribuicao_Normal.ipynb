## 📝 Introdução

A distribuição normal, também conhecida como distribuição Gaussiana, é a mais importante e mais utilizada em toda a estatística. Sua curva característica em forma de sino aparece naturalmente em inúmeros fenômenos: alturas de pessoas, erros de medição, QI, pressão arterial, e muitos outros.

O nome "Gaussiana" homenageia **Carl Friedrich Gauss** (1777-1855), que a usou para analisar dados astronômicos. Porém, a distribuição já havia sido estudada antes por **Abraham de Moivre** em 1733.

## 📚 Explicação Detalhada

### 1. O que é a Distribuição Normal?

A distribuição normal é uma distribuição de probabilidade contínua, simétrica em torno da média, descrita pela função:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

Onde:
- **μ (mu)**: Média — define o centro da curva
- **σ (sigma)**: Desvio padrão — define a dispersão (largura) da curva

### 2. Parâmetros μ e σ

- **μ** controla a posição do pico da curva. Se μ aumenta, a curva se desloca para a direita.
- **σ** controla achatamento da curva. σ pequeno → curva mais alta e estreita (dados concentrados). σ grande → curva mais baixa e larga (dados dispersos).

### 3. Propriedades da Curva Normal

1. **Simetria**: A curva é simétrica em torno da média (μ).
2. **Média = Mediana = Moda**: Os três coincidem no centro.
3. **Assíntotas**: A curva nunca toca o eixo x — aproxima-se dele infinitamente.
4. **Área total = 1**: A área sob a curva representa 100% da probabilidade.
5. **Inflection points**: A curva muda de concavidade em x = μ ± σ.

### 4. Regra Empírica (68-95-99.7)

Para qualquer distribuição normal:
- **68%** dos dados estão a 1 desvio padrão da média (μ ± 1σ)
- **95%** dos dados estão a 2 desvios padrão da média (μ ± 2σ)
- **99.7%** dos dados estão a 3 desvios padrão da média (μ ± 3σ)

```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Bibliotecas importadas com sucesso!")
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Plotando a curva normal com diferentes parâmetros
# Mostramos como μ e σ afetam a forma da distribuição

x = np.linspace(-10, 20, 500)

# Diferentes combinações de μ e σ
parametros = [(0, 1), (5, 1), (5, 2), (8, 1.5)]
cores = ['blue', 'red', 'green', 'purple']
legendas = []

plt.figure(figsize=(12, 5))

for (mu, sigma), cor in zip(parametros, cores):
    y = stats.norm.pdf(x, mu, sigma)
    plt.plot(x, y, color=cor, lw=2)
    legendas.append(f'μ={mu}, σ={sigma}')

plt.title('Curvas Normais com Diferentes Parâmetros μ e σ', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Densidade de Probabilidade (PDF)', fontsize=12)
plt.legend(legendas)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Exemplo 2: Gerando dados normalmente distribuídos
# Usamos np.random.normal para simular alturas de pessoas

np.random.seed(42)

# Altura de homens brasileiros: μ ≈ 1.73m, σ ≈ 0.07m
altura_media = 1.73
desvio_altura = 0.07
n_pessoas = 10000

# Geramos os dados
alturas = np.random.normal(loc=altura_media, scale=desvio_altura, size=n_pessoas)

# Estatísticas
media_obs = np.mean(alturas)
desvio_obs = np.std(alturas)

print(f"Média observada: {media_obs:.4f} m (esperado: {altura_media})")
print(f"Desvio padrão observado: {desvio_obs:.4f} m (esperado: {desvio_altura})")

# Histograma vs curva teórica
plt.figure(figsize=(10, 5))
count, bins, _ = plt.hist(alturas, bins=50, density=True, alpha=0.6, 
                           color='skyblue', edgecolor='black', label='Dados simulados')

x = np.linspace(alturas.min(), alturas.max(), 200)
y = stats.norm.pdf(x, altura_media, desvio_altura)
plt.plot(x, y, 'r-', lw=2, label='Curva normal teórica')

plt.title('Distribuição das Alturas (Dados Simulados)', fontsize=14)
plt.xlabel('Altura (m)', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Exemplo 3: Demonstração da Regra Empírica (68-95-99.7)
# Verificamos quantos dados caem dentro de 1, 2 e 3 desvios padrão

np.random.seed(42)

mu = 100  # Média (ex.: QI)
sigma = 15  # Desvio padrão
dados = np.random.normal(loc=mu, scale=sigma, size=100000)

# Calculamos as proporções observadas
dentro_1sigma = np.sum(np.abs(dados - mu) <= 1 * sigma) / len(dados) * 100
dentro_2sigma = np.sum(np.abs(dados - mu) <= 2 * sigma) / len(dados) * 100
dentro_3sigma = np.sum(np.abs(dados - mu) <= 3 * sigma) / len(dados) * 100

print("=== Regra Empírica (68-95-99.7) ===")
print(f"μ ± 1σ: Esperado 68%, Observado {dentro_1sigma:.2f}%")
print(f"μ ± 2σ: Esperado 95%, Observado {dentro_2sigma:.2f}%")
print(f"μ ± 3σ: Esperado 99.7%, Observado {dentro_3sigma:.2f}%")

# Visualização
fig, ax = plt.subplots(figsize=(12, 5))

# Histograma
ax.hist(dados, bins=80, density=True, alpha=0.5, color='gray', edgecolor='black')

# Curva normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 300)
y = stats.norm.pdf(x, mu, sigma)
ax.plot(x, y, 'b-', lw=2)

# Regiões coloridas
cores = ['gold', 'orange', 'red']
labels = ['68% (1σ)', '95% (2σ)', '99.7% (3σ)']

for i, k in enumerate([1, 2, 3]):
    mask = (x >= mu - k*sigma) & (x <= mu + k*sigma)
    ax.fill_between(x, y, where=mask, color=cores[i], alpha=0.3, label=labels[i])

ax.axvline(mu, color='black', linestyle='--', label=f'μ = {mu}')
ax.set_title('Regra Empírica (68-95-99.7) na Distribuição Normal', fontsize=14)
ax.set_xlabel('Valor', fontsize=12)
ax.set_ylabel('Densidade', fontsize=12)
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```

```python
# Exemplo 4: Probabilidades na distribuição normal
# Usando scipy.stats.norm para calcular áreas sob a curva

from scipy import stats

mu = 100  # QI médio
sigma = 15

# Probabilidade de QI < 120
prob_menos_120 = stats.norm.cdf(120, mu, sigma)

# Probabilidade de QI > 130
prob_mais_130 = 1 - stats.norm.cdf(130, mu, sigma)

# Probabilidade de QI entre 85 e 115
prob_entre_85_115 = stats.norm.cdf(115, mu, sigma) - stats.norm.cdf(85, mu, sigma)

print(f"P(QI < 120) = {prob_menos_120:.4f} ({prob_menos_120*100:.2f}%)")
print(f"P(QI > 130) = {prob_mais_130:.4f} ({prob_mais_130*100:.2f}%)")
print(f"P(85 < QI < 115) = {prob_entre_85_115:.4f} ({prob_entre_85_115*100:.2f}%)")
```

```python
# Exemplo 5: Visualizando a área sob a curva para uma probabilidade

mu, sigma = 100, 15
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = stats.norm.pdf(x, mu, sigma)

# Região: QI > 130
limite = 130
prob = 1 - stats.norm.cdf(limite, mu, sigma)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', lw=2, label='Curva normal')

mask = x >= limite
plt.fill_between(x, y, where=mask, color='coral', alpha=0.5, 
                 label=f'P(X > {limite}) = {prob:.4f}')

plt.axvline(limite, color='red', linestyle='--', lw=1.5)
plt.title('Probabilidade como Área sob a Curva Normal', fontsize=14)
plt.xlabel('QI', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## ✅ Exercícios Resolvidos

**Exercício 1:** A pressão arterial sistólica de adultos segue uma distribuição normal com μ = 120 mmHg e σ = 15 mmHg. Qual a porcentagem de adultos com pressão entre 105 e 135 mmHg?

```python
# Solução Exercício 1
from scipy import stats

mu = 120
sigma = 15

# P(105 < X < 135)
prob = stats.norm.cdf(135, mu, sigma) - stats.norm.cdf(105, mu, sigma)
print(f"P(105 < X < 135) = {prob:.4f} = {prob*100:.2f}%")
print(f"Note: 105 e 135 estão a ±1σ da média, então ≈ 68%")
```

**Exercício 2:** O tempo de vida de uma lâmpada segue distribuição normal com μ = 2000 horas e σ = 250 horas. Qual a probabilidade de uma lâmpada durar mais de 2500 horas?

```python
# Solução Exercício 2
mu = 2000
sigma = 250

prob_mais_2500 = 1 - stats.norm.cdf(2500, mu, sigma)
print(f"P(X > 2500) = {prob_mais_2500:.4f} = {prob_mais_2500*100:.2f}%")

# Visualização
x = np.linspace(1000, 3000, 400)
y = stats.norm.pdf(x, mu, sigma)

plt.figure(figsize=(10, 4))
plt.plot(x, y, 'b-', lw=2)
mask = x >= 2500
plt.fill_between(x, y, where=mask, color='coral', alpha=0.5, label=f'{prob_mais_2500*100:.2f}%')
plt.axvline(mu, color='gray', linestyle='--', alpha=0.5)
plt.title('Probabilidade de durar mais de 2500 horas')
plt.xlabel('Horas de vida')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 📌 Resumo

- A **distribuição normal** é uma distribuição contínua, simétrica, em forma de sino.
- Parâmetros: **μ** (centro) e **σ** (dispersão).
- Propriedades: simétrica, média=mediana=moda, assíntotas, área total = 1.
- **Regra empírica**: 68% (1σ), 95% (2σ), 99.7% (3σ) dos dados.
- Muitos fenômenos naturais seguem (aproximadamente) a distribuição normal.

## 💡 Curiosidades

A distribuição normal aparece em lugares surpreendentes: a distribuição das alturas de soldados na França do século XIX (estudada por Laplace), os erros de tiro em competições de tiro ao alvo, e até mesmo na distribuição de características biológicas como o tamanho de folhas de uma mesma espécie de planta. O matemático Francis Galton chamou a curva normal de "a rainha das distribuições" pela sua onipresença na natureza.

## ✅ Boas Práticas

1. Antes de assumir normalidade, faça um teste visual (histograma + curva) ou formal (Shapiro-Wilk).
2. A regra empírica é uma aproximação — use CDF para cálculos precisos.
3. Mencione sempre μ e σ ao descrever dados com distribuição normal.
4. Cuidado: dados reais raramente são perfeitamente normais — normalmente são "aproximadamente normais".
5. Use `np.random.normal` com `seed` para garantir reprodutibilidade.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Assumir que todos os dados são normais | Nem tudo segue distribuição normal — receita, renda, tempo de espera são assimétricos. |
| Confundir a altura da curva com probabilidade | A probabilidade é a **área**, não a altura em um ponto. |
| Usar regra empírica sem verificar normalidade | A regra 68-95-99.7 só vale para distribuições normais. |
| Achar que a média define a forma da curva | A forma é definida por σ; μ apenas desloca a curva. |
