## 📝 Introdução

O **box plot** (gráfico de caixa ou diagrama de caixa e bigodes) é uma poderosa ferramenta visual que resume a distribuição de dados numéricos através de cinco medidas: **mínimo**, **primeiro quartil (Q1)**, **mediana (Q2)**, **terceiro quartil (Q3)** e **máximo** (ou limites dos bigodes).

Criado por **John Tukey** na década de 1970, o box plot é excelente para:
- Visualizar a dispersão dos dados de forma compacta
- Identificar outliers (valores atípicos)
- Comparar distribuições entre grupos
- Avaliar simetria e assimetria

## 📚 Explicação Detalhada

### 1. Estrutura do Box Plot

Uma caixa (box) representa o intervalo interquartil (IQR = Q3 - Q1), que contém 50% dos dados:
- **Linha inferior da caixa**: Q1 (25º percentil)
- **Linha no meio da caixa**: Mediana (Q2, 50º percentil)
- **Linha superior da caixa**: Q3 (75º percentil)
- **Bigodes (whiskers)**: vão até 1.5×IQR além de Q1 e Q3
- **Pontos além dos bigodes**: outliers

### 2. Interpretação

- **Caixa curta**: dados concentrados (baixa variabilidade)
- **Caixa longa**: dados dispersos (alta variabilidade)
- **Mediana centralizada**: distribuição simétrica
- **Mediana deslocada**: distribuição assimétrica
- **Muitos outliers**: possíveis valores atípicos ou caudas pesadas

### 3. Vantagens do Box Plot

- Ocupa pouco espaço
- Permite comparar muitos grupos lado a lado
- Robusto a outliers (diferente da média/desvio padrão)
- Revela assimetria, dispersão e valores extremos simultaneamente

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
np.random.seed(42)
print('Bibliotecas importadas!')
```

```python
# Criando dados de exemplo
# Notas de alunos de 3 turmas diferentes
turma_a = np.random.normal(loc=70, scale=10, size=100)
turma_b = np.random.normal(loc=60, scale=15, size=100)
turma_c = np.concatenate([
    np.random.normal(loc=75, scale=5, size=80),
    np.random.normal(loc=40, scale=5, size=20)  # grupo com notas baixas
])

dados = [turma_a, turma_b, turma_c]
rotulos = ['Turma A', 'Turma B', 'Turma C']

for turma, dados_t in zip(rotulos, dados):
    print(f'{turma}: Média={dados_t.mean():.1f}, DP={dados_t.std():.1f}, Mediana={np.median(dados_t):.1f}')
```

```python
# Box plot básico com matplotlib
plt.figure(figsize=(10, 6))
bp = plt.boxplot(dados, labels=rotulos, patch_artist=True,
                 boxprops=dict(facecolor='lightblue', alpha=0.7),
                 medianprops=dict(color='red', linewidth=2),
                 whiskerprops=dict(linewidth=1.5),
                 capprops=dict(linewidth=1.5))
plt.title('Box Plot — Notas por Turma')
plt.ylabel('Nota')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Box plot com seaborn (mais bonito)
df_box = pd.DataFrame({
    'Nota': np.concatenate([turma_a, turma_b, turma_c]),
    'Turma': ['A']*100 + ['B']*100 + ['C']*100
})

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_box, x='Turma', y='Nota', palette='Set2')
plt.title('Box Plot — Notas por Turma (seaborn)')
plt.ylabel('Nota')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Box plot horizontal
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_box, y='Turma', x='Nota', palette='Set2', orient='h')
plt.title('Box Plot Horizontal — Notas por Turma')
plt.xlabel('Nota')
plt.grid(axis='x', alpha=0.3)
plt.show()
```

```python
# Interpretação detalhada de um box plot
np.random.seed(42)
dados_ex = np.random.normal(loc=50, scale=15, size=200)
# Adicionando alguns outliers
dados_ex = np.append(dados_ex, [5, 10, 95, 98])

q1 = np.percentile(dados_ex, 25)
q2 = np.median(dados_ex)
q3 = np.percentile(dados_ex, 75)
iqr = q3 - q1
lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr
outliers = dados_ex[(dados_ex < lim_inf) | (dados_ex > lim_sup)]

print('=== Análise do Box Plot ===')
print(f'Q1 (25%): {q1:.2f}')
print(f'Mediana (Q2): {q2:.2f}')
print(f'Q3 (75%): {q3:.2f}')
print(f'IQR: {iqr:.2f}')
print(f'Limite inferior (bigode): {lim_inf:.2f}')
print(f'Limite superior (bigode): {lim_sup:.2f}')
print(f'Número de outliers: {len(outliers)}')
print(f'Outliers: {np.sort(outliers)}')

plt.figure(figsize=(8, 4))
plt.boxplot(dados_ex, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot com Anotação dos Componentes')
plt.xlabel('Valor')
plt.grid(axis='x', alpha=0.3)
# Anotações
plt.annotate('Outliers', xy=(5, 1.05), fontsize=10, ha='center', color='red')
plt.annotate(f'Q1={q1:.1f}', xy=(q1, 1.15), fontsize=9, ha='center')
plt.annotate(f'Mediana={q2:.1f}', xy=(q2, 1.15), fontsize=9, ha='center')
plt.annotate(f'Q3={q3:.1f}', xy=(q3, 1.15), fontsize=9, ha='center')
plt.show()
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Comparando múltiplos grupos
np.random.seed(42)
grupos = ['Grupo A', 'Grupo B', 'Grupo C', 'Grupo D', 'Grupo E']
medias = [50, 55, 60, 65, 70]
dados_grupos = [np.random.normal(loc=loc, scale=12, size=80) for loc in medias]

df_grupos = pd.DataFrame({
    'Valor': np.concatenate(dados_grupos),
    'Grupo': np.repeat(grupos, 80)
})

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_grupos, x='Grupo', y='Valor', palette='viridis')
plt.title('Comparação de 5 Grupos com Box Plot')
plt.ylabel('Valor')
plt.grid(axis='y', alpha=0.3)
plt.show()
print('Observe como a mediana aumenta de A para E, mas a dispersão é similar.')
```

```python
# Exemplo 2: Box plot com distribuição sobreposta (violin plot)
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_box, x='Turma', y='Nota', palette='Set2', inner='box')
plt.title('Violin Plot — Distribuição + Box Plot')
plt.ylabel('Nota')
plt.grid(axis='y', alpha=0.3)
plt.show()
print('O violin plot combina box plot com KDE, mostrando a forma completa da distribuição.')
```

```python
# Exemplo 3: Identificando assimetria com box plot
np.random.seed(7)
simetrico = np.random.normal(loc=50, scale=10, size=200)
assimetrico_pos = np.random.exponential(scale=15, size=200)
assimetrico_neg = 100 - np.random.exponential(scale=15, size=200)

fig, axes = plt.subplots(1, 3, figsize=(14, 5))
for ax, dados, tit in zip(axes, [simetrico, assimetrico_pos, assimetrico_neg],
                           ['Simétrica', 'Assimetria Positiva', 'Assimetria Negativa']):
    ax.boxplot(dados, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
    ax.set_title(tit)
    ax.set_ylabel('Valor')
    ax.grid(axis='y', alpha=0.3)
    # Mostrar média vs mediana
    ax.axhline(np.mean(dados), color='red', linestyle='--', linewidth=1, label=f'Média={np.mean(dados):.1f}')
    ax.axhline(np.median(dados), color='green', linestyle='-', linewidth=1, label=f'Mediana={np.median(dados):.1f}')
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Gere 4 distribuições normais com médias 40, 50, 60, 70 e desvio padrão 10 (n=100 cada). Crie um box plot comparando os 4 grupos. Interprete os resultados.

```python
# Solução do Exercício 1
np.random.seed(42)
medias = [40, 50, 60, 70]
grupos = [f'Grupo {i+1} (μ={m})' for i, m in enumerate(medias)]
dados_g = [np.random.normal(loc=m, scale=10, size=100) for m in medias]

plt.figure(figsize=(10, 6))
plt.boxplot(dados_g, labels=grupos, patch_artist=True,
            boxprops=dict(facecolor='skyblue', alpha=0.7),
            medianprops=dict(color='red', linewidth=2))
plt.title('Comparação de 4 Grupos com Box Plot')
plt.ylabel('Valor')
plt.grid(axis='y', alpha=0.3)
plt.show()
print('Interpretação: As medianas aumentam gradualmente.')
print('A dispersão é similar entre grupos (mesmo σ=10).')
print('A assimetria é pequena em todos os grupos (distribuições simétricas).')
```

**Exercício 2:** Crie um dataset com 3 grupos, sendo um deles com outliers intencionais. Identifique os outliers visualmente.

```python
# Solução do Exercício 2
np.random.seed(42)
g1 = np.random.normal(60, 8, 100)
g2 = np.random.normal(65, 10, 100)
g3 = np.random.normal(70, 6, 100)
# Adicionando outliers
g3 = np.append(g3, [95, 98, 100, 30, 25])

df_out = pd.DataFrame({
    'Valor': np.concatenate([g1, g2, g3]),
    'Grupo': ['G1 (sem outliers)']*100 + ['G2 (sem outliers)']*100 + ['G3 (com outliers)']*105
})

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_out, x='Grupo', y='Valor', palette='Set2')
plt.title('Box Plot — Identificando Outliers')
plt.ylabel('Valor')
plt.grid(axis='y', alpha=0.3)
plt.show()
total_out = len(df_out[df_out['Grupo'] == 'G3 (com outliers)'])
print(f'G3: {total_out} observações, incluindo 5 outliers intencionais.')
print('Os outliers aparecem como pontos além dos bigodes no box plot.')
```

## 📌 Resumo

- **Box plot** resume a distribuição em 5 números: mínimo, Q1, mediana, Q3, máximo.
- **Caixa (box)**: IQR (intervalo interquartil — 50% dos dados).
- **Bigodes (whiskers)**: até 1.5×IQR além de Q1 e Q3.
- **Outliers**: pontos além dos bigodes.
- **Mediana centralizada** = simetria. **Deslocada** = assimetria.
- Excelente para **comparar múltiplos grupos** lado a lado.
- `plt.boxplot()` e `sns.boxplot()` no Python.

## 💡 Curiosidades

O **box plot** foi criado por **John Tukey** (1915-2000), um dos maiores estatísticos do século XX. Tukey também inventou:
- O termo **"bit"** (binary digit)
- O algoritmo **FFT** (Fast Fourier Transform)
- O teste **Tukey HSD** para comparações múltiplas
- A técnica **Jackknife** para estimação de viés

Tukey era conhecido por sua filosofia prática: "A combinação de algumas visualizações e algumas cálculos geralmente é pior ou melhor do que qualquer um deles sozinho." Ele defendia a **Análise Exploratória de Dados** (EDA) como abordagem fundamental.

## ✅ Boas Práticas

1. **Sempre inclua o box plot** ao comparar grupos — ele revela mais que médias isoladas.
2. **Use box plots horizontais** para muitos grupos (melhor leitura dos rótulos).
3. **Combine com swarm plot** para pequenas amostras (ver todos os pontos).
4. **Considere violin plot** quando a forma da distribuição for importante.
5. **Documente** se removeu outliers e por quê.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Ignorar outliers no box plot | Investigue cada outlier — pode ser erro ou descoberta. |
| Confundir bigodes com min/max | Bigodes são limitados a 1.5×IQR, não ao valor extremo. |
| Usar box plot com amostras muito pequenas (<10) | Prefira gráfico de pontos ou swarm plot. |
| Não verificar a escala dos eixos | Eixos com escalas diferentes podem enganar. |
