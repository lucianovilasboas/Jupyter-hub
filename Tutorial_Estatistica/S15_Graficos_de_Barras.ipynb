## 📝 Introdução

Os **gráficos de barras** representam dados categóricos através de barras retangulares cuja altura (ou comprimento) é proporcional ao valor que representam. Diferente dos histogramas (dados contínuos), os gráficos de barras comparam **categorias distintas**.

Neste capítulo, vamos explorar os tipos mais comuns de gráficos de barras e como criá-los em Python.

## 📚 Explicação Detalhada

### 1. Barras Verticais vs Horizontais

- **Verticais**: comuns, boas para nomes curtos
- **Horizontais**: ideais para muitas categorias ou nomes longos

### 2. Barras Simples

Cada barra = uma categoria. Útil para comparar magnitudes.

### 3. Barras Agrupadas (Grouped)

Duas ou mais séries lado a lado. Para comparações do tipo "antes vs depois".

### 4. Barras Empilhadas (Stacked)

Séries empilhadas. Mostram a composição total de cada categoria.

### 5. Quando Usar

- Comparar valores entre categorias distintas
- Mostrar rankings (ordene as barras!)
- Apresentar contagens ou proporções
- Comparar partes de um todo (barras empilhadas)

**Evite quando:** dados contínuos (use histograma), mais de 10-15 categorias, diferenças muito pequenas.

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
np.random.seed(42)
print('Bibliotecas importadas!')
```

```python
# Criando dados de exemplo — vendas por produto
produtos = ['Notebook', 'Smartphone', 'Tablet', 'Monitor', 'Teclado', 'Mouse', 'Webcam']
vendas_2023 = np.random.randint(50, 300, size=len(produtos))
vendas_2024 = (vendas_2023 * np.random.uniform(1.05, 1.35, size=len(produtos))).astype(int)

df = pd.DataFrame({'Produto': produtos, 'Vendas_2023': vendas_2023, 'Vendas_2024': vendas_2024})
print(df)
```

```python
# Barras verticais simples
plt.figure(figsize=(10, 5))
plt.bar(df['Produto'], df['Vendas_2024'], color='royalblue', edgecolor='black')
plt.title('Vendas por Produto — 2024')
plt.xlabel('Produto')
plt.ylabel('Unidades Vendidas')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Barras horizontais (melhor para muitos itens)
plt.figure(figsize=(10, 5))
plt.barh(df['Produto'], df['Vendas_2024'], color='coral', edgecolor='black')
plt.title('Vendas por Produto — 2024')
plt.xlabel('Unidades Vendidas')
plt.grid(axis='x', alpha=0.3)
plt.show()
```

```python
# Barras ordenadas (ranking)
df_ordenado = df.sort_values('Vendas_2024', ascending=False)
plt.figure(figsize=(10, 5))
plt.bar(df_ordenado['Produto'], df_ordenado['Vendas_2024'], color='seagreen', edgecolor='black')
plt.title('Ranking de Vendas — 2024')
plt.xlabel('Produto')
plt.ylabel('Unidades Vendidas')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Barras agrupadas (grouped)
x = np.arange(len(produtos))
largura = 0.35

plt.figure(figsize=(10, 5))
plt.bar(x - largura/2, df['Vendas_2023'], largura, label='2023', color='lightblue', edgecolor='black')
plt.bar(x + largura/2, df['Vendas_2024'], largura, label='2024', color='royalblue', edgecolor='black')
plt.xticks(x, produtos)
plt.title('Comparação de Vendas: 2023 vs 2024')
plt.xlabel('Produto')
plt.ylabel('Unidades Vendidas')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Barras empilhadas (stacked)
plt.figure(figsize=(10, 5))
plt.bar(produtos, df['Vendas_2023'], label='2023', color='lightcoral', edgecolor='black')
plt.bar(produtos, df['Vendas_2024'], bottom=df['Vendas_2023'], label='2024', color='indianred', edgecolor='black')
plt.title('Vendas Acumuladas: 2023 + 2024')
plt.xlabel('Produto')
plt.ylabel('Total de Unidades')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
print('A barra total mostra a soma dos dois anos para cada produto.')
```

```python
# Barplot com seaborn (mostra média com barra de erro)
# Dados no formato longo para seaborn
df_long = pd.DataFrame({
    'Produto': produtos * 2,
    'Ano': ['2023'] * 7 + ['2024'] * 7,
    'Vendas': list(df['Vendas_2023']) + list(df['Vendas_2024'])
})

plt.figure(figsize=(10, 5))
sns.barplot(data=df_long, x='Produto', y='Vendas', hue='Ano', palette='Set2')
plt.title('Vendas por Produto e Ano (seaborn barplot)')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Countplot com seaborn (frequência de categorias)
np.random.seed(42)
categorias = np.random.choice(['Péssimo', 'Ruim', 'Regular', 'Bom', 'Ótimo'], size=200, p=[0.05, 0.10, 0.20, 0.40, 0.25])

plt.figure(figsize=(8, 5))
ordem = ['Péssimo', 'Ruim', 'Regular', 'Bom', 'Ótimo']
sns.countplot(y=categorias, order=ordem, palette='RdYlGn')
plt.title('Distribuição de Respostas — Pesquisa de Satisfação')
plt.xlabel('Contagem')
plt.ylabel('Categoria')
plt.show()
print('Countplot mostra a frequência de cada categoria automaticamente.')
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Vendas por região (barras horizontais)
regioes = ['Norte', 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Sul']
vendas_reg = np.random.randint(100, 500, size=5)

plt.figure(figsize=(10, 5))
cores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
plt.barh(regioes, vendas_reg, color=cores, edgecolor='black')
plt.title('Vendas por Região')
plt.xlabel('Unidades Vendidas')
for i, v in enumerate(vendas_reg):
    plt.text(v + 5, i, str(v), va='center', fontweight='bold')
plt.grid(axis='x', alpha=0.3)
plt.show()
```

```python
# Exemplo 2: Proporções com barras empilhadas normalizadas
np.random.seed(100)
turnos = ['Manhã', 'Tarde', 'Noite']
generos_turno = {t: np.random.choice(['Feminino', 'Masculino'], size=100, p=[0.5, 0.5]) for t in turnos}

df_turno = pd.DataFrame({t: pd.Series(g).value_counts() for t, g in generos_turno.items()}).T
df_turno_pct = df_turno.div(df_turno.sum(axis=1), axis=0) * 100

plt.figure(figsize=(10, 5))
b1 = plt.bar(turnos, df_turno_pct['Feminino'], label='Feminino', color='lightpink', edgecolor='black')
b2 = plt.bar(turnos, df_turno_pct['Masculino'], bottom=df_turno_pct['Feminino'], label='Masculino', color='lightblue', edgecolor='black')
plt.title('Proporção de Gênero por Turno')
plt.ylabel('Porcentagem (%)')
plt.legend()
for i, (t, row) in enumerate(df_turno_pct.iterrows()):
    plt.text(i, row['Feminino']/2, f'{row["Feminino"]:.0f}%', ha='center', fontweight='bold')
    plt.text(i, row['Feminino'] + row['Masculino']/2, f'{row["Masculino"]:.0f}%', ha='center', fontweight='bold')
plt.show()
```

```python
# Solução do Exercício 1
paises = ['Brasil', 'EUA', 'China', 'Índia', 'Indonésia']
pop = [213, 331, 1412, 1380, 273]
df_pop = pd.DataFrame({'País': paises, 'População': pop}).sort_values('População', ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(df_pop['País'], df_pop['População'], color=['gold', 'coral', 'royalblue', 'seagreen', 'orange'], edgecolor='black')
plt.title('População por País (milhões)')
plt.ylabel('População (milhões)')
for i, v in enumerate(df_pop['População']):
    plt.text(i, v + 10, f'{v}M', ha='center', fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## 📌 Resumo

- **Gráficos de barras** comparam categorias distintas.
- **Verticais**: padrão. **Horizontais**: melhores para muitas categorias.
- **Agrupadas**: comparam séries lado a lado.
- **Empilhadas**: mostram composição total.
- Seaborn: `countplot` (frequência), `barplot` (média + erro).
- Sempre **ordene** barras em rankings.

## 💡 Curiosidades

**William Playfair** (1759-1823) inventou o gráfico de barras em 1786 no livro "The Commercial and Political Atlas". Ele também inventou o gráfico de linhas e o gráfico de pizza! Na época, seus gráficos foram criticados por serem "muito simples" — hoje são onipresentes na visualização de dados.

## ✅ Boas Práticas

1. **Comece do zero** no eixo y para evitar distorções.
2. **Ordene** barras em rankings (maior para menor).
3. **Use cores consistentes** para a mesma categoria em múltiplos gráficos.
4. **Adicione rótulos de valor** nas barras para facilitar a leitura.
5. **Prefira barras horizontais** quando os nomes das categorias são longos.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Eixo y não começar do zero | Sempre comece do zero em gráficos de barras. |
| Muitas categorias (crowding) | Limite a 10-15 categorias ou use barras horizontais. |
| Cores confusas sem legenda | Use legendas claras e cores distintas. |
| Usar barras 3D desnecessariamente | 3D distorce a percepção — prefira 2D. |
