## 📝 Introdução

Em estatística, raramente temos acesso a todos os elementos de um grupo que desejamos estudar. Por questões de custo, tempo ou até impossibilidade física, precisamos trabalhar com um subconjunto representativo — a **amostra** (sample). A **população** (population) é o conjunto completo de todos os elementos que compartilham uma característica de interesse. A arte de selecionar uma amostra que reflita fielmente a população é a base da inferência estatística.

## 📚 Explicação Detalhada

### 1. População vs Amostra

A **população** (ou universo) é o conjunto total de indivíduos, objetos ou medidas que possuem alguma característica comum. Por exemplo, todos os eleitores brasileiros, todas as lâmpadas produzidas por uma fábrica em um mês, ou todas as estrelas da Via Láctea.

A **amostra** é um subconjunto finito e representativo extraído da população. Idealmente, a amostra deve preservar as propriedades da população para que possamos generalizar (inferir) os resultados obtidos na amostra para toda a população.

**Exemplo prático:** Se queremos saber a altura média dos brasileiros, a população seria todos os brasileiros. Coletar a altura de cada um seria impraticável, então medimos uma amostra de, digamos, 10.000 pessoas distribuídas por todas as regiões do país.

### 2. Por que Amostrar?

Existem razões práticas e teóricas para usar amostras:
- **Custo reduzido**: Coletar dados de uma amostra é muito mais barato que de toda a população
- **Viabilidade**: Algumas populações são infinitas ou inacessíveis (ex: todos os peixes do oceano)
- **Tempo**: Amostras permitem análises mais rápidas
- **Precisão**: Com recursos concentrados em menos elementos, podemos medir com maior acurácia
- **Destrutividade**: Em testes destrutivos (ex: resistência de um lote de parafusos), testar todos destruiria o lote inteiro

```python
# Diferença entre população e amostra com numpy
import numpy as np

# População simulada: 1.000.000 de valores de renda familiar (R$)
# Usamos distribuição log-normal para simular renda real
np.random.seed(42)
populacao = np.random.lognormal(mean=8.0, sigma=0.6, size=1_000_000)

# Parâmetro real da população (média verdadeira)
media_populacional = np.mean(populacao)
print(f'Média populacional verdadeira: R$ {media_populacional:.2f}')

# Extraindo uma amostra aleatória simples de 1000 elementos
amostra = np.random.choice(populacao, size=1000, replace=False)
media_amostral = np.mean(amostra)
print(f'Média amostral (n=1000): R$ {media_amostral:.2f}')
print(f'Diferença: R$ {abs(media_populacional - media_amostral):.2f}')

# Quanto maior a amostra, mais próxima a média amostral fica da populacional
for n in [10, 50, 100, 500, 2000, 10000]:
    amostra_n = np.random.choice(populacao, size=n, replace=False)
    diff = abs(media_populacional - np.mean(amostra_n))
    print(f'n={n:5d} → erro amostral: R$ {diff:.2f}')
```

### 3. Viés de Amostragem

O **viés de amostragem** (sampling bias) ocorre quando a amostra não representa adequadamente a população, tendendo a favorecer certos subgrupos em detrimento de outros. É o maior inimigo de uma pesquisa estatística confiável.

**Tipos comuns de viés:**
- **Viés de seleção**: certos grupos têm maior probabilidade de serem incluídos
- **Viés de voluntário**: pessoas que se voluntariam diferem das que não se voluntariam
- **Viés de sobrevivência**: analisamos apenas os que "sobreviveram" a um processo
- **Viés de resposta**: pessoas mentem ou distorcem respostas
- **Viés de confirmação**: o pesquisador busca dados que confirmem suas hipóteses

```python
# Demonstração de viés de amostragem
np.random.seed(42)

# População: 60% mulheres, 40% homens, com médias salariais diferentes
pop_mulheres = np.random.normal(5000, 1000, 600_000)  # 60%
pop_homens = np.random.normal(5500, 1200, 400_000)    # 40%
populacao = np.concatenate([pop_mulheres, pop_homens])

# Amostra tendenciosa: apenas homens (viés de seleção)
amostra_venciosa = np.random.choice(pop_homens, size=500, replace=False)
print(f'Média populacional real: R$ {np.mean(populacao):.2f}')
print(f'Média amostra viesada (só homens): R$ {np.mean(amostra_venciosa):.2f}')

# Amostra estratificada proporcional (corrige o viés)
amostra_mulheres = np.random.choice(pop_mulheres, size=300, replace=False)
amostra_homens = np.random.choice(pop_homens, size=200, replace=False)
amostra_balanceada = np.concatenate([amostra_mulheres, amostra_homens])
print(f'Média amostra balanceada: R$ {np.mean(amostra_balanceada):.2f}')
print()
print('A amostra balanceada (estratificada) estima melhor a média real.')
```

### 4. Tamanho da Amostra

O **tamanho da amostra** (sample size) influencia diretamente a precisão das estimativas. Amostras maiores reduzem o erro amostral, mas com custos crescentes. Existe um ponto de equilíbrio entre precisão e custo.

**Fatores que influenciam o tamanho da amostra:**
- **Variabilidade populacional**: populações mais heterogêneas exigem amostras maiores
- **Nível de confiança desejado**: maior confiança → maior amostra
- **Margem de erro aceitável**: menor erro → maior amostra
- **Tamanho da população**: para populações pequenas, o tamanho da amostra precisa ser proporcionalmente maior

A fórmula básica para amostra aleatória simples (população infinita) é:
n = (Z² × σ²) / E²
onde Z é o escore crítico, σ o desvio padrão e E a margem de erro.

```python
# Simulação do efeito do tamanho da amostra na precisão
import matplotlib.pyplot as plt

np.random.seed(42)
populacao = np.random.normal(100, 15, 100_000)
media_verdadeira = np.mean(populacao)

tamanhos = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
erros = []

for n in tamanhos:
    # Repetimos 1000 vezes para cada tamanho
    erros_n = []
    for _ in range(1000):
        amostra = np.random.choice(populacao, size=n, replace=False)
        erros_n.append(np.mean(amostra) - media_verdadeira)
    # Erro padrão (desvio padrão das médias amostrais)
    erros.append(np.std(erros_n))

for n, erro in zip(tamanhos, erros):
    print(f'n={n:5d} → erro padrão: {erro:.3f}')

print()
print('O erro padrão diminui com o aumento da amostra (Lei dos Grandes Números).')
```

## 💻 Exemplos Práticos

```python
# Exemplo prático: amostragem aleatória de uma base de clientes
import pandas as pd

np.random.seed(42)

# Simulando uma base de 50.000 clientes
clientes = pd.DataFrame({
    'id_cliente': range(1, 50001),
    'idade': np.random.randint(18, 80, 50000),
    'renda': np.random.lognormal(8, 0.5, 50000),
    'estado': np.random.choice(['SP', 'RJ', 'MG', 'RS', 'BA', 'PR', 'PE', 'CE', 'SC', 'DF'], 50000),
    'ativo': np.random.choice([True, False], 50000, p=[0.7, 0.3])
})

print('Base de clientes — população completa:')
print(clientes.head(10))
print(f'\nTotal de clientes: {len(clientes)}')

# Amostragem aleatória simples de 2000 clientes
amostra_aleatoria = clientes.sample(n=2000, random_state=42)
print(f'\nAmostra aleatória simples (n=2000):')
print(amostra_aleatoria.head(10))

# Comparando parâmetros
print(f'\n--- Comparação População vs Amostra ---')
print(f'Idade média (população): {clientes["idade"].mean():.1f}')
print(f'Idade média (amostra):   {amostra_aleatoria["idade"].mean():.1f}')
print(f'Renda média (população): R$ {clientes["renda"].mean():.2f}')
print(f'Renda média (amostra):   R$ {amostra_aleatoria["renda"].mean():.2f}')
print(f'Proporção ativos (pop):  {clientes["ativo"].mean():.3f}')
print(f'Proporção ativos (amost):{amostra_aleatoria["ativo"].mean():.3f}')
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Uma loja online quer estimar o ticket médio de suas 15.000 vendas no último mês. Por que ela deve usar uma amostra em vez de analisar todas as vendas? Que cuidados deve tomar na seleção da amostra?

```python
# Solução:
# Razões para amostrar:
# 1. Economia de recursos computacionais e tempo de processamento
# 2. Se os dados forem coletados manualmente, reduz custos operacionais
# 3. Permite análise exploratória rápida antes de processar tudo
#
# Cuidados na seleção:
# - Garantir que a amostra inclua todos os dias da semana
# - Não excluir vendas de alto valor (evitar viés)
# - Usar amostragem aleatória estratificada por categoria de produto
#
# Demonstração:
np.random.seed(42)
vendas = np.random.lognormal(5, 0.8, 15000)  # população de 15.000 vendas

amostra = np.random.choice(vendas, size=500, replace=False)
print(f'Ticket médio real: R$ {np.mean(vendas):.2f}')
print(f'Ticket médio amostral: R$ {np.mean(amostra):.2f}')
print(f'Erro: R$ {abs(np.mean(vendas) - np.mean(amostra)):.2f}')
print(f'Com apenas 500 amostras (3,3%), o erro foi pequeno!')
```

## 📌 Resumo
- **População**: conjunto completo de elementos com uma característica comum
- **Amostra**: subconjunto representativo da população
- Amostramos por economia, viabilidade, tempo e precisão
- **Viés de amostragem**: distorção sistemática que compromete a representatividade
- **Tamanho da amostra**: influencia diretamente a precisão; maior amostra → menor erro
- A **Lei dos Grandes Números** garante que a média amostral converge para a média populacional conforme n aumenta

## 💡 Curiosidades
- O censo demográfico brasileiro (IBGE) tenta entrevistar todos os domicílios do país — mais de 70 milhões — mas mesmo assim usa técnicas de amostragem para questões detalhadas
- O famoso erro da revista Literary Digest em 1936 (que previu a derrota de Roosevelt) foi causado por viés de amostragem: entrevistaram apenas seus assinantes, que eram mais ricos que a média
- Netflix usa amostragem para decidir quais séries renovar: analisa o comportamento de uma amostra representativa de usuários antes de commitr recursos

## ✅ Boas Práticas
- Sempre documente o método de amostragem utilizado
- Declare o tamanho da amostra e a margem de erro em pesquisas
- Use amostragem estratificada quando a população tiver subgrupos relevantes
- Evite amostras de conveniência sempre que possível
- Calcule o tamanho mínimo necessário antes de coletar dados
- Reproduza a análise em múltiplas amostras para verificar consistência

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Achar que a amostra precisa ser grande apenas em números absolutos | Avaliar a amostra em relação à heterogeneidade da população |
| Ignorar o viés de seleção | Questionar quem está sendo excluído da amostra |
| Generalizar resultados de amostras não-representativas | Verificar se o método de amostragem é probabilístico |
| Confundir amostra grande com amostra representativa | Uma amostra grande mas viesada continua sendo ruim |
