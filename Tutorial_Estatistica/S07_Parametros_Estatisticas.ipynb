## 📝 Introdução

Na estatística inferencial, usamos dados de uma amostra para fazer afirmações sobre uma população. Para isso, precisamos de uma linguagem clara que diferencie os **valores reais** (desconhecidos) da população dos **valores calculados** a partir da amostra. Os primeiros são chamados de **parâmetros** (parameters) e os segundos de **estatísticas** (statistics). A notação distinta nos ajuda a manter essa diferença sempre visível.

## 📚 Explicação Detalhada

### 1. Parâmetros Populacionais

**Parâmetros** (parameters) são medidas que descrevem uma característica de toda a **população**. Como raramente temos acesso à população completa, os parâmetros são geralmente **desconhecidos** e representam o valor verdadeiro que buscamos estimar.

**Notação grega (padrão):**
- **μ** (mi): média populacional
- **σ** (sigma): desvio padrão populacional
- **σ²**: variância populacional
- **p**: proporção populacional
- **N**: tamanho da população
- **ρ** (rô): coeficiente de correlação populacional
- **β** (beta): coeficiente de regressão populacional

**Exemplos:**
- μ = altura média de todos os brasileiros
- p = proporção de eleitores que votariam em um candidato

### 2. Estatísticas Amostrais

**Estatísticas** (statistics) são medidas calculadas a partir de uma **amostra**. Elas servem como **estimativas** dos parâmetros populacionais. Diferentes amostras produzem diferentes estatísticas — essa variabilidade é natural e mensurável.

**Notação latina (padrão):**
- **x̄** (x barra): média amostral
- **s**: desvio padrão amostral
- **s²**: variância amostral
- **p̂** (p chapéu): proporção amostral
- **n**: tamanho da amostra
- **r**: coeficiente de correlação amostral
- **b**: coeficiente de regressão amostral

**Relação entre parâmetro e estatística:**
- μ é fixo (mas desconhecido); x̄ varia conforme a amostra
- x̄ é um **estimador** de μ — com boas propriedades, x̄ se aproxima de μ

```python
# Parâmetros populacionais vs estatísticas amostrais — simulação
import numpy as np

np.random.seed(42)

# População de 500.000 pessoas com altura (cm) distribuída normalmente
populacao_altura = np.random.normal(170, 10, 500_000)

# PARÂMETROS POPULACIONAIS (valores verdadeiros, geralmente desconhecidos)
mu = np.mean(populacao_altura)      # μ - média populacional
sigma = np.std(populacao_altura)    # σ - desvio padrão populacional
print('PARÂMETROS POPULACIONAIS (μ, σ):')
print(f'μ (média populacional): {mu:.2f} cm')
print(f'σ (desvio padrão pop.): {sigma:.2f} cm')
print()

# EXTRAINDO MÚLTIPLAS AMOSTRAS E CALCULANDO ESTATÍSTICAS
print('ESTATÍSTICAS AMOSTRAIS (x̄, s) para diferentes amostras:')
for i in range(5):
    amostra = np.random.choice(populacao_altura, size=100, replace=False)
    x_bar = np.mean(amostra)   # média amostral
    s = np.std(amostra, ddof=1) # desvio padrão amostral (correção de Bessel)
    erro = abs(x_bar - mu)
    print(f'Amostra {i+1}: x̄ = {x_bar:.2f} cm, s = {s:.2f} cm, erro = {erro:.2f} cm')
```

### 3. Propriedades dos Estimadores

Um **estimador** (estimator) é uma regra ou fórmula para calcular uma estatística que estima um parâmetro. Nem todo estimador é bom. Três propriedades fundamentais definem a qualidade de um estimador:

1. **Não-viesado (unbiased)**: O valor esperado do estimador é igual ao parâmetro verdadeiro. E[x̄] = μ.
2. **Consistente (consistent)**: Conforme n aumenta, o estimador converge para o parâmetro.
3. **Eficiente (efficient)**: Entre estimadores não-viesados, o de menor variância é o mais eficiente.

A média amostral (x̄) é um estimador não-viesado, consistente e eficiente para μ. Já a variância amostral com correção de Bessel (s² = Σ(xi-x̄)²/(n-1)) é um estimador não-viesado para σ².

```python
# Verificando a propriedade de não-viesamento
np.random.seed(42)

populacao = np.random.normal(100, 15, 200_000)
mu_real = np.mean(populacao)

# Simulação de Monte Carlo: extraímos 10.000 amostras e calculamos x̄
n = 30
n_simulacoes = 10000
medias_amostrais = []

for _ in range(n_simulacoes):
    amostra = np.random.choice(populacao, size=n, replace=False)
    medias_amostrais.append(np.mean(amostra))

media_das_medias = np.mean(medias_amostrais)
print(f'Parâmetro verdadeiro (μ): {mu_real:.4f}')
print(f'Média das 10.000 médias amostrais: {media_das_medias:.4f}')
print(f'Diferença (viés): {media_das_medias - mu_real:.6f}')
print()
print(f'O estimador x̄ é não-viesado: E[x̄] ≈ μ')
```

### 4. Estimativas Pontuais vs Intervalares

Uma **estimativa pontual** (point estimate) é um único valor numérico que estima o parâmetro. Exemplo: "a média amostral é 172 cm".

Uma **estimativa intervalar** (interval estimate) fornece um intervalo de valores plausíveis para o parâmetro, com um nível de confiança associado. Exemplo: "a média populacional está entre 169 cm e 175 cm, com 95% de confiança".

Estimativas intervalares são mais informativas porque comunicam a incerteza inerente ao processo de amostragem. O **erro padrão** (standard error) é o desvio padrão da distribuição amostral da estatística:
- Erro padrão da média: EP = σ / √n
- Quando σ é desconhecido, usamos s / √n (erro padrão estimado)

```python
# Estimativa pontual e intervalar da média
from scipy import stats

np.random.seed(42)

# População
populacao = np.random.normal(100, 15, 200_000)
mu_real = np.mean(populacao)

# Uma única amostra de tamanho 50
amostra = np.random.choice(populacao, size=50, replace=False)
n = len(amostra)
x_bar = np.mean(amostra)
s = np.std(amostra, ddof=1)
ep = s / np.sqrt(n)  # erro padrão estimado

# Estimativa pontual
print(f'Estimativa pontual: x̄ = {x_bar:.2f}')  
print(f'Parâmetro real: μ = {mu_real:.2f}')
print(f'Erro padrão estimado: EP = {ep:.3f}')
print()

# Estimativa intervalar com 95% de confiança (distribuição t-Student)
confianca = 0.95
alpha = 1 - confianca
t_critico = stats.t.ppf(1 - alpha/2, df=n-1)
margem_erro = t_critico * ep
intervalo = (x_bar - margem_erro, x_bar + margem_erro)

print(f'Intervalo de confiança de 95%:')
print(f'({intervalo[0]:.2f}, {intervalo[1]:.2f})')
print(f'O parâmetro μ = {mu_real:.2f} está dentro do intervalo? ', end='')
print('Sim' if intervalo[0] <= mu_real <= intervalo[1] else 'Não')
```

## 💻 Exemplos Práticos

```python
# Exemplo completo: estimando múltiplos parâmetros de uma população
import pandas as pd

np.random.seed(42)

# População simulada de uma cidade (200.000 habitantes)
renda = np.random.lognormal(mean=8.2, sigma=0.7, size=200_000)
idade = np.random.normal(35, 12, 200_000).astype(int)
escolaridade = np.random.choice(['Fundamental', 'Médio', 'Superior', 'Pós-graduação'],
                                 size=200_000, p=[0.25, 0.35, 0.30, 0.10])

# PARÂMETROS POPULACIONAIS (μ, σ, p)
mu_renda = np.mean(renda)
sigma_renda = np.std(renda)
p_superior = np.mean(escolaridade == 'Superior')

print('PARÂMETROS POPULACIONAIS (verdadeiros):')
print(f'μ renda = R$ {mu_renda:.2f}')
print(f'σ renda = R$ {sigma_renda:.2f}')
print(f'p (Superior) = {p_superior:.3f}')
print()

# Amostra de 500 habitantes
amostra_idx = np.random.choice(200_000, size=500, replace=False)
renda_amostra = renda[amostra_idx]
idade_amostra = idade[amostra_idx]
escol_amostra = escolaridade[amostra_idx]

# ESTATÍSTICAS AMOSTRAIS (x̄, s, p̂)
x_bar = np.mean(renda_amostra)
s_renda = np.std(renda_amostra, ddof=1)
p_hat = np.mean(escol_amostra == 'Superior')

print('ESTATÍSTICAS AMOSTRAIS (estimativas):')
print(f'x̄ renda = R$ {x_bar:.2f} (erro: R$ {abs(x_bar - mu_renda):.2f})')
print(f's renda = R$ {s_renda:.2f} (erro: R$ {abs(s_renda - sigma_renda):.2f})')
print(f'p̂ (Superior) = {p_hat:.3f} (erro: {abs(p_hat - p_superior):.4f})')
print()
print('As estatísticas amostrais se aproximam dos parâmetros populacionais.')
```

## 📌 Resumo
- **Parâmetros** (μ, σ, p) descrevem a **população** — são fixos, geralmente desconhecidos
- **Estatísticas** (x̄, s, p̂) são calculadas da **amostra** — variam entre amostras
- **Estimador**: regra ou fórmula (ex: x̄ é estimador de μ)
- **Estimativa pontual**: valor numérico único
- **Estimativa intervalar**: intervalo com nível de confiança
- Bons estimadores são: não-viesados, consistentes e eficientes

## 💡 Curiosidades
- O uso de letras gregas para parâmetros e latinas para estatísticas foi padronizado por Ronald Fisher no início do século XX
- A correção de Bessel (n-1 no denominador) recebeu esse nome em homenagem ao astrônomo alemão Friedrich Bessel, que a propôs em 1818 para corrigir erros sistemáticos em observações astronômicas
- O "chapéu" (^) em p̂ é universal em estatística para indicar "estimativa de": lê-se "p chapéu"

## ✅ Boas Práticas
- Sempre identifique claramente se um valor é parâmetro ou estatística nos relatórios
- Use a notação correta (μ vs x̄, σ vs s) para evitar ambiguidade
- Reporte sempre o erro padrão junto com a estimativa pontual
- Prefira estimativas intervalares a apenas pontuais
- Documente o tamanho da amostra (n) em toda análise

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Tratar x̄ como se fosse μ | Lembre-se: x̄ é uma estimativa, não o valor verdadeiro |
| Usar σ quando deveria usar s (ou vice-versa) | σ para população, s para amostra — sempre verificar o contexto |
| Esquecer o ddof=1 no numpy | np.std(amostra, ddof=1) para s (amostral) vs ddof=0 para σ (populacional) |
| Ignorar o erro padrão | Sempre calcule e reporte EP = s/√n junto com x̄ |
