## Juri no Tribunal

Um julgamento esta acontecendo.

O juiz diz ao juri:
*"O reu e inocente ate que se prove o contrario."*

Essa e a **presuncao de inocencia** — o ponto de partida.

A acusacao apresenta evidencias:
- DNA, testemunhas, motivos, oportunidade
- Se as evidencias sao **fortes o suficiente**, o juri pode
  "rejeitar a presuncao de inocencia" e condenar o reu

Mas o juri nunca tem **100% de certeza**.
Sempre existe o risco de condenar um inocente
ou absolver um culpado.

**E e exatamente assim que funcionam as conclusoes estatisticas!**


## 📊 Traduzindo

| Juri no Tribunal | Estatistica |
|-|-|
| Presuncao de inocencia | Hipotesese nula (H0) — o "padrao" que assumimos |
| Evidencias da acusacao | Dados coletados na amostra |
| Evidencias suficientemente fortes | Resultado estatisticamente significativo |
| Condenar o reu | Rejeitar H0 — concluir que ha efeito |
| Absolver o reu | Nao rejeitar H0 — nao ha evidencias suficientes |
| Condenar um inocente | Erro tipo I (falso positivo) |
| Absolver um culpado | Erro tipo II (falso negativo) |


## 📏 O Ponto de Partida: Presuncao de Inocencia

Na estatistica, sempre comecamos com uma **hipotese nula (H0)**:
o "padrao", a suposicao de que nada mudou.

Exemplos:
- H0: a media e 100 (nada mudou)
- H0: o tratamento nao tem efeito
- H0: a moeda e honesta

So rejeitamos H0 se as evidencias (dados) forem fortes o suficiente.
Assim como o juri so condena se as provas forem solidas.


## 📏 A Duvida Sempre Existe

No tribunal, o juri pode errar:
- **Condenar um inocente** → erro tipo I (falso positivo)
- **Absolver um culpado** → erro tipo II (falso negativo)

Na estatistica, a mesma coisa:
- Rejeitar H0 quando ela e verdadeira (falso positivo)
- Nao rejeitar H0 quando ela e falsa (falso negativo)

Nunca eliminamos o erro — apenas **controlamos** a chance de comete-lo.
O nivel de significancia (α = 0,05) significa:
"aceitamos 5% de chance de condenar um inocente".


```python
# O juri decidindo com dados
import random, statistics as stat
random.seed(42)

# Simulando: Uma fabrica diz que suas latas tem 350 mL
media_verdadeira = 350  # H0: a media e 350
amostra = [random.gauss(348, 5) for _ in range(30)]
media_obs = stat.mean(amostra)
print(f"Media observada na amostra: {media_obs:.1f} mL")
print(f"H0 afirma: a media e {media_verdadeira} mL")
print(f"Diferenca: {abs(media_obs - media_verdadeira):.1f} mL")

# O juri (nos) precisa decidir: evidencia forte o bastante?
# Se a diferenca e grande, rejeitamos H0 (condenamos)
# Se e pequena, mantemos H0 (absolvemos)
```

## 📌 Resumo

| Juri no Tribunal | Estatistica |
|-|-|
| Presuncao de inocencia | H0 — hipotese nula |
| Evidencias da acusacao | Dados da amostra |
| Condenar | Rejeitar H0 |
| Absolver | Nao rejeitar H0 |
| Condenar inocente | Erro tipo I |
| Absolver culpado | Erro tipo II |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
