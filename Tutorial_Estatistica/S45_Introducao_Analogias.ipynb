## Detetive

Imagine a seguinte cena: um detetive chega a uma cena de crime.
Ele nao viu o crime acontecer. Tudo o que ele tem sao pistas.

- Uma pegada aqui
- Um fio de cabelo ali
- Um horario anotado em um papel
- O depoimento de uma testemunha

O detetive **coleta** essas pistas, **organiza** as informacoes,
**analisa** os fatos e **tira conclusoes** sobre o que aconteceu.

Essa e exatamente a mesma funcao da Estatistica!


## 📊 Traduzindo

| Detetive | Estatistica |
|-|-|
| Pista isolada na cena | Dado bruto |
| Todas as pistas reunidas | Conjunto de dados |
| Detetive organizando o quadro | Organizacao e limpeza dos dados |
| Detetive analisando padroes | Analise estatistica |
| Conclusao do caso | Conclusao baseada em dados |
| Detetive admite duvida | Inferencia estatistica — margem de erro |


## 📏 Coletar as Pistas

O detetive nao inventa pistas — ele as encontra no mundo real.
Da mesma forma, a estatistica comeca com **dados reais**:
pesquisas, experimentos, sensores, questionarios.

Se as pistas forem falsas (dados de ma qualidade),
a conclusao sera errada — nao importa o quao bom seja o detetive.


## 📏 Organizar o Quadro de Evidencias

O detetive coloca fotos, fios e anotacoes em um quadro.
Ele **organiza** as pistas para enxergar o quadro geral.

Nos fazemos o mesmo com tabelas, graficos e resumos.
Dados organizados revelam padroes que dados brutos escondem.


## 📏 Analisar e Concluir (Inferencia)

Com as pistas organizadas, o detetive pergunta:
*"O que esses fatos me dizem sobre o que aconteceu?"*

Na estatistica, essa e a **inferencia**:
usar dados de uma amostra (as pistas que temos)
para concluir algo sobre a populacao (o que realmente aconteceu).

E o detetive nunca tem **certeza absoluta** —
assim como nos nunca temos 100% de confianca.
Sempre existe uma margem de erro, um risco.


```python
# Vamos simular o detetive analisando horarios de avistamentos
horarios = [14, 15, 14, 16, 15]  # horas do dia
print("Pistas (horarios):", horarios)
print("Total de pistas:", len(horarios))
print()
print("--- Analise do Detetive ---")
print("Horario medio:", sum(horarios)/len(horarios))
print("Horario mais comum:", max(set(horarios), key=horarios.count))
print("Menor:", min(horarios), "| Maior:", max(horarios))
```

```python
# O detetive sabe que nao ha certeza absoluta
import random
random.seed(42)
todos = [random.gauss(15, 2) for _ in range(1000)]
media_real = sum(todos)/len(todos)
print("Media REAL (desconhecida):", f"{media_real:.1f}h")
for n in [3, 10, 50]:
    amostra = random.sample(todos, n)
    media_est = sum(amostra)/len(amostra)
    erro = abs(media_est - media_real)
    print(f"n={n:2d} -> estimativa: {media_est:.1f}h, erro: {erro:.2f}h")
```

## 📌 Resumo

| Detetive | Estatistica |
|-|-|
| Detetive coletando pistas | Coleta de dados |
| Detetive organizando o quadro | Organizacao e descricao |
| Detetive analisando padroes | Analise estatistica |
| Detetive concluindo o caso | Tomada de decisao |
| Detetive admite margem de duvida | Inferencia e margem de erro |

A analogia do detetive mostra que:

> A estatistica esta em todo lugar — so precisamos do olhar certo para enxerga-la.
