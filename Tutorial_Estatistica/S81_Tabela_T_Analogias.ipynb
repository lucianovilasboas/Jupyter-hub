## Cardapio Especial

A **tabela T** e como um cardapio que muda
conforme o **numero de pessoas na mesa** (graus de liberdade).

- Mesa com 2 pessoas (gl=1): cardapio limitado, precos altos
  (caudas largas, valores criticos altos)
- Mesa com 30 pessoas (gl=29): cardapio normal
  (caudas proximas da normal)

Quanto mais pessoas na mesa (mais dados),
mais o cardapio T se parece com o cardapio normal (Z).

Com poucas pessoas, os "precos" (valores criticos) sao mais altos
para compensar a incerteza.


## 📊 Traduzindo

| Cardapio Especial | Estatistica |
|-|-|
| Cardapio que muda | Tabela T varia com gl |
| 2 pessoas na mesa | gl=1: valores criticos altos |
| 30 pessoas na mesa | gl=29: proximo da normal |
| Preco alto | Valor critico maior |


```python
# Cardapio T em Python
from scipy import stats as sp_stats
confianca = 0.95
grazinhos = [1, 2, 5, 10, 30, 100]
print("gl  | Valor critico T (95%)")
print("-"*25)
for gl in grazinhos:
    t = sp_stats.t.ppf(0.975, df=gl)
    print(f"{gl:3d} | {t:.4f}")
print(f"\nPara comparar, Z (normal): {sp_stats.norm.ppf(0.975):.4f}")
```

## 📌 Resumo

| Cardapio Especial | Estatistica |
|-|-|
| Cardapio varia | Tabela T varia com gl |
| 2 pessoas: caro | gl pequeno: T grande |
| 30 pessoas: normal | gl grande: T ≈ Z |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
