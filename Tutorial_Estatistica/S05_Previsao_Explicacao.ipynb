## 📝 Introdução

Na estatística, frequentemente queremos dois tipos de resposta: **prever** algo que não conhecemos ou **explicar** como as variáveis se relacionam. Embora pareçam semelhantes, previsão e explicação têm objetivos diferentes e requerem abordagens distintas.

A previsão busca acertar o valor de uma variável de interesse (ex: preço de uma casa amanhã). A explicação busca entender relações causais entre variáveis (ex: aumentar a escolaridade aumenta a renda?). Neste capítulo, exploraremos ambas e veremos como construir modelos preditivos simples.

## 📚 Explicação Detalhada

### 1. Previsão vs Explicação

| Aspecto | Previsão | Explicação |
|---------|----------|------------|
| Objetivo | Acertar o valor futuro | Entender relações causais |
| Métrica | Erro de previsão (RMSE, MAE) | Significância dos coeficientes |
| Complexidade | Modelos complexos são bem-vindos | Modelos simples e interpretáveis |
| Exemplo | Preço de ações amanhã | Efeito de propaganda nas vendas |

### 2. Correlação vs Causalidade

Um dos erros mais comuns em estatística é confundir correlação com causalidade. **Correlação** indica que duas variáveis variam juntas. **Causalidade** significa que uma variável causa a outra.

Exemplo clássico: o número de afogamentos aumenta com as vendas de sorvete. Isso não significa que sorvete causa afogamento — ambas são causadas pelo calor (variável de confusão).

### 3. Regressão Linear Simples

A regressão linear modela a relação entre uma variável dependente (Y) e uma independente (X):

**Y = a + bX + ε**

Onde:
- a = intercepto (valor de Y quando X = 0)
- b = coeficiente angular (mudança em Y para cada unidade de X)
- ε = erro (resíduo)

```python
# Exemplo: correlacao nao implica causalidade
# Simulando dados onde ambas variaveis sao influenciadas por uma terceira
import random
import matplotlib.pyplot as plt

random.seed(42)

# Variavel de confusao: temperatura
temperatura = [random.uniform(20, 40) for _ in range(50)]

# Vendas de sorvete aumentam com a temperatura
sorvetes = [t * 3 + random.gauss(0, 10) for t in temperatura]

# Afogamentos aumentam com a temperatura (mais pessoas na praia)
afogamentos = [t * 0.5 + random.gauss(0, 5) for t in temperatura]

print("Correlação entre sorvetes e afogamentos (dados simulados):")

# Calculo simples de correlacao de Pearson
n = len(sorvetes)
media_s = sum(sorvetes)/n
media_a = sum(afogamentos)/n
num = sum((s - media_s)*(a - media_a) for s, a in zip(sorvetes, afogamentos))
den = (sum((s - media_s)**2 for s in sorvetes) * sum((a - media_a)**2 for a in afogamentos))**0.5
corr = num / den
print(f"Correlação: {corr:.3f}")
print("Ambas são causadas pela temperatura, não uma pela outra!")
```

```python
# Visualizando a correlacao espuria
plt.scatter(sorvetes, afogamentos, alpha=0.6)
plt.title('Sorvetes vs Afogamentos (Correlação Espúria)')
plt.xlabel('Vendas de Sorvete')
plt.ylabel('Número de Afogamentos')
plt.grid(True, alpha=0.3)
plt.show()
```

## 💻 Exemplos Práticos

```python
# Exemplo prático: regressão linear simples
# Prevendo peso (kg) a partir da altura (cm)
import math

# Dados simulados de altura e peso de 30 pessoas
alturas = [165, 170, 172, 158, 180, 175, 168, 182, 160, 176,
           163, 178, 155, 170, 185, 169, 174, 177, 162, 171,
           166, 179, 164, 173, 181, 167, 176, 159, 183, 168]
pesos =   [62, 70, 74, 55, 85, 78, 65, 88, 52, 80,
           58, 82, 50, 72, 90, 68, 76, 79, 56, 71,
           60, 84, 59, 73, 87, 63, 77, 54, 89, 67]

# Calculando a regressão linear: peso = a + b * altura
n = len(alturas)
media_x = sum(alturas) / n
media_y = sum(pesos) / n

# Coeficiente angular (b)
num_b = sum((alturas[i] - media_x) * (pesos[i] - media_y) for i in range(n))
den_b = sum((alturas[i] - media_x)**2 for i in range(n))
b = num_b / den_b

# Intercepto (a)
a = media_y - b * media_x

print("=== Regressão Linear: Peso ~ Altura ===")
print(f"Equação: peso = {a:.2f} + {b:.4f} * altura")
print(f"Interpretação: para cada 1cm a mais, o peso aumenta {b:.2f}kg em média")
```

```python
# Avaliando o modelo: prevendo e calculando erro
# Fazendo previsões
previsoes = [a + b * h for h in alturas]

# Erro quadrático médio (RMSE)
erros = [(pesos[i] - previsoes[i])**2 for i in range(n)]
mse = sum(erros) / n
rmse = math.sqrt(mse)

# R² (coeficiente de determinação)
ss_total = sum((pesos[i] - media_y)**2 for i in range(n))
ss_res = sum((pesos[i] - previsoes[i])**2 for i in range(n))
r2 = 1 - ss_res / ss_total

print(f"RMSE (erro médio): {rmse:.2f} kg")
print(f"R²: {r2:.4f} (a altura explica {r2*100:.1f}% da variação do peso)")

# Visualizando
plt.figure(figsize=(8, 5))
plt.scatter(alturas, pesos, color='blue', alpha=0.6, label='Dados reais')
plt.plot(alturas, previsoes, color='red', linewidth=2, label='Linha de regressão')
plt.title('Regressão Linear: Peso vs Altura')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Fazendo previsões para novos valores
novas_alturas = [164, 177, 186]
previsoes_pesos = [a + b * h for h in novas_alturas]

print("=== Previsões ===")
for i in range(len(novas_alturas)):
    print(f"Altura {novas_alturas[i]}cm → Peso previsto: {previsoes_pesos[i]:.1f}kg")
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Dados X = [1, 2, 3, 4, 5] e Y = [2, 4, 5, 4, 5], calcule a correlação entre X e Y.

```python
# Solução do Exercício 1
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
n = len(x)
mx = sum(x)/n
my = sum(y)/n
num = sum((x[i]-mx)*(y[i]-my) for i in range(n))
den = (sum((x[i]-mx)**2 for i in range(n)) * sum((y[i]-my)**2 for i in range(n)))**0.5
corr_xy = num / den
print(f"Correlação entre X e Y: {corr_xy:.4f}")
if abs(corr_xy) > 0.7:
    print("Correlação forte")
elif abs(corr_xy) > 0.5:
    print("Correlação moderada")
else:
    print("Correlação fraca")
```

**Exercício 2:** Usando os dados de altura e peso, qual o peso previsto para uma pessoa com 170cm?

```python
# Solução do Exercício 2
altura_nova = 170
peso_previsto = a + b * altura_nova
print(f"Para altura = {altura_nova}cm:")
print(f"Peso previsto = {peso_previsto:.1f}kg")
```

## 📌 Resumo

- **Previsão**: foco em acertar valores futuros. **Explicação**: foco em entender relações causais.
- **Correlação ≠ Causalidade**: variáveis podem estar correlacionadas sem relação causal.
- **Regressão linear**: modelo simples para prever Y a partir de X.
- **R²**: mede quanto da variação de Y é explicada pelo modelo.
- Python permite implementar regressão linear manualmente ou com bibliotecas.

## 💡 Curiosidades

O termo **regressão** foi cunhado por Francis Galton no século XIX. Galton estudava a altura de pais e filhos e observou que pais muito altos tendiam a ter filhos mais baixos que eles, e pais muito baixos tendiam a ter filhos mais altos — um fenômeno que ele chamou de "regressão à média" (regression to the mean). Esse conceito é fundamental para entender porque eventos extremos tendem a ser seguidos por eventos menos extremos.

3. Prefira modelos simples e interpretáveis para fins de explicação.
4. Visualize sempre os resíduos para verificar suposições do modelo.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Confundir correlação com causalidade | Lembre-se: correlação não implica causalidade. |
| Overfitting (modelo muito complexo) | Use validação cruzada e prefira modelos simples. |
| Ignorar resíduos do modelo | Sempre analise os erros do modelo. |
| Extrapolar além dos dados observados | Previsões fora do intervalo dos dados são arriscadas. |
