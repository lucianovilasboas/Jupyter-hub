## 📝 Introdução

Na prática, raramente temos acesso a todos os dados de uma população inteira. Coletar dados de milhões de pessoas é caro e demorado. A inferência estatística nos permite tirar conclusões sobre uma população inteira usando apenas uma amostra (subconjunto) dos dados.

É como provar uma sopa: você não precisa comer a panela inteira para saber se está salgada — uma colher (amostra) bem mexida é suficiente para inferir o sabor de toda a sopa (população).

## 📚 Explicação Detalhada

### 1. População vs Amostra

- **População**: Conjunto completo de todos os elementos que queremos estudar (ex.: todos os brasileiros).
- **Amostra**: Subconjunto representativo da população (ex.: 2000 brasileiros entrevistados).
- **Parâmetro**: Valor numérico que descreve uma característica da **população** (ex.: μ, σ).
- **Estatística**: Valor numérico que descreve uma característica da **amostra** (ex.: x̄, s).

### 2. Estimação Pontual e Intervalar

- **Estimação pontual**: Um único valor usado para estimar o parâmetro populacional (ex.: a média amostral x̄ = 25 anos).
- **Estimação intervalar**: Um intervalo de valores plausíveis para o parâmetro, com um certo nível de confiança (ex.: IC 95% = [23.5, 26.5] anos).

### 3. Teste de Hipóteses

Procedimento para decidir se uma afirmação sobre a população é apoiada pela evidência amostral:
- **H₀ (Hipótese nula)**: Afirmação padrão, geralmente "sem efeito"
- **H₁ (Hipótese alternativa)**: O que queremos provar
- **Nível de significância (α)**: Probabilidade de rejeitar H₀ quando ela é verdadeira
- **p-valor**: Probabilidade de observar os dados (ou algo mais extremo) se H₀ for verdadeira

### 4. Lei dos Grandes Números (LGN)

Conforme o tamanho da amostra aumenta, a média amostral se aproxima da média populacional. Quanto mais dados coletamos, mais precisa é nossa estimativa.

### 5. Teorema Central do Limite (TCL)

Para amostras grandes (n ≥ 30), a distribuição das médias amostrais tende a uma distribuição normal, independentemente da forma da distribuição populacional original. Este é um dos resultados mais importantes da estatística!

## 💻 Exemplos Práticos

```python
# Exemplo 1: Estimação pontual da média populacional
# Suponha que a altura média real dos brasileiros seja 1.70m (parâmetro populacional)
# Vamos coletar uma amostra e estimar essa média

np.random.seed(42)  # Para reprodutibilidade

# Simulamos a população: alturas seguem distribuição normal com μ=1.70, σ=0.10
populacao = np.random.normal(loc=1.70, scale=0.10, size=100000)

# Coletamos uma amostra de 50 pessoas
amostra = np.random.choice(populacao, size=50, replace=False)

# Estimativa pontual da média
media_amostral = np.mean(amostra)
media_real = np.mean(populacao)

print(f"Média real da população (μ): {media_real:.4f} m")
print(f"Média da amostra (x̄): {media_amostral:.4f} m")
print(f"Erro de estimação: {abs(media_real - media_amostral):.4f} m")
```

```python
# Exemplo 2: Lei dos Grandes Números — demonstrando convergência
# Quanto maior a amostra, mais próxima a média amostral fica da média populacional

np.random.seed(42)

# População com distribuição exponencial (assimétrica, para mostrar que funciona)
pop = np.random.exponential(scale=5, size=100000)
media_pop = np.mean(pop)

# Coletamos amostras de tamanhos crescentes
tamanhos = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
medias = []

for n in tamanhos:
    amostra_n = np.random.choice(pop, size=n, replace=False)
    medias.append(np.mean(amostra_n))

# Plotamos a convergência
plt.figure(figsize=(10, 6))
plt.plot(tamanhos, medias, 'bo-', label='Média amostral')
plt.axhline(y=media_pop, color='r', linestyle='--', label=f'Média populacional (μ = {media_pop:.2f})')
plt.xscale('log')
plt.xlabel('Tamanho da amostra (n)', fontsize=12)
plt.ylabel('Média amostral (x̄)', fontsize=12)
plt.title('Lei dos Grandes Números: Convergência da Média Amostral', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Exemplo 3: Teorema Central do Limite — Simulação Monte Carlo
# Demonstramos que a distribuição das médias amostrais tende à normal
# mesmo quando a população original não é normal

np.random.seed(42)

# População com distribuição uniforme (NÃO é normal)
pop_uniforme = np.random.uniform(low=0, high=10, size=100000)

# Parâmetros da simulação
n_amostras = 10000  # Número de amostras que coletamos
tamanho_amostra = 30  # Tamanho de cada amostra

# Guardamos a média de cada amostra
medias_amostrais = []

for _ in range(n_amostras):
    amostra = np.random.choice(pop_uniforme, size=tamanho_amostra, replace=True)
    medias_amostrais.append(np.mean(amostra))

medias_amostrais = np.array(medias_amostrais)

# Visualizamos
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histograma da população original
axes[0].hist(pop_uniforme, bins=50, color='skyblue', edgecolor='black', density=True)
axes[0].set_title('Distribuição da População (Uniforme)', fontsize=13)
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Densidade')

# Histograma das médias amostrais
axes[1].hist(medias_amostrais, bins=50, color='salmon', edgecolor='black', density=True)
axes[1].set_title(f'Distribuição das Médias Amostrais (n = {tamanho_amostra})', fontsize=13)
axes[1].set_xlabel('Média Amostral')
axes[1].set_ylabel('Densidade')

# Sobrepomos a curva normal teórica
from scipy import stats
mu_teorico = np.mean(pop_uniforme)
sigma_teorico = np.std(pop_uniforme) / np.sqrt(tamanho_amostra)
x = np.linspace(min(medias_amostrais), max(medias_amostrais), 100)
axes[1].plot(x, stats.norm.pdf(x, mu_teorico, sigma_teorico), 'b-', lw=2, label='Curva Normal (TCL)')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f"Média das médias amostrais: {np.mean(medias_amostrais):.4f}")
print(f"Média populacional: {np.mean(pop_uniforme):.4f}")
print(f"Desvio padrão das médias (erro padrão): {np.std(medias_amostrais):.4f}")
```

```python
# Exemplo 4: TCL com diferentes tamanhos de amostra
# Mostramos como a distribuição fica mais "normal" conforme n aumenta

np.random.seed(42)

# População exponencial (fortemente assimétrica)
pop_exp = np.random.exponential(scale=2, size=100000)

tamanhos = [2, 5, 10, 30, 100]
n_amostras = 5000

fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i, n in enumerate(tamanhos):
    medias_n = []
    for _ in range(n_amostras):
        amostra = np.random.choice(pop_exp, size=n, replace=True)
        medias_n.append(np.mean(amostra))

    axes[i].hist(medias_n, bins=40, density=True, color='lightgreen', edgecolor='black')
    axes[i].set_title(f'n = {n}', fontsize=12)
    axes[i].set_xlabel('Média amostral')

plt.suptitle('TCL: Efeito do Tamanho da Amostra na Distribuição das Médias', fontsize=14)
plt.tight_layout()
plt.show()
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Uma população de notas tem distribuição desconhecida com μ = 70 e σ = 15. Se coletarmos amostras de n = 36 alunos, qual será a distribuição aproximada das médias amostrais segundo o TCL?

```python
# Solução Exercício 1
# Pelo TCL, a distribuição das médias será aproximadamente normal com:
# μ_x̄ = μ = 70
# σ_x̄ = σ / √n = 15 / √36 = 15 / 6 = 2.5

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

mu = 70
sigma = 15
n = 36
erro_padrao = sigma / np.sqrt(n)

print(f"Média das médias amostrais (μ_x̄) = {mu}")
print(f"Erro padrão (σ_x̄ = σ/√n) = {sigma}/{np.sqrt(n)} = {erro_padrao:.2f}")

# Visualização
x = np.linspace(mu - 4*erro_padrao, mu + 4*erro_padrao, 200)
y = stats.norm.pdf(x, mu, erro_padrao)

plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b-', lw=2)
plt.fill_between(x, y, alpha=0.3, color='skyblue')
plt.axvline(mu, color='red', linestyle='--', label=f'μ = {mu}')
plt.xlabel('Média amostral')
plt.ylabel('Densidade')
plt.title('Distribuição das Médias Amostrais (TCL)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

**Exercício 2:** Uma máquina produz parafusos com diâmetro médio populacional de 10 mm e desvio padrão de 0.5 mm. Coletamos uma amostra de 100 parafusos e obtemos média de 10.08 mm. Qual a margem de erro (para 95% de confiança)?

```python
# Solução Exercício 2
# Margem de erro = z* × σ/√n
# Para 95% de confiança, z* ≈ 1.96

sigma = 0.5  # desvio padrão populacional
n = 100
z_95 = 1.96  # valor crítico para 95%

erro_padrao = sigma / np.sqrt(n)
margem_erro = z_95 * erro_padrao

print(f"Erro padrão: {erro_padrao:.4f} mm")
print(f"Margem de erro (95%): {margem_erro:.4f} mm")
print(f"Intervalo de confiança: [10.08 - {margem_erro:.4f}, 10.08 + {margem_erro:.4f}]")
print(f"IC 95%: [{10.08 - margem_erro:.4f}, {10.08 + margem_erro:.4f}]")
```

## 📌 Resumo

- **Inferência estatística** usa amostras para tirar conclusões sobre populações.
- **Estimação pontual** dá um único valor; **estimação intervalar** dá um intervalo com nível de confiança.
- **Teste de hipóteses** avalia evidências contra uma hipótese nula.
- **Lei dos Grandes Números**: a média amostral converge para a média populacional com n grande.
- **Teorema Central do Limite**: a distribuição das médias amostrais tende à normal quando n ≥ 30.

## 💡 Curiosidades

O Teorema Central do Limite foi formalizado por **Abraham de Moivre** no século XVIII e posteriormente generalizado por **Pierre-Simon Laplace** e **Aleksandr Lyapunov**. Diz a lenda que Laplace usou o TCL para estimar a população da França com base em amostras de nascimentos, ajudando o governo napoleônico no planejamento de recursos. Hoje, o TCL é usado em tudo: desde previsões eleitorais até algoritmos de machine learning!

## ✅ Boas Práticas

1. Sempre verifique se sua amostra é representativa (aleatória, não viesada).
2. Para usar o TCL, certifique-se de que n ≥ 30 (ou que a população é aproximadamente normal).
3. Reporte sempre o intervalo de confiança junto com a estimativa pontual.
4. Nunca generalize além da população da qual a amostra foi retirada.
5. Use a simulação Monte Carlo para verificar a validade dos seus resultados.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Achar que o TCL se aplica aos dados originais (e não às médias amostrais) | Lembre-se: o TCL fala da **distribuição das médias amostrais**, não dos dados brutos. |
| Usar n < 30 sem verificar a normalidade da população | Para n pequeno, a população precisa ser aproximadamente normal. |
| Confundir erro padrão com desvio padrão | Erro padrão = σ/√n; desvio padrão = σ. |
| Ignorar o viés de seleção da amostra | Amostra não aleatória → inferência inválida. |
