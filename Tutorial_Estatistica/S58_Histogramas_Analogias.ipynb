## Mapa de Calor da Cidade

Imagine um mapa de calor da sua cidade.
Areas vermelhas = muito movimento. Azuis = pouco movimento.

O **histograma** e o mapa de calor dos seus dados!

Em vez de ruas, ele tem **barras** (intervalos de valores).
Em vez de cores, ele tem **alturas** (quantidade de dados).

O histograma mostra:
- **Onde os dados se concentram** (o "centro quente")
- **Ate onde se espalham** (as "periferias frias")
- **Se ha lacunas** (areas sem dados)
- **A forma** da distribuicao (simetrica, inclinada...)


## 📊 Traduzindo

| Mapa de Calor da Cidade | Estatistica |
|-|-|
| Mapa de calor da cidade | Histograma dos dados |
| Areas vermelhas (muita concentracao) | Barras altas (muitos dados ali) |
| Areas azuis (pouca concentracao) | Barras baixas (poucos dados ali) |
| Bairros com ruas | Intervalos / bins do histograma |
| Centro quente | Media / mediana dos dados |


## 📏 O que o Histograma Revela

Um histograma bem feito responde:

1. **Onde esta o centro?** (em torno de que valor)
2. **Qual a dispersao?** (dados concentrados ou espalhados)
3. **A forma e simetrica?** (curva normal ou assimetrica)
4. **Ha outliers?** (barras isoladas nas pontas)

E tudo num unico grafico!


```python
# Mapa de calor dos dados
import random, matplotlib.pyplot as plt
random.seed(42)
# Temperaturas de uma cidade durante um ano
temperaturas = [random.gauss(25, 5) for _ in range(365)]

plt.figure(figsize=(8, 4))
plt.hist(temperaturas, bins=15, color="tomato", edgecolor="black", alpha=0.7)
plt.title("Mapa de Calor — Temperaturas da Cidade")
plt.xlabel("Temperatura (C)")
plt.ylabel("Numero de dias")
plt.grid(alpha=0.3)
plt.show()
print(f"Media: {sum(temperaturas)/len(temperaturas):.1f}C")
print(f"Min: {min(temperaturas):.1f}C | Max: {max(temperaturas):.1f}C")
```

## 📌 Resumo

| Mapa de Calor da Cidade | Estatistica |
|-|-|
| Areas vermelhas | Barras altas (concentracao) |
| Areas azuis | Barras baixas (poucos dados) |
| Bairros | Intervalos / bins |
| Centro quente | Centro dos dados |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
