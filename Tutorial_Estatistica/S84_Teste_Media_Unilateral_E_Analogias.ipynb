## Conta de Luz

A distribuidora de energia PROMETE que a conta de luz media
na sua regiao e de **R$ 120**.

Voce suspeita que esta vindo **MENOS** que isso
(talvez estejam subestimando o consumo medio).

Voce coleta 25 contas: media de **R$ 112**, desvio R$ 18.

Teste **unilateral a esquerda**:
H0: media = 120
H1: media < 120 (so nos importa se e MENOR)

Se for maior, problema da distribuidora.
Voce so quer saber se esta recebendo menos do que o prometido.


## 📊 Traduzindo

| Conta de Luz | Estatistica |
|-|-|
| Promessa: R$ 120 | H0: media = 120 |
| Contas: media R$ 112 | Media amostral |
| So importa se e MENOR | Teste unilateral esquerda |
| Se for maior, ignoramos | Cauda direita nao importa |


```python
# Conta de luz
import math, statistics as stat
from scipy import stats as sp_stats
contas = [115, 108, 112, 120, 105, 111, 118, 107, 113, 116,
          109, 114, 110, 117, 106, 112, 115, 108, 113, 111,
          119, 107, 114, 110, 116]
media = stat.mean(contas)
desvio = stat.stdev(contas)
n = len(contas)
t = (media - 120) / (desvio / math.sqrt(n))
p = sp_stats.t.cdf(t, df=n-1)
print(f"Media das contas: R$ {media:.1f}")
print(f"H0: media = R$ 120")
print(f"T: {t:.3f}")
print(f"p-valor (unilateral esq): {p:.4f}")
if p < 0.05:
    print("CONCLUSAO: Contas MENORES que o prometido!")
else:
    print("CONCLUSAO: Evidencias insuficientes")
```

## 📌 Resumo

| Conta de Luz | Estatistica |
|-|-|
| Promessa: R$ 120 | H0 |
| Contas: R$ 112 | Media amostral |
| So menor importa | Teste unilateral esquerda |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
