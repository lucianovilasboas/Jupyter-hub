## 📝 Introdução

A **Estatística Descritiva** é o ramo da estatística responsável por organizar, resumir e apresentar dados de forma clara e informativa. Enquanto a Estatística Inferencial busca fazer generalizações sobre populações a partir de amostras, a Descritiva se preocupa em **descrever** o que os dados mostram.

Imagine que você recebe uma planilha com 10.000 registros de vendas de uma loja. De que forma você poderia entender o que esses dados dizem? Você poderia calcular o **valor médio** das vendas (centro), verificar se as vendas são muito **variáveis** (dispersão) e analisar se há **dias excepcionais** ou **padrões sazonais** (forma). Essas três dimensões — centro, variação e forma — são a base da Estatística Descritiva.

## 📚 Explicação Detalhada

### 1. O que é Estatística Descritiva?

A Estatística Descritiva resume e descreve as características principais de um conjunto de dados. Ela não faz inferências ou generalizações — apenas **organiza e apresenta** os dados de forma compreensível.

**Principais ferramentas:**
- **Tabelas de frequência** — organizam os dados em categorias ou intervalos
- **Gráficos** — histogramas, barras, pizza, box plots, etc.
- **Medidas numéricas** — média, mediana, desvio padrão, quartis, etc.

### 2. Medidas de Centro (Location)

As medidas de centro buscam identificar um valor **típico** ou **central**:

- **Média (Mean)**: Soma dos valores dividida pela quantidade. Sensível a outliers.
- **Mediana (Median)**: Valor do meio quando os dados estão ordenados. Robusta a outliers.
- **Moda (Mode)**: Valor que mais se repete. Útil para dados categóricos.

### 3. Medidas de Variação (Scale)

Indicam o quanto os dados estão **espalhados**:

- **Amplitude (Range)**: Diferença entre o máximo e o mínimo.
- **Variância (Variance)**: Média dos desvios quadráticos em relação à média.
- **Desvio Padrão (Standard Deviation)**: Raiz quadrada da variância.
- **Intervalo Interquartil (IQR)**: Diferença entre o 3º e o 1º quartil.

### 4. Medidas de Forma (Shape)

Descrevem o **formato** da distribuição:

- **Assimetria (Skewness)**: Indica se a distribuição é simétrica ou tende para um lado.
  - *Positiva*: cauda à direita
  - *Negativa*: cauda à esquerda
  - *Simétrica*: dados equilibrados
- **Curtose (Kurtosis)**: Mede o "achatamento" da distribuição.
  - *Mesocúrtica*: normal (≈ 0)
  - *Leptocúrtica*: pico alto, caudas pesadas
  - *Platicúrtica*: pico baixo, caudas leves

```python
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuração para gráficos em português
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
print('Bibliotecas importadas com sucesso!')
```

```python
# Criando um dataset sintético para demonstração
# Simulamos as notas (0-100) de 50 alunos em uma prova
np.random.seed(42)
notas = np.random.normal(loc=65, scale=15, size=50)
notas = np.clip(notas, 0, 100).round(1)

# Criando um DataFrame
df = pd.DataFrame({'Aluno': [f'Aluno_{i+1:02d}' for i in range(50)], 'Nota': notas})
print('Dataset criado:')
print(df.head(10))
```

```python
# Overview descritivo completo com pandas
print('=' * 50)
print('RESUMO DESCRITIVO DAS NOTAS')
print('=' * 50)

desc = df['Nota'].describe()
print(desc)

print('\n--- Medidas de Centro ---')
print(f'Média: {df["Nota"].mean():.2f}')
print(f'Mediana: {df["Nota"].median():.2f}')
print(f'Moda: {df["Nota"].mode().values}')

print('\n--- Medidas de Variação ---')
print(f'Amplitude (Range): {df["Nota"].max() - df["Nota"].min():.2f}')
print(f'Variância: {df["Nota"].var():.4f}')
print(f'Desvio Padrão: {df["Nota"].std():.4f}')
print(f'IQR: {df["Nota"].quantile(0.75) - df["Nota"].quantile(0.25):.4f}')

print('\n--- Medidas de Forma ---')
print(f'Assimetria (Skewness): {df["Nota"].skew():.4f}')
print(f'Curtose (Kurtosis): {df["Nota"].kurtosis():.4f}')
```

```python
# Visualizando a distribuição com histograma e box plot
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['Nota'], bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(df['Nota'].mean(), color='red', linestyle='--', linewidth=2, label=f'Média ({df["Nota"].mean():.1f})')
plt.axvline(df['Nota'].median(), color='green', linestyle='-', linewidth=2, label=f'Mediana ({df["Nota"].median():.1f})')
plt.title('Distribuição das Notas')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.legend()

plt.subplot(1, 2, 2)
plt.boxplot(df['Nota'], vert=True, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot das Notas')
plt.ylabel('Nota')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Comparando distribuições com diferentes formas
np.random.seed(123)

# Distribuição simétrica (normal)
simetrica = np.random.normal(loc=50, scale=10, size=1000)

# Distribuição assimétrica à direita (exponencial)
assimetrica_dir = np.random.exponential(scale=20, size=1000)

# Distribuição assimétrica à esquerda
assimetrica_esq = 100 - np.random.exponential(scale=20, size=1000)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].hist(simetrica, bins=20, color='steelblue', edgecolor='white')
axes[0].set_title('Distribuição Simétrica')
axes[0].set_xlabel('Valor')

axes[1].hist(assimetrica_dir, bins=20, color='coral', edgecolor='white')
axes[1].set_title('Assimetria Positiva (cauda à direita)')
axes[1].set_xlabel('Valor')

axes[2].hist(assimetrica_esq, bins=20, color='mediumseagreen', edgecolor='white')
axes[2].set_title('Assimetria Negativa (cauda à esquerda)')
axes[2].set_xlabel('Valor')

plt.tight_layout()
plt.show()
```

```python
# Exemplo 2: Impacto da variação na distribuição
np.random.seed(456)

baixa_var = np.random.normal(loc=70, scale=5, size=200)
media_var = np.random.normal(loc=70, scale=15, size=200)
alta_var = np.random.normal(loc=70, scale=30, size=200)

print('Comparação de Desvio Padrão:')
print(f'Baixa variação (std=5):  Média={baixa_var.mean():.1f}, Std={baixa_var.std():.1f}')
print(f'Média variação (std=15): Média={media_var.mean():.1f}, Std={media_var.std():.1f}')
print(f'Alta variação (std=30):  Média={alta_var.mean():.1f}, Std={alta_var.std():.1f}')

plt.figure(figsize=(10, 5))
plt.hist(baixa_var, bins=15, alpha=0.5, label='Std=5', color='blue')
plt.hist(media_var, bins=15, alpha=0.5, label='Std=15', color='orange')
plt.hist(alta_var, bins=15, alpha=0.5, label='Std=30', color='red')
plt.title('Mesma Média, Diferentes Variações')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Dado o conjunto de vendas diárias (em R$) de uma loja durante 15 dias: [1200, 850, 950, 1100, 1300, 900, 1050, 1150, 1250, 800, 1000, 1400, 950, 1100, 1200], calcule as principais medidas descritivas.

```python
# Solução do Exercício 1
vendas = [1200, 850, 950, 1100, 1300, 900, 1050, 1150, 1250, 800, 1000, 1400, 950, 1100, 1200]
s = pd.Series(vendas)
print('=== Medidas Descritivas das Vendas ===')
print(f'Média: R$ {s.mean():.2f}')
print(f'Mediana: R$ {s.median():.2f}')
print(f'Desvio Padrão: R$ {s.std():.2f}')
print(f'Mínimo: R$ {s.min():.2f}')
print(f'Máximo: R$ {s.max():.2f}')
print(f'Amplitude: R$ {s.max() - s.min():.2f}')
print(f'Assimetria: {s.skew():.4f}')
```

**Exercício 2:** Gere 100 números aleatórios de uma distribuição uniforme entre 0 e 100. Calcule média, desvio padrão e verifique se a assimetria está próxima de zero.

```python
# Solução do Exercício 2
np.random.seed(789)
uniformes = np.random.uniform(0, 100, 100)
print('=== Distribuição Uniforme ===')
print(f'Média: {uniformes.mean():.2f}')
print(f'Mediana: {np.median(uniformes):.2f}')
print(f'Desvio Padrão: {uniformes.std():.2f}')
sk = pd.Series(uniformes).skew()
print(f'Assimetria: {sk:.4f}')
print(f'Assimetria próxima de zero? {"Sim" if abs(sk) < 0.5 else "Não"}')
```

## 📌 Resumo

- A **Estatística Descritiva** organiza, resume e apresenta dados.
- Três dimensões: **centro**, **variação** e **forma**.
- **Medidas de centro**: média (sensível), mediana (robusta) e moda (frequência).
- **Medidas de variação**: amplitude, variância, desvio padrão e IQR.
- **Medidas de forma**: assimetria (skewness) e curtose (kurtosis).

## 💡 Curiosidades

Você sabia que o termo **"Estatística Descritiva"** foi popularizado por **Ronald Fisher** na década de 1920? Fisher também desenvolveu a ANOVA e o conceito de máxima verossimilhança. Outra curiosidade: **John Tukey**, criador do box plot, também inventou o termo **"bit"** (binary digit)!

## ✅ Boas Práticas

1. **Sempre comece com a descritiva** — antes de qualquer modelo, explore os dados.
2. **Nunca confie em uma única medida** — média sem desvio padrão pode enganar.
3. **Visualize sempre** — gráficos revelam padrões que números escondem.
4. **Documente outliers** — podem indicar erros ou descobertas importantes.
5. **Use sementes (seed)** para reprodutibilidade.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Reportar apenas a média sem dispersão | Sempre inclua desvio padrão ou IQR. |
| Ignorar a forma da distribuição | Um histograma revela assimetrias que a média esconde. |
| Confundir assimetria com curtose | Assimetria é direção da cauda; curtose é peso das caudas. |
| Achar que descritiva é simples demais | Análises descritivas bem feitas são a base da ciência de dados. |
