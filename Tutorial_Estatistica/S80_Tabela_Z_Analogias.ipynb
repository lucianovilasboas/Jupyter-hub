## Cardapio de Probabilidades

A **tabela Z** e como o **cardapio** de um restaurante.

Voce chega com um Z-score (seu pedido):
"Qual a probabilidade de Z ser menor que 1,96?"

Olha no cardapio (tabela Z):
Linha 1,9 + coluna 0,06 = 0,9750.

Pronto: 97,5% dos dados estao abaixo de Z=1,96.

Nao precisa decorar — precisa saber **consultar**.
Assim como voce nao decorar o cardapio,
mas sabe ler e encontrar o que quer.


## 📊 Traduzindo

| Cardapio de Probabilidades | Estatistica |
|-|-|
| Cardapio do restaurante | Tabela Z |
| "Quero um prato Z=1,96" | Z-score como pedido |
| Preco: R$ 0,9750 | Probabilidade acumulada |
| Linha 1,9 + coluna 0,06 | Como ler a tabela Z |


## 📏 Como Usar a Tabela Z

1. Calcule o Z-score
2. Na tabela, encontre a linha da parte inteira + 1 decimal
3. Encontre a coluna do 2o decimal
4. A celula e a probabilidade de Z ser menor que aquele valor

Exemplo: Z = 1,96
- Linha 1,9 | Coluna 0,06
- Valor: 0,9750
- Significa: P(Z < 1,96) = 97,5%


```python
# Cardapio Z em Python
from scipy import stats as sp_stats
valores_z = [0, 1.0, 1.64, 1.96, 2.5, 3.0]
print("Z   | P(Z < z)")
print("-"*15)
for z in valores_z:
    p = sp_stats.norm.cdf(z)
    print(f"{z:3.2f} | {p:.4f} ({p*100:.1f}%)")
```

## 📌 Resumo

| Cardapio de Probabilidades | Estatistica |
|-|-|
| Cardapio | Tabela Z |
| Pedido: Z=1,96 | Z-score |
| Preco: 0,9750 | Probabilidade |
| Saber consultar | Nao precisa decorar |
