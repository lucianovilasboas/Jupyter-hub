## Check-up Medico

Voce vai ao medico para um **check-up** anual.
O medico mede:
- **Pressao arterial** (media)
- **Peso** (mediana — o "meio" do seu peso)
- **Temperatura** (moda — o valor mais comum)
- **Batimentos cardiacos** (variacao ao longo do dia)

Cada medida conta algo diferente sobre sua saude.
Juntas, formam um **retrato completo** do seu estado fisico.

A **estatistica descritiva** e isso: um check-up dos seus dados!
Em vez de mostrar todos os 10.000 numeros brutos,
ela calcula medidas-resumo que contam a historia principal.


## 📊 Traduzindo

| Check-up Medico | Estatistica |
|-|-|
| Medir pressao | Calcular a media |
| Peso: valor do meio | Calcular a mediana |
| Temperatura mais frequente | Calcular a moda |
| Variacao dos batimentos | Calcular desvio padrao e variancia |
| Boletim medico com tudo resumido | Tabela de estatisticas descritivas |


## 📏 As Medidas do Check-up

**Media**: o centro dos dados (equilibrio)
**Mediana**: o valor do meio (robusta a extremos)
**Moda**: o valor mais comum (tipico)
**Desvio padrao**: a dispersao (quanto os dados variam)
**Minimo e Maximo**: os extremos

Cada medida e como um exame diferente no check-up.
Nenhum Exame sozinho conta a historia completa!


```python
# Check-up dos dados
import random, statistics as stat
random.seed(42)
# Simulando gastos mensais de 50 pessoas (em R$)
gastos = [random.gauss(2000, 500) for _ in range(50)]
print("=== CHECK-UP DOS DADOS ===
")
print(f"Pressao (media):      R$ {stat.mean(gastos):.2f}")
print(f"Peso (mediana):       R$ {stat.median(gastos):.2f}")
print(f"Temperatura (moda):   R$ {stat.mode(gastos):.2f}")
print(f"Variacao (desvio):    R$ {stat.stdev(gastos):.2f}")
print(f"Minimo:               R$ {min(gastos):.2f}")
print(f"Maximo:               R$ {max(gastos):.2f}")
```

## 📌 Resumo

| Check-up Medico | Estatistica |
|-|-|
| Medir pressao | Media |
| Peso do meio | Mediana |
| Temperatura mais comum | Moda |
| Variacao dos batimentos | Desvio padrao |
| Boletim medico | Tabela descritiva |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
