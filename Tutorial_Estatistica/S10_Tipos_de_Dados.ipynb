## 📝 Introdução

Nem todos os dados são iguais. Antes de qualquer análise, precisamos classificar os dados com que estamos trabalhando, pois o tipo de dado determina quais operações matemáticas e estatísticas são apropriadas. A classificação fundamental divide os dados em **qualitativos** (categóricos) e **quantitativos** (numéricos), cada um com suas subcategorias.

## 📚 Explicação Detalhada

### 1. Dados Qualitativos (Categóricos)

Dados **qualitativos** (qualitative/categorical data) descrevem qualidades, categorias ou atributos que não podem ser medidos numericamente no sentido matemático.

**Subtipos:**
- **Nominais** (nominal): categorias sem ordem intrínseca. Ex: cor dos olhos, estado civil, marca de carro
- **Ordinais** (ordinal): categorias com ordem definida, mas distâncias desiguais ou desconhecidas. Ex: nível de escolaridade, classe social, avaliação (ruim/médio/bom)

**Operações permitidas:**
- Nominais: contagem (frequência), moda
- Ordinais: contagem, moda, mediana, percentis
- **Não** se aplica: média, desvio padrão (não faz sentido aritmético)

### 2. Dados Quantitativos (Numéricos)

Dados **quantitativos** (quantitative/numerical data) representam quantidades mensuráveis e permitem operações aritméticas.

**Subtipos:**
- **Discretos** (discrete): valores inteiros e contáveis. Ex: número de filhos, quantidade de carros na família, contagem de acessos a um site
- **Contínuos** (continuous): valores em um intervalo contínuo, podem assumir qualquer valor real (com precisão limitada pelo instrumento). Ex: altura, peso, temperatura, tempo

**Operações permitidas:**
- Todas as operações aritméticas e estatísticas: média, mediana, desvio padrão, correlação
- Diferença fundamental: contínuos podem ser medidos em qualquer nível de precisão; discretos são contagens

```python
# Classificando diferentes tipos de dados
import pandas as pd
import numpy as np

np.random.seed(42)

# Criando um dataset com diversos tipos de dados
dados = pd.DataFrame({
    # QUALITATIVO NOMINAL
    'cor_preferida': np.random.choice(['Azul', 'Verde', 'Vermelho', 'Amarelo', 'Preto'], 20),
    'estado_civil': np.random.choice(['Solteiro', 'Casado', 'Divorciado'], 20),
    # QUALITATIVO ORDINAL
    'escolaridade': np.random.choice(['Fundamental', 'Médio', 'Superior', 'Pós-graduação'], 20),
    'satisfacao': np.random.choice(['Muito Ruim', 'Ruim', 'Regular', 'Bom', 'Excelente'], 20),
    # QUANTITATIVO DISCRETO
    'num_filhos': np.random.randint(0, 5, 20),
    'acessos_site': np.random.poisson(15, 20),
    # QUANTITATIVO CONTÍNUO
    'altura_cm': np.random.normal(170, 10, 20).round(1),
    'temperatura': np.random.uniform(36.0, 37.5, 20).round(1)
})

print('DATASET COM DIFERENTES TIPOS DE DADOS')
print(dados.to_string(index=False))
print()

# Verificando tipos no pandas
print('\nTipos detectados pelo pandas:')
print(dados.dtypes)
```

### 3. Como Identificar o Tipo de Dado

Para classificar corretamente um dado, faça as seguintes perguntas em ordem:

1. **É um número?**
   - Não → **Qualitativo**
   - Sim → **Quantitativo** (vá para pergunta 3)
2. **Se qualitativo, tem ordem?**
   - Não → **Nominal**
   - Sim → **Ordinal**
3. **Se quantitativo, pode assumir qualquer valor em um intervalo?**
   - Não (valores inteiros, contáveis) → **Discreto**
   - Sim → **Contínuo**

**Atenção para armadilhas comuns:**
- CEP é número mas é nominal (categoria sem ordem)
- Ano é número mas frequentemente é tratado como ordinal ou nominal
- Nota de 0 a 10 pode ser ordinal (conceitos) ou quantitativo discreto (pontuação exata)

```python
# Identificando tipos de dados — exemplos práticos

dados_exemplo = {
    'CEP': 12345678,           # NOMINAL (categoria, apesar de numérico)
    'Ano_nascimento': 1990,    # ORDINAL (tem ordem, mas diferença não é uniforme para análise temporal)
    'Nota_ENC': 8.5,           # QUANTITATIVO CONTÍNUO (ou discreto, depende)
    'Genero': 'Feminino',      # NOMINAL
    'Classificacao': 'Prata',  # ORDINAL (Bronze < Prata < Ouro)
    'Temp_Corporal': 36.8,     # QUANTITATIVO CONTÍNUO
    'Qtd_Irmaos': 2,           # QUANTITATIVO DISCRETO
    'Time_Futebol': 'Flamengo',# NOMINAL
    'Grau_Instrucao': 3,       # ORDINAL (1-Fund, 2-Médio, 3-Superior)
    'Renda_Mensal': 4500.00    # QUANTITATIVO CONTÍNUO
}

print('IDENTIFICAÇÃO DE TIPOS DE DADOS')
print('-' * 50)
for var, valor in dados_exemplo.items():
    if type(valor) == str:
        tipo = 'Qualitativo Nominal'
    elif var in ['CEP', 'Time_Futebol']:
        tipo = 'Qualitativo Nominal (categórico numérico)'
    elif var in ['Ano_nascimento', 'Grau_Instrucao', 'Classificacao']:
        tipo = 'Qualitativo Ordinal'
    elif type(valor) == int and valor < 100:
        tipo = 'Quantitativo Discreto'
    else:
        tipo = 'Quantitativo Contínuo'
    print(f'{var:20s} = {str(valor):10s} → {tipo}')
```

### 4. Importância de Classificar Corretamente

Classificar dados corretamente é crucial porque:

1. **Determina as análises possíveis**:
   - Nominais: teste qui-quadrado, proporções
   - Ordinais: mediana, correlação de Spearman
   - Quantitativos: média, teste t, correlação de Pearson, regressão

2. **Evita erros estatísticos**:
   - Calcular média de dados ordinais (ex: média de avaliações 'Ruim/Médio/Bom')
   - Tratar CEP como número para regressão

3. **Guia a visualização**:
   - Categóricos → barras, pizza
   - Numéricos → histograma, boxplot, dispersão

4. **Define a codificação para machine learning**:
   - Nominais → one-hot encoding
   - Ordinais → label encoding ordenado
   - Numéricos → uso direto ou normalização

```python
# Visualizações apropriadas para cada tipo de dado

print('VISUALIZAÇÕES POR TIPO DE DADO')
print()
print('Qualitativo Nominal:    gráfico de barras, gráfico de pizza')
print('Qualitativo Ordinal:    gráfico de barras ordenado')
print('Quantitativo Discreto:  gráfico de barras, histograma')
print('Quantitativo Contínuo:  histograma, boxplot, densidade, dispersão')
print()
print('ANÁLISES ESTATÍSTICAS POR TIPO DE DADO')
print()
print('Nominal:    frequência, moda, teste qui-quadrado')
print('Ordinal:    mediana, percentis, correlação de Spearman')
print('Discreto:   média, mediana, moda, desvio padrão')
print('Contínuo:   média, mediana, desvio padrão, correlação de Pearson, regressão')
print()
print('ATENÇÃO: Escolher a análise errada para o tipo de dado')
print('pode levar a conclusões inválidas!')
```

## 💻 Exemplos Práticos

```python
# Exemplo prático: classificação e manipulação de tipos de dados
np.random.seed(42)

# Base de dados de uma pesquisa de mercado
pesquisa = pd.DataFrame({
    'id_cliente': range(1, 101),
    'genero': np.random.choice(['Masculino', 'Feminino', 'Outro'], 100, p=[0.48, 0.48, 0.04]),
    'faixa_etaria': np.random.choice(['18-25', '26-35', '36-50', '51+'], 100),
    'estado': np.random.choice(['SP', 'RJ', 'MG', 'RS', 'BA'], 100),
    'renda_mensal': np.random.lognormal(8.0, 0.6, 100).round(0),
    'num_compras_mes': np.random.poisson(3, 100),
    'satisfacao': np.random.choice([1, 2, 3, 4, 5], 100, p=[0.05, 0.10, 0.20, 0.35, 0.30]),
    'possui_cartao': np.random.choice([True, False], 100)
})

print('BASE DE PESQUISA DE MERCADO — CLASSIFICAÇÃO:')
print('-' * 60)

classificacao = {
    'id_cliente': 'Quantitativo Discreto (identificador — tratar como nominal)',
    'genero': 'Qualitativo Nominal',
    'faixa_etaria': 'Qualitativo Ordinal',
    'estado': 'Qualitativo Nominal',
    'renda_mensal': 'Quantitativo Contínuo',
    'num_compras_mes': 'Quantitativo Discreto',
    'satisfacao': 'Qualitativo Ordinal (escala Likert 1-5)',
    'possui_cartao': 'Qualitativo Nominal (binário)'
}

for col, tipo in classificacao.items():
    print(f'{col:20s} → {tipo}')

print('\nESTATÍSTICAS APROPRIADAS PARA CADA COLUNA:')
print(f'Gênero (nominal): moda = {pesquisa["genero"].mode()[0]}')
print(f'Renda (contínuo): média = R$ {pesquisa["renda_mensal"].mean():.2f}')
print(f'Satisfação (ordinal): mediana = {pesquisa["satisfacao"].median()}')
print(f'Compras/mês (discreto): média = {pesquisa["num_compras_mes"].mean():.1f}')
```

## 📌 Resumo
- **Qualitativos**: categorias ou qualidades — nominais (sem ordem) ou ordinais (com ordem)
- **Quantitativos**: números — discretos (contáveis) ou contínuos (mensuráveis)
- O tipo de dado determina as operações estatísticas válidas
- Números nem sempre são quantitativos: CEP, códigos, etc. são categóricos
- Classificar corretamente previne erros de análise e visualização

## 💡 Curiosidades
- O nível de medida mais simples, o nominal, é o mais comum em pesquisas de marketing e ciências sociais
- Variáveis ordinais são frequentemente tratadas como numéricas em análises — isso é controverso e pode levar a erros
- Dados contínuos na prática são sempre discretizados pela precisão do instrumento: uma régua milimetrada torna a altura "discreta" em mm

## ✅ Boas Práticas
- Sempre classifique suas variáveis antes de começar a análise
- Use o tipo `category` do pandas para variáveis categóricas (economiza memória)
- Nunca calcule média de variáveis ordinais sem justificativa
- Documente a classificação de cada variável no dicionário de dados
- Em machine learning, aplique encoding adequado para cada tipo

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Tratar variáveis ordinais como numéricas | Verificar se as distâncias entre categorias são iguais |
| Calcular média de dados nominais codificados | Ex: média de 1=Azul, 2=Verde não faz sentido |
| Interpretar CEP ou CPF como números | São identificadores, use como string |
| Ignorar a diferença entre discreto e contínuo | Discreto para contagens, contínuo para medidas |
