## 📝 Introdução

O **range** (ou amplitude total) é a medida de dispersão mais simples: é a **diferença entre o maior e o menor valor** dos dados.

$$\text{Range} = X_{\text{máximo}} - X_{\text{mínimo}}$$

Apesar da simplicidade, o range é útil em:
- Previsão do tempo (amplitude térmica)
- Controle de qualidade (tolerância)
- Esportes (diferença entre primeiro e último)
- Finanças (máxima e mínima de uma ação)

> 🧠 **Para refletir:** Se a temperatura máxima for 35°C e a mínima 20°C, o range é 15°C. Isso dá a amplitude térmica, mas não diz como a temperatura variou durante o dia.

## 📚 Explicação Detalhada

### 1. Cálculo do Range

Range = **Valor Máximo - Valor Mínimo**

**Exemplo:** `{3, 8, 12, 15, 22}` → Máx=22, Mín=3, Range=19

Em Python: `np.ptp()` (peak-to-peak) ou `max() - min()`.

### 2. Vantagens
- **Simplicidade:** fácil de calcular e entender
- **Interpretação direta:** "os dados variam dentro de X unidades"

### 3. Limitações
- **Sensível a outliers:** um valor extremo distorce o range
- **Ignora a distribuição interna:** não diz nada sobre como os dados se distribuem
- **Depende do tamanho da amostra:** quanto mais dados, maior o range
- **Não é robusto:** dois conjuntos podem ter o mesmo range mas distribuições diferentes

## 💻 Exemplos Práticos

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Exemplo 1: Calculo do range
dados = np.array([3, 8, 12, 15, 22])
range_manual = np.max(dados) - np.min(dados)
range_numpy = np.ptp(dados)
print(f"Dados: {dados}")
print(f"Maximo: {np.max(dados)}")
print(f"Minimo: {np.min(dados)}")
print(f"Range (manual): {range_manual}")
print(f"Range (np.ptp): {range_numpy}")
```

```python
# Exemplo 2: Sensibilidade a outliers
sem_outlier = np.array([10, 12, 15, 18, 20, 22, 25])
com_outlier = np.array([10, 12, 15, 18, 20, 22, 25, 100])
print("Sem outlier:", sem_outlier)
print(f"  Range: {np.ptp(sem_outlier)}")
print("Com outlier (100):", com_outlier)
print(f"  Range: {np.ptp(com_outlier)}")
print("Um unico outlier aumentou o range de 15 para 90!")
```

```python
# Exemplo 3: Range vs IQR — robustez
dados_base = np.random.normal(50, 10, 100)
dados_com_outlier = np.append(dados_base, 500)
range_base = np.ptp(dados_base)
range_com = np.ptp(dados_com_outlier)
iqr_base = np.percentile(dados_base, 75) - np.percentile(dados_base, 25)
iqr_com = np.percentile(dados_com_outlier, 75) - np.percentile(dados_com_outlier, 25)
print(f"Range original:    {range_base:.2f}")
print(f"Range com outlier: {range_com:.2f}  (mudou!)")
print(f"IQR original:      {iqr_base:.2f}")
print(f"IQR com outlier:   {iqr_com:.2f}  (nao mudou!)")
print("O IQR e resistente a outliers; o range nao.")
```

```python
# Exemplo 4: Mesmo range, distribuicoes diferentes
a = np.array([0, 0, 0, 0, 20])
b = np.array([0, 5, 10, 15, 20])
c = np.array([0, 19, 19.5, 19.8, 20])
d = np.random.normal(10, 2, 100)
d = d - np.min(d)
d = d / np.max(d) * 20
fig, axes = plt.subplots(1, 4, figsize=(16, 3))
todos = [a, b, c, d]
tits = ["Todos no minimo", "Uniforme", "Todos no maximo", "Normal"]
for i, (dado, tit) in enumerate(zip(todos, tits)):
    axes[i].hist(dado, bins=5, color="skyblue", edgecolor="black")
    axes[i].set_title(f"{tit}\nRange = {np.ptp(dado):.1f}")
    axes[i].set_xlabel("Valor")
    axes[i].set_ylabel("Frequencia")
plt.tight_layout(); plt.show()
print("Todos os 4 conjuntos tem range ~20, mas as distribuicoes sao diferentes!")
```

```python
# Exemplo 5: Aplicacao — amplitude termica anual
dias = np.arange(365)
temp = 25 + 10 * np.sin(2 * np.pi * dias / 365 - np.pi/2) + np.random.normal(0, 3, 365)
print("Resumo das temperaturas do ano:")
print(f"  Media: {np.mean(temp):.1f}C")
print(f"  Maxima: {np.max(temp):.1f}C")
print(f"  Minima: {np.min(temp):.1f}C")
print(f"  Range (amplitude termica anual): {np.ptp(temp):.1f}C")
plt.figure(figsize=(12, 4))
plt.plot(dias, temp, '-', alpha=0.7, linewidth=0.5)
plt.axhline(np.max(temp), color='red', linestyle='--', label=f'Max: {np.max(temp):.1f}C')
plt.axhline(np.min(temp), color='blue', linestyle='--', label=f'Min: {np.min(temp):.1f}C')
plt.fill_between(dias, np.min(temp), np.max(temp), alpha=0.1, color='yellow')
plt.xlabel('Dia do Ano'); plt.ylabel('Temperatura (C)')
plt.title('Amplitude Termica Anual'); plt.legend(); plt.grid(True, alpha=0.3); plt.show()
```

```python
# Exercicio 1: Range de precos de acoes
precos = np.array([15.20, 15.80, 14.90, 16.10, 15.50, 14.70, 16.30])
range_precos = np.ptp(precos)
print(f"Precos: {precos}")
print(f"Preco max: R$ {np.max(precos):.2f}")
print(f"Preco min: R$ {np.min(precos):.2f}")
print(f"Range: R$ {range_precos:.2f}")
```

```python
# Exercicio 2: Comparando range
conj_a = np.random.normal(50, 5, 100)
conj_b = np.random.normal(50, 15, 100)
print(f"Conj A — Media={np.mean(conj_a):.1f}, Range={np.ptp(conj_a):.1f}")
print(f"Conj B — Media={np.mean(conj_b):.1f}, Range={np.ptp(conj_b):.1f}")
print("Conjunto B tem maior range (maior dispersao).")
```

## 📝 Resumo do Capítulo

- **Range** = valor máximo - valor mínimo
- É a medida de dispersão **mais simples**
- **Vantagem:** fácil de calcular e interpretar
- **Limitação:** extremamente sensível a outliers
- **Limitação:** ignora a distribuição interna
- Use `np.ptp()` para calcular em Python
- Prefira **IQR** ou **DP** para análises robustas

## 🧐 Curiosidades

- O range é usado em **cartas de controle** (R-chart) na indústria.
- Na meteorologia, o range de temperatura é a **amplitude térmica**.
- Walter Shewhart criou os gráficos de controle em 1924.
- Em esportes, o range entre primeiro e último é o **gap**.

## ✅ Boas Práticas

- Use range para **comunicação rápida** com público não-técnico
- Reporte **mínimo e máximo** junto com o range
- Não use range como **única medida** de dispersão
- Em controle de qualidade, range é útil para **amostras pequenas** (n < 10)

## ⚠️ Erros Comuns

| Erro | Consequência | Correção |
|------|-------------|----------|
| Range como única medida | Análise incompleta | Complete com IQR ou DP |
| Ignorar outliers no range | Superestima dispersão | Use IQR |
| Comparar ranges de amostras diferentes | Viés de tamanho | Use DP ou CV |
| Confundir range com IQR | Medidas diferentes | Range=max-min, IQR=Q3-Q1 |
