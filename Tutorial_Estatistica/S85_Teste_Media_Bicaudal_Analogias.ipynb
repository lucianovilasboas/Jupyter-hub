## Balanca da Farmacia

A balanca da farmacia diz que voce pesa **70 kg**.
Voce desconfia que ela esta **desregulada**.

Pode estar marcando MAIS ou MENOS que o real.
Nao importa a direcao — importa se e DIFERENTE.

Voce se pesa 10 vezes em balancas diferentes:
media: 71,5 kg. Desvio: 2 kg.

Teste **bicaudal**:
H0: media = 70 (balanca esta certa)
H1: media 70 (balanca esta errada — para qualquer lado)

O erro pode estar para cima ou para baixo.
Ambos os lados importam igualmente.


## 📊 Traduzindo

| Balanca da Farmacia | Estatistica |
|-|-|
| Balanca: 70 kg | H0: media = 70 |
| 10 medicoes: 71,5 kg | Media amostral |
| Pode ser mais OU menos | Teste bicaudal |
| Nao importa a direcao | Duas caudas |


```python
# Balanca da farmacia
import math, statistics as stat
from scipy import stats as sp_stats
pesagens = [71, 73, 70, 72, 71, 74, 70, 72, 71, 70]
media = stat.mean(pesagens)
desvio = stat.stdev(pesagens)
n = len(pesagens)
t = (media - 70) / (desvio / math.sqrt(n))
p = 2 * (1 - sp_stats.t.cdf(abs(t), df=n-1))
print(f"Media das pesagens: {media:.1f} kg")
print(f"H0: media = 70 kg")
print(f"T: {t:.3f}")
print(f"p-valor (bicaudal): {p:.4f}")
if p < 0.05:
    print("CONCLUSAO: Balanca desregulada!")
else:
    print("CONCLUSAO: Sem evidencias de desregulagem")
```

## 📌 Resumo

| Balanca da Farmacia | Estatistica |
|-|-|
| Balanca: 70 kg | H0 |
| Medicoes: 71,5 kg | Media amostral |
| Mais OU menos | Bicaudal |
| Ambos lados | Duas caudas |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
