## Chute no Escuro

Imagine um pote de vidro cheio de balas.
Alguem pergunta: **quantas balas tem no pote?**

Voce pode:
1. **Chutar um numero**: "Acho que tem 150 balas"
   → **estimativa pontual** (um unico valor)
2. **Dar uma faixa**: "Entre 140 e 160 balas"
   → **estimacao intervalar** (intervalo de confianca)

A estimativa intervalar e mais **honesta**:
ela admite que voce nao tem certeza.
A faixa tem mais chance de acertar o valor verdadeiro.


## 📊 Traduzindo

| Chute no Escuro | Estatistica |
|-|-|
| "Acho que tem 150 balas" | Estimativa pontual — chute unico |
| "Entre 140 e 160 balas" | Estimativa intervalar — faixa |
| Valor real: 153 balas | Parametro verdadeiro (desconhecido) |
| Quanto maior o pote | Maior a incerteza |


## 📏 Pontual vs Intervalar

**Estimativa pontual (x):**
- Unico valor, facil de entender
- Quase nunca acerta exatamente

**Intervalo de confianca (IC):**
- Faixa de valores plausiveis
- Quanto maior a confianca (95% vs 99%), maior a faixa
- Mais informacao: da o erro junto com a estimativa


```python
# Chute no escuro
import random
random.seed(42)
valor_real = 153  # ninguem sabe

# Simulando 5 pessoas chutando
chutes = [random.randint(140, 165) for _ in range(5)]
print(f"Valor real de balas: {valor_real}")
print(f"Chutes individuais: {chutes}")
print(f"Estimativa pontual (media dos chutes): {sum(chutes)/len(chutes):.0f}")
print(f"Estimativa intervalar: [{min(chutes)}, {max(chutes)}]")
```

## 📌 Resumo

| Chute no Escuro | Estatistica |
|-|-|
| "Chuto 150" | Estimativa pontual |
| "Entre 140 e 160" | Estimativa intervalar |
| Valor real no pote | Parametro verdadeiro |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
