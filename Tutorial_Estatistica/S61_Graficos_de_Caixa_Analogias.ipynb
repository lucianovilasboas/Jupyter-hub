## Raio-X do Salario

Imagine um **raio-X** dos salarios de uma empresa.
O medico (ou estatistico) quer ver:

- **Minimo**: o menor salario (estagiario: R$ 1.500)
- **Q1 (25%)**: 25% ganham ate R$ 2.500
- **Mediana (50%)**: o salario do meio: R$ 4.000
- **Q3 (75%)**: 75% ganham ate R$ 6.500
- **Maximo**: o maior salario (CEO: R$ 25.000)

O **box plot** (grafico de caixa) mostra isso de uma vez!
A "caixa" sao os 50% centrais. Os "bigodes" sao os extremos.
E os pontos fora sao os **outliers** — como o salario do CEO.


## 📊 Traduzindo

| Raio-X do Salario | Estatistica |
|-|-|
| Raio-X do corpo | Box plot dos dados |
| Ossos (estrutura central) | A caixa (Q1 a Q3) |
| Coracao (centro) | Mediana (Q2) |
| Bracos e pernas (extremidades) | Bigodes do box plot |
| Fratura fora do lugar | Outlier — valor extremo |


## 📏 Lendo o Box Plot

O box plot e poderoso porque mostra 5 informacoes num unico desenho:
1. **Minimo** (ponta do bigode inferior)
2. **Q1** (inicio da caixa)
3. **Mediana** (linha no meio da caixa)
4. **Q3** (fim da caixa)
5. **Maximo** (ponta do bigode superior)

E ainda destaca **outliers** — pontos que fogem do padrao.


```python
# Raio-X dos salarios
import random, matplotlib.pyplot as plt
random.seed(42)
salarios = [random.gauss(4000, 1500) for _ in range(100)]
salarios.append(25000)  # outlier: CEO

plt.figure(figsize=(6, 4))
plt.boxplot(salarios, vert=True, patch_artist=True,
            boxprops=dict(facecolor="lightblue"),
            medianprops=dict(color="red", linewidth=2))
plt.title("Raio-X dos Salarios")
plt.ylabel("Salario (R$)")
plt.grid(axis="y", alpha=0.3)
plt.show()
print("O ponto isolado e o CEO!")
```

## 📌 Resumo

| Raio-X do Salario | Estatistica |
|-|-|
| Raio-X | Box plot |
| Ossos (caixa) | Q1 a Q3 — 50% centrais |
| Coracao (centro) | Mediana |
| Bracos (bigodes) | Minimo e maximo |
| Fratura | Outlier |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
