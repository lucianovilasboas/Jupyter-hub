## Fita Metrica Curta

Imagine que voce so tem uma **fita metrica curta** (30 cm)
para medir uma sala. Sua medicao e menos precisa.

Agora imagine que voce tem uma **fita de 50 m**.
Sua medicao e muito mais precisa.

A **distribuicao T** e a "fita curta" da estatistica.
Quando voce tem **poucos dados** (amostra pequena),
a incerteza e maior — entao a distribuicao T tem
**caudas mais largas** que a normal, para compensar.

Conforme voce ganha mais dados (a fita cresce),
a distribuicao T se aproxima da normal.
Com n > 30, a diferenca e quase imperceptivel.


## 📊 Traduzindo

| Fita Metrica Curta | Estatistica |
|-|-|
| Fita metrica de 30 cm | Amostra pequena (poucos dados) |
| Fita de 50 m | Amostra grande (muitos dados) |
| Medicao imprecisa | Maior incerteza — caudas mais largas |
| Fita crescendo | n aumenta, T se aproxima da normal |
| Fita de 50 m = fita de 30 cm? | n > 30: T ≈ Normal |


## 📏 Quando Usar T

Use a **distribuicao T** quando:
- Amostra pequena (n < 30)
- Desvio padrao populacional (sigma) DESCONHECIDO

Use a **distribuicao Normal** quando:
- Amostra grande (n ≥ 30)
- OU desvio padrao populacional CONHECIDO


```python
# Fita curta vs fita longa
import math, statistics as stat
from scipy import stats as sp_stats
# Amostra pequena (n=10)
dados_pequenos = [5.2, 4.8, 5.1, 5.3, 4.9, 5.0, 5.2, 4.7, 5.1, 5.0]
n = len(dados_pequenos)
media = stat.mean(dados_pequenos)
erro = stat.stdev(dados_pequenos) / math.sqrt(n)
t_critico = sp_stats.t.ppf(0.975, df=n-1)  # 95% confianca
ic = (media - t_critico*erro, media + t_critico*erro)
print(f"n = {n}")
print(f"IC 95% usando T: [{ic[0]:.3f}, {ic[1]:.3f}]")
# Comparar com normal
z_critico = 1.96
ic_z = (media - z_critico*erro, media + z_critico*erro)
print(f"IC 95% usando Z: [{ic_z[0]:.3f}, {ic_z[1]:.3f}] (mais estreito!)")
```

## 📌 Resumo

| Fita Metrica Curta | Estatistica |
|-|-|
| Fita curta (30 cm) | Amostra pequena (n < 30) |
| Fita longa (50 m) | Amostra grande (n > 30) |
| Caudas mais largas | T tem mais incerteza |
| Fita cresce | T se aproxima da Normal |

> A estatistica esta em todo lugar — so precisamos do olhar certo.
