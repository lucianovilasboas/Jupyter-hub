## 📝 Introdução

Depois de coletar os dados, o próximo passo é **descrevê-los**. A Estatística Descritiva nos fornece ferramentas para resumir grandes conjuntos de dados em números e gráficos que revelam padrões, tendências e peculiaridades.

Descrever dados é como fazer um raio-X do seu dataset: você identifica valores típicos, verifica a variabilidade, detecta valores extremos e entende a forma da distribuição. Sem essa etapa, qualquer análise posterior pode ser enganosa ou incompleta.

## 📚 Explicação Detalhada

### 1. Medidas de Tendência Central

As medidas de tendência central indicam um valor típico ou central dos dados:

- **Média (mean)**: Soma de todos os valores dividida pela quantidade. Sensível a outliers.
- **Mediana (median)**: Valor do meio quando os dados estão ordenados. Robusta a outliers.
- **Moda (mode)**: Valor que mais aparece no conjunto de dados.

### 2. Medidas de Dispersão (Variabilidade)

Mostram o quanto os dados se espalham em torno da tendência central:

- **Amplitude (range)**: Diferença entre o maior e o menor valor.
- **Variância**: Média dos quadrados dos desvios em relação à média.
- **Desvio Padrão (standard deviation)**: Raiz quadrada da variância. Na mesma unidade dos dados originais.
- **Quartis e IQR**: Dividem os dados em quatro partes iguais. O IQR (intervalo interquartil) é a diferença entre Q3 e Q1.

### 3. Distribuição dos Dados

A distribuição descreve a frequência de cada valor. Podemos visualizá-la com:
- **Histograma**: Barras que mostram frequências em faixas de valores.
- **Boxplot**: Mostra mediana, quartis e outliers.
- **Gráfico de densidade**: Curva suave da distribuição.

```python
# Exemplo: calculando medidas descritivas em Python
dados = [12, 15, 14, 10, 18, 15, 13, 16, 14, 17, 15, 20, 11, 14, 13]

# Medidas de tendência central
media = sum(dados) / len(dados)
dados_ordenados = sorted(dados)
n = len(dados_ordenados)
if n % 2 == 0:
    mediana = (dados_ordenados[n//2 - 1] + dados_ordenados[n//2]) / 2
else:
    mediana = dados_ordenados[n//2]

# Moda (valor que mais aparece)
frequencias = {}
for v in dados:
    frequencias[v] = frequencias.get(v, 0) + 1
max_freq = max(frequencias.values())
moda = [k for k, v in frequencias.items() if v == max_freq]

print("=== Dados ===")
print(dados)
print(f"\nMédia: {media:.2f}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda} (frequência: {max_freq})")
```

```python
# Medidas de dispersão
# Variância e desvio padrão (populacional)
variancia = sum((x - media)**2 for x in dados) / len(dados)
desvio_padrao = variancia ** 0.5
amplitude = max(dados) - min(dados)

print("=== Medidas de Dispersão ===")
print(f"Amplitude: {amplitude}")
print(f"Variância: {variancia:.2f}")
print(f"Desvio Padrão: {desvio_padrao:.2f}")

# Quartis
def quartil(dados_ord, q):
    # q de 0 a 1
    pos = q * (len(dados_ord) - 1)
    inf = int(pos)
    frac = pos - inf
    return dados_ord[inf] * (1 - frac) + dados_ord[min(inf + 1, len(dados_ord)-1)] * frac

q1 = quartil(dados_ordenados, 0.25)
q3 = quartil(dados_ordenados, 0.75)
iqr = q3 - q1
print(f"Q1 (25%): {q1}")
print(f"Q3 (75%): {q3}")
print(f"IQR (intervalo interquartil): {iqr}")
```

## 💻 Exemplos Práticos

```python
# Exemplo prático: análise descritiva de um dataset de vendas
import matplotlib.pyplot as plt

# Vendas diárias de uma loja (em R$) durante 30 dias
vendas = [1200, 850, 950, 1100, 1300, 900, 1050, 1150, 980, 1250,
          1080, 920, 1400, 1020, 1180, 960, 1350, 890, 1120, 1050,
          1280, 940, 1000, 1220, 1150, 980, 1320, 910, 1070, 1200]

# Estatísticas descritivas
media_vendas = sum(vendas) / len(vendas)
vendas_ord = sorted(vendas)
mediana_vendas = vendas_ord[len(vendas_ord)//2]
dp_vendas = (sum((x - media_vendas)**2 for x in vendas) / len(vendas)) ** 0.5

print("=== Análise de Vendas (30 dias) ===")
print(f"Média: R$ {media_vendas:.2f}")
print(f"Mediana: R$ {mediana_vendas:.2f}")
print(f"Desvio Padrão: R$ {dp_vendas:.2f}")
print(f"Mínimo: R$ {min(vendas):.2f}")
print(f"Máximo: R$ {max(vendas):.2f}")
```

```python
# Visualizando a distribuição das vendas
plt.figure(figsize=(12, 4))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(vendas, bins=8, color='skyblue', edgecolor='black')
plt.axvline(media_vendas, color='red', linestyle='--', label=f'Média: R$ {media_vendas:.0f}')
plt.axvline(mediana_vendas, color='green', linestyle='--', label=f'Mediana: R$ {mediana_vendas:.0f}')
plt.title('Histograma das Vendas')
plt.xlabel('Vendas (R$)')
plt.ylabel('Frequência')
plt.legend()

# Boxplot
plt.subplot(1, 2, 2)
plt.boxplot(vendas, vert=True)
plt.title('Boxplot das Vendas')
plt.ylabel('Vendas (R$)')
plt.grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Dado o conjunto de notas [6, 7, 8, 9, 10, 6, 7, 8, 9, 10, 5, 6, 7], calcule média, mediana e moda.

```python
# Solução do Exercício 1
notas = [6, 7, 8, 9, 10, 6, 7, 8, 9, 10, 5, 6, 7]
media_n = sum(notas) / len(notas)
notas_ord = sorted(notas)
mediana_n = notas_ord[len(notas_ord)//2]

freq_n = {}
for v in notas:
    freq_n[v] = freq_n.get(v, 0) + 1
max_f = max(freq_n.values())
moda_n = [k for k, v in freq_n.items() if v == max_f]

print(f"Notas: {notas}")
print(f"Média: {media_n:.2f}")
print(f"Mediana: {mediana_n}")
print(f"Moda: {moda_n} (frequência: {max_f})")
```

**Exercício 2:** Os dados [2, 4, 4, 4, 5, 5, 7, 9] têm um outlier? Calcule o IQR e verifique.

```python
# Solução do Exercício 2
dados2 = [2, 4, 4, 4, 5, 5, 7, 9]
d2_ord = sorted(dados2)

# Quartis
q1_2 = d2_ord[len(d2_ord)//4]
q3_2 = d2_ord[(3*len(d2_ord))//4]
iqr_2 = q3_2 - q1_2
limite_inf = q1_2 - 1.5 * iqr_2
limite_sup = q3_2 + 1.5 * iqr_2

print(f"Q1: {q1_2}, Q3: {q3_2}, IQR: {iqr_2}")
print(f"Limites: [{limite_inf}, {limite_sup}]")
outliers = [x for x in dados2 if x < limite_inf or x > limite_sup]
if outliers:
    print(f"Outliers encontrados: {outliers}")
else:
    print("Nenhum outlier encontrado.")
```

## 💡 Curiosidades

O **desvio padrão** foi introduzido por Karl Pearson em 1894. Pearson foi um dos fundadores da estatística moderna e também criou o coeficiente de correlação. Uma curiosidade: em uma distribuição normal (sino), aproximadamente 68% dos dados estão dentro de 1 desvio padrão da média, 95% dentro de 2 desvios e 99.7% dentro de 3 desvios. Essa é a chamada **regra empírica** ou regra 68-95-99.7.

## ✅ Boas Práticas

1. Sempre visualize os dados antes de calcular estatísticas.
2. Use a mediana em vez da média quando houver outliers.
3. Reporte sempre a média acompanhada do desvio padrão.
4. Prefira o IQR ao desvio padrão para descrever distribuições assimétricas.
5. Documente as unidades de medida de todas as variáveis.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Usar a média em dados com outliers | Prefira a mediana para dados assimétricos. |
| Ignorar a variabilidade dos dados | Sempre reporte a média com uma medida de dispersão. |
| Confundir desvio padrão com erro padrão | Desvio padrão mede dispersão dos dados; erro padrão mede precisão da estimativa. |
| Interpretar a moda em dados contínuos | A moda é mais útil para dados categóricos ou discretos. |
