## Fabrica de Refrigerante

Uma fabrica diz que cada lata de refrigerante tem **350 mL**.

Voce desconfia que esta vindo **menos**.
Compra 20 latas e mede o conteudo:
Media: 345 mL. Desvio: 8 mL.

H0: a media verdadeira e 350 (a fabrica esta certa).
H1: a media verdadeira e menor que 350 (a fabrica engana).

O **teste T para media** responde:
se H0 fosse verdadeira, qual a chance de uma amostra
de 20 latas ter media 345 ou menos?

Se essa chance (p-valor) for pequena ( < 0,05 ),
rejeitamos H0: a fabrica esta enganando!


## 📊 Traduzindo

| Fabrica de Refrigerante | Estatistica |
|-|-|
| Fabrica: 350 mL por lata | H0: media = 350 |
| Voce mede 20 latas: 345 mL | Amostra: n=20, x=345 |
| Diferenca de 5 mL | Diferenca observada |
| E so azar ou e enganacao? | Teste T responde |


```python
# Teste da fabrica
import math, statistics as stat
from scipy import stats as sp_stats
amostra = [348, 342, 347, 344, 346, 340, 349, 343, 345, 347,
           341, 348, 344, 346, 342, 349, 345, 343, 347, 346]
media = stat.mean(amostra)
desvio = stat.stdev(amostra)
n = len(amostra)
t = (media - 350) / (desvio / math.sqrt(n))
from scipy import stats as sp_stats
p_valor = sp_stats.t.cdf(t, df=n-1)
print(f"Media amostral: {media:.1f} mL")
print(f"H0: media = 350 mL")
print(f"Estatistica T: {t:.3f}")
print(f"p-valor: {p_valor:.4f}")
if p_valor < 0.05:
    print("CONCLUSAO: Rejeitamos H0 — fabrica engana!")
else:
    print("CONCLUSAO: Nao rejeitamos H0")
```

## 📌 Resumo

| Fabrica de Refrigerante | Estatistica |
|-|-|
| Fabrica: 350 mL | H0: media = 350 |
| Medicao: 345 mL | Media amostral |
| Diferenca | Estatistica T |
| Chance de ser acaso | p-valor |
