## 📝 Introdução

Em 1946, o psicólogo Stanley Smith Stevens publicou um artigo que se tornaria fundamental para a estatística: "On the Theory of Scales of Measurement". Ele propôs que as medições podem ser classificadas em quatro **níveis** (scales/levels of measurement), cada um com propriedades matemáticas específicas. Esta classificação é mais refinada que a simples divisão entre qualitativo e quantitativo e determina exatamente quais operações estatísticas são válidas para cada variável.

## 📚 Explicação Detalhada

### 1. Escala Nominal

A **escala nominal** (nominal scale) é o nível mais básico de medição. Os números ou rótulos são usados apenas para **nomear** ou **categorizar** — não há ordem, magnitude ou distância entre as categorias.

**Propriedades:**
- Identidade: cada valor é único e identifica uma categoria
- Sem ordem: as categorias podem ser reorganizadas livremente
- Sem distância: não existe diferença quantitativa entre categorias

**Exemplos:** gênero, estado civil, cor dos olhos, time do coração, país de origem, tipo sanguíneo, CEP

**Operações permitidas:**
- Contagem (frequência absoluta e relativa)
- Moda (categoria mais frequente)
- Teste qui-quadrado

**Operações PROIBIDAS:**
- Média, mediana (não fazem sentido)
- Soma, subtração, multiplicação, divisão

### 2. Escala Ordinal

Na **escala ordinal** (ordinal scale), as categorias têm uma **ordem** definida, mas as distâncias entre elas não são conhecidas ou uniformes. Sabemos quem é maior/menor, mas não o quanto maior/menor.

**Propriedades:**
- Identidade + Ordem
- As categorias são comparáveis em termos de "maior que" e "menor que"
- A magnitude das diferenças entre categorias não é constante

**Exemplos:** nível de escolaridade (Fundamental < Médio < Superior), classe social (Baixa < Média < Alta), avaliação (Ruim < Regular < Bom < Excelente), estágio de doença

**Operações permitidas:**
- Tudo do nível nominal +
- Mediana
- Percentis, quartis
- Correlação de Spearman

**Operações PROIBIDAS:**
- Média aritmética (controverso — usada em alguns contextos com cautela)
- Desvio padrão

```python
# Demonstração: Nominal vs Ordinal
import numpy as np
import pandas as pd

np.random.seed(42)

# Escala Nominal: cores preferidas (sem ordem)
cores = np.random.choice(['Azul', 'Verde', 'Vermelho', 'Amarelo'], 100)

# Escala Ordinal: nível de satisfação (com ordem)
satisfacao = np.random.choice(['Muito Ruim', 'Ruim', 'Regular', 'Bom', 'Muito Bom'], 100,
                               p=[0.05, 0.10, 0.20, 0.40, 0.25])

# Ordem intrínseca da satisfação
ordem_satisfacao = ['Muito Ruim', 'Ruim', 'Regular', 'Bom', 'Muito Bom']

print('ESCALA NOMINAL — CORES (sem ordem):')
print(pd.Series(cores).value_counts())
print(f'Moda: {pd.Series(cores).mode()[0]})')
print()

print('ESCALA ORDINAL — SATISFAÇÃO (com ordem):')
contagem = pd.Series(satisfacao).value_counts()
for nivel in ordem_satisfacao:
    print(f'  {nivel:15s}: {contagem.get(nivel, 0)}')
print(f'Moda: {pd.Series(satisfacao).mode()[0]}')
print(f'Categoria mediana (50º percentil): {ordem_satisfacao[2]}')
print()
print('Nota: na escala ordinal podemos calcular mediana, mas não média.')
```

### 3. Escala Intervalar

Na **escala intervalar** (interval scale), além da ordem, as **distâncias** entre valores são conhecidas e iguais. No entanto, **não existe um zero verdadeiro** — o zero é arbitrário.

**Propriedades:**
- Identidade + Ordem + Distância igual
- Intervalos iguais representam diferenças iguais na grandeza medida
- Zero arbitrário (não significa ausência da grandeza)

**Exemplos:**
- Temperatura em °C ou °F (0°C não significa "sem temperatura")
- QI (quociente de inteligência)
- Ano (0 d.C. é arbitrário)
- Notas em testes padronizados

**Operações permitidas:**
- Tudo do nível ordinal +
- Média aritmética
- Desvio padrão, variância
- Correlação de Pearson
- Teste t, ANOVA

**Operações PROIBIDAS:**
- Multiplicação e divisão (não faz sentido dizer que 20°C é "duas vezes mais quente" que 10°C)
- Razões (ratio) — pois não há zero verdadeiro

### 4. Escala de Razão

A **escala de razão** (ratio scale) é o nível mais completo de medição. Possui todas as propriedades anteriores **mais um zero absoluto** (verdadeiro), que representa a ausência total da grandeza.

**Propriedades:**
- Identidade + Ordem + Distância igual + Zero absoluto
- Permite todas as operações aritméticas, inclusive razões

**Exemplos:**
- Altura, peso, distância (0 cm = ausência de altura)
- Renda (R$ 0 = nenhuma renda)
- Tempo de reação (0 ms = nenhum tempo)
- Número de filhos, vendas, clientes

**Operações permitidas:**
- Todas: média, mediana, moda, desvio padrão
- **Razões**: faz sentido dizer que 200 kg é o dobro de 100 kg
- Coeficiente de variação (CV = σ/μ)
- Todas as análises estatísticas e operações aritméticas

```python
# Comparação: Intervalar vs Razão
import numpy as np

np.random.seed(42)

# Dados INTERVALARES: temperatura em °C (zero arbitrário)
temp_outono = np.random.normal(18, 5, 30)   # °C
term_primavera = np.random.normal(21, 5, 30)  # °C

# Dados de RAZÃO: altura em cm (zero absoluto)
altura_homens = np.random.normal(175, 8, 30)   # cm
altura_mulheres = np.random.normal(162, 7, 30) # cm

print('ESCALA INTERVALAR — TEMPERATURA (°C):')
print(f'Outono: média = {np.mean(temp_outono):.1f}°C')
print(f'Primavera: média = {np.mean(term_primavera):.1f}°C')
print(f'Diferença = {np.mean(term_primavera) - np.mean(temp_outono):.1f}°C')
# Proibido: "Primavera é X vezes mais quente"
# print(f'Razão = {np.mean(term_primavera) / np.mean(temp_outono):.2f}') # ERRADO!
print(f'(Razão não é válida para escala intervalar — 0°C é arbitrário)')
print()

print('ESCALA DE RAZÃO — ALTURA (cm):')
print(f'Homens: média = {np.mean(altura_homens):.1f} cm')
print(f'Mulheres: média = {np.mean(altura_mulheres):.1f} cm')
print(f'Diferença = {np.mean(altura_homens) - np.mean(altura_mulheres):.1f} cm')
# Permitido: "Homens são X vezes mais altos" (zero absoluto)
razao = np.mean(altura_homens) / np.mean(altura_mulheres)
print(f'Razão (homens/mulheres) = {razao:.3f}')
```

### 5. Tabela Resumo dos Níveis de Medição

| Propriedade | Nominal | Ordinal | Intervalar | Razão |
|-------------|---------|---------|------------|-------|
| Identidade (igual/diferente) | ✅ | ✅ | ✅ | ✅ |
| Ordem (maior/menor) | ❌ | ✅ | ✅ | ✅ |
| Distância igual entre valores | ❌ | ❌ | ✅ | ✅ |
| Zero absoluto | ❌ | ❌ | ❌ | ✅ |

**Operações estatísticas por nível:**
- Nominal: moda, frequência, qui-quadrado
- Ordinal: + mediana, percentis, Spearman
- Intervalar: + média, desvio padrão, Pearson, t, ANOVA
- Razão: + coeficiente de variação, razões, todas as operações aritméticas

```python
# Determinando o nível de medição de várias variáveis

def classificar_nivel(var_nome, descricao):
    """
    Classifica o nível de medição e explica o critério.
    """
    print(f'{var_nome:25s} → ', end='')
    
    # Regras de classificação
    if descricao.get('zero_absoluto') and descricao.get('distancia_igual'):
        nivel = 'Razão'
    elif descricao.get('distancia_igual'):
        nivel = 'Intervalar'
    elif descricao.get('ordem'):
        nivel = 'Ordinal'
    else:
        nivel = 'Nominal'
    
    print(f'{nivel}')
    return nivel

variaveis = {
    'Gênero': {'ordem': False, 'distancia_igual': False, 'zero_absoluto': False},
    'Grau de instrução': {'ordem': True, 'distancia_igual': False, 'zero_absoluto': False},
    'Temperatura (°C)': {'ordem': True, 'distancia_igual': True, 'zero_absoluto': False},
    'Peso (kg)': {'ordem': True, 'distancia_igual': True, 'zero_absoluto': True},
    'Ano de nascimento': {'ordem': True, 'distancia_igual': True, 'zero_absoluto': False},
    'Renda mensal (R$)': {'ordem': True, 'distancia_igual': True, 'zero_absoluto': True},
    'Nível de dor (1-10)': {'ordem': True, 'distancia_igual': False, 'zero_absoluto': False},
    'Cor dos olhos': {'ordem': False, 'distancia_igual': False, 'zero_absoluto': False},
    'Distância (km)': {'ordem': True, 'distancia_igual': True, 'zero_absoluto': True},
    'QI (Quociente Int.)': {'ordem': True, 'distancia_igual': True, 'zero_absoluto': False},
}

print('CLASSIFICAÇÃO POR NÍVEL DE MEDIÇÃO')
print('=' * 50)
for var, desc in variaveis.items():
    classificar_nivel(var, desc)
```

## 💻 Exemplos Práticos

```python
# Exemplo prático: aplicando os níveis de medição a um dataset real

np.random.seed(42)

# Dataset de pacientes
pacientes = pd.DataFrame({
    'id': range(1, 51),
    'sexo': np.random.choice(['F', 'M'], 50),                          # NOMINAL
    'grupo_sanguineo': np.random.choice(['A', 'B', 'AB', 'O'], 50),    # NOMINAL
    'estagio_cancer': np.random.choice(['I', 'II', 'III', 'IV'], 50),  # ORDINAL
    'nivel_dor': np.random.randint(1, 11, 50),                         # ORDINAL (escala subjetiva 1-10)
    'temp_corporal': np.random.normal(36.5, 0.5, 50).round(1),         # INTERVALAR
    'pressao_sist': np.random.normal(120, 15, 50).round(0),            # RAZÃO (zero absoluto)
    'altura_cm': np.random.normal(165, 10, 50).round(1),               # RAZÃO
    'peso_kg': np.random.normal(70, 15, 50).round(1),                  # RAZÃO
    'qtd_consultas': np.random.poisson(3, 50)                          # RAZÃO (discreta)
})

print('DATASET DE PACIENTES — NÍVEIS DE MEDIÇÃO')
print('=' * 60)
for col in pacientes.columns:
    if col == 'id':
        print(f'{col:20s} → Razão (mas tratado como identificador nominal)')
    elif col in ['sexo', 'grupo_sanguineo']:
        print(f'{col:20s} → Nominal')
    elif col in ['estagio_cancer', 'nivel_dor']:
        print(f'{col:20s} → Ordinal')
    elif col == 'temp_corporal':
        print(f'{col:20s} → Intervalar')
    else:
        print(f'{col:20s} → Razão')

print()
print('ANÁLISES APROPRIADAS:')
print(f'Sexo (nominal): {pacientes["sexo"].value_counts().to_dict()}')
print(f'Estágio câncer (ordinal): mediana = {pacientes["estagio_cancer"].mode()[0]}')
print(f'Temperatura (intervalar): média = {pacientes["temp_corporal"].mean():.1f}°C, dp = {pacientes["temp_corporal"].std():.1f}')
print(f'Altura (razão): média = {pacientes["altura_cm"].mean():.1f} cm, CV = {pacientes["altura_cm"].std()/pacientes["altura_cm"].mean()*100:.1f}%')
```

## 📌 Resumo
- **Nominal**: identidade apenas — categorias sem ordem. Ex: cor, gênero
- **Ordinal**: identidade + ordem — categorias ordenadas. Ex: escolaridade, cargo
- **Intervalar**: identidade + ordem + distância igual — zero arbitrário. Ex: °C, QI
- **Razão**: tudo anterior + zero absoluto. Ex: altura, peso, renda
- Cada nível herda as operações dos níveis anteriores e adiciona novas
- Escolher a análise errada para o nível de medição leva a conclusões inválidas

## 💡 Curiosidades
- Stanley Smith Stevens (1906-1973) era psicólogo, não estatístico — sua classificação foi inicialmente criticada, mas tornou-se padrão
- A escala Kelvin foi proposta por Lord Kelvin em 1848 e é a única escala de temperatura com zero absoluto (0 K = -273,15°C)
- O coeficiente de variação (CV = σ/μ) só é válido para escala de razão, pois depende do zero absoluto para ser interpretado

## ✅ Boas Práticas
- Identifique o nível de medição de cada variável antes de qualquer análise
- Escolha o teste estatístico baseado no nível mais baixo das variáveis envolvidas
- Ao criar questionários, defina o nível de medição desejado para cada questão
- Para variáveis ordinais, prefira medianas e percentis a médias
- Documente o nível de medição no dicionário de dados do projeto

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Calcular média de variáveis ordinais | Usar mediana para variáveis ordinais |
| Tratar escala Likert como intervalar automaticamente | Verificar se os intervalos são percebidos como iguais |
| Multiplicar/dividir temperaturas Celsius | Lembrar que 0°C é arbitrário — usar Kelvin para razões |
| Ignorar o nível de medição na escolha do teste qui-quadrado | Qui-quadrado é para nominais; Spearman para ordinais; Pearson para intervalar/razão |
