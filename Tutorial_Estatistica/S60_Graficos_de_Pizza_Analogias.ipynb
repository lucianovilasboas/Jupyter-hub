## Conta do Restaurante

Quatro amigos vao a um restaurante.
A conta total deu R$ 100.

Cada um comeu algo diferente:
- Ana: R$ 30 (30% do total)
- Bia: R$ 25 (25%)
- Carlos: R$ 35 (35%)
- Dani: R$ 10 (10%)

Se voce desenhar um circulo e dividir nesses pedacos,
tem um **grafico de pizza** (ou grafico de setores)!

O grafico de pizza mostra **partes de um todo**.
Cada fatia e uma categoria, e o tamanho mostra a proporcao.


## 📊 Traduzindo

| Conta do Restaurante | Estatistica |
|-|-|
| Conta total de R$ 100 | Total dos dados (100%) |
| Quanto cada um comeu | Proporcao de cada categoria |
| Fatia de 30% | 30% dos dados pertencem a essa categoria |
| Pizza inteira | O conjunto completo dos dados |


## 📏 Quando Usar (e Quando EVITAR)

**Use quando:**
- Voce tem poucas categorias (2-5)
- Quer mostrar proporcoes em relacao ao todo

**Evite quando:**
- Tem muitas categorias (vira uma pizza confusa)
- As proporcoes sao muito parecidas (dificil comparar)
- Voce quer comparar valores absolutos (use barras)


```python
# A conta do restaurante
import matplotlib.pyplot as plt
pessoas = ["Ana", "Bia", "Carlos", "Dani"]
valores = [30, 25, 35, 10]
cores = ["gold", "lightcoral", "lightskyblue", "lightgreen"]

plt.figure(figsize=(6, 6))
plt.pie(valores, labels=pessoas, autopct="%1.0f%%", colors=cores, startangle=90)
plt.title("Divisao da Conta do Restaurante")
plt.show()
```

## 📌 Resumo

| Conta do Restaurante | Estatistica |
|-|-|
| Conta total | 100% dos dados |
| Fatia de cada um | Proporcao de cada categoria |
| Pizza inteira | Conjunto completo |
