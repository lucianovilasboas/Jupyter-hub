## Ranking da Turma

Na escola, o professor afixa o **ranking** das notas no quadro.
Cada aluno e uma barra. A altura da barra e a nota.

Com uma olhada, voce ja sabe:
- Quem foi muito bem (barras altas)
- Quem foi mal (barras baixas)
- A diferenca entre o 1o e o ultimo

O **grafico de barras** e exatamente isso:
uma comparacao visual entre categorias.

Nao precisa ler numeros — os olhos veem as alturas.


## 📊 Traduzindo

| Ranking da Turma | Estatistica |
|-|-|
| Ranking da turma | Grafico de barras |
| Altura da barra = nota | Altura = valor da variavel |
| Cada aluno | Cada categoria / grupo |
| Diferenca entre 1o e ultimo | Comparacao visual entre categorias |


## 📏 Barras Verticais vs Horizontais

Barras **verticais**: boas para poucas categorias.
Barras **horizontais**: melhores para muitas categorias
(os rotulos cabem melhor).

Regra de ouro: comece do zero no eixo Y!
Senao, uma barra de 10 parece ter metade do tamanho de uma de 12.


```python
# Ranking da turma em Python
import matplotlib.pyplot as plt
alunos = ["Ana", "Bia", "Carlos", "Dani", "Edu"]
notas = [8.5, 6.0, 9.2, 7.8, 5.5]

plt.figure(figsize=(8, 4))
plt.bar(alunos, notas, color=["green" if n>=7 else "tomato" for n in notas])
plt.axhline(y=7, color="gray", linestyle="--", label="Media (7.0)")
plt.title("Ranking da Turma")
plt.ylabel("Nota")
plt.ylim(0, 10)
plt.legend()
plt.grid(axis="y", alpha=0.3)
plt.show()
print(f"Media da turma: {sum(notas)/len(notas):.1f}")
```

## 📌 Resumo

| Ranking da Turma | Estatistica |
|-|-|
| Ranking da turma | Grafico de barras |
| Altura = nota | Altura = valor |
| Cada aluno | Cada categoria |
