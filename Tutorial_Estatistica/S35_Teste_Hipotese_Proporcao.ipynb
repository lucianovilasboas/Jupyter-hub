## 📝 Introdução

Em muitas situações práticas, estamos interessados em testar hipóteses sobre proporções — a fração de indivíduos ou itens que possuem uma determinada característica. Por exemplo: "Mais de 60% dos eleitores aprovam o governo?" ou "A proporção de peças defeituosas diminuiu após o ajuste da máquina?".

O **teste Z para proporção** é a ferramenta adequada para essas situações. Utilizamos a distribuição normal como aproximação da distribuição binomial (válida para amostras grandes).

## 📚 Explicação Detalhada

### 1. Quando Usar o Teste Z para Proporção

O teste Z para proporção é usado quando:
- Os dados são categóricos (sucesso/falha, sim/não)
- A amostra é suficientemente grande: n·p₀ ≥ 10 e n·(1-p₀) ≥ 10
- As observações são independentes (amostragem aleatória simples)
- Estamos testando H₀: p = p₀, H₁: p > p₀, p < p₀, ou p ≠ p₀

### 2. Estatística Z para Proporção

A fórmula da estatística Z é:

Z = (p̂ - p₀) / √(p₀(1-p₀)/n)

Onde:
- p̂ = x/n é a proporção amostral (x = número de sucessos)
- p₀ é a proporção sob H₀
- n é o tamanho amostral

### 3. Tipos de Teste

| Teste | H₀ | H₁ | Região de Rejeição |
|-------|-----|-----|-------------------|
| Bicaudal | p = p₀ | p ≠ p₀ | Z < -Z_{α/2} ou Z > Z_{α/2} |
| Unilateral Direita | p ≤ p₀ | p > p₀ | Z > Z_{α} |
| Unilateral Esquerda | p ≥ p₀ | p < p₀ | Z < -Z_{α} |

### 4. Abordagem com p-valor vs Valor Crítico

Na abordagem do **p-valor**: calculamos a probabilidade de observar uma proporção tão extrema quanto a obtida, assumindo H₀ verdadeira. Se p-valor < α, rejeitamos H₀.

Na abordagem do **valor crítico**: comparamos Z calculado com Z crítico da tabela normal. Se |Z| > Z_{crítico}, rejeitamos H₀.

## 💻 Exemplos Práticos

```python
# Importando bibliotecas
import numpy as np
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
import matplotlib.pyplot as plt

print("Bibliotecas carregadas com sucesso!")
```

```python
# Exemplo 1: Teste Z para proporcao (bicaudal) - calculo manual
# Um candidato afirma que 40% dos eleitores o apoiam.
# Uma pesquisa com 200 eleitores mostrou 68 apoiadores.
# Teste a 5% de significancia.

n = 200
x = 68  # numero de "sucessos" (apoiadores)
p0 = 0.40  # proporcao sob H0
alpha = 0.05

# Proporcao amostral
p_hat = x / n

# Erro padrao sob H0
erro_padrao = np.sqrt(p0 * (1 - p0) / n)

# Estatistica Z
z = (p_hat - p0) / erro_padrao

# Valor critico (bicaudal)
z_crit = stats.norm.ppf(1 - alpha / 2)

# p-valor
p_valor = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"n = {n}, x (sucessos) = {x}")
print(f"Proporção amostral (p̂) = {p_hat:.3f}")
print(f"Proporção sob H₀ (p₀) = {p0}")
print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
print(f"Erro padrão = {erro_padrao:.4f}")
print(f"Estatística Z = {z:.4f}")
print(f"Valor crítico Z = ±{z_crit:.4f}")
print(f"p-valor = {p_valor:.4f}")
print()
if abs(z) > z_crit or p_valor < alpha:
    print("✅ Rejeitamos H₀: a proporção de apoiadores é diferente de 40%.")
else:
    print("❌ Não rejeitamos H₀: não há evidência de diferença significativa.")
```

```python
# Exemplo 2: Usando o proportions_ztest do statsmodels
# Verificando se a proporcao de pessoas que preferem a marca A é maior que 50%
# 120 de 200 entrevistados preferem a marca A

n = 200
x = 120
p0 = 0.50
alpha = 0.05

# proportions_ztest(count, nobs, value) faz teste bicaudal por padrao
z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0)

print("Usando statsmodels.proportions_ztest:")
print(f"x = {x}, n = {n}")
print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
print(f"Estatística Z = {z_stat:.4f}")
print(f"p-valor = {p_val:.4f}")
if p_val < alpha:
    print(f"✅ p-valor {p_val:.4f} < α = {alpha} → Rejeitamos H₀")
else:
    print(f"❌ p-valor {p_val:.4f} ≥ α = {alpha} → Não rejeitamos H₀")
```

```python
# Exemplo 3: Teste unilateral a direita
# Um programa de vacinacao afirma que pelo menos 90% da populacao esta vacinada.
# Uma amostra de 500 pessoas mostra que 460 estao vacinadas.
# Teste se a proporcao é maior que 90%.

n = 500
x = 460
p0 = 0.90
alpha = 0.05

# Para teste unilateral a direita: alternative='larger'
z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0, alternative='larger')

print(f"x = {x}, n = {n}")
print(f"Proporção amostral = {x/n:.3f}")
print(f"H₀: p ≤ {p0} | H₁: p > {p0}")
print(f"Estatística Z = {z_stat:.4f}")
print(f"p-valor = {p_val:.4f}")
if p_val < alpha:
    print(f"✅ Rejeitamos H₀: a proporção de vacinados é maior que 90%.")
else:
    print(f"❌ Não rejeitamos H₀: não há evidência suficiente.")
```

```python
# Exemplo 4: Visualizacao do teste de proporcao
p_hat_val = x / n
erro = np.sqrt(p0 * (1 - p0) / n)
z_val = (p_hat_val - p0) / erro

z_vals = np.linspace(-4, 4, 1000)
densidade = stats.norm.pdf(z_vals, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(z_vals, densidade, 'b-', label='Normal Padrão')

# Regiao de rejeicao (unilateral direita)
z_crit_uni = stats.norm.ppf(1 - alpha)
x_fill = np.linspace(z_crit_uni, 4, 100)
plt.fill_between(x_fill, stats.norm.pdf(x_fill), color='red', alpha=0.3, label='Região de Rejeição')

plt.axvline(z_val, color='green', linestyle='--', lw=2, label=f'Z = {z_val:.2f}')
plt.axvline(z_crit_uni, color='red', linestyle=':', lw=2, label=f'Z crítico = {z_crit_uni:.2f}')
plt.title('Teste de Proporção Unilateral à Direita')
plt.xlabel('Z')
plt.ylabel('Densidade')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Exemplo 5: Funcao completa de teste de proporcao
def teste_proporcao(x, n, p0, alpha=0.05, alternativa='two-sided'):
    """Realiza teste Z para proporcao.
    alternativa: 'two-sided', 'larger', 'smaller'"""
    from statsmodels.stats.proportion import proportions_ztest
    z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0, alternative=alternativa)
    rejeita = p_val < alpha
    
    print(f"Resultado do Teste de Proporção:")
    print(f"  Proporção amostral: {x/n:.4f}")
    print(f"  Proporção H₀: {p0}")
    print(f"  Estatística Z: {z_stat:.4f}")
    print(f"  p-valor: {p_val:.4f}")
    print(f"  α: {alpha}")
    print(f"  {'✅ Rejeitamos H₀' if rejeita else '❌ Não rejeitamos H₀'}")
    return {"z": z_stat, "p_valor": p_val, "rejeita": rejeita}

# Testando a funcao
teste_proporcao(x=68, n=200, p0=0.40, alpha=0.05)
print()
teste_proporcao(x=120, n=200, p0=0.50, alpha=0.05, alternativa='larger')
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Uma universidade afirma que 55% dos seus alunos são do sexo feminino. Em uma amostra de 300 alunos, 144 são mulheres. Teste a afirmação a 5% de significância.

```python
# Solucao do Exercicio 1
n = 300
x = 144
p0 = 0.55
alpha = 0.05

p_hat = x / n
erro = np.sqrt(p0 * (1 - p0) / n)
z = (p_hat - p0) / erro
p_val = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Proporção amostral: {p_hat:.3f}")
print(f"H₀: p = {p0} | H₁: p ≠ {p0}")
print(f"Z = {z:.4f}, p-valor = {p_val:.4f}")
if p_val < alpha:
    print("✅ Rejeitamos H₀: a proporção de mulheres é diferente de 55%.")
else:
    print("❌ Não rejeitamos H₀.")
```

**Exercício 2:** Uma empresa afirma que menos de 5% das suas peças são defeituosas. Em 400 peças testadas, 14 apresentaram defeitos. Teste a afirmação a 1% de significância (teste unilateral).

```python
# Solucao do Exercicio 2
n = 400
x = 14
p0 = 0.05
alpha = 0.01

z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0, alternative='smaller')

print(f"Proporção amostral: {x/n:.4f}")
print(f"H₀: p ≥ {p0} | H₁: p < {p0}")
print(f"Z = {z_stat:.4f}, p-valor = {p_val:.4f}")
if p_val < alpha:
    print("✅ Rejeitamos H₀: a proporção de defeitos é menor que 5%.")
else:
    print("❌ Não rejeitamos H₀: não há evidência suficiente.")
```

**Exercício 3:** Uma pesquisa eleitoral com 1000 eleitores mostra que 530 votariam no candidato X. O candidato precisa de mais de 50% para vencer. Teste a 5% se ele tem mais de 50% dos votos.

```python
# Solucao do Exercicio 3
n = 1000
x = 530
p0 = 0.50
alpha = 0.05

z_stat, p_val = proportions_ztest(count=x, nobs=n, value=p0, alternative='larger')

print(f"Proporção amostral: {x/n:.3f}")
print(f"H₀: p ≤ {p0} | H₁: p > {p0}")
print(f"Z = {z_stat:.4f}, p-valor = {p_val:.4f}")
if p_val < alpha:
    print("✅ Rejeitamos H₀: o candidato tem mais de 50% dos votos.")
else:
    print("❌ Não rejeitamos H₀: não há evidência de vantagem.")
```

## 🏆 Desafios

1. Pesquise sobre o **teste exato de Fisher** (Fisher's exact test). Quando ele é preferível ao teste Z para proporção?
2. Implemente uma simulação de Monte Carlo para estimar o p-valor de um teste de proporção, sem usar a aproximação normal.
3. Investigue como o tamanho amostral afeta a potência do teste de proporção. Crie um gráfico de potência vs n.

```python
# Desafio 2: Simulacao Monte Carlo para p-valor
np.random.seed(123)
n_sim = 10000
n = 200
p0 = 0.40
x_obs = 68

# Gerando dados sob H0
contagem_extrema = 0
for i in range(n_sim):
    sim = np.random.binomial(n=n, p=p0)
    if abs(sim - n*p0) >= abs(x_obs - n*p0):
        contagem_extrema += 1

p_val_mc = contagem_extrema / n_sim
print(f"p-valor (Monte Carlo): {p_val_mc:.4f}")
print(f"p-valor (aproximação normal): {2 * (1 - stats.norm.cdf(abs((x_obs/n - p0) / np.sqrt(p0*(1-p0)/n)))):.4f}")
```

## 📌 Resumo

- O **teste Z para proporção** avalia hipóteses sobre a proporção populacional.
- A estatística Z = (p̂ - p₀) / √(p₀(1-p₀)/n) segue distribuição normal.
- Condições: n·p₀ ≥ 10 e n·(1-p₀) ≥ 10, amostra aleatória independente.
- Podemos usar o **proportions_ztest** do statsmodels ou calcular manualmente.
- O teste pode ser bicaudal, unilateral à direita ou unilateral à esquerda.
- A interpretação segue a mesma lógica de p-valor e valor crítico dos testes de hipótese.

## 💡 Curiosidades

O teste Z para proporção é uma aplicação direta do **Teorema Central do Limite**: a distribuição amostral da proporção se aproxima de uma normal quando n é grande. Na prática, a regra n·p₀ ≥ 10 garante uma aproximação razoável. Para amostras pequenas, devemos usar o **teste exato binomial**, que calcula a probabilidade diretamente da distribuição binomial sem aproximação.

## ✅ Boas Práticas

1. Sempre verifique as condições de aplicação (n·p₀ ≥ 10) antes de usar o teste Z.
2. Escolha o tipo de teste (bicaudal ou unilateral) com base na pergunta de pesquisa.
3. Use o proportions_ztest do statsmodels para menos código e menos erros.
4. Reporte o intervalo de confiança da proporção junto com o teste.
5. Para amostras pequenas, considere o teste exato binomial.

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Usar teste Z com amostra pequena | Verifique n·p₀ ≥ 10 e n·(1-p₀) ≥ 10 |
| Confundir proporção amostral com populacional | p̂ é da amostra, p₀ é da hipótese |
| Esquecer de usar o erro padrão sob H₀ | O erro usa p₀, não p̂ |
| Não especificar o tipo de teste | Sempre defina se é bicaudal ou unilateral |
| Interpretar o p-valor como P(H₀|dados) | p-valor é P(dados|H₀) |
