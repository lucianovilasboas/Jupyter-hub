## 📝 Introdução

O **desvio padrão** é a medida de dispersão mais usada em estatística. Ele nos diz, em média, **quanto cada valor se desvia da média**.

**Por que é especial?**
- Está na **mesma unidade** dos dados
- Base para a **regra empírica** (68-95-99.7)
- Fundamental para **inferência estatística**
- Usado no **Z-score** e testes de hipótese

> 🧠 **Para refletir:** Se a altura média é 170cm com DP=10cm, ~68% dos alunos têm entre 160cm e 180cm.

**Notação:**
- σ (sigma) — desvio padrão **populacional**
- s — desvio padrão **amostral**
- σ² — variância populacional
- s² — variância amostral

## 📚 Explicação Detalhada

### 1. Variância

$$\sigma^2 = \frac{\sum (x_i - \mu)^2}{n}$$

**Problema:** está em unidades **ao quadrado** (ex: cm²).

### 2. Desvio Padrão

$$\sigma = \sqrt{\sigma^2}$$

Volta para a **unidade original**.

### 3. Cálculo Passo a Passo

Dados: `{5, 7, 9, 11, 13}`
1. Média: 9
2. Desvios: -4, -2, 0, 2, 4
3. Desvios²: 16, 4, 0, 4, 16
4. Soma: 40
5. Variância: 40/5 = 8
6. DP: √8 = 2.83

### 4. Regra Empírica (68-95-99.7)

Para distribuições **normais**:
- **68%** a ±1σ da média
- **95%** a ±2σ da média
- **99.7%** a ±3σ da média

### 5. σ vs s

| | População | Amostra |
|--------|-----------|---------|
| Variância | σ² = Σ(xi-μ)²/N | s² = Σ(xi-x̄)²/(n-1) |
| DP | σ | s |

Na amostra, dividimos por n-1 (correção de Bessel).

### 6. Coeficiente de Variação

$$CV = \frac{s}{\bar{x}} \times 100\%$$

Compara dispersão relativa entre conjuntos com unidades diferentes.

## 💻 Exemplos Práticos

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Exemplo 1: Calculo passo a passo
dados = np.array([5, 7, 9, 11, 13])
media = np.mean(dados)
desvios = dados - media
desvios2 = desvios ** 2
variancia = np.mean(desvios2)
desvio_padrao = np.sqrt(variancia)
print("Calculo passo a passo:")
print(f"Dados: {dados}")
print(f"Media: {media}")
print(f"Desvios: {desvios}")
print(f"Desvios²: {desvios2}")
print(f"Soma dos desvios²: {sum(desvios2)}")
print(f"Variancia (pop): {variancia:.2f}")
print(f"Desvio padrao: {desvio_padrao:.2f}")
print(f"\nConfirmando com numpy — np.var: {np.var(dados):.2f}, np.std: {np.std(dados):.2f}")
```

```python
# Exemplo 2: Populacional vs Amostral
populacao = np.random.normal(100, 15, 100000)
amostra = np.random.choice(populacao, 30)
print("Desvio Padrao: Populacao vs Amostra")
print(f"Populacao — sigma real: {np.std(populacao):.2f}")
print(f"Amostra — s (ddof=0): {np.std(amostra, ddof=0):.2f}  (dividindo por n)")
print(f"Amostra — s (ddof=1): {np.std(amostra, ddof=1):.2f}  (dividindo por n-1)")
print("ddof=1 (amostral) e mais proximo do valor real da populacao.")
```

```python
# Exemplo 3: Regra Empirica (68-95-99.7)
normal = np.random.normal(100, 15, 100000)
media = np.mean(normal); std = np.std(normal)
dentro_1 = np.sum(np.abs(normal - media) <= std)
dentro_2 = np.sum(np.abs(normal - media) <= 2 * std)
dentro_3 = np.sum(np.abs(normal - media) <= 3 * std)
total = len(normal)
print("Regra Empirica — Dados Normais:")
print(f"  Dentro de 1σ (±{std:.1f}): {dentro_1/total*100:.1f}%  (esperado: 68%)")
print(f"  Dentro de 2σ (±{2*std:.1f}): {dentro_2/total*100:.1f}%  (esperado: 95%)")
print(f"  Dentro de 3σ (±{3*std:.1f}): {dentro_3/total*100:.1f}%  (esperado: 99.7%)")
```

```python
# Exemplo 4: Visualizacao da regra empirica
plt.figure(figsize=(10, 6))
count, bins, _ = plt.hist(normal, bins=60, color='skyblue', edgecolor='black', alpha=0.7, density=True)
plt.axvline(media - std, color='red', linestyle='--', linewidth=2, label=f'±1σ ({media-std:.1f} a {media+std:.1f})')
plt.axvline(media + std, color='red', linestyle='--', linewidth=2)
plt.axvline(media - 2*std, color='green', linestyle='-.', linewidth=2, label=f'±2σ')
plt.axvline(media + 2*std, color='green', linestyle='-.', linewidth=2)
plt.axvline(media - 3*std, color='purple', linestyle=':', linewidth=2, label=f'±3σ')
plt.axvline(media + 3*std, color='purple', linestyle=':', linewidth=2)
plt.axvline(media, color='black', linewidth=2, label=f'Media = {media:.1f}')
plt.fill_between(bins, 0, 1, where=(bins >= media-std) & (bins <= media+std), color='red', alpha=0.05)
plt.fill_between(bins, 0, 1, where=(bins >= media-2*std) & (bins <= media+2*std), color='green', alpha=0.05)
plt.fill_between(bins, 0, 1, where=(bins >= media-3*std) & (bins <= media+3*std), color='purple', alpha=0.05)
plt.xlabel('Valor'); plt.ylabel('Densidade'); plt.title('Regra Empirica (68-95-99.7) — Distribuicao Normal')
plt.legend(); plt.grid(True, alpha=0.3); plt.show()
```

```python
# Exemplo 5: Variancia vs Desvio Padrao
dados = np.array([10, 12, 15, 18, 20, 22, 25])
var = np.var(dados); std = np.std(dados)
print(f"Dados: {dados}")
print(f"Variancia: {var:.2f} (unidade² — dificil interpretar)")
print(f"Desvio Padrao: {std:.2f} (mesma unidade — facil interpretar)")
print(f"\nSe os dados estao em cm, a variancia esta em cm² e o DP em cm.")
```

```python
# Exemplo 6: Coeficiente de Variacao
alturas = np.random.normal(170, 10, 100)
pesos = np.random.normal(70, 8, 100)
cv_alt = np.std(alturas) / np.mean(alturas) * 100
cv_peso = np.std(pesos) / np.mean(pesos) * 100
print(f"Altura — Media: {np.mean(alturas):.1f}cm, DP: {np.std(alturas):.1f}cm, CV: {cv_alt:.1f}%")
print(f"Peso — Media: {np.mean(pesos):.1f}kg, DP: {np.std(pesos):.1f}kg, CV: {cv_peso:.1f}%")
print("\nEmbora o DP da altura seja maior, o CV mostra que o peso tem maior")
print("variabilidade relativa (proporcional a media).")
```

## 📝 Exercícios Resolvidos

```python
# Exercicio 1: DP de temperaturas
temps = np.array([28, 30, 27, 31, 29, 33, 26])
print(f"Temperaturas: {temps}")
print(f"Media: {np.mean(temps):.1f}C")
print(f"Variancia: {np.var(temps):.2f}C²")
print(f"Desvio Padrao: {np.std(temps):.2f}C")
print(f"Regra empirica: 68% entre {np.mean(temps)-np.std(temps):.1f}C e {np.mean(temps)+np.std(temps):.1f}C")
```

```python
# Exercicio 2: Comparando DP e CV
grupo1 = np.random.normal(50, 5, 100)
grupo2 = np.random.normal(200, 20, 100)
print(f"Grupo 1 — Media={np.mean(grupo1):.1f}, DP={np.std(grupo1):.1f}, CV={np.std(grupo1)/np.mean(grupo1)*100:.1f}%")
print(f"Grupo 2 — Media={np.mean(grupo2):.1f}, DP={np.std(grupo2):.1f}, CV={np.std(grupo2)/np.mean(grupo2)*100:.1f}%")
print("\nDP do grupo 2 e maior, mas o CV e similar — a variabilidade relativa e parecida.")
```

## 📝 Resumo do Capítulo

- **Desvio Padrão (DP)**: raiz quadrada da variância
- **Variância**: média dos desvios ao quadrado (unidade²)
- **Regra empírica**: 68%-95%-99.7% para dados normais
- σ = DP populacional, s = DP amostral (ddof=1)
- **CV**: coeficiente de variação (DP/média × 100%)
- O DP é a base para inferência estatística e Z-score

## 🧐 Curiosidades

- O símbolo σ foi popularizado por **Karl Pearson** em 1894.
- Em finanças, o DP é chamado de **volatilidade** — o risco de um ativo.
- A **regra empírica** é tão precisa que é usada para controle de qualidade em indústrias.
- O Six Sigma (6σ) busca ter apenas 3.4 defeitos por milhão de oportunidades.

## ✅ Boas Práticas

- Sempre reporte **média ± DP**
- Use **ddof=1** para amostras
- Use **CV** para comparar dispersão entre escalas diferentes
- Verifique a **normalidade** antes de aplicar a regra empírica
- O DP é sensível a outliers — use IQR como alternativa robusta

## ⚠️ Erros Comuns

| Erro | Consequência | Correção |
|------|-------------|----------|
| Confundir σ e s | Viés na estimativa | Use ddof=1 para amostra |
| Ignorar a regra empírica | Interpretação errada | Só vale para dados normais |
| Não usar √ da variância | Unidade ao quadrado | Use np.std() |
| Comparar DP de escalas diferentes | Inválido | Use CV |
| Achar que DP zero = dados bons | Pode ser trivial | Verifique os dados |
