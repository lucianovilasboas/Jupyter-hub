## 📝 Introdução

Muitas vezes queremos estimar a **proporção** (percentual) de uma população que possui uma determinada característica: a intenção de voto em um candidato, a proporção de clientes satisfeitos, o percentual de itens defeituosos em uma fábrica.

Neste capítulo, aprenderemos a construir intervalos de confiança para proporções populacionais — uma das técnicas mais usadas em pesquisas de opinião, controle de qualidade e epidemiologia.

## 📚 Explicação Detalhada

### 1. Proporção Populacional (p)

- **p**: Proporção verdadeira na população (desconhecida, o que queremos estimar)
- **p̂** (p-hat): Proporção amostral (estimativa pontual)

$$\hat{p} = \frac{\text{número de sucessos na amostra}}{n}$$

### 2. Intervalo de Confiança para Proporção

$$\hat{p} \pm z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

Onde:
- $z_{\alpha/2}$ é o valor crítico da normal padrão
- $\sqrt{\hat{p}(1-\hat{p})/n}$ é o erro padrão da proporção

### 3. Condições de Aplicação

Para que o IC seja válido, precisamos:
- **Amostra aleatória simples**
- **n × p̂ ≥ 10** (pelo menos 10 sucessos esperados)
- **n × (1 - p̂) ≥ 10** (pelo menos 10 fracassos esperados)
- A população deve ser pelo menos 20× maior que a amostra (independência)

### 4. Margem de Erro

A margem de erro é a metade da largura do IC:

$$ME = z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

É comum vermos "margem de erro de 2 pontos percentuais" em pesquisas eleitorais.

```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Bibliotecas importadas com sucesso!")
```

## 💻 Exemplos Práticos

```python
# Exemplo 1: Pesquisa eleitoral
# 1200 eleitores entrevistados, 540 disseram votar no candidato A

n = 1200
votos_a = 540
p_hat = votos_a / n

print("=== Pesquisa Eleitoral ===")
print(f"Total de entrevistados: {n}")
print(f"Intenção de voto em A: {votos_a} ({p_hat:.1%})")

# Verificando condições
sucessos = n * p_hat
fracassos = n * (1 - p_hat)
print(f"\nVerificação das condições:")
print(f"n × p̂ = {sucessos:.0f} ≥ 10? {sucessos >= 10}")
print(f"n × (1-p̂) = {fracassos:.0f} ≥ 10? {fracassos >= 10}")

# IC 95%
z_crit = stats.norm.ppf(0.975)
ep_prop = np.sqrt(p_hat * (1 - p_hat) / n)
margem = z_crit * ep_prop

ic_inf = p_hat - margem
ic_sup = p_hat + margem

print(f"\nProporção amostral (p̂): {p_hat:.4f} ({p_hat:.1%})")
print(f"Erro padrão: {ep_prop:.4f}")
print(f"Valor crítico Z (95%): {z_crit:.4f}")
print(f"Margem de erro: {margem:.4f} ({margem:.1%})")
print(f"IC 95% para intenção de voto: [{ic_inf:.4f}, {ic_sup:.4f}]")
print(f"IC 95%: [{ic_inf:.1%}, {ic_sup:.1%}]")
```

```python
# Exemplo 2: Margem de erro em pesquisas
# Mostramos como a margem de erro depende do tamanho da amostra

tamanhos = [100, 250, 500, 800, 1000, 1500, 2000, 3000, 5000]
p_hat_exemplo = 0.5  # pior caso (margem máxima)

print("=== Efeito do Tamanho da Amostra na Margem de Erro ===")
print(f"{'n':<8} {'Margem (95%)':<16} {'Margem (99%)':<16}")
print("-" * 40)

for n_atual in tamanhos:
    ep_atual = np.sqrt(p_hat_exemplo * (1 - p_hat_exemplo) / n_atual)
    margem_95 = stats.norm.ppf(0.975) * ep_atual
    margem_99 = stats.norm.ppf(0.995) * ep_atual
    print(f"{n_atual:<8} ±{margem_95:<5.1%}          ±{margem_99:<5.1%}")
```

```python
# Exemplo 3: Simulando uma pesquisa eleitoral
# Criamos uma população com proporção real e verificamos o IC

np.random.seed(42)

proporcao_real = 0.45  # 45% dos eleitores preferem o candidato
populacao_eleitores = np.random.binomial(1, proporcao_real, size=100000)

# Coletamos amostras de diferentes tamanhos
for n_amostra in [100, 400, 1000]:
    amostra_eleitores = np.random.choice(populacao_eleitores, n_amostra, replace=False)
    p_hat_sim = np.mean(amostra_eleitores)
    
    ep_sim = np.sqrt(p_hat_sim * (1 - p_hat_sim) / n_amostra)
    z_crit = stats.norm.ppf(0.975)
    margem_sim = z_crit * ep_sim
    
    ic_sim = (p_hat_sim - margem_sim, p_hat_sim + margem_sim)
    contem = ic_sim[0] <= proporcao_real <= ic_sim[1]
    
    print(f"n = {n_amostra:4d}: p̂ = {p_hat_sim:.3f}, IC 95% = [{ic_sim[0]:.3f}, {ic_sim[1]:.3f}], "
          f"contém p? {contem}")
```

```python
# Exemplo 4: Visualização do IC para proporção
# Vários cenários de proporção amostral

n = 500
cenarios = [("Candidato A", 0.52), ("Candidato B", 0.48), ("Aberto", 0.35), (
            "Rejeição", 0.12)]

fig, ax = plt.subplots(figsize=(10, 5))

for i, (nome, p_hat_cen) in enumerate(cenarios):
    ep_cen = np.sqrt(p_hat_cen * (1 - p_hat_cen) / n)
    margem_cen = stats.norm.ppf(0.975) * ep_cen
    
    ax.errorbar(p_hat_cen, i, xerr=margem_cen, fmt='o', capsize=5, capthick=2,
                 markersize=8)
    ax.text(p_hat_cen, i + 0.15, f'{p_hat_cen:.1%} ± {margem_cen:.1%}', 
            ha='center', fontsize=10)

ax.set_yticks(range(len(cenarios)))
ax.set_yticklabels([c[0] for c in cenarios])
ax.set_xlabel('Proporção', fontsize=12)
ax.set_title('IC 95% para Diferentes Cenários (n = 500)', fontsize=13)
ax.axvline(0.5, color='gray', linestyle=':', alpha=0.5, label='50% (maioria)')
ax.grid(True, alpha=0.3)
ax.legend()
plt.show()
```

```python
# Exemplo 5: Cálculo do tamanho de amostra necessário
# Qual n para uma margem de erro de 2% (IC 95%)?

margem_desejada = 0.02
p_previsto = 0.5  # pior caso (máxima variância)
z = stats.norm.ppf(0.975)

n_necessario = (z**2 * p_previsto * (1 - p_previsto)) / (margem_desejada**2)
n_necessario = int(np.ceil(n_necessario))

print("=== Cálculo do Tamanho da Amostra ===")
print(f"Margem de erro desejada: {margem_desejada:.1%}")
print(f"Nível de confiança: 95%")
print(f"Proporção estimada (pior caso): {p_previsto}")
print(f"n necessário: {n_necessario} eleitores")
print()
print("Para diferentes margens:")
for margem in [0.01, 0.02, 0.03, 0.05]:
    n_req = int(np.ceil((z**2 * 0.5 * 0.5) / margem**2))
    print(f"Margem ±{margem:.0%}: n = {n_req}")
```

```python
# Exemplo 6: Controle de qualidade — proporção de defeitos
# 50 peças inspecionadas, 3 apresentaram defeito

n_pecas = 50
defeitos = 3
p_def = defeitos / n_pecas

print("=== Controle de Qualidade ===")
print(f"Peças inspecionadas: {n_pecas}")
print(f"Peças com defeito: {defeitos}")
print(f"Proporção de defeitos (p̂): {p_def:.4f} ({p_def:.2%})")

# Verificação das condições
sucessos_def = n_pecas * p_def
fracassos_def = n_pecas * (1 - p_def)
print(f"\nCondições: n×p̂ = {sucessos_def:.0f}, n×(1-p̂) = {fracassos_def:.0f}")
print(f"Condições satisfeitas? {sucessos_def >= 10 and fracassos_def >= 10}")

if sucessos_def >= 10 and fracassos_def >= 10:
    ep_def = np.sqrt(p_def * (1 - p_def) / n_pecas)
    margem_def = stats.norm.ppf(0.975) * ep_def
    print(f"IC 95% para % de defeitos: [{p_def - margem_def:.4f}, {p_def + margem_def:.4f}]")
    print(f"IC 95%: [{max(0, p_def - margem_def):.2%}, {min(1, p_def + margem_def):.2%}]")
else:
    print("Condições não satisfeitas para IC aproximado.")
    print("Com poucos defeitos, usar método exato (binomial).")
```

## ✅ Exercícios Resolvidos

**Exercício 1:** Uma pesquisa com 800 eleitores mostra que 320 votariam no candidato X. Construa o IC de 95% para a verdadeira proporção de votos do candidato.

```python
# Solução Exercício 1
n = 800
sucesso = 320
p_hat = sucesso / n

print(f"p̂ = {sucesso}/{n} = {p_hat:.4f}")
print(f"n×p̂ = {n*p_hat:.0f} ≥ 10? {n*p_hat >= 10}")
print(f"n×(1-p̂) = {n*(1-p_hat):.0f} ≥ 10? {n*(1-p_hat) >= 10}")

ep = np.sqrt(p_hat * (1 - p_hat) / n)
z = stats.norm.ppf(0.975)
margem = z * ep

print(f"\nErro padrão: {ep:.4f}")
print(f"Margem de erro: {margem:.4f} ({margem:.1%})")
print(f"IC 95%: [{p_hat - margem:.4f}, {p_hat + margem:.4f}]")
print(f"IC 95%: [{p_hat - margem:.1%}, {p_hat + margem:.1%}]")
```

**Exercício 2:** Quantas pessoas devem ser entrevistadas para uma margem de erro de 3% (IC 95%) supondo p = 0.5?

```python
# Solução Exercício 2
margem_alvo = 0.03
p_est = 0.5
z_alpha = stats.norm.ppf(0.975)

n_necessario = (z_alpha**2 * p_est * (1 - p_est)) / (margem_alvo**2)
n_necessario = int(np.ceil(n_necessario))

print(f"Margem desejada: {margem_alvo:.0%}")
print(f"Z crítico (95%): {z_alpha:.4f}")
print(f"n necessário: {n_necessario}")
print(f"\nNota: pesquisas eleitorais sérias costumam entrevistar ~1067 pessoas
para margem de ±3%.")
```

## 📌 Resumo

- **Proporção amostral (p̂)**: estimativa pontual da proporção populacional.
- **IC para proporção**: $\hat{p} \pm z \sqrt{\hat{p}(1-\hat{p})/n}$.
- **Condições**: np̂ ≥ 10 e n(1-p̂) ≥ 10.
- **Margem de erro**: depende de n, nível de confiança e p̂.
- **Pior caso**: p̂ = 0.5 → margem máxima → usar para calcular n.

## 💡 Curiosidades

## ✅ Boas Práticas

1. Sempre verifique as condições np̂ ≥ 10 e n(1-p̂) ≥ 10 antes de usar o IC aproximado.
2. Use p̂ = 0.5 para calcular o tamanho da amostra quando não tem ideia da proporção.
3. Reporte a margem de erro junto com a estimativa pontual.
4. Lembre-se: a margem de erro não cobre viés de seleção ou erros não-amostrais.
5. Para proporções próximas de 0 ou 1, considere métodos alternativos (Wilson, exato binomial).

## ⚠️ Erros Comuns

| Erro | Como Evitar |
|------|-------------|
| Aplicar o IC sem verificar as condições | Sempre confira np̂ ≥ 10 e n(1-p̂) ≥ 10. |
| Interpretar margem de erro como garantia absoluta | A margem considera apenas erro amostral, não erros de coleta. |
| Usar Z ao invés de t | Para proporção usamos Z (não t), pois o erro padrão usa a própria estimativa. |
| Esquecer que p̂ = 0 ou p̂ = 1 dão erro padrão zero | Isso não significa que a estimativa é perfeita! |
