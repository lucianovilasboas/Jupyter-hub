## Enquete do Campeonato

Faltam 3 rodadas para o fim do campeonato.
Um instituto de pesquisa entrevista 1000 torcedores:
**600 dizem que o time A vai ser campeao.**

Proporcao na amostra: 600 / 1000 = **60%**

Mas 60% e o valor real? Ou pode ser 58%? 62%?

O **intervalo de confianca para proporcao** responde:
"Com 95% de confianca, a verdadeira intencao de torcida
esta entre 57% e 63%".

Quanto mais pessoas entrevistadas, menor a margem de erro.


## 📊 Traduzindo

| Enquete do Campeonato | Estatistica |
|-|-|
| 600 de 1000 torcedores | Proporcao amostral (p) |
| Intencao REAL de todos | Proporcao populacional (P) |
| Margem de erro da pesquisa | Erro padrao da proporcao |
| "Entre 57% e 63%" | Intervalo de confianca |


## 📏 Calculo da Margem de Erro

A margem de erro depende de:
1. **Tamanho da amostra (n)**: quanto maior, menor o erro
2. **Proporcao observada (p)**: quanto mais perto de 50%,
   maior a variabilidade (maior o erro)
3. **Nivel de confianca**: 99% da margem maior que 95%


```python
# Enquete do campeonato
import math
n = 1000
p = 600 / n  # proporcao amostral
erro = 1.96 * math.sqrt(p * (1-p) / n)  # IC 95%
ic_inf = p - erro
ic_sup = p + erro
print(f"Proporcao observada: {p*100:.1f}%")
print(f"Margem de erro: {erro*100:.1f}%")
print(f"IC 95%: [{ic_inf*100:.1f}%, {ic_sup*100:.1f}%]")
```

## 📌 Resumo

| Enquete do Campeonato | Estatistica |
|-|-|
| 600 de 1000 | Proporcao amostral |
| Intencao real | Proporcao populacional |
| Margem de erro | Erro padrao |
| 57% a 63% | Intervalo de confianca |
