## 📝 Introdução

Quando temos um conjunto de dados, uma das primeiras perguntas que fazemos é: **qual é o valor típico?** O que chamamos de "average" (média) é na verdade um conceito mais amplo: queremos um número que represente o centro dos dados.

A palavra **média** no dia a dia geralmente se refere à média aritmética (mean). Mas em estatística, existem **três medidas principais** de tendência central:

| Medida | Nome em português | O que representa |
|--------|-------------------|-----------------|
| **Mean** | Média aritmética | Soma dos valores dividida pela quantidade |
| **Median** | Mediana | Valor que divide os dados ao meio (50%) |
| **Mode** | Moda | Valor que mais aparece nos dados |

Cada uma conta uma história diferente sobre os dados. A escolha correta depende da **forma da distribuição** e da **presença de outliers**.

> 🧠 **Para refletir:** Se você tem 5 amigos com salários de R$ 2.000, R$ 2.200, R$ 2.500, R$ 3.000 e R$ 50.000, qual valor representa "o salário típico do grupo"? A média é R$ 11.940, mas será que isso descreve bem o grupo?

## 📚 Explicação Detalhada

### 1. O Conceito de Tendência Central

Tendência central é a tendência dos dados de se agruparem em torno de um valor central. As medidas de tendência central nos dão um **resumo numérico** do conjunto de dados.

Imagine que você mediu a altura de 100 estudantes. As alturas variam, mas existe uma faixa onde a maioria se concentra. A medida de tendência central captura essa faixa.

### 2. As Três Medidas

**Mean (Média Aritmética):**
- Soma de todos os valores dividida pela quantidade
- Representada por $\bar{x}$ (amostra) ou $\mu$ (população)
- Sensível a outliers (valores extremos)

**Median (Mediana):**
- Valor do meio quando os dados estão ordenados
- Se n é ímpar: valor central
- Se n é par: média dos dois valores centrais
- Robusta a outliers

**Mode (Moda):**
- Valor que ocorre com maior frequência
- Pode haver mais de uma moda (bimodal, multimodal)
- Única medida que funciona para dados qualitativos

### 3. Simetria e Assimetria

A relação entre mean, median e mode revela a **forma da distribuição**:

| Distribuição | Relação | Exemplo |
|-------------|---------|---------|
| **Simétrica** (Normal) | Mean ≈ Median ≈ Mode | Altura de adultos |
| **Assimetria Positiva** (à direita) | Mean > Median > Mode | Renda, salários |
| **Assimetria Negativa** (à esquerda) | Mean < Median < Mode | Idade de aposentadoria |

Quando a distribuição é **simétrica**, as três medidas coincidem. Quando há **assimetria**, elas se separam: a mean é puxada na direção da cauda, enquanto a median permanece no centro.

### 4. Quando Usar Cada Medida

| Situação | Medida Recomendada | Motivo |
|----------|-------------------|--------|
| Dados simétricos sem outliers | Mean | Mais eficiente estatisticamente |
| Dados com outliers ou assimétricos | Median | Robusta a valores extremos |
| Dados qualitativos (categorias) | Mode | Única medida aplicável |
| Distribuição bimodal | Mode (ambas) | Revela grupos nos dados |
| Relatórios financeiros/salários | Median | Não é distorcida por extremos |

## 💻 Exemplos Práticos

```python
# Importando as bibliotecas necessárias
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Configurando a semente para reprodutibilidade
np.random.seed(42)
```

```python
# Exemplo 1: Distribuição simétrica (Normal)
# Gerando dados de altura (cm) de 1000 pessoas
alturas = np.random.normal(loc=170, scale=10, size=1000)

# Calculando as três medidas
mean_alt = np.mean(alturas)
median_alt = np.median(alturas)
mode_alt = stats.mode(alturas, keepdims=True)[0][0]

print("Dados de Altura (Distribuição Simétrica)")
print(f"Mean:   {mean_alt:.2f} cm")
print(f"Median: {median_alt:.2f} cm")
print(f"Mode:   {mode_alt:.2f} cm")
print()
print("Observação: Em uma distribuição simétrica,")
print("as três medidas são muito próximas.")
```

```python
# Exemplo 2: Distribuição assimétrica (Renda)
# Gerando dados de renda (reais) com assimetria positiva
renda = np.random.exponential(scale=3000, size=1000)

mean_renda = np.mean(renda)
median_renda = np.median(renda)
mode_renda = stats.mode(np.round(renda, -2), keepdims=True)[0][0]

print("Dados de Renda (Distribuição Assimétrica à Direita)")
print(f"Mean:   R$ {mean_renda:.2f}")
print(f"Median: R$ {median_renda:.2f}")
print(f"Mode:   R$ {mode_renda:.2f}")
print()
print("Observação: Mean > Median, pois a média é")
print("puxada para a direita pelos valores altos.")
```

```python
# Exemplo 3: Visualização das três medidas
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histograma - Altura (simétrica)
axes[0].hist(alturas, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].axvline(mean_alt, color='red', linestyle='--', linewidth=2, label=f'Mean={mean_alt:.1f}')
axes[0].axvline(median_alt, color='green', linestyle='-.', linewidth=2, label=f'Median={median_alt:.1f}')
axes[0].axvline(mode_alt, color='purple', linestyle=':', linewidth=2, label=f'Mode={mode_alt:.1f}')
axes[0].set_title('Distribuição Simétrica — Altura (cm)')
axes[0].set_xlabel('Altura (cm)')
axes[0].set_ylabel('Frequência')
axes[0].legend()

# Histograma - Renda (assimétrica)
axes[1].hist(renda, bins=30, color='salmon', edgecolor='black', alpha=0.7)
axes[1].axvline(mean_renda, color='red', linestyle='--', linewidth=2, label=f'Mean={mean_renda:.0f}')
axes[1].axvline(median_renda, color='green', linestyle='-.', linewidth=2, label=f'Median={median_renda:.0f}')
axes[1].axvline(mode_renda, color='purple', linestyle=':', linewidth=2, label=f'Mode={mode_renda:.0f}')
axes[1].set_title('Distribuição Assimétrica — Renda (R$)')
axes[1].set_xlabel('Renda (R$)')
axes[1].set_ylabel('Frequência')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Exemplo 4: Dados com outliers
# Simulando notas de uma turma com um aluno que tirou nota muito baixa
notas = np.array([7.0, 7.5, 8.0, 6.5, 7.2, 8.5, 6.8, 7.8, 2.0, 7.3])

print("Notas da Turma (com outlier: 2.0)")
print(f"Notas: {notas}")
print(f"Mean:   {np.mean(notas):.2f}")
print(f"Median: {np.median(notas):.2f}")

# Removendo o outlier
notas_sem_outlier = notas[notas > 3.0]
print()
print("Sem o outlier:")
print(f"Notas: {notas_sem_outlier}")
print(f"Mean:   {np.mean(notas_sem_outlier):.2f}")
print(f"Median: {np.median(notas_sem_outlier):.2f}")
print()
print("A mediana foi menos afetada pelo outlier do que a média.")
```

## 📝 Exercícios Resolvidos

```python
# Exercício Resolvido 1
# Calcule mean, median e mode para as idades de um grupo de 15 pessoas

idades = np.array([22, 25, 29, 22, 30, 35, 28, 22, 31, 27, 26, 24, 29, 23, 40])

media = np.mean(idades)
mediana = np.median(idades)
moda = stats.mode(idades, keepdims=True)[0][0]

print(f"Idades: {idades}")
print(f"Mean:   {media:.2f}")
print(f"Median: {mediana:.1f}")
print(f"Mode:   {moda}")
print()
print("A moda é 22 anos, que aparece 3 vezes.")
```

```python
# Exercício Resolvido 2
# Verifique se a distribuição abaixo é simétrica ou assimétrica

dados = np.random.gamma(shape=2, scale=2, size=1000)

media = np.mean(dados)
mediana = np.median(dados)

print(f"Mean:   {media:.2f}")
print(f"Median: {mediana:.2f}")

if abs(media - mediana) < 0.1 * media:
    print("A distribuição é aproximadamente simétrica.")
elif media > mediana:
    print("Assimetria positiva (à direita) — Mean > Median.")
else:
    print("Assimetria negativa (à esquerda) — Mean < Median.")
```

## 📝 Resumo do Capítulo

- **Average** é o conceito geral de medida de tendência central
- As três principais medidas são **mean** (média), **median** (mediana) e **mode** (moda)
- **Mean**: soma/quantidade, sensível a outliers
- **Median**: valor do meio, robusta a outliers
- **Mode**: valor mais frequente, funciona para dados qualitativos
- Distribuições **simétricas**: três medidas coincidem
- Distribuições **assimétricas**: mean é puxada na direção da cauda
- A escolha da medida correta depende da forma dos dados e da presença de outliers

## 🧐 Curiosidades

- O termo "average" vem do árabe **awarīyah** (mercadoria danificada), usado no comércio marítimo medieval para dividir prejuízos entre comerciantes.
- A **mediana** era conhecida como "a média dos pobres" no século XIX, pois era usada para calcular impostos de forma mais justa para a população de baixa renda.
- Uma distribuição pode ter **mais de uma moda**. O recorde é de uma distribuição com 15 modas diferentes em um estudo geológico.
- No esporte, o ** batting average** do beisebol é na verdade uma proporção, não uma média no sentido estatístico.

## ✅ Boas Práticas

- Sempre calcule **mais de uma medida** de tendência central para entender melhor os dados
- Antes de escolher a medida, faça um **gráfico** (histograma ou box plot)
- **Documente** qual medida foi usada e por quê
- Prefira a **mediana** para dados de renda, preços de imóveis e outras variáveis financeiras
- Use a **moda** para dados categóricos (cores, marcas, categorias)
- Desconfie da média se os dados tiverem **outliers** ou **assimetria**

## ⚠️ Erros Comuns

| Erro | Consequência | Correção |
|------|-------------|----------|
| Usar mean sem verificar outliers | Resultado distorcido | Calcule median também |
| Ignorar a forma da distribuição | Escolha errada da medida | Faça um histograma |
| Achar que average = média aritmética | Confusão conceitual | Lembre-se: average é o conceito geral |
| Usar mode para dados contínuos | Cada valor pode ser único | Arredonde ou agrupe os dados |
| Confundir simetria com normalidade | Distribuições simétricas podem não ser normais | Verifique curtose e testes |
